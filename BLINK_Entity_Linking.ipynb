```python
import blink.main_dense as main_dense
import argparse

# Assume BLINK models and data are set up correctly
models_path = "models/"  # Update this path to where your BLINK models are stored

# Configuration for BLINK
config = {
    "test_entities": None,
    "test_mentions": None,
    "interactive": False,
    "top_k": 3,
    "biencoder_model": models_path + "biencoder_wiki_large.bin",
    "biencoder_config": models_path + "biencoder_wiki_large.json",
    "entity_catalogue": models_path + "entity.jsonl",
    "entity_encoding": models_path + "all_entities_large.t7",
    "crossencoder_model": models_path + "crossencoder_wiki_large.bin",
    "crossencoder_config": models_path + "crossencoder_wiki_large.json",
    "fast": False,  # Set to True for faster results but possibly less accurate
    "output_path": "logs/"  # Logging directory
}

args = argparse.Namespace(**config)

# Load BLINK models based on the configuration
models = main_dense.load_models(args, logger=None)

# Use the 'formatted_data' from Part 1 as 'data_to_link' here
data_to_link = formatted_data

# Run BLINK entity linking on the formatted data
_, _, _, _, _, predictions, scores = main_dense.run(args, None, *models, test_data=data_to_link)


results_to_save = []
for i, prediction in enumerate(predictions):
    entity_mention = data_to_link[i]['mention']
    entity_type = data_to_link[i]['entity_type']
    prediction_details = [{
        "entity": entity,
        "confidence_score": str(score)
    } for entity, score in zip(prediction, scores[i])]

    results_to_save.append({
        "mention": entity_mention,
        "type": entity_type,
        "predictions": prediction_details
    })

# Define a path for the output file
output_path = 'predictions.json'

# Write the results to a JSON file
with open(output_path, 'w', encoding='utf-8') as f:
    json.dump(results_to_save, f, indent=4)

print(f"Predictions saved to {output_path}")
```

```python
def identify_long_tail_entities_from_predictions(predictions, threshold=-2):
    """
    Identify long-tail entities based on the highest confidence score of predicted links.

    :param predictions: A list of dicts with 'Entity Mention' and 'Predicted Links'.
    :param threshold: Score threshold for identifying long-tail entities.
    :return: A list of long-tail entity mentions.
    """
    long_tail_entities = []
    for prediction in predictions:
        # Get the highest score among the predicted links for the current entity
        highest_score = max(link['score'] for link in prediction['Predicted Links'])

        if highest_score < threshold:
            long_tail_entities.append(prediction['Entity Mention'])

    return long_tail_entities

# Example data structure
predictions = [
    {
        "Entity Mention": "tradingcharts",
        "Predicted Links": [
            {"link": "Online trading community", "score": -6.958751201629639},
            {"link": "Exchange (organized market)", "score": -7.257745265960693},
            # More predicted links...
        ]
    },
    {
        "Entity Mention": "exchange",
        "Predicted Links": [
            {"link": "Exchange (organized market)", "score": -4.462727069854736},
            {"link": "Stock exchange", "score": -4.596463203430176},
            # More predicted links...
        ]
    },
    # Add more entities and their predicted links...
]

long_tail_entities = identify_long_tail_entities_from_predictions(predictions, threshold=-10)
print("Long-tail entities:", long_tail_entities)
```

#Entity Mention
This is the text extracted from your input that BLINK has attempted to link to a Wikipedia entity. For example, "semtech corp" is identified as an entity mention in your text.

#Predicted Links
For each entity mention, BLINK provides a list of possible Wikipedia entities (or other linked data) that could correspond to the mention. These predictions are based on the similarity of the context in your input text to the information associated with potential Wikipedia entities.

#Scores
Each predicted link comes with a score. This score represents the model's confidence in the association between the entity mention and the Wikipedia entity. Higher scores indicate higher confidence. It's worth noting that the score can sometimes be negative, which typically happens when the model's confidence is lower. For some mentions, especially common names or terms, the highest score might be quite low, reflecting the ambiguity of the link.

#Interpretation Tips
# High Confidence Links:
Predictions with a significantly higher score compared to others for the same mention suggest a strong match. For example, if "Peta Credlin" has a high score linked to the correct Wikipedia page, it's a reliable match.
#Multiple Predictions:
When there are several closely scored predictions, it indicates ambiguity. This might require manual verification to determine the most accurate link.
Negative Scores: Even if a score is negative, if it's the highest score for that entity, it may still be the best match among the given options. Negative scores primarily indicate relative confidence.
# Context Matters:
The context in which the entity mention appears (words or phrases before and after the mention) plays a crucial role in determining the match. The same entity mention could be linked to different Wikipedia pages based on different contexts.
#Practical Use
Enhancing Knowledge Bases: Use the linked entities to enrich your database or knowledge base with reliable, referenced information.
Disambiguation: Manually review cases with close scores or high ambiguity to disambiguate entities.

#Data Cleaning:
The model can help in cleaning and structuring datasets by identifying and linking entities to a standardized knowledge base like Wikipedia.

```python
import requests
import json
import re


def fetch_wikipedia_summary(title):
    """Fetch the summary of a Wikipedia page given its title."""
    url = f"https://en.wikipedia.org/api/rest_v1/page/summary/{title.replace(' ', '_')}"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json().get('extract', '')
    return None

def infer_entity_type_from_content(content):
    """Infer entity type from Wikipedia page content using expanded and context-sensitive keyword checks."""

    # Regex patterns to identify keywords with flexible surrounding context
    person_patterns = [
        r'\b(born)\b',
        r'\b(died)\b',
        r'\b(politician)\b',
        r'\b(actor|actress)\b',
        r'\b(singer)\b',
        r'\b(athlete)\b',
        r'\b(author)\b',
        r'\b(artist)\b',
        r'\b(director)\b',
        r'\b(philosopher)\b'
    ]

    org_patterns = [
        r'\b(corporation)\b',
        r'\b(company)\b',
        r'\b(organization)\b',
        r'\b(institution)\b',
        r'\b(association)\b',
        r'\b(foundation)\b',
        r'\b(university)\b',
        r'\b(college)\b',
        r'\b(enterprise)\b',
        r'\b(bank)\b'
    ]

    loc_patterns = [
        r'\b(city)\b',
        r'\b(village)\b',
        r'\b(town)\b',
        r'\b(county)\b',
        r'\b(state)\b',
        r'\b(country)\b',
        r'\b(nation)\b',
        r'\b(province)\b',
        r'\b(region)\b',
        r'\b(district)\b'
    ]

    # Check if any of the patterns match and return the corresponding entity type
    if any(re.search(pattern, content, re.IGNORECASE) for pattern in person_patterns):
        return "PERSON"
    elif any(re.search(pattern, content, re.IGNORECASE) for pattern in org_patterns):
        return "ORG"
    elif any(re.search(pattern, content, re.IGNORECASE) for pattern in loc_patterns):
        return "LOC"

    return None

def check_entity_type(predictions):
    results = []
    for item in predictions:
        mention = item["mention"]
        expected_type = item["type"]
        top_entity = item["predictions"][0]["entity"]

        # Fetch the Wikipedia summary of the top entity
        summary = fetch_wikipedia_summary(top_entity)
        if summary is None:
            result = f"No summary found for {top_entity}."
            results.append({"mention": mention, "result": result})
            continue

        # Infer the type from the Wikipedia summary
        inferred_type = infer_entity_type_from_content(summary.lower())

        # Check if the inferred type matches the expected type
        if inferred_type == expected_type:
            result = "MATCH"
        else:
            result = "NO MATCH"

        results.append({
            "mention": mention,
            "expected_type": expected_type,
            "top_entity": top_entity,
            "inferred_type": inferred_type,
            "result": result
        })

    # Save the results to a JSON file
    with open('entity_check_results.json', 'w', encoding='utf-8') as f:
        json.dump(results, f, indent=4, ensure_ascii=False)

    print("Results saved to entity_check_results.json")

# Example predictions as provided
predictions = [
    {
        "mention": "chillicothe's",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Chillicothe, Missouri",
                "confidence_score": "-5.2446394"
            },
            {
                "entity": "Chillicothe, Illinois",
                "confidence_score": "-6.2553144"
            },
            {
                "entity": "Chillicothe, Ohio",
                "confidence_score": "-6.428744"
            }
        ]
    },
    {
        "mention": "rob weingartner",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Rob Weingartner",
                "confidence_score": "-0.047881465"
            },
            {
                "entity": "Weingarten Realty",
                "confidence_score": "-11.251331"
            },
            {
                "entity": "Richard Weingardt",
                "confidence_score": "-11.762729"
            }
        ]
    },
    {
        "mention": "pump house",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Pumping station",
                "confidence_score": "-3.9107552"
            },
            {
                "entity": "Spring house",
                "confidence_score": "-7.273636"
            },
            {
                "entity": "Tankhouse",
                "confidence_score": "-8.472154"
            }
        ]
    },
    {
        "mention": "the pump house center for the arts",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Luther Burbank Center for the Arts",
                "confidence_score": "-11.539211"
            },
            {
                "entity": "Sangre de Cristo Arts and Conference Center",
                "confidence_score": "-12.047555"
            },
            {
                "entity": "Goat Farm Arts Center",
                "confidence_score": "-12.308259"
            }
        ]
    },
    {
        "mention": "kevin coleman",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kevin Coleman (aviator)",
                "confidence_score": "-9.057244"
            },
            {
                "entity": "Kevin Coleman (soccer)",
                "confidence_score": "-10.637237"
            },
            {
                "entity": "Kevin Coleman",
                "confidence_score": "-10.730374"
            }
        ]
    },
    {
        "mention": "chillicothe & circleville",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Circleville, Ohio",
                "confidence_score": "-9.384942"
            },
            {
                "entity": "Circleville, Utah",
                "confidence_score": "-11.699418"
            },
            {
                "entity": "Circleville",
                "confidence_score": "-12.236474"
            }
        ]
    },
    {
        "mention": "rob weingartner",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Rob Weingartner",
                "confidence_score": "-0.047881465"
            },
            {
                "entity": "Weingarten Realty",
                "confidence_score": "-11.251331"
            },
            {
                "entity": "Richard Weingardt",
                "confidence_score": "-11.762729"
            }
        ]
    },
    {
        "mention": "oscar",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Academy Awards",
                "confidence_score": "2.9287302"
            },
            {
                "entity": "OScar",
                "confidence_score": "-7.5889173"
            },
            {
                "entity": "Oscar (bull)",
                "confidence_score": "-11.212821"
            }
        ]
    },
    {
        "mention": "everest",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Mount Everest",
                "confidence_score": "-5.034682"
            },
            {
                "entity": "Everest (1998 film)",
                "confidence_score": "-7.707544"
            },
            {
                "entity": "Ski jumping at the 2018 Winter Olympics \u2013 Men's large hill team",
                "confidence_score": "-11.99105"
            }
        ]
    },
    {
        "mention": "everest",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Mount Everest",
                "confidence_score": "-5.034682"
            },
            {
                "entity": "Everest (1998 film)",
                "confidence_score": "-7.707544"
            },
            {
                "entity": "Ski jumping at the 2018 Winter Olympics \u2013 Men's large hill team",
                "confidence_score": "-11.99105"
            }
        ]
    },
    {
        "mention": "jason clarke",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jason Clarke",
                "confidence_score": "10.829769"
            },
            {
                "entity": "Jason Clarke (writer)",
                "confidence_score": "-4.473886"
            },
            {
                "entity": "Jason Clarke (ice hockey)",
                "confidence_score": "-8.998358"
            }
        ]
    },
    {
        "mention": "rob hall",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Rob Hall",
                "confidence_score": "4.9627304"
            },
            {
                "entity": "Rich Hall",
                "confidence_score": "-10.500752"
            },
            {
                "entity": "Bob Hall (musician)",
                "confidence_score": "-11.219953"
            }
        ]
    },
    {
        "mention": "einar wegener",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Gerda Wegener",
                "confidence_score": "-1.7132143"
            },
            {
                "entity": "Emmy Wegener",
                "confidence_score": "-10.654565"
            },
            {
                "entity": "Ulli Wegner",
                "confidence_score": "-11.86225"
            }
        ]
    },
    {
        "mention": "lili elbe",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Lili Elbe",
                "confidence_score": "12.226471"
            },
            {
                "entity": "Lili Bech",
                "confidence_score": "-11.21326"
            },
            {
                "entity": "Jenny Elbe",
                "confidence_score": "-11.836555"
            }
        ]
    },
    {
        "mention": "alicia vikander",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Alicia Vikander",
                "confidence_score": "7.7965236"
            },
            {
                "entity": "Alicia Vignoli",
                "confidence_score": "-11.423744"
            },
            {
                "entity": "Viomak",
                "confidence_score": "-12.586442"
            }
        ]
    },
    {
        "mention": "gerda wegener",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Gerda Wegener",
                "confidence_score": "5.5195546"
            },
            {
                "entity": "Ingo Wegener",
                "confidence_score": "-10.890681"
            },
            {
                "entity": "Lisa Wegner",
                "confidence_score": "-12.014977"
            }
        ]
    },
    {
        "mention": "jennifer lawrence",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jennifer Lawrence",
                "confidence_score": "12.278758"
            },
            {
                "entity": "Jennifer Jason Leigh",
                "confidence_score": "-9.586303"
            },
            {
                "entity": "Jennifer Lopez",
                "confidence_score": "-10.717816"
            }
        ]
    },
    {
        "mention": "bradley cooper",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Bradley Cooper",
                "confidence_score": "12.290053"
            },
            {
                "entity": "Bradley Cooper (athlete)",
                "confidence_score": "-5.6047487"
            },
            {
                "entity": "Gary Cooper",
                "confidence_score": "-8.848578"
            }
        ]
    },
    {
        "mention": "david o. russell",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "David O. Russell",
                "confidence_score": "11.249957"
            },
            {
                "entity": "David F. O. Russell",
                "confidence_score": "-4.831576"
            },
            {
                "entity": "David Oyelowo",
                "confidence_score": "-11.446107"
            }
        ]
    },
    {
        "mention": "oscar",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Academy Awards",
                "confidence_score": "2.9287302"
            },
            {
                "entity": "OScar",
                "confidence_score": "-7.5889173"
            },
            {
                "entity": "Oscar (bull)",
                "confidence_score": "-11.212821"
            }
        ]
    },
    {
        "mention": "lawrence",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Lawrence Elion",
                "confidence_score": "-6.7262163"
            },
            {
                "entity": "Lawrence Grant",
                "confidence_score": "-7.1517196"
            },
            {
                "entity": "Lawrence Brooks",
                "confidence_score": "-9.781568"
            }
        ]
    },
    {
        "mention": "joy",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Joy Brook",
                "confidence_score": "-7.3752456"
            },
            {
                "entity": "Joy Nichols",
                "confidence_score": "-7.830309"
            },
            {
                "entity": "Joy Osmanski",
                "confidence_score": "-10.075492"
            }
        ]
    },
    {
        "mention": "tom hardy",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Tom Hardy",
                "confidence_score": "8.108426"
            },
            {
                "entity": "Tom Hardy (soccer)",
                "confidence_score": "-9.03538"
            },
            {
                "entity": "TomSka",
                "confidence_score": "-10.449981"
            }
        ]
    },
    {
        "mention": "the kray twins",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Kray twins",
                "confidence_score": "1.1147673"
            },
            {
                "entity": "Mitchell and Webb",
                "confidence_score": "-11.455656"
            },
            {
                "entity": "Ray brothers",
                "confidence_score": "-12.016515"
            }
        ]
    },
    {
        "mention": "kray",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kray twins",
                "confidence_score": "-1.2573466"
            },
            {
                "entity": "The Krays (film)",
                "confidence_score": "-9.055235"
            },
            {
                "entity": "The Krays",
                "confidence_score": "-12.304819"
            }
        ]
    },
    {
        "mention": "kemp",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Jenny Kemp",
                "confidence_score": "-10.142403"
            },
            {
                "entity": "Kemp",
                "confidence_score": "-11.133258"
            },
            {
                "entity": "Kemp, Ohio",
                "confidence_score": "-11.19892"
            }
        ]
    },
    {
        "mention": "patricia highsmith",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Patricia Highsmith",
                "confidence_score": "12.286905"
            },
            {
                "entity": "Carol M. Highsmith",
                "confidence_score": "-8.170243"
            },
            {
                "entity": "Jim Highsmith",
                "confidence_score": "-11.313494"
            }
        ]
    },
    {
        "mention": "rooney mara",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Rooney Mara",
                "confidence_score": "11.925614"
            },
            {
                "entity": "Kate Mara",
                "confidence_score": "1.9138248"
            },
            {
                "entity": "Mickey Rooney",
                "confidence_score": "-5.363707"
            }
        ]
    },
    {
        "mention": "rooney mara",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Rooney Mara",
                "confidence_score": "11.925614"
            },
            {
                "entity": "Kate Mara",
                "confidence_score": "1.9138248"
            },
            {
                "entity": "Mickey Rooney",
                "confidence_score": "-5.363707"
            }
        ]
    },
    {
        "mention": "oscar",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Academy Awards",
                "confidence_score": "2.9287302"
            },
            {
                "entity": "OScar",
                "confidence_score": "-7.5889173"
            },
            {
                "entity": "Oscar (bull)",
                "confidence_score": "-11.212821"
            }
        ]
    },
    {
        "mention": "rooney",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mickey Rooney",
                "confidence_score": "0.5553501"
            },
            {
                "entity": "Rooney family",
                "confidence_score": "-5.59094"
            },
            {
                "entity": "Wayne Rooney",
                "confidence_score": "-6.4846096"
            }
        ]
    },
    {
        "mention": "oscar",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Academy Awards",
                "confidence_score": "2.9287302"
            },
            {
                "entity": "OScar",
                "confidence_score": "-7.5889173"
            },
            {
                "entity": "Oscar (bull)",
                "confidence_score": "-11.212821"
            }
        ]
    },
    {
        "mention": "hugh glass",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Hugh Glass",
                "confidence_score": "3.4003975"
            },
            {
                "entity": "Adam Glass",
                "confidence_score": "-10.516348"
            },
            {
                "entity": "Jimmy Glass",
                "confidence_score": "-11.011116"
            }
        ]
    },
    {
        "mention": "whitey bulger",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Whitey Bulger",
                "confidence_score": "12.145009"
            },
            {
                "entity": "Whitey: United States of America v. James J. Bulger",
                "confidence_score": "-7.8552628"
            },
            {
                "entity": "Jay Bulger",
                "confidence_score": "-11.043739"
            }
        ]
    },
    {
        "mention": "benedict cumberbatch",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Benedict Cumberbatch",
                "confidence_score": "2.1088939"
            },
            {
                "entity": "List of A Nightmare on Elm Street characters",
                "confidence_score": "-10.994975"
            },
            {
                "entity": "List of Lucchese crime family mobsters",
                "confidence_score": "-11.605229"
            }
        ]
    },
    {
        "mention": "fbi",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Federal Bureau of Investigation",
                "confidence_score": "9.418494"
            },
            {
                "entity": "FBI Police",
                "confidence_score": "-5.331061"
            },
            {
                "entity": "FBI (TV series)",
                "confidence_score": "-8.448312"
            }
        ]
    },
    {
        "mention": "steve jobs",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Steve Jobs",
                "confidence_score": "11.6573305"
            },
            {
                "entity": "Steve Jobs (film)",
                "confidence_score": "-2.893585"
            },
            {
                "entity": "Steve Jobs (clothing company)",
                "confidence_score": "-7.980975"
            }
        ]
    },
    {
        "mention": "sony",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sony",
                "confidence_score": "10.630389"
            },
            {
                "entity": "PlayStation",
                "confidence_score": "-6.2952456"
            },
            {
                "entity": "PlayStation (console)",
                "confidence_score": "-7.820069"
            }
        ]
    },
    {
        "mention": "michael fassbender",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Michael Fassbender",
                "confidence_score": "2.4499412"
            },
            {
                "entity": "Rainer Werner Fassbinder",
                "confidence_score": "-9.537158"
            },
            {
                "entity": "Joseph Fassbender",
                "confidence_score": "-10.606246"
            }
        ]
    },
    {
        "mention": "steve jobs",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Steve Jobs",
                "confidence_score": "11.6573305"
            },
            {
                "entity": "Steve Jobs (film)",
                "confidence_score": "-2.893585"
            },
            {
                "entity": "Steve Jobs (clothing company)",
                "confidence_score": "-7.980975"
            }
        ]
    },
    {
        "mention": "jobs",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Job",
                "confidence_score": "7.691918"
            },
            {
                "entity": "Employment",
                "confidence_score": "-1.3834543"
            },
            {
                "entity": "Jobs for a Change",
                "confidence_score": "-8.5092"
            }
        ]
    },
    {
        "mention": "maud watts",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Kate Watts",
                "confidence_score": "-10.434687"
            },
            {
                "entity": "Mary Stanbery Watts",
                "confidence_score": "-10.589912"
            },
            {
                "entity": "Maud Ward",
                "confidence_score": "-11.169714"
            }
        ]
    },
    {
        "mention": "carey mulligan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Carey Mulligan",
                "confidence_score": "12.278102"
            },
            {
                "entity": "George Mulligan",
                "confidence_score": "-11.174512"
            },
            {
                "entity": "Mary Mulligan",
                "confidence_score": "-11.421451"
            }
        ]
    },
    {
        "mention": "meryl streep",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Meryl Streep",
                "confidence_score": "12.702667"
            },
            {
                "entity": "Meryl Streep on screen and stage",
                "confidence_score": "1.490675"
            },
            {
                "entity": "Mary Wilkinson Streep",
                "confidence_score": "-8.188468"
            }
        ]
    },
    {
        "mention": "emmeline pankhurst",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Emmeline Pankhurst",
                "confidence_score": "8.254952"
            },
            {
                "entity": "Statue of Emmeline Pankhurst",
                "confidence_score": "-6.753363"
            },
            {
                "entity": "Sylvia Pankhurst",
                "confidence_score": "-9.208009"
            }
        ]
    },
    {
        "mention": "shakespeare",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "William Shakespeare",
                "confidence_score": "7.219626"
            },
            {
                "entity": "Edmund Shakespeare",
                "confidence_score": "-6.8099236"
            },
            {
                "entity": "John Shakespeare",
                "confidence_score": "-7.85892"
            }
        ]
    },
    {
        "mention": "michael fassbender",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Michael Fassbender",
                "confidence_score": "2.4499412"
            },
            {
                "entity": "Rainer Werner Fassbinder",
                "confidence_score": "-9.537158"
            },
            {
                "entity": "Joseph Fassbender",
                "confidence_score": "-10.606246"
            }
        ]
    },
    {
        "mention": "marion cotillard",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Marion Cotillard",
                "confidence_score": "11.856938"
            },
            {
                "entity": "Marion Coutts",
                "confidence_score": "-11.746547"
            },
            {
                "entity": "Marion Arnott",
                "confidence_score": "-11.944566"
            }
        ]
    },
    {
        "mention": "fassbender",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Michael Fassbender",
                "confidence_score": "-4.6443315"
            },
            {
                "entity": "Rainer Werner Fassbinder",
                "confidence_score": "-5.6304193"
            },
            {
                "entity": "Hedwig Fassbender",
                "confidence_score": "-6.6002936"
            }
        ]
    },
    {
        "mention": "julianne moore",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Julianne Moore",
                "confidence_score": "11.603573"
            },
            {
                "entity": "Julianne Nicholson",
                "confidence_score": "-10.49235"
            },
            {
                "entity": "Julianne Hough",
                "confidence_score": "-11.252386"
            }
        ]
    },
    {
        "mention": "laurel hester",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Laurel Hester",
                "confidence_score": "12.090917"
            },
            {
                "entity": "Laurel Cronin",
                "confidence_score": "-11.251886"
            },
            {
                "entity": "Carolyn Hester",
                "confidence_score": "-11.715314"
            }
        ]
    },
    {
        "mention": "stacie andree",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Stacie Lynn Renna",
                "confidence_score": "-10.795066"
            },
            {
                "entity": "Stacie Foster",
                "confidence_score": "-10.828129"
            },
            {
                "entity": "Stacie Randall",
                "confidence_score": "-11.016896"
            }
        ]
    },
    {
        "mention": "ellen page",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ellen Page",
                "confidence_score": "11.127679"
            },
            {
                "entity": "Ellen Brooks",
                "confidence_score": "-11.080144"
            },
            {
                "entity": "Ellen Goosenberg Kent",
                "confidence_score": "-11.777549"
            }
        ]
    },
    {
        "mention": "cannes",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Cannes",
                "confidence_score": "8.212696"
            },
            {
                "entity": "2019 Cannes Film Festival",
                "confidence_score": "-7.503466"
            },
            {
                "entity": "1989 Cannes Film Festival",
                "confidence_score": "-8.076613"
            }
        ]
    },
    {
        "mention": "michael caine",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Michael Caine",
                "confidence_score": "12.066547"
            },
            {
                "entity": "Michael Harris Caine",
                "confidence_score": "-3.829782"
            },
            {
                "entity": "Michael Caine (racing driver)",
                "confidence_score": "-4.357123"
            }
        ]
    },
    {
        "mention": "fred ballinger",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Adrian Ballinger",
                "confidence_score": "-10.5318365"
            },
            {
                "entity": "Bill S. Ballinger",
                "confidence_score": "-10.869629"
            },
            {
                "entity": "Jeffrey Ballinger",
                "confidence_score": "-11.536318"
            }
        ]
    },
    {
        "mention": "alps",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Alps",
                "confidence_score": "1.4700428"
            },
            {
                "entity": "Alps (film)",
                "confidence_score": "-8.524772"
            },
            {
                "entity": "The Alps (band)",
                "confidence_score": "-9.795535"
            }
        ]
    },
    {
        "mention": "mick boyle",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Frankie Boyle",
                "confidence_score": "-9.869424"
            },
            {
                "entity": "Mick Napier",
                "confidence_score": "-11.287291"
            },
            {
                "entity": "Micky Flanagan",
                "confidence_score": "-11.825775"
            }
        ]
    },
    {
        "mention": "harvey keitel",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Harvey Keitel",
                "confidence_score": "12.7008915"
            },
            {
                "entity": "Wilhelm Keitel",
                "confidence_score": "-10.136067"
            },
            {
                "entity": "Harvey Levin",
                "confidence_score": "-11.275096"
            }
        ]
    },
    {
        "mention": "zomato",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Zomato",
                "confidence_score": "10.992883"
            },
            {
                "entity": "Zomaron",
                "confidence_score": "-10.259671"
            },
            {
                "entity": "Arbusto Energy",
                "confidence_score": "-11.490374"
            }
        ]
    },
    {
        "mention": "sequoia india",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sequoia Capital",
                "confidence_score": "-6.7231884"
            },
            {
                "entity": "IBM India",
                "confidence_score": "-10.91781"
            },
            {
                "entity": "Coca-Cola India",
                "confidence_score": "-10.935952"
            }
        ]
    },
    {
        "mention": "vy capital",
        "type": "ORG",
        "predictions": [
            {
                "entity": "E.ventures",
                "confidence_score": "-10.503558"
            },
            {
                "entity": "ING Vysya Bank",
                "confidence_score": "-10.786627"
            },
            {
                "entity": "Vaibhav Global",
                "confidence_score": "-10.996903"
            }
        ]
    },
    {
        "mention": "zomato",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Zomato",
                "confidence_score": "8.684343"
            },
            {
                "entity": "Zomaron",
                "confidence_score": "-11.254109"
            },
            {
                "entity": "Zeto",
                "confidence_score": "-13.03151"
            }
        ]
    },
    {
        "mention": "vy capital",
        "type": "ORG",
        "predictions": [
            {
                "entity": "E.ventures",
                "confidence_score": "-10.503558"
            },
            {
                "entity": "ING Vysya Bank",
                "confidence_score": "-10.786627"
            },
            {
                "entity": "Vaibhav Global",
                "confidence_score": "-10.996903"
            }
        ]
    },
    {
        "mention": "zomato",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Zomato",
                "confidence_score": "10.992883"
            },
            {
                "entity": "Zomaron",
                "confidence_score": "-10.259671"
            },
            {
                "entity": "Arbusto Energy",
                "confidence_score": "-11.490374"
            }
        ]
    },
    {
        "mention": "zomato founder",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "List of Internet entrepreneurs",
                "confidence_score": "-7.58396"
            },
            {
                "entity": "MBA Fakhro Group",
                "confidence_score": "-11.412689"
            },
            {
                "entity": "Bambi Francisco",
                "confidence_score": "-11.903372"
            }
        ]
    },
    {
        "mention": "deepinder goyal",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Des Raj Goyal",
                "confidence_score": "-8.819308"
            },
            {
                "entity": "Shipra Goyal",
                "confidence_score": "-9.278876"
            },
            {
                "entity": "Ajay Goyal",
                "confidence_score": "-11.174516"
            }
        ]
    },
    {
        "mention": "zomato",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Zomato",
                "confidence_score": "10.992883"
            },
            {
                "entity": "Zomaron",
                "confidence_score": "-10.259671"
            },
            {
                "entity": "Arbusto Energy",
                "confidence_score": "-11.490374"
            }
        ]
    },
    {
        "mention": "zomato",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Zomato",
                "confidence_score": "10.992883"
            },
            {
                "entity": "Zomaron",
                "confidence_score": "-10.259671"
            },
            {
                "entity": "Arbusto Energy",
                "confidence_score": "-11.490374"
            }
        ]
    },
    {
        "mention": "sequoia india",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sequoia Capital",
                "confidence_score": "-6.7231884"
            },
            {
                "entity": "IBM India",
                "confidence_score": "-10.91781"
            },
            {
                "entity": "Coca-Cola India",
                "confidence_score": "-10.935952"
            }
        ]
    },
    {
        "mention": "vy capital",
        "type": "ORG",
        "predictions": [
            {
                "entity": "E.ventures",
                "confidence_score": "-10.503558"
            },
            {
                "entity": "ING Vysya Bank",
                "confidence_score": "-10.786627"
            },
            {
                "entity": "Vaibhav Global",
                "confidence_score": "-10.996903"
            }
        ]
    },
    {
        "mention": "goyal",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Naresh Goyal",
                "confidence_score": "-5.3655577"
            },
            {
                "entity": "Ajay Goyal",
                "confidence_score": "-5.4327"
            },
            {
                "entity": "LC Goyal",
                "confidence_score": "-7.7647424"
            }
        ]
    },
    {
        "mention": "zomato",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Zomato",
                "confidence_score": "10.992883"
            },
            {
                "entity": "Zomaron",
                "confidence_score": "-10.259671"
            },
            {
                "entity": "Arbusto Energy",
                "confidence_score": "-11.490374"
            }
        ]
    },
    {
        "mention": "chelsea",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Chelsea F.C.",
                "confidence_score": "0.9543292"
            },
            {
                "entity": "Chelsea (band)",
                "confidence_score": "-5.952766"
            },
            {
                "entity": "Chelsea F.C. Women",
                "confidence_score": "-7.2783732"
            }
        ]
    },
    {
        "mention": "aitor karanka",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Aitor Karanka",
                "confidence_score": "3.9784498"
            },
            {
                "entity": "Aitor Ari\u00f1o",
                "confidence_score": "-12.094859"
            },
            {
                "entity": "Aitor Fern\u00e1ndez (footballer, born 1991)",
                "confidence_score": "-12.154846"
            }
        ]
    },
    {
        "mention": "jose mourinho",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jos\u00e9 Mourinho",
                "confidence_score": "4.1757812"
            },
            {
                "entity": "Jo\u00e3o Moutinho",
                "confidence_score": "-10.907227"
            },
            {
                "entity": "Jos\u00e9 Mota (footballer, born 1979)",
                "confidence_score": "-11.688601"
            }
        ]
    },
    {
        "mention": "karanka",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Karanka",
                "confidence_score": "-6.5544124"
            },
            {
                "entity": "Aitor Karanka",
                "confidence_score": "-7.402139"
            },
            {
                "entity": "David Karanka",
                "confidence_score": "-10.885568"
            }
        ]
    },
    {
        "mention": "mourinho",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Jos\u00e9 Mourinho",
                "confidence_score": "-3.9383771"
            },
            {
                "entity": "Jo\u00e3o Moutinho",
                "confidence_score": "-8.484281"
            },
            {
                "entity": "Jo\u00e3o Moutinho (footballer, born 1998)",
                "confidence_score": "-10.966695"
            }
        ]
    },
    {
        "mention": "chelsea",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Chelsea F.C.",
                "confidence_score": "-1.250842"
            },
            {
                "entity": "Chelsea F.C. Women",
                "confidence_score": "-8.080724"
            },
            {
                "entity": "Chelsea Olivia",
                "confidence_score": "-10.535999"
            }
        ]
    },
    {
        "mention": "chelsea",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Chelsea F.C.",
                "confidence_score": "-1.250842"
            },
            {
                "entity": "Chelsea F.C. Women",
                "confidence_score": "-8.080724"
            },
            {
                "entity": "Chelsea Olivia",
                "confidence_score": "-10.535999"
            }
        ]
    },
    {
        "mention": "ame info staff",
        "type": "ORG",
        "predictions": [
            {
                "entity": "AME Accounting Software",
                "confidence_score": "-10.947708"
            },
            {
                "entity": "Ameritech Interactive Media Services",
                "confidence_score": "-11.852916"
            },
            {
                "entity": "Army Model and Simulation Executive Council",
                "confidence_score": "-12.170522"
            }
        ]
    },
    {
        "mention": "ppia",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Professional Photographers of America",
                "confidence_score": "-10.027242"
            },
            {
                "entity": "Pakistan Institute of International Affairs",
                "confidence_score": "-10.612904"
            },
            {
                "entity": "Pakistan Press International",
                "confidence_score": "-11.019932"
            }
        ]
    },
    {
        "mention": "matt",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Matt Damon",
                "confidence_score": "-3.689821"
            },
            {
                "entity": "Matt Drews",
                "confidence_score": "-11.411744"
            },
            {
                "entity": "Matt Gold",
                "confidence_score": "-11.5561075"
            }
        ]
    },
    {
        "mention": "red rock canyon'",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Red Rock Canyon State Park (California)",
                "confidence_score": "-7.0642753"
            },
            {
                "entity": "Red Rock (Wyoming)",
                "confidence_score": "-10.667834"
            },
            {
                "entity": "Red Canyon (Fremont County, Wyoming)",
                "confidence_score": "-11.15699"
            }
        ]
    },
    {
        "mention": "kraft boulders'",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Kraft Group",
                "confidence_score": "-9.100753"
            },
            {
                "entity": "Black Rock City, LLC",
                "confidence_score": "-11.273794"
            },
            {
                "entity": "Kraft, Kentucky",
                "confidence_score": "-11.760942"
            }
        ]
    },
    {
        "mention": "matt",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Matt Damon",
                "confidence_score": "-3.689821"
            },
            {
                "entity": "Matt Drews",
                "confidence_score": "-11.411744"
            },
            {
                "entity": "Matt Gold",
                "confidence_score": "-11.5561075"
            }
        ]
    },
    {
        "mention": "two sizes",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Canoeing at the 2016 Summer Olympics \u2013 Men's K-2 200 metres",
                "confidence_score": "-12.196137"
            },
            {
                "entity": "Paragliding at the 2018 Asian Games \u2013 Women's individual accuracy",
                "confidence_score": "-12.439982"
            },
            {
                "entity": "N-Formylmethionine (data page)",
                "confidence_score": "-12.7245245"
            }
        ]
    },
    {
        "mention": "new colgate total daily repair",
        "type": "ORG",
        "predictions": [
            {
                "entity": "All Pro Sound",
                "confidence_score": "-11.041748"
            },
            {
                "entity": "New, Improved Recording",
                "confidence_score": "-11.739537"
            },
            {
                "entity": "A Few Small Repairs",
                "confidence_score": "-12.390899"
            }
        ]
    },
    {
        "mention": "monoprice hi-fi dj style acoustic pro studio headphones",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Monoprice",
                "confidence_score": "-7.9346685"
            },
            {
                "entity": "Professional audio",
                "confidence_score": "-9.008237"
            },
            {
                "entity": "High fidelity",
                "confidence_score": "-9.013612"
            }
        ]
    },
    {
        "mention": "credit line",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Credit card",
                "confidence_score": "-3.838872"
            },
            {
                "entity": "Credit rating",
                "confidence_score": "-7.5874653"
            },
            {
                "entity": "Credit derivative",
                "confidence_score": "-8.194542"
            }
        ]
    },
    {
        "mention": "queensland's",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Queensland",
                "confidence_score": "8.801793"
            },
            {
                "entity": "Queensland rugby league team",
                "confidence_score": "-3.9636805"
            },
            {
                "entity": "Queensland borders",
                "confidence_score": "-4.8139405"
            }
        ]
    },
    {
        "mention": "the gold coast",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Gold Coast, Queensland",
                "confidence_score": "4.5411906"
            },
            {
                "entity": "City of Gold Coast",
                "confidence_score": "-2.8693502"
            },
            {
                "entity": "TAFE Queensland Gold Coast",
                "confidence_score": "-6.5645375"
            }
        ]
    },
    {
        "mention": "the gold coast",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Gold Coast, Queensland",
                "confidence_score": "4.5411906"
            },
            {
                "entity": "City of Gold Coast",
                "confidence_score": "-2.8693502"
            },
            {
                "entity": "TAFE Queensland Gold Coast",
                "confidence_score": "-6.5645375"
            }
        ]
    },
    {
        "mention": "intuit inc.",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Intuit",
                "confidence_score": "9.700464"
            },
            {
                "entity": "Intuit: The Center for Intuitive and Outsider Art",
                "confidence_score": "-9.959895"
            },
            {
                "entity": "Intuit Canada",
                "confidence_score": "-10.329904"
            }
        ]
    },
    {
        "mention": "intuit workforce",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Intuit",
                "confidence_score": "-4.8245764"
            },
            {
                "entity": "Intuitor",
                "confidence_score": "-10.760281"
            },
            {
                "entity": "Intuit Canada",
                "confidence_score": "-10.785452"
            }
        ]
    },
    {
        "mention": "deliv",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Deliv",
                "confidence_score": "9.213618"
            },
            {
                "entity": "PROIV",
                "confidence_score": "-11.505514"
            },
            {
                "entity": "DIIV",
                "confidence_score": "-11.705367"
            }
        ]
    },
    {
        "mention": "orderup",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "OrderUp",
                "confidence_score": "8.679609"
            },
            {
                "entity": "OpenUP",
                "confidence_score": "-11.695452"
            },
            {
                "entity": "U.P. Town Center",
                "confidence_score": "-12.134152"
            }
        ]
    },
    {
        "mention": "intuit",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Intuit",
                "confidence_score": "12.278062"
            },
            {
                "entity": "Intuit: The Center for Intuitive and Outsider Art",
                "confidence_score": "-2.4283082"
            },
            {
                "entity": "Intuit Canada",
                "confidence_score": "-9.856866"
            }
        ]
    },
    {
        "mention": "intuit inc",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Intuit",
                "confidence_score": "11.465545"
            },
            {
                "entity": "Intuit Canada",
                "confidence_score": "-10.429128"
            },
            {
                "entity": "Intrexon",
                "confidence_score": "-11.117709"
            }
        ]
    },
    {
        "mention": "intuit inc",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Intuit",
                "confidence_score": "11.465545"
            },
            {
                "entity": "Intuit Canada",
                "confidence_score": "-10.429128"
            },
            {
                "entity": "Intrexon",
                "confidence_score": "-11.117709"
            }
        ]
    },
    {
        "mention": "intuit inc",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Intuit",
                "confidence_score": "11.465545"
            },
            {
                "entity": "Intuit Canada",
                "confidence_score": "-10.429128"
            },
            {
                "entity": "Intrexon",
                "confidence_score": "-11.117709"
            }
        ]
    },
    {
        "mention": "related stocks",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Related party transaction",
                "confidence_score": "-9.29873"
            },
            {
                "entity": "The Related Companies",
                "confidence_score": "-9.47582"
            },
            {
                "entity": "List of mergers of securities firms",
                "confidence_score": "-10.029923"
            }
        ]
    },
    {
        "mention": "pope francis",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Pope Francis",
                "confidence_score": "11.410729"
            },
            {
                "entity": "Pope Francis's visit to the Philippines",
                "confidence_score": "-7.4476953"
            },
            {
                "entity": "Francis Francis",
                "confidence_score": "-12.172629"
            }
        ]
    },
    {
        "mention": "francis",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Pope Francis",
                "confidence_score": "9.714959"
            },
            {
                "entity": "Francis Rooney",
                "confidence_score": "-7.8511076"
            },
            {
                "entity": "Francis Francis",
                "confidence_score": "-11.663054"
            }
        ]
    },
    {
        "mention": "eu",
        "type": "ORG",
        "predictions": [
            {
                "entity": "European Union",
                "confidence_score": "8.82327"
            },
            {
                "entity": "Council of Europe",
                "confidence_score": "-1.8394974"
            },
            {
                "entity": "European Parliament",
                "confidence_score": "-2.4300888"
            }
        ]
    },
    {
        "mention": "donald tusk",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Donald Tusk",
                "confidence_score": "7.318099"
            },
            {
                "entity": "Donald Trump",
                "confidence_score": "-6.7520595"
            },
            {
                "entity": "Donald Winkler",
                "confidence_score": "-11.668759"
            }
        ]
    },
    {
        "mention": "europe",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Europe",
                "confidence_score": "4.80574"
            },
            {
                "entity": "Council of Europe",
                "confidence_score": "-4.198911"
            },
            {
                "entity": "Time in Europe",
                "confidence_score": "-7.8469768"
            }
        ]
    },
    {
        "mention": "tusk",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Donald Tusk",
                "confidence_score": "-0.50873744"
            },
            {
                "entity": "Bradley Tusk",
                "confidence_score": "-8.189271"
            },
            {
                "entity": "Tusk (2014 film)",
                "confidence_score": "-8.227561"
            }
        ]
    },
    {
        "mention": "kolinda grabar-kitarovic",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kolinda Grabar-Kitarovi\u0107",
                "confidence_score": "11.856862"
            },
            {
                "entity": "Zofia Kirkor-Kiedroniowa",
                "confidence_score": "-11.832198"
            },
            {
                "entity": "Nikolina Grabovac",
                "confidence_score": "-11.878356"
            }
        ]
    },
    {
        "mention": "xenophobic",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Xenophobia",
                "confidence_score": "7.575575"
            },
            {
                "entity": "Homophobia",
                "confidence_score": "-8.480961"
            },
            {
                "entity": "Ethnic nationalism",
                "confidence_score": "-10.012016"
            }
        ]
    },
    {
        "mention": "angela merkel",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Angela Merkel",
                "confidence_score": "11.385147"
            },
            {
                "entity": "Therese Merkel",
                "confidence_score": "-10.817756"
            },
            {
                "entity": "Angela Visser",
                "confidence_score": "-11.481587"
            }
        ]
    },
    {
        "mention": "europe",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Europe",
                "confidence_score": "4.80574"
            },
            {
                "entity": "Council of Europe",
                "confidence_score": "-4.198911"
            },
            {
                "entity": "Time in Europe",
                "confidence_score": "-7.8469768"
            }
        ]
    },
    {
        "mention": "europe",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Europe",
                "confidence_score": "4.80574"
            },
            {
                "entity": "Council of Europe",
                "confidence_score": "-4.198911"
            },
            {
                "entity": "Time in Europe",
                "confidence_score": "-7.8469768"
            }
        ]
    },
    {
        "mention": "tusk",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Donald Tusk",
                "confidence_score": "-0.50873744"
            },
            {
                "entity": "Bradley Tusk",
                "confidence_score": "-8.189271"
            },
            {
                "entity": "Tusk (2014 film)",
                "confidence_score": "-8.227561"
            }
        ]
    },
    {
        "mention": "eu",
        "type": "ORG",
        "predictions": [
            {
                "entity": "European Union",
                "confidence_score": "8.82327"
            },
            {
                "entity": "Council of Europe",
                "confidence_score": "-1.8394974"
            },
            {
                "entity": "European Parliament",
                "confidence_score": "-2.4300888"
            }
        ]
    },
    {
        "mention": "croatia",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Croatia",
                "confidence_score": "6.6205053"
            },
            {
                "entity": "Socialist Republic of Croatia",
                "confidence_score": "-2.9952147"
            },
            {
                "entity": "Croats",
                "confidence_score": "-3.8070939"
            }
        ]
    },
    {
        "mention": "european union",
        "type": "ORG",
        "predictions": [
            {
                "entity": "European Union",
                "confidence_score": "9.689422"
            },
            {
                "entity": "European Union and the United Nations",
                "confidence_score": "-3.5153482"
            },
            {
                "entity": "Council of Europe",
                "confidence_score": "-3.839314"
            }
        ]
    },
    {
        "mention": "grabar-kitarovic",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kolinda Grabar-Kitarovi\u0107",
                "confidence_score": "5.1088285"
            },
            {
                "entity": "Berislavi\u0107 family of Graberje",
                "confidence_score": "-11.413875"
            },
            {
                "entity": "Grabovica, Kotor Varo\u0161",
                "confidence_score": "-11.923662"
            }
        ]
    },
    {
        "mention": "mediterranean",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Mediterranean Sea",
                "confidence_score": "2.0298774"
            },
            {
                "entity": "Adriatic Sea",
                "confidence_score": "-3.7762926"
            },
            {
                "entity": "Mediterranean climate",
                "confidence_score": "-4.8675737"
            }
        ]
    },
    {
        "mention": "europe",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Europe",
                "confidence_score": "4.80574"
            },
            {
                "entity": "Council of Europe",
                "confidence_score": "-4.198911"
            },
            {
                "entity": "Time in Europe",
                "confidence_score": "-7.8469768"
            }
        ]
    },
    {
        "mention": "the international organization for migration",
        "type": "ORG",
        "predictions": [
            {
                "entity": "International Organization for Migration",
                "confidence_score": "-1.8067393"
            },
            {
                "entity": "International migration",
                "confidence_score": "-8.878758"
            },
            {
                "entity": "United Nations High Commissioner for Refugees",
                "confidence_score": "-9.8502035"
            }
        ]
    },
    {
        "mention": "eu",
        "type": "ORG",
        "predictions": [
            {
                "entity": "European Union",
                "confidence_score": "8.82327"
            },
            {
                "entity": "Council of Europe",
                "confidence_score": "-1.8394974"
            },
            {
                "entity": "European Parliament",
                "confidence_score": "-2.4300888"
            }
        ]
    },
    {
        "mention": "beyond security",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Beyond Fear: Thinking Sensibly About Security in an Uncertain World",
                "confidence_score": "-4.483824"
            },
            {
                "entity": "Beyond the Boundary",
                "confidence_score": "-8.727658"
            },
            {
                "entity": "Beyond Limits",
                "confidence_score": "-10.786933"
            }
        ]
    },
    {
        "mention": "alp",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "9.75735"
            },
            {
                "entity": "Liberal Party of Australia",
                "confidence_score": "-0.30665222"
            },
            {
                "entity": "Australian Labor Party Caucus",
                "confidence_score": "-1.8082628"
            }
        ]
    },
    {
        "mention": "bill shorten",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bill Shorten",
                "confidence_score": "11.175131"
            },
            {
                "entity": "Bill Brown (Australian politician)",
                "confidence_score": "-11.409501"
            },
            {
                "entity": "Bill Stretch",
                "confidence_score": "-11.620874"
            }
        ]
    },
    {
        "mention": "malcolm turnbull",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Malcolm Turnbull",
                "confidence_score": "10.568268"
            },
            {
                "entity": "Malcolm Fraser",
                "confidence_score": "-7.4552507"
            },
            {
                "entity": "Malcolm Andrews",
                "confidence_score": "-10.548671"
            }
        ]
    },
    {
        "mention": "the morgan poll",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Roy Morgan Research",
                "confidence_score": "-7.930781"
            },
            {
                "entity": "Opinion polling for the next Australian federal election",
                "confidence_score": "-10.010151"
            },
            {
                "entity": "Harper Polling",
                "confidence_score": "-10.813744"
            }
        ]
    },
    {
        "mention": "labor",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "4.0723853"
            },
            {
                "entity": "Australian Labor Party (New South Wales Branch)",
                "confidence_score": "-1.6283982"
            },
            {
                "entity": "Australian Labor Party (Australian Capital Territory Branch)",
                "confidence_score": "-2.7244446"
            }
        ]
    },
    {
        "mention": "bill shorten",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bill Shorten",
                "confidence_score": "11.175131"
            },
            {
                "entity": "Bill Brown (Australian politician)",
                "confidence_score": "-11.409501"
            },
            {
                "entity": "Bill Stretch",
                "confidence_score": "-11.620874"
            }
        ]
    },
    {
        "mention": "morgan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Charles Morgan (Australian politician)",
                "confidence_score": "-5.300947"
            },
            {
                "entity": "Arthur Morgan (Australian politician)",
                "confidence_score": "-7.485084"
            },
            {
                "entity": "Alf Morgans",
                "confidence_score": "-9.045464"
            }
        ]
    },
    {
        "mention": "alp",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "9.75735"
            },
            {
                "entity": "Liberal Party of Australia",
                "confidence_score": "-0.30665222"
            },
            {
                "entity": "Australian Labor Party Caucus",
                "confidence_score": "-1.8082628"
            }
        ]
    },
    {
        "mention": "shorten",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Bill Shorten",
                "confidence_score": "2.9178472"
            },
            {
                "entity": "Anthony Shorten",
                "confidence_score": "-3.6046405"
            },
            {
                "entity": "Division of Shortland",
                "confidence_score": "-8.91212"
            }
        ]
    },
    {
        "mention": "labor",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "4.0723853"
            },
            {
                "entity": "Australian Labor Party (New South Wales Branch)",
                "confidence_score": "-1.6283982"
            },
            {
                "entity": "Australian Labor Party (Australian Capital Territory Branch)",
                "confidence_score": "-2.7244446"
            }
        ]
    },
    {
        "mention": "yarralumla",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Yarralumla, Australian Capital Territory",
                "confidence_score": "0.2666576"
            },
            {
                "entity": "Yarrowlumla Shire",
                "confidence_score": "-6.0518813"
            },
            {
                "entity": "Electoral district of Yaralla",
                "confidence_score": "-10.557759"
            }
        ]
    },
    {
        "mention": "nationals",
        "type": "LOC",
        "predictions": [
            {
                "entity": "National Party of Australia",
                "confidence_score": "8.716648"
            },
            {
                "entity": "National Party of Australia \u2013 NSW",
                "confidence_score": "6.115891"
            },
            {
                "entity": "National Party of Australia \u2013 Victoria",
                "confidence_score": "1.8994992"
            }
        ]
    },
    {
        "mention": "nationals",
        "type": "LOC",
        "predictions": [
            {
                "entity": "National Party of Australia",
                "confidence_score": "8.716648"
            },
            {
                "entity": "National Party of Australia \u2013 NSW",
                "confidence_score": "6.115891"
            },
            {
                "entity": "National Party of Australia \u2013 Victoria",
                "confidence_score": "1.8994992"
            }
        ]
    },
    {
        "mention": "labor",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "4.0723853"
            },
            {
                "entity": "Australian Labor Party (New South Wales Branch)",
                "confidence_score": "-1.6283982"
            },
            {
                "entity": "Australian Labor Party (Australian Capital Territory Branch)",
                "confidence_score": "-2.7244446"
            }
        ]
    },
    {
        "mention": "reachtel",
        "type": "ORG",
        "predictions": [
            {
                "entity": "ReachTEL",
                "confidence_score": "9.550408"
            },
            {
                "entity": "TPG Telecom",
                "confidence_score": "-9.268029"
            },
            {
                "entity": "Transperth",
                "confidence_score": "-10.560846"
            }
        ]
    },
    {
        "mention": "western australia\u2019s",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Western Australia",
                "confidence_score": "7.843019"
            },
            {
                "entity": "Government of Western Australia",
                "confidence_score": "-4.192222"
            },
            {
                "entity": "Perth",
                "confidence_score": "-4.664821"
            }
        ]
    },
    {
        "mention": "labor",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "4.0723853"
            },
            {
                "entity": "Australian Labor Party (New South Wales Branch)",
                "confidence_score": "-1.6283982"
            },
            {
                "entity": "Australian Labor Party (Australian Capital Territory Branch)",
                "confidence_score": "-2.7244446"
            }
        ]
    },
    {
        "mention": "alp",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "9.75735"
            },
            {
                "entity": "Liberal Party of Australia",
                "confidence_score": "-0.30665222"
            },
            {
                "entity": "Australian Labor Party Caucus",
                "confidence_score": "-1.8082628"
            }
        ]
    },
    {
        "mention": "hawke",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bob Hawke",
                "confidence_score": "4.7292747"
            },
            {
                "entity": "Hawke Government",
                "confidence_score": "-0.73986286"
            },
            {
                "entity": "Electoral district of Hawkesbury",
                "confidence_score": "-8.415814"
            }
        ]
    },
    {
        "mention": "canning",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Division of Canning",
                "confidence_score": "1.3687484"
            },
            {
                "entity": "Electoral district of Canning",
                "confidence_score": "-2.6417148"
            },
            {
                "entity": "Electoral district of Cannington",
                "confidence_score": "-5.2056365"
            }
        ]
    },
    {
        "mention": "liberal mp",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Liberal Party of Australia",
                "confidence_score": "0.29603356"
            },
            {
                "entity": "List of New Zealand Liberal Party MPs",
                "confidence_score": "-6.766188"
            },
            {
                "entity": "Liberal Party of Australia (Tasmanian Division)",
                "confidence_score": "-8.266706"
            }
        ]
    },
    {
        "mention": "don randall",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Don Randall",
                "confidence_score": "12.53306"
            },
            {
                "entity": "Don Cody",
                "confidence_score": "-12.283795"
            },
            {
                "entity": "Don Darling",
                "confidence_score": "-12.319475"
            }
        ]
    },
    {
        "mention": "liberals",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Liberal Party of Australia (New South Wales Division)",
                "confidence_score": "7.5084767"
            },
            {
                "entity": "Liberal Party of Australia",
                "confidence_score": "6.149262"
            },
            {
                "entity": "Liberal Party of Canada",
                "confidence_score": "-8.332208"
            }
        ]
    },
    {
        "mention": "howard",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "John Howard",
                "confidence_score": "3.7991908"
            },
            {
                "entity": "Howard Government",
                "confidence_score": "-2.429742"
            },
            {
                "entity": "Michael Howard",
                "confidence_score": "-7.9599032"
            }
        ]
    },
    {
        "mention": "canning",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Division of Canning",
                "confidence_score": "1.3687484"
            },
            {
                "entity": "Electoral district of Canning",
                "confidence_score": "-2.6417148"
            },
            {
                "entity": "Electoral district of Cannington",
                "confidence_score": "-5.2056365"
            }
        ]
    },
    {
        "mention": "lindsay",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Division of Lindsay",
                "confidence_score": "3.1409073"
            },
            {
                "entity": "Robert Lindsay (Australian politician)",
                "confidence_score": "-4.8408933"
            },
            {
                "entity": "1996 Lindsay by-election",
                "confidence_score": "-8.629483"
            }
        ]
    },
    {
        "mention": "nsw",
        "type": "ORG",
        "predictions": [
            {
                "entity": "New South Wales",
                "confidence_score": "7.0234013"
            },
            {
                "entity": "Transdev NSW",
                "confidence_score": "-9.391757"
            },
            {
                "entity": "NSW TrainLink",
                "confidence_score": "-9.591588"
            }
        ]
    },
    {
        "mention": "galaxy",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Galaxy (Australian television)",
                "confidence_score": "1.002422"
            },
            {
                "entity": "Galaxy Resources",
                "confidence_score": "-5.135199"
            },
            {
                "entity": "Galaxy (radio network)",
                "confidence_score": "-7.2699904"
            }
        ]
    },
    {
        "mention": "tony abbott",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Tony Abbott",
                "confidence_score": "11.302775"
            },
            {
                "entity": "Anthony Abbott",
                "confidence_score": "-5.048757"
            },
            {
                "entity": "Tony Abbott (Canadian politician)",
                "confidence_score": "-5.979717"
            }
        ]
    },
    {
        "mention": "turnbull",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Malcolm Turnbull",
                "confidence_score": "-0.58703876"
            },
            {
                "entity": "Turnbull Government",
                "confidence_score": "-3.1242805"
            },
            {
                "entity": "Malcolm Fraser",
                "confidence_score": "-8.084025"
            }
        ]
    },
    {
        "mention": "abbott",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Abbott Government",
                "confidence_score": "-3.1683836"
            },
            {
                "entity": "Tony Abbott",
                "confidence_score": "-3.4744928"
            },
            {
                "entity": "Steven Marshall",
                "confidence_score": "-9.844067"
            }
        ]
    },
    {
        "mention": "alp",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "9.75735"
            },
            {
                "entity": "Liberal Party of Australia",
                "confidence_score": "-0.30665222"
            },
            {
                "entity": "Australian Labor Party Caucus",
                "confidence_score": "-1.8082628"
            }
        ]
    },
    {
        "mention": "shorten",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bill Shorten",
                "confidence_score": "3.940631"
            },
            {
                "entity": "Anthony Shorten",
                "confidence_score": "-3.7404323"
            },
            {
                "entity": "Division of Shortland",
                "confidence_score": "-8.627759"
            }
        ]
    },
    {
        "mention": "labor",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "4.0723853"
            },
            {
                "entity": "Australian Labor Party (New South Wales Branch)",
                "confidence_score": "-1.6283982"
            },
            {
                "entity": "Australian Labor Party (Australian Capital Territory Branch)",
                "confidence_score": "-2.7244446"
            }
        ]
    },
    {
        "mention": "labor",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "4.0723853"
            },
            {
                "entity": "Australian Labor Party (New South Wales Branch)",
                "confidence_score": "-1.6283982"
            },
            {
                "entity": "Australian Labor Party (Australian Capital Territory Branch)",
                "confidence_score": "-2.7244446"
            }
        ]
    },
    {
        "mention": "labor",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "4.0723853"
            },
            {
                "entity": "Australian Labor Party (New South Wales Branch)",
                "confidence_score": "-1.6283982"
            },
            {
                "entity": "Australian Labor Party (Australian Capital Territory Branch)",
                "confidence_score": "-2.7244446"
            }
        ]
    },
    {
        "mention": "alp",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "9.75735"
            },
            {
                "entity": "Liberal Party of Australia",
                "confidence_score": "-0.30665222"
            },
            {
                "entity": "Australian Labor Party Caucus",
                "confidence_score": "-1.8082628"
            }
        ]
    },
    {
        "mention": "abbott",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Abbott Government",
                "confidence_score": "-3.1683836"
            },
            {
                "entity": "Tony Abbott",
                "confidence_score": "-3.4744928"
            },
            {
                "entity": "Steven Marshall",
                "confidence_score": "-9.844067"
            }
        ]
    },
    {
        "mention": "shorten",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bill Shorten",
                "confidence_score": "3.940631"
            },
            {
                "entity": "Anthony Shorten",
                "confidence_score": "-3.7404323"
            },
            {
                "entity": "Division of Shortland",
                "confidence_score": "-8.627759"
            }
        ]
    },
    {
        "mention": "shorten",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bill Shorten",
                "confidence_score": "3.940631"
            },
            {
                "entity": "Anthony Shorten",
                "confidence_score": "-3.7404323"
            },
            {
                "entity": "Division of Shortland",
                "confidence_score": "-8.627759"
            }
        ]
    },
    {
        "mention": "labor",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "4.0723853"
            },
            {
                "entity": "Australian Labor Party (New South Wales Branch)",
                "confidence_score": "-1.6283982"
            },
            {
                "entity": "Australian Labor Party (Australian Capital Territory Branch)",
                "confidence_score": "-2.7244446"
            }
        ]
    },
    {
        "mention": "labor",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "4.0723853"
            },
            {
                "entity": "Australian Labor Party (New South Wales Branch)",
                "confidence_score": "-1.6283982"
            },
            {
                "entity": "Australian Labor Party (Australian Capital Territory Branch)",
                "confidence_score": "-2.7244446"
            }
        ]
    },
    {
        "mention": "shorten",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bill Shorten",
                "confidence_score": "3.940631"
            },
            {
                "entity": "Anthony Shorten",
                "confidence_score": "-3.7404323"
            },
            {
                "entity": "Division of Shortland",
                "confidence_score": "-8.627759"
            }
        ]
    },
    {
        "mention": "labor",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "4.0723853"
            },
            {
                "entity": "Australian Labor Party (New South Wales Branch)",
                "confidence_score": "-1.6283982"
            },
            {
                "entity": "Australian Labor Party (Australian Capital Territory Branch)",
                "confidence_score": "-2.7244446"
            }
        ]
    },
    {
        "mention": "kevin rudd",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kevin Rudd",
                "confidence_score": "11.032903"
            },
            {
                "entity": "Kevin Moss",
                "confidence_score": "-11.466459"
            },
            {
                "entity": "Kevin Stewart (Australian politician)",
                "confidence_score": "-11.759232"
            }
        ]
    },
    {
        "mention": "shorten",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bill Shorten",
                "confidence_score": "3.940631"
            },
            {
                "entity": "Anthony Shorten",
                "confidence_score": "-3.7404323"
            },
            {
                "entity": "Division of Shortland",
                "confidence_score": "-8.627759"
            }
        ]
    },
    {
        "mention": "alp",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "9.75735"
            },
            {
                "entity": "Liberal Party of Australia",
                "confidence_score": "-0.30665222"
            },
            {
                "entity": "Australian Labor Party Caucus",
                "confidence_score": "-1.8082628"
            }
        ]
    },
    {
        "mention": "labor",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "4.0723853"
            },
            {
                "entity": "Australian Labor Party (New South Wales Branch)",
                "confidence_score": "-1.6283982"
            },
            {
                "entity": "Australian Labor Party (Australian Capital Territory Branch)",
                "confidence_score": "-2.7244446"
            }
        ]
    },
    {
        "mention": "labor",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "4.0723853"
            },
            {
                "entity": "Australian Labor Party (New South Wales Branch)",
                "confidence_score": "-1.6283982"
            },
            {
                "entity": "Australian Labor Party (Australian Capital Territory Branch)",
                "confidence_score": "-2.7244446"
            }
        ]
    },
    {
        "mention": "shorten",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bill Shorten",
                "confidence_score": "3.940631"
            },
            {
                "entity": "Anthony Shorten",
                "confidence_score": "-3.7404323"
            },
            {
                "entity": "Division of Shortland",
                "confidence_score": "-8.627759"
            }
        ]
    },
    {
        "mention": "alp",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "9.75735"
            },
            {
                "entity": "Liberal Party of Australia",
                "confidence_score": "-0.30665222"
            },
            {
                "entity": "Australian Labor Party Caucus",
                "confidence_score": "-1.8082628"
            }
        ]
    },
    {
        "mention": "labor",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "4.0723853"
            },
            {
                "entity": "Australian Labor Party (New South Wales Branch)",
                "confidence_score": "-1.6283982"
            },
            {
                "entity": "Australian Labor Party (Australian Capital Territory Branch)",
                "confidence_score": "-2.7244446"
            }
        ]
    },
    {
        "mention": "labor",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "4.0723853"
            },
            {
                "entity": "Australian Labor Party (New South Wales Branch)",
                "confidence_score": "-1.6283982"
            },
            {
                "entity": "Australian Labor Party (Australian Capital Territory Branch)",
                "confidence_score": "-2.7244446"
            }
        ]
    },
    {
        "mention": "canning",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Division of Canning",
                "confidence_score": "1.3687484"
            },
            {
                "entity": "Electoral district of Canning",
                "confidence_score": "-2.6417148"
            },
            {
                "entity": "Electoral district of Cannington",
                "confidence_score": "-5.2056365"
            }
        ]
    },
    {
        "mention": "the house of representatives",
        "type": "ORG",
        "predictions": [
            {
                "entity": "United States House of Representatives",
                "confidence_score": "-2.7429712"
            },
            {
                "entity": "Oregon House of Representatives",
                "confidence_score": "-4.0283885"
            },
            {
                "entity": "Arizona House of Representatives",
                "confidence_score": "-8.782681"
            }
        ]
    },
    {
        "mention": "alp",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "9.75735"
            },
            {
                "entity": "Liberal Party of Australia",
                "confidence_score": "-0.30665222"
            },
            {
                "entity": "Australian Labor Party Caucus",
                "confidence_score": "-1.8082628"
            }
        ]
    },
    {
        "mention": "john howard",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "John Howard",
                "confidence_score": "10.323403"
            },
            {
                "entity": "John Howard (Australian actor)",
                "confidence_score": "-9.177613"
            },
            {
                "entity": "John Howard (Southampton Test MP)",
                "confidence_score": "-10.726227"
            }
        ]
    },
    {
        "mention": "labor",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "4.0723853"
            },
            {
                "entity": "Australian Labor Party (New South Wales Branch)",
                "confidence_score": "-1.6283982"
            },
            {
                "entity": "Australian Labor Party (Australian Capital Territory Branch)",
                "confidence_score": "-2.7244446"
            }
        ]
    },
    {
        "mention": "canning",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Division of Canning",
                "confidence_score": "1.3687484"
            },
            {
                "entity": "Electoral district of Canning",
                "confidence_score": "-2.6417148"
            },
            {
                "entity": "Electoral district of Cannington",
                "confidence_score": "-5.2056365"
            }
        ]
    },
    {
        "mention": "labor",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "4.0723853"
            },
            {
                "entity": "Australian Labor Party (New South Wales Branch)",
                "confidence_score": "-1.6283982"
            },
            {
                "entity": "Australian Labor Party (Australian Capital Territory Branch)",
                "confidence_score": "-2.7244446"
            }
        ]
    },
    {
        "mention": "labor",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "4.0723853"
            },
            {
                "entity": "Australian Labor Party (New South Wales Branch)",
                "confidence_score": "-1.6283982"
            },
            {
                "entity": "Australian Labor Party (Australian Capital Territory Branch)",
                "confidence_score": "-2.7244446"
            }
        ]
    },
    {
        "mention": "labor",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "4.0723853"
            },
            {
                "entity": "Australian Labor Party (New South Wales Branch)",
                "confidence_score": "-1.6283982"
            },
            {
                "entity": "Australian Labor Party (Australian Capital Territory Branch)",
                "confidence_score": "-2.7244446"
            }
        ]
    },
    {
        "mention": "alp",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "9.75735"
            },
            {
                "entity": "Liberal Party of Australia",
                "confidence_score": "-0.30665222"
            },
            {
                "entity": "Australian Labor Party Caucus",
                "confidence_score": "-1.8082628"
            }
        ]
    },
    {
        "mention": "alp",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "9.75735"
            },
            {
                "entity": "Liberal Party of Australia",
                "confidence_score": "-0.30665222"
            },
            {
                "entity": "Australian Labor Party Caucus",
                "confidence_score": "-1.8082628"
            }
        ]
    },
    {
        "mention": "alp",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "9.75735"
            },
            {
                "entity": "Liberal Party of Australia",
                "confidence_score": "-0.30665222"
            },
            {
                "entity": "Australian Labor Party Caucus",
                "confidence_score": "-1.8082628"
            }
        ]
    },
    {
        "mention": "alp",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "9.75735"
            },
            {
                "entity": "Liberal Party of Australia",
                "confidence_score": "-0.30665222"
            },
            {
                "entity": "Australian Labor Party Caucus",
                "confidence_score": "-1.8082628"
            }
        ]
    },
    {
        "mention": "labor",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "4.0723853"
            },
            {
                "entity": "Australian Labor Party (New South Wales Branch)",
                "confidence_score": "-1.6283982"
            },
            {
                "entity": "Australian Labor Party (Australian Capital Territory Branch)",
                "confidence_score": "-2.7244446"
            }
        ]
    },
    {
        "mention": "alp",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "9.75735"
            },
            {
                "entity": "Liberal Party of Australia",
                "confidence_score": "-0.30665222"
            },
            {
                "entity": "Australian Labor Party Caucus",
                "confidence_score": "-1.8082628"
            }
        ]
    },
    {
        "mention": "labor",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "4.0723853"
            },
            {
                "entity": "Australian Labor Party (New South Wales Branch)",
                "confidence_score": "-1.6283982"
            },
            {
                "entity": "Australian Labor Party (Australian Capital Territory Branch)",
                "confidence_score": "-2.7244446"
            }
        ]
    },
    {
        "mention": "labor",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "4.0723853"
            },
            {
                "entity": "Australian Labor Party (New South Wales Branch)",
                "confidence_score": "-1.6283982"
            },
            {
                "entity": "Australian Labor Party (Australian Capital Territory Branch)",
                "confidence_score": "-2.7244446"
            }
        ]
    },
    {
        "mention": "labor",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "4.0723853"
            },
            {
                "entity": "Australian Labor Party (New South Wales Branch)",
                "confidence_score": "-1.6283982"
            },
            {
                "entity": "Australian Labor Party (Australian Capital Territory Branch)",
                "confidence_score": "-2.7244446"
            }
        ]
    },
    {
        "mention": "rudd",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kevin Rudd",
                "confidence_score": "0.9712579"
            },
            {
                "entity": "Rudd Government (2007\u20132010)",
                "confidence_score": "-3.074466"
            },
            {
                "entity": "Philip Ruddock",
                "confidence_score": "-9.005422"
            }
        ]
    },
    {
        "mention": "wayne swan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Wayne Swan",
                "confidence_score": "12.203911"
            },
            {
                "entity": "Wayne Martin (judge)",
                "confidence_score": "-11.387193"
            },
            {
                "entity": "Wayne Berry",
                "confidence_score": "-11.447394"
            }
        ]
    },
    {
        "mention": "labor",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "4.0723853"
            },
            {
                "entity": "Australian Labor Party (New South Wales Branch)",
                "confidence_score": "-1.6283982"
            },
            {
                "entity": "Australian Labor Party (Australian Capital Territory Branch)",
                "confidence_score": "-2.7244446"
            }
        ]
    },
    {
        "mention": "rudd",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kevin Rudd",
                "confidence_score": "0.9712579"
            },
            {
                "entity": "Rudd Government (2007\u20132010)",
                "confidence_score": "-3.074466"
            },
            {
                "entity": "Philip Ruddock",
                "confidence_score": "-9.005422"
            }
        ]
    },
    {
        "mention": "julia gillard",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Julia Gillard",
                "confidence_score": "11.079709"
            },
            {
                "entity": "Gillard Government",
                "confidence_score": "-3.544491"
            },
            {
                "entity": "Julia Banks",
                "confidence_score": "-11.10387"
            }
        ]
    },
    {
        "mention": "actu",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Council of Trade Unions",
                "confidence_score": "4.713805"
            },
            {
                "entity": "ACT New Zealand",
                "confidence_score": "-6.8611875"
            },
            {
                "entity": "ACT Government",
                "confidence_score": "-6.955899"
            }
        ]
    },
    {
        "mention": "labor",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "4.0723853"
            },
            {
                "entity": "Australian Labor Party (New South Wales Branch)",
                "confidence_score": "-1.6283982"
            },
            {
                "entity": "Australian Labor Party (Australian Capital Territory Branch)",
                "confidence_score": "-2.7244446"
            }
        ]
    },
    {
        "mention": "rudd",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kevin Rudd",
                "confidence_score": "0.9712579"
            },
            {
                "entity": "Rudd Government (2007\u20132010)",
                "confidence_score": "-3.074466"
            },
            {
                "entity": "Philip Ruddock",
                "confidence_score": "-9.005422"
            }
        ]
    },
    {
        "mention": "labor",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "4.0723853"
            },
            {
                "entity": "Australian Labor Party (New South Wales Branch)",
                "confidence_score": "-1.6283982"
            },
            {
                "entity": "Australian Labor Party (Australian Capital Territory Branch)",
                "confidence_score": "-2.7244446"
            }
        ]
    },
    {
        "mention": "labor",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "4.0723853"
            },
            {
                "entity": "Australian Labor Party (New South Wales Branch)",
                "confidence_score": "-1.6283982"
            },
            {
                "entity": "Australian Labor Party (Australian Capital Territory Branch)",
                "confidence_score": "-2.7244446"
            }
        ]
    },
    {
        "mention": "labor",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "4.0723853"
            },
            {
                "entity": "Australian Labor Party (New South Wales Branch)",
                "confidence_score": "-1.6283982"
            },
            {
                "entity": "Australian Labor Party (Australian Capital Territory Branch)",
                "confidence_score": "-2.7244446"
            }
        ]
    },
    {
        "mention": "labor",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "4.0723853"
            },
            {
                "entity": "Australian Labor Party (New South Wales Branch)",
                "confidence_score": "-1.6283982"
            },
            {
                "entity": "Australian Labor Party (Australian Capital Territory Branch)",
                "confidence_score": "-2.7244446"
            }
        ]
    },
    {
        "mention": "labor",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "4.0723853"
            },
            {
                "entity": "Australian Labor Party (New South Wales Branch)",
                "confidence_score": "-1.6283982"
            },
            {
                "entity": "Australian Labor Party (Australian Capital Territory Branch)",
                "confidence_score": "-2.7244446"
            }
        ]
    },
    {
        "mention": "howard",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "John Howard",
                "confidence_score": "3.7991908"
            },
            {
                "entity": "Howard Government",
                "confidence_score": "-2.429742"
            },
            {
                "entity": "Michael Howard",
                "confidence_score": "-7.9599032"
            }
        ]
    },
    {
        "mention": "kindy",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Alex Kindy",
                "confidence_score": "-9.07692"
            },
            {
                "entity": "Ben Chifley",
                "confidence_score": "-9.183324"
            },
            {
                "entity": "Results of the 2013 Australian federal election in territories",
                "confidence_score": "-11.610268"
            }
        ]
    },
    {
        "mention": "alp",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "9.75735"
            },
            {
                "entity": "Liberal Party of Australia",
                "confidence_score": "-0.30665222"
            },
            {
                "entity": "Australian Labor Party Caucus",
                "confidence_score": "-1.8082628"
            }
        ]
    },
    {
        "mention": "john black",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "John Black (Australian politician)",
                "confidence_score": "6.4515595"
            },
            {
                "entity": "John Black (director)",
                "confidence_score": "-10.166463"
            },
            {
                "entity": "John Black (Georgia politician)",
                "confidence_score": "-11.881625"
            }
        ]
    },
    {
        "mention": "labor",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "4.0723853"
            },
            {
                "entity": "Australian Labor Party (New South Wales Branch)",
                "confidence_score": "-1.6283982"
            },
            {
                "entity": "Australian Labor Party (Australian Capital Territory Branch)",
                "confidence_score": "-2.7244446"
            }
        ]
    },
    {
        "mention": "australian development strategies",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Committee for Economic Development of Australia",
                "confidence_score": "-8.79475"
            },
            {
                "entity": "Australian Republic Movement",
                "confidence_score": "-9.243781"
            },
            {
                "entity": "Engineered Australia Plan Party",
                "confidence_score": "-9.922927"
            }
        ]
    },
    {
        "mention": "alp",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "9.75735"
            },
            {
                "entity": "Liberal Party of Australia",
                "confidence_score": "-0.30665222"
            },
            {
                "entity": "Australian Labor Party Caucus",
                "confidence_score": "-1.8082628"
            }
        ]
    },
    {
        "mention": "davis county school district",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Davis School District",
                "confidence_score": "0.26024666"
            },
            {
                "entity": "Davis County, Utah",
                "confidence_score": "-5.546171"
            },
            {
                "entity": "Jeff Davis County School District",
                "confidence_score": "-10.779643"
            }
        ]
    },
    {
        "mention": "dui",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Driving under the influence",
                "confidence_score": "3.0896108"
            },
            {
                "entity": "Dud",
                "confidence_score": "-8.460576"
            },
            {
                "entity": "Duisburg",
                "confidence_score": "-8.680082"
            }
        ]
    },
    {
        "mention": "brightgreen",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Shades of yellow",
                "confidence_score": "-11.037421"
            },
            {
                "entity": "GelGreen",
                "confidence_score": "-11.137072"
            },
            {
                "entity": "BrightVolt",
                "confidence_score": "-11.805461"
            }
        ]
    },
    {
        "mention": "the d700+",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Flyer New Look bus",
                "confidence_score": "-10.521388"
            },
            {
                "entity": "70,000",
                "confidence_score": "-10.596213"
            },
            {
                "entity": "2000 yen note",
                "confidence_score": "-11.274036"
            }
        ]
    },
    {
        "mention": "tru-colour",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "8-bit color",
                "confidence_score": "-7.0598793"
            },
            {
                "entity": "RGB color model",
                "confidence_score": "-10.345664"
            },
            {
                "entity": "Tristimulus colorimeter",
                "confidence_score": "-11.437182"
            }
        ]
    },
    {
        "mention": "jeremy corbyn",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jeremy Corbyn",
                "confidence_score": "12.129742"
            },
            {
                "entity": "Labour Party leadership of Jeremy Corbyn",
                "confidence_score": "-6.0791974"
            },
            {
                "entity": "Shadow Cabinet of Jeremy Corbyn",
                "confidence_score": "-7.130388"
            }
        ]
    },
    {
        "mention": "labour",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Labour Party (UK)",
                "confidence_score": "2.7017815"
            },
            {
                "entity": "Labour Party (Ireland)",
                "confidence_score": "-7.014437"
            },
            {
                "entity": "New Labour",
                "confidence_score": "-7.3273387"
            }
        ]
    },
    {
        "mention": "labour",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Labour Party (UK)",
                "confidence_score": "2.7017815"
            },
            {
                "entity": "Labour Party (Ireland)",
                "confidence_score": "-7.014437"
            },
            {
                "entity": "New Labour",
                "confidence_score": "-7.3273387"
            }
        ]
    },
    {
        "mention": "corbyn",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jeremy Corbyn",
                "confidence_score": "0.5464574"
            },
            {
                "entity": "Labour Party leadership of Jeremy Corbyn",
                "confidence_score": "-6.71769"
            },
            {
                "entity": "Corbynmania",
                "confidence_score": "-8.093617"
            }
        ]
    },
    {
        "mention": "corbyn",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jeremy Corbyn",
                "confidence_score": "0.5464574"
            },
            {
                "entity": "Labour Party leadership of Jeremy Corbyn",
                "confidence_score": "-6.71769"
            },
            {
                "entity": "Corbynmania",
                "confidence_score": "-8.093617"
            }
        ]
    },
    {
        "mention": "jez",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jez",
                "confidence_score": "-9.403179"
            },
            {
                "entity": "Jez Smith",
                "confidence_score": "-9.855446"
            },
            {
                "entity": "Jerzy Je\u017c",
                "confidence_score": "-10.687274"
            }
        ]
    },
    {
        "mention": "corbyn",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jeremy Corbyn",
                "confidence_score": "0.5464574"
            },
            {
                "entity": "Labour Party leadership of Jeremy Corbyn",
                "confidence_score": "-6.71769"
            },
            {
                "entity": "Corbynmania",
                "confidence_score": "-8.093617"
            }
        ]
    },
    {
        "mention": "corbyn",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jeremy Corbyn",
                "confidence_score": "0.5464574"
            },
            {
                "entity": "Labour Party leadership of Jeremy Corbyn",
                "confidence_score": "-6.71769"
            },
            {
                "entity": "Corbynmania",
                "confidence_score": "-8.093617"
            }
        ]
    },
    {
        "mention": "corbyn",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jeremy Corbyn",
                "confidence_score": "0.5464574"
            },
            {
                "entity": "Labour Party leadership of Jeremy Corbyn",
                "confidence_score": "-6.71769"
            },
            {
                "entity": "Corbynmania",
                "confidence_score": "-8.093617"
            }
        ]
    },
    {
        "mention": "green party",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Green party",
                "confidence_score": "5.3362203"
            },
            {
                "entity": "Green Party (UK)",
                "confidence_score": "3.8751528"
            },
            {
                "entity": "Green Party of England and Wales",
                "confidence_score": "0.06545478"
            }
        ]
    },
    {
        "mention": "natalie bennett",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Natalie Bennett",
                "confidence_score": "12.218274"
            },
            {
                "entity": "Natalie Elphicke",
                "confidence_score": "-10.461393"
            },
            {
                "entity": "Natalie Massenet",
                "confidence_score": "-10.998551"
            }
        ]
    },
    {
        "mention": "david cameron",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "David Cameron",
                "confidence_score": "10.9060545"
            },
            {
                "entity": "Premiership of David Cameron",
                "confidence_score": "-6.4903407"
            },
            {
                "entity": "John Cameron, Lord Cameron",
                "confidence_score": "-11.321237"
            }
        ]
    },
    {
        "mention": "europe",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Europe",
                "confidence_score": "3.2753286"
            },
            {
                "entity": "European Union",
                "confidence_score": "-1.5593975"
            },
            {
                "entity": "Council of Europe",
                "confidence_score": "-2.6994913"
            }
        ]
    },
    {
        "mention": "labour",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Labour Party (UK)",
                "confidence_score": "2.7017815"
            },
            {
                "entity": "Labour Party (Ireland)",
                "confidence_score": "-7.014437"
            },
            {
                "entity": "New Labour",
                "confidence_score": "-7.3273387"
            }
        ]
    },
    {
        "mention": "carol turner",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Carol I. Turner",
                "confidence_score": "-10.652766"
            },
            {
                "entity": "Mary Turner (trade unionist)",
                "confidence_score": "-10.715254"
            },
            {
                "entity": "Carol Bellamy",
                "confidence_score": "-11.365939"
            }
        ]
    },
    {
        "mention": "stop the war coalition",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Stop the War Coalition",
                "confidence_score": "10.414976"
            },
            {
                "entity": "Stop the War Coalition (Australia)",
                "confidence_score": "-5.6734443"
            },
            {
                "entity": "Arms Reduction Coalition",
                "confidence_score": "-9.634313"
            }
        ]
    },
    {
        "mention": "parliament",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Parliament",
                "confidence_score": "7.3973947"
            },
            {
                "entity": "Parliament of the United Kingdom",
                "confidence_score": "4.891293"
            },
            {
                "entity": "Parliament of England",
                "confidence_score": "-4.1363077"
            }
        ]
    },
    {
        "mention": "walter wolfgang",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Walter Wolfgang",
                "confidence_score": "12.015021"
            },
            {
                "entity": "Walter Heller",
                "confidence_score": "-6.216104"
            },
            {
                "entity": "Walter Freud",
                "confidence_score": "-9.0312"
            }
        ]
    },
    {
        "mention": "cnd",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Campaign for Nuclear Disarmament",
                "confidence_score": "4.9109015"
            },
            {
                "entity": "Christian CND",
                "confidence_score": "-4.6788287"
            },
            {
                "entity": "Centre for Law and Democracy",
                "confidence_score": "-9.36251"
            }
        ]
    },
    {
        "mention": "wolfgang",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Walter Wolfgang",
                "confidence_score": "7.731377"
            },
            {
                "entity": "Wolfgang",
                "confidence_score": "-0.7777078"
            },
            {
                "entity": "Wolfgang Schuster",
                "confidence_score": "-9.278899"
            }
        ]
    },
    {
        "mention": "download) civil services mentor magazine",
        "type": "ORG",
        "predictions": [
            {
                "entity": "25 Magazine",
                "confidence_score": "-11.536889"
            },
            {
                "entity": "Specialized Systems Consultants",
                "confidence_score": "-11.701355"
            },
            {
                "entity": "Society of National Association Publications",
                "confidence_score": "-11.889682"
            }
        ]
    },
    {
        "mention": "pdf yojana magazine",
        "type": "ORG",
        "predictions": [
            {
                "entity": "PDF",
                "confidence_score": "-4.29032"
            },
            {
                "entity": "Yojana Malayalam",
                "confidence_score": "-10.763352"
            },
            {
                "entity": "Society of National Association Publications",
                "confidence_score": "-11.901158"
            }
        ]
    },
    {
        "mention": "current public administration magazine",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Public Administration (journal)",
                "confidence_score": "-10.280338"
            },
            {
                "entity": "Currentweek",
                "confidence_score": "-10.343081"
            },
            {
                "entity": "Current Affairs (magazine)",
                "confidence_score": "-10.858498"
            }
        ]
    },
    {
        "mention": "modi government",
        "type": "ORG",
        "predictions": [
            {
                "entity": "First Modi ministry",
                "confidence_score": "-3.1535122"
            },
            {
                "entity": "Narendra Modi",
                "confidence_score": "-3.9032998"
            },
            {
                "entity": "Modi ministry",
                "confidence_score": "-11.468361"
            }
        ]
    },
    {
        "mention": "aparajita priyadarshini behera",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Aparna Dixit",
                "confidence_score": "-11.457622"
            },
            {
                "entity": "Aparajita Goppi",
                "confidence_score": "-11.538665"
            },
            {
                "entity": "Priyadarshini Raje Scindia",
                "confidence_score": "-11.655857"
            }
        ]
    },
    {
        "mention": "indian economic service",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Indian Economic Service",
                "confidence_score": "11.819936"
            },
            {
                "entity": "Indian Revenue Service",
                "confidence_score": "-8.392486"
            },
            {
                "entity": "Indian Statistical Service",
                "confidence_score": "-9.539835"
            }
        ]
    },
    {
        "mention": "upsc exams",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Union Public Service Commission",
                "confidence_score": "-3.0394003"
            },
            {
                "entity": "Uttar Pradesh Subordinate Services Selection Commission",
                "confidence_score": "-6.3360825"
            },
            {
                "entity": "Uttar Pradesh Public Service Commission",
                "confidence_score": "-6.5125074"
            }
        ]
    },
    {
        "mention": "mahesh sharma",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mahesh Sharma",
                "confidence_score": "10.007316"
            },
            {
                "entity": "Mahesh Joshi",
                "confidence_score": "-11.039451"
            },
            {
                "entity": "Mahesh Joshi (Madhya Pradesh politician)",
                "confidence_score": "-12.039932"
            }
        ]
    },
    {
        "mention": "\u0930\u093e\u0937\u094d\u091f\u094d\u0930\u092e\u0902\u0921\u0932 \u0916\u0947\u0932",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Jhala Nath Khanal",
                "confidence_score": "-9.743552"
            },
            {
                "entity": "Khanal cabinet",
                "confidence_score": "-11.335571"
            },
            {
                "entity": "Khandkhel",
                "confidence_score": "-11.445943"
            }
        ]
    },
    {
        "mention": "download) civil services mentor magazine",
        "type": "ORG",
        "predictions": [
            {
                "entity": "25 Magazine",
                "confidence_score": "-11.536889"
            },
            {
                "entity": "Specialized Systems Consultants",
                "confidence_score": "-11.701355"
            },
            {
                "entity": "Society of National Association Publications",
                "confidence_score": "-11.889682"
            }
        ]
    },
    {
        "mention": "amazon click",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Amazon (company)",
                "confidence_score": "-5.560904"
            },
            {
                "entity": "Amazon Echo",
                "confidence_score": "-10.150369"
            },
            {
                "entity": "Apache Click",
                "confidence_score": "-11.264038"
            }
        ]
    },
    {
        "mention": "digital india socio economic",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Digital India",
                "confidence_score": "-4.382664"
            },
            {
                "entity": "Economy of India",
                "confidence_score": "-8.516699"
            },
            {
                "entity": "Digital Society Day",
                "confidence_score": "-9.509406"
            }
        ]
    },
    {
        "mention": "national issues international issues economy india & the world science & technology sports awards & prizes persons",
        "type": "ORG",
        "predictions": [
            {
                "entity": "2015 National Games of India",
                "confidence_score": "-8.429058"
            },
            {
                "entity": "National Games of India",
                "confidence_score": "-9.760227"
            },
            {
                "entity": "2015 National Games of India medal table",
                "confidence_score": "-10.541377"
            }
        ]
    },
    {
        "mention": "yojana magazine",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Yojana Malayalam",
                "confidence_score": "-7.3992524"
            },
            {
                "entity": "SA Rugby Magazine",
                "confidence_score": "-10.947701"
            },
            {
                "entity": "Yolk magazine",
                "confidence_score": "-11.785766"
            }
        ]
    },
    {
        "mention": "current public administration magazine",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Public Administration (journal)",
                "confidence_score": "-10.280338"
            },
            {
                "entity": "Currentweek",
                "confidence_score": "-10.343081"
            },
            {
                "entity": "Current Affairs (magazine)",
                "confidence_score": "-10.858498"
            }
        ]
    },
    {
        "mention": "modi government posted",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Narendra Modi",
                "confidence_score": "-3.3587415"
            },
            {
                "entity": "First Modi ministry",
                "confidence_score": "-7.311114"
            },
            {
                "entity": "Modi ministry",
                "confidence_score": "-12.308445"
            }
        ]
    },
    {
        "mention": "current public administration magazine welfare administration major",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Currentweek",
                "confidence_score": "-10.805299"
            },
            {
                "entity": "Public Administration (journal)",
                "confidence_score": "-11.07009"
            },
            {
                "entity": "Journal of Current Southeast Asian Affairs",
                "confidence_score": "-12.060736"
            }
        ]
    },
    {
        "mention": "modi government ever",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Narendra Modi",
                "confidence_score": "-4.494273"
            },
            {
                "entity": "Premiership of Narendra Modi",
                "confidence_score": "-6.5817175"
            },
            {
                "entity": "First Modi ministry (Gujarat)",
                "confidence_score": "-10.175757"
            }
        ]
    },
    {
        "mention": "modi government",
        "type": "ORG",
        "predictions": [
            {
                "entity": "First Modi ministry",
                "confidence_score": "-3.1535122"
            },
            {
                "entity": "Narendra Modi",
                "confidence_score": "-3.9032998"
            },
            {
                "entity": "Modi ministry",
                "confidence_score": "-11.468361"
            }
        ]
    },
    {
        "mention": "lok sabha",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Lok Sabha",
                "confidence_score": "10.685716"
            },
            {
                "entity": "Rajya Sabha",
                "confidence_score": "-1.6677055"
            },
            {
                "entity": "14th Lok Sabha",
                "confidence_score": "-7.9358997"
            }
        ]
    },
    {
        "mention": "pradhan mantri",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Pradhanmantri",
                "confidence_score": "8.048137"
            },
            {
                "entity": "Pradhan Mantri",
                "confidence_score": "7.419612"
            },
            {
                "entity": "K. N. Pradhan",
                "confidence_score": "-10.706172"
            }
        ]
    },
    {
        "mention": "parliament",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Parliament of India",
                "confidence_score": "8.404705"
            },
            {
                "entity": "Parliament",
                "confidence_score": "5.0738215"
            },
            {
                "entity": "Parliament of Pakistan",
                "confidence_score": "-6.196754"
            }
        ]
    },
    {
        "mention": "sansad",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sansad",
                "confidence_score": "1.8994746"
            },
            {
                "entity": "Sansad Adarsh Gram Yojana",
                "confidence_score": "1.0814773"
            },
            {
                "entity": "Scheduled Castes and Scheduled Tribes",
                "confidence_score": "-6.391504"
            }
        ]
    },
    {
        "mention": "gram yojana",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Sansad Adarsh Gram Yojana",
                "confidence_score": "3.4763048"
            },
            {
                "entity": "Pradhan Mantri Gramin Awaas Yojana",
                "confidence_score": "-2.3751888"
            },
            {
                "entity": "Gram panchayat",
                "confidence_score": "-8.762727"
            }
        ]
    },
    {
        "mention": "greenery",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Landscaping",
                "confidence_score": "-2.6385872"
            },
            {
                "entity": "Natural landscaping",
                "confidence_score": "-7.3776646"
            },
            {
                "entity": "Greenery",
                "confidence_score": "-11.598724"
            }
        ]
    },
    {
        "mention": "namami ganga yojana",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Pradhan Mantri Gramin Awaas Yojana",
                "confidence_score": "-11.141009"
            },
            {
                "entity": "Uma Oya Hydropower Complex",
                "confidence_score": "-11.94069"
            },
            {
                "entity": "Nambikkai",
                "confidence_score": "-12.112962"
            }
        ]
    },
    {
        "mention": "the integrated ganga conservation mission",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Integrated Conservation and Development Project",
                "confidence_score": "-11.5035515"
            },
            {
                "entity": "International Gorilla Conservation Programme",
                "confidence_score": "-12.169415"
            },
            {
                "entity": "Okapi Conservation Project",
                "confidence_score": "-12.482703"
            }
        ]
    },
    {
        "mention": "namami ganga yojana'",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jogi",
                "confidence_score": "-9.663323"
            },
            {
                "entity": "Naaya Aayam Multi-Disciplinary Institute",
                "confidence_score": "-12.015794"
            },
            {
                "entity": "Nambikkai",
                "confidence_score": "-12.097835"
            }
        ]
    },
    {
        "mention": "the modi government",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Narendra Modi",
                "confidence_score": "-5.288714"
            },
            {
                "entity": "First Modi ministry",
                "confidence_score": "-7.3127785"
            },
            {
                "entity": "Modi ministry",
                "confidence_score": "-12.524251"
            }
        ]
    },
    {
        "mention": "namami ganga'",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Namli",
                "confidence_score": "-10.992118"
            },
            {
                "entity": "Naaya Aayam Multi-Disciplinary Institute",
                "confidence_score": "-12.029527"
            },
            {
                "entity": "Nambikkai",
                "confidence_score": "-12.055055"
            }
        ]
    },
    {
        "mention": "pradhan mantri",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Pradhan Mantri",
                "confidence_score": "7.1750093"
            },
            {
                "entity": "Pradhanmantri",
                "confidence_score": "3.171644"
            },
            {
                "entity": "K. N. Pradhan",
                "confidence_score": "-11.190413"
            }
        ]
    },
    {
        "mention": "krishi sinchayee",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Krishnendu Adhikari",
                "confidence_score": "-11.106698"
            },
            {
                "entity": "Sukrit Wisetkaew",
                "confidence_score": "-11.947045"
            },
            {
                "entity": "Krish J. Sathaar",
                "confidence_score": "-12.04358"
            }
        ]
    },
    {
        "mention": "yojana",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Yojana",
                "confidence_score": "7.5459175"
            },
            {
                "entity": "Mahajanas",
                "confidence_score": "-9.598613"
            },
            {
                "entity": "Sajjana",
                "confidence_score": "-12.02983"
            }
        ]
    },
    {
        "mention": "pradhan mantri",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Pradhan Mantri",
                "confidence_score": "7.1750093"
            },
            {
                "entity": "Pradhanmantri",
                "confidence_score": "3.171644"
            },
            {
                "entity": "K. N. Pradhan",
                "confidence_score": "-11.190413"
            }
        ]
    },
    {
        "mention": "krishi sinchayee",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Krishnendu Adhikari",
                "confidence_score": "-11.106698"
            },
            {
                "entity": "Sukrit Wisetkaew",
                "confidence_score": "-11.947045"
            },
            {
                "entity": "Krish J. Sathaar",
                "confidence_score": "-12.04358"
            }
        ]
    },
    {
        "mention": "random inspection scheme",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Random assignment",
                "confidence_score": "-10.538868"
            },
            {
                "entity": "Points-based immigration system",
                "confidence_score": "-11.012437"
            },
            {
                "entity": "Random testing",
                "confidence_score": "-11.336345"
            }
        ]
    },
    {
        "mention": "provident fund",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Provident fund",
                "confidence_score": "9.231428"
            },
            {
                "entity": "Central Provident Fund",
                "confidence_score": "-2.3899734"
            },
            {
                "entity": "Public Provident Fund (India)",
                "confidence_score": "-4.583775"
            }
        ]
    },
    {
        "mention": "apprentice protsahan yojana:",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Swarna Jayanti Shahari Rozgar Yojana",
                "confidence_score": "-10.846993"
            },
            {
                "entity": "Sampoorna Grameen Rozgar Yojana",
                "confidence_score": "-11.422941"
            },
            {
                "entity": "Mukhya Mantri Yuva Swarozgar Yojana",
                "confidence_score": "-11.452571"
            }
        ]
    },
    {
        "mention": "rashtriya swasthya bima yojana",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Rashtriya Swasthya Bima Yojana",
                "confidence_score": "8.013553"
            },
            {
                "entity": "Pradhan Mantri Jeevan Jyoti Bima Yojana",
                "confidence_score": "-8.765374"
            },
            {
                "entity": "Rashtriya Madhyamik Shiksha Abhiyan",
                "confidence_score": "-10.671638"
            }
        ]
    },
    {
        "mention": "a smart card",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Smart card",
                "confidence_score": "5.4725575"
            },
            {
                "entity": "Digital card",
                "confidence_score": "-10.458941"
            },
            {
                "entity": "South African identity card",
                "confidence_score": "-11.108408"
            }
        ]
    },
    {
        "mention": "aparajita priyadarshini behera",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Aparna Dixit",
                "confidence_score": "-11.457622"
            },
            {
                "entity": "Aparajita Goppi",
                "confidence_score": "-11.538665"
            },
            {
                "entity": "Priyadarshini Raje Scindia",
                "confidence_score": "-11.655857"
            }
        ]
    },
    {
        "mention": "indian economic service",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Indian Economic Service",
                "confidence_score": "11.819936"
            },
            {
                "entity": "Indian Revenue Service",
                "confidence_score": "-8.392486"
            },
            {
                "entity": "Indian Statistical Service",
                "confidence_score": "-9.539835"
            }
        ]
    },
    {
        "mention": "aparajita priyadarshini behera",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Aparna Dixit",
                "confidence_score": "-11.457622"
            },
            {
                "entity": "Aparajita Goppi",
                "confidence_score": "-11.538665"
            },
            {
                "entity": "Priyadarshini Raje Scindia",
                "confidence_score": "-11.655857"
            }
        ]
    },
    {
        "mention": "indian economic service",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Indian Economic Service",
                "confidence_score": "11.819936"
            },
            {
                "entity": "Indian Revenue Service",
                "confidence_score": "-8.392486"
            },
            {
                "entity": "Indian Statistical Service",
                "confidence_score": "-9.539835"
            }
        ]
    },
    {
        "mention": "ies",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Indian Economic Service",
                "confidence_score": "11.033611"
            },
            {
                "entity": "Institute for Employment Studies",
                "confidence_score": "-3.7548587"
            },
            {
                "entity": "Indian Administrative Service",
                "confidence_score": "-6.881179"
            }
        ]
    },
    {
        "mention": "the indian economic service",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Indian Economic Service",
                "confidence_score": "9.033747"
            },
            {
                "entity": "Indian Revenue Service",
                "confidence_score": "-7.977242"
            },
            {
                "entity": "Indian Trade Service",
                "confidence_score": "-8.57201"
            }
        ]
    },
    {
        "mention": "upsc",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Union Public Service Commission",
                "confidence_score": "7.877064"
            },
            {
                "entity": "Uttar Pradesh Public Service Commission",
                "confidence_score": "-8.023665"
            },
            {
                "entity": "UP Stock Exchange",
                "confidence_score": "-11.585762"
            }
        ]
    },
    {
        "mention": "mahakalapada",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Mahakalapada",
                "confidence_score": "4.680041"
            },
            {
                "entity": "Mahamahopadhyaya",
                "confidence_score": "-9.830241"
            },
            {
                "entity": "Alapadamba",
                "confidence_score": "-12.248522"
            }
        ]
    },
    {
        "mention": "phd",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Doctor of Philosophy",
                "confidence_score": "10.021392"
            },
            {
                "entity": "Doctorate",
                "confidence_score": "-3.8302166"
            },
            {
                "entity": "Doctor of Divinity",
                "confidence_score": "-9.886741"
            }
        ]
    },
    {
        "mention": "jawaharlal nehru university",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Jawaharlal Nehru University",
                "confidence_score": "10.6941595"
            },
            {
                "entity": "Jawaharlal Nehru",
                "confidence_score": "-5.3543134"
            },
            {
                "entity": "Jawaharlal Nehru Technological University",
                "confidence_score": "-12.205911"
            }
        ]
    },
    {
        "mention": "aparajita",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Aparajita",
                "confidence_score": "0.86721766"
            },
            {
                "entity": "Aparajita Mohanty",
                "confidence_score": "-5.385344"
            },
            {
                "entity": "Aparajito",
                "confidence_score": "-8.45177"
            }
        ]
    },
    {
        "mention": "kapileswar high school",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Government High School, Karwar",
                "confidence_score": "-11.090369"
            },
            {
                "entity": "Nepali Kalyan High School",
                "confidence_score": "-12.145613"
            },
            {
                "entity": "Jharia Gujarati Hindi High School",
                "confidence_score": "-12.349246"
            }
        ]
    },
    {
        "mention": "post-graduation in economics",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Master of Economics",
                "confidence_score": "-9.636912"
            },
            {
                "entity": "Postgraduate education",
                "confidence_score": "-10.0965395"
            },
            {
                "entity": "Post Leaving Certificate",
                "confidence_score": "-11.229852"
            }
        ]
    },
    {
        "mention": "ira singhal",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ira Singhal",
                "confidence_score": "12.27744"
            },
            {
                "entity": "Singhal",
                "confidence_score": "-10.195976"
            },
            {
                "entity": "Ashok Singhal",
                "confidence_score": "-11.510916"
            }
        ]
    },
    {
        "mention": "indian administrative service",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Indian Administrative Service",
                "confidence_score": "11.102447"
            },
            {
                "entity": "All India Services",
                "confidence_score": "-0.5961854"
            },
            {
                "entity": "Indian Revenue Service",
                "confidence_score": "-5.109149"
            }
        ]
    },
    {
        "mention": "vandana rao",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Vandana Rao",
                "confidence_score": "5.897549"
            },
            {
                "entity": "C. Vidyasagar Rao",
                "confidence_score": "-11.620967"
            },
            {
                "entity": "N. Bhaskara Rao",
                "confidence_score": "-11.672618"
            }
        ]
    },
    {
        "mention": "renu raj",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Renu Kumari Singh",
                "confidence_score": "-10.885899"
            },
            {
                "entity": "Renu Kumari",
                "confidence_score": "-11.065326"
            },
            {
                "entity": "Renu Margrate",
                "confidence_score": "-12.126997"
            }
        ]
    },
    {
        "mention": "suharsha bhagat",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Sudarshan Bhagat",
                "confidence_score": "-10.554279"
            },
            {
                "entity": "Sukhdeo Bhagat",
                "confidence_score": "-10.98663"
            },
            {
                "entity": "Sujatha Singh",
                "confidence_score": "-11.503991"
            }
        ]
    },
    {
        "mention": "department of personnel and training",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Ministry of Personnel, Public Grievances and Pensions",
                "confidence_score": "-6.2770805"
            },
            {
                "entity": "Department of Appointment and Personnel",
                "confidence_score": "-8.276475"
            },
            {
                "entity": "RAF Personnel and Training Command",
                "confidence_score": "-10.844527"
            }
        ]
    },
    {
        "mention": "nidhi gupta",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Nidhi Singh",
                "confidence_score": "-11.313328"
            },
            {
                "entity": "Nidhi Sunil",
                "confidence_score": "-11.388279"
            },
            {
                "entity": "Nitish Jain",
                "confidence_score": "-11.685103"
            }
        ]
    },
    {
        "mention": "haryana",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Haryana",
                "confidence_score": "8.280659"
            },
            {
                "entity": "Government of Haryana",
                "confidence_score": "-6.5648227"
            },
            {
                "entity": "Haryana Police",
                "confidence_score": "-7.699554"
            }
        ]
    },
    {
        "mention": "discuss your answer",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Question",
                "confidence_score": "-7.3469243"
            },
            {
                "entity": "Discussion group",
                "confidence_score": "-8.35656"
            },
            {
                "entity": "Any Answers?",
                "confidence_score": "-8.775854"
            }
        ]
    },
    {
        "mention": "comment box",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Comment (computer programming)",
                "confidence_score": "-8.036764"
            },
            {
                "entity": "Comments section",
                "confidence_score": "-8.16006"
            },
            {
                "entity": "Text box",
                "confidence_score": "-10.449292"
            }
        ]
    },
    {
        "mention": "hamid ansari's",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mohammad Hamid Ansari",
                "confidence_score": "-4.345567"
            },
            {
                "entity": "Faridul Haq Ansari",
                "confidence_score": "-9.550763"
            },
            {
                "entity": "Salim Ansari",
                "confidence_score": "-10.480372"
            }
        ]
    },
    {
        "mention": "the special investigation team",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Inter-Services Intelligence",
                "confidence_score": "-8.863917"
            },
            {
                "entity": "Crime Investigation Department (Pakistan)",
                "confidence_score": "-11.241536"
            },
            {
                "entity": "Special Task Force",
                "confidence_score": "-11.313462"
            }
        ]
    },
    {
        "mention": "participatory notes",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Participatory note",
                "confidence_score": "7.417099"
            },
            {
                "entity": "Participatory media",
                "confidence_score": "-9.2998495"
            },
            {
                "entity": "Participatory budgeting",
                "confidence_score": "-10.855769"
            }
        ]
    },
    {
        "mention": "p-notes",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Participatory note",
                "confidence_score": "3.3666077"
            },
            {
                "entity": "Note-taking",
                "confidence_score": "-8.2424555"
            },
            {
                "entity": "Graduate",
                "confidence_score": "-9.901853"
            }
        ]
    },
    {
        "mention": "g subba rao",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "G. S. R. Subba Rao",
                "confidence_score": "-0.2877108"
            },
            {
                "entity": "I. V. Subba Rao",
                "confidence_score": "-9.859145"
            },
            {
                "entity": "B. A. Subba Rao",
                "confidence_score": "-10.090159"
            }
        ]
    },
    {
        "mention": "p n roy",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Samarendra Nath Roy",
                "confidence_score": "-7.9348793"
            },
            {
                "entity": "M. N. Roy",
                "confidence_score": "-8.135347"
            },
            {
                "entity": "Probir Roy",
                "confidence_score": "-10.056556"
            }
        ]
    },
    {
        "mention": "iii & iv",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Indian Institutes of Science Education and Research",
                "confidence_score": "-10.38115"
            },
            {
                "entity": "2012 PSA World Series Finals",
                "confidence_score": "-11.782668"
            },
            {
                "entity": "Paragliding at the 2018 Asian Games \u2013 Women's individual accuracy",
                "confidence_score": "-12.377837"
            }
        ]
    },
    {
        "mention": "daily questions challenge",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Helsinki Challenge",
                "confidence_score": "-10.7572975"
            },
            {
                "entity": "DES Challenges",
                "confidence_score": "-11.574565"
            },
            {
                "entity": "AI Challenge",
                "confidence_score": "-11.762808"
            }
        ]
    },
    {
        "mention": "discuss your answer",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Question",
                "confidence_score": "-7.3469243"
            },
            {
                "entity": "Discussion group",
                "confidence_score": "-8.35656"
            },
            {
                "entity": "Any Answers?",
                "confidence_score": "-8.775854"
            }
        ]
    },
    {
        "mention": "comment box",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Comment (computer programming)",
                "confidence_score": "-8.036764"
            },
            {
                "entity": "Comments section",
                "confidence_score": "-8.16006"
            },
            {
                "entity": "Text box",
                "confidence_score": "-10.449292"
            }
        ]
    },
    {
        "mention": "tamil nadu",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Tamil Nadu",
                "confidence_score": "6.1177793"
            },
            {
                "entity": "Chennai",
                "confidence_score": "-0.7596286"
            },
            {
                "entity": "Coastline of Tamil Nadu",
                "confidence_score": "-7.730553"
            }
        ]
    },
    {
        "mention": "jaduguda",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jadugora",
                "confidence_score": "3.0001957"
            },
            {
                "entity": "Jaduguda uranium mine",
                "confidence_score": "2.710908"
            },
            {
                "entity": "Nidugunda",
                "confidence_score": "-11.274503"
            }
        ]
    },
    {
        "mention": "pressurized water",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Pressurized water reactor",
                "confidence_score": "6.6550784"
            },
            {
                "entity": "Pressurized heavy-water reactor",
                "confidence_score": "-5.488706"
            },
            {
                "entity": "Pressurizer",
                "confidence_score": "-5.532977"
            }
        ]
    },
    {
        "mention": "pressurized heavy",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Pressurized heavy-water reactor",
                "confidence_score": "4.3804383"
            },
            {
                "entity": "Pressurized water reactor",
                "confidence_score": "-5.4452624"
            },
            {
                "entity": "Pressurizer",
                "confidence_score": "-6.7267804"
            }
        ]
    },
    {
        "mention": "fast breeder",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Fast Breeder Test Reactor",
                "confidence_score": "-4.5922112"
            },
            {
                "entity": "Breeder reactor",
                "confidence_score": "-6.711451"
            },
            {
                "entity": "Fast-neutron reactor",
                "confidence_score": "-8.9336815"
            }
        ]
    },
    {
        "mention": "advanced heavy",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Advanced heavy-water reactor",
                "confidence_score": "-7.387637"
            },
            {
                "entity": "AVIC Advanced Heavy Lifter",
                "confidence_score": "-8.708627"
            },
            {
                "entity": "All That Is Heavy",
                "confidence_score": "-11.912709"
            }
        ]
    },
    {
        "mention": "kamini",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "KAMINI",
                "confidence_score": "7.050175"
            },
            {
                "entity": "Kamini Roy",
                "confidence_score": "-7.6860356"
            },
            {
                "entity": "Kamini Jain",
                "confidence_score": "-11.620673"
            }
        ]
    },
    {
        "mention": "rohini click",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Rohini Kesavan Srihari",
                "confidence_score": "-10.762311"
            },
            {
                "entity": "Rohini Mariam Idicula",
                "confidence_score": "-11.21457"
            },
            {
                "entity": "Roshni Prakash",
                "confidence_score": "-11.797088"
            }
        ]
    },
    {
        "mention": "current affairs mcq",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Global Affairs Canada",
                "confidence_score": "-10.564147"
            },
            {
                "entity": "Coordinating Minister for National Security (Singapore)",
                "confidence_score": "-11.983417"
            },
            {
                "entity": "MCQ",
                "confidence_score": "-12.07381"
            }
        ]
    },
    {
        "mention": "upsc exams",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Union Public Service Commission",
                "confidence_score": "-3.0394003"
            },
            {
                "entity": "Uttar Pradesh Subordinate Services Selection Commission",
                "confidence_score": "-6.3360825"
            },
            {
                "entity": "Uttar Pradesh Public Service Commission",
                "confidence_score": "-6.5125074"
            }
        ]
    },
    {
        "mention": "upsc exams",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Union Public Service Commission",
                "confidence_score": "-3.0394003"
            },
            {
                "entity": "Uttar Pradesh Subordinate Services Selection Commission",
                "confidence_score": "-6.3360825"
            },
            {
                "entity": "Uttar Pradesh Public Service Commission",
                "confidence_score": "-6.5125074"
            }
        ]
    },
    {
        "mention": "diclofenac",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Diclofenac",
                "confidence_score": "12.24802"
            },
            {
                "entity": "Diclofenamide",
                "confidence_score": "-9.881673"
            },
            {
                "entity": "Alclofenac",
                "confidence_score": "-11.090948"
            }
        ]
    },
    {
        "mention": "asia",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Asia",
                "confidence_score": "3.6320422"
            },
            {
                "entity": "Kabaddi at the 2018 Asian Games \u2013 Men",
                "confidence_score": "-12.179671"
            },
            {
                "entity": "Kabaddi at the 2018 Asian Games \u2013 Women",
                "confidence_score": "-12.255481"
            }
        ]
    },
    {
        "mention": "current affairs for",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Current affairs (news format)",
                "confidence_score": "-4.1939354"
            },
            {
                "entity": "Current Affairs (magazine)",
                "confidence_score": "-6.503863"
            },
            {
                "entity": "RT\u00c9 News and Current Affairs",
                "confidence_score": "-9.099768"
            }
        ]
    },
    {
        "mention": "greenpeace india stating",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Greenpeace India",
                "confidence_score": "2.6043289"
            },
            {
                "entity": "Greenpeace",
                "confidence_score": "-3.5989876"
            },
            {
                "entity": "Greenpeace (disambiguation)",
                "confidence_score": "-12.285451"
            }
        ]
    },
    {
        "mention": "fcra",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Fair Credit Reporting Act",
                "confidence_score": "3.878489"
            },
            {
                "entity": "Chinese Association for Relief and Ensuing Services",
                "confidence_score": "-8.0355425"
            },
            {
                "entity": "Forum for Fact-finding Documentation and Advocacy",
                "confidence_score": "-10.658574"
            }
        ]
    },
    {
        "mention": "the ministry of home affairs",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Ministry of Home Affairs (India)",
                "confidence_score": "0.2794178"
            },
            {
                "entity": "Minister of Home Affairs (India)",
                "confidence_score": "-5.373281"
            },
            {
                "entity": "Ministry of Home Affairs (Sri Lanka)",
                "confidence_score": "-7.9640565"
            }
        ]
    },
    {
        "mention": "ngo",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Non-governmental organization",
                "confidence_score": "7.5985384"
            },
            {
                "entity": "International non-governmental organization",
                "confidence_score": "-2.4419703"
            },
            {
                "entity": "Nonprofit organization",
                "confidence_score": "-5.26467"
            }
        ]
    },
    {
        "mention": "the foreign contribution regulation act",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Foreign Contribution (Regulation) Act, 2010",
                "confidence_score": "5.941937"
            },
            {
                "entity": "Foreign Exchange Management Act",
                "confidence_score": "-9.323585"
            },
            {
                "entity": "Foreign Corrupt Practices Act",
                "confidence_score": "-10.181523"
            }
        ]
    },
    {
        "mention": "the delhi high court",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Delhi High Court",
                "confidence_score": "4.571439"
            },
            {
                "entity": "Punjab and Haryana High Court",
                "confidence_score": "-10.549688"
            },
            {
                "entity": "Uttarakhand High Court",
                "confidence_score": "-11.31017"
            }
        ]
    },
    {
        "mention": "mha",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Master of Health Administration",
                "confidence_score": "-2.552178"
            },
            {
                "entity": "Rajya Sabha",
                "confidence_score": "-4.2541146"
            },
            {
                "entity": "MHA (Methodist Homes for the Aged)",
                "confidence_score": "-8.809649"
            }
        ]
    },
    {
        "mention": "fcra",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Fair Credit Reporting Act",
                "confidence_score": "3.878489"
            },
            {
                "entity": "Chinese Association for Relief and Ensuing Services",
                "confidence_score": "-8.0355425"
            },
            {
                "entity": "Forum for Fact-finding Documentation and Advocacy",
                "confidence_score": "-10.658574"
            }
        ]
    },
    {
        "mention": "mha",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Master of Health Administration",
                "confidence_score": "-2.552178"
            },
            {
                "entity": "Rajya Sabha",
                "confidence_score": "-4.2541146"
            },
            {
                "entity": "MHA (Methodist Homes for the Aged)",
                "confidence_score": "-8.809649"
            }
        ]
    },
    {
        "mention": "greenpeace india",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Greenpeace India",
                "confidence_score": "10.40537"
            },
            {
                "entity": "Greenpeace",
                "confidence_score": "-3.3043106"
            },
            {
                "entity": "Greenpeace USA",
                "confidence_score": "-8.513196"
            }
        ]
    },
    {
        "mention": "mha",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Master of Health Administration",
                "confidence_score": "-2.552178"
            },
            {
                "entity": "Rajya Sabha",
                "confidence_score": "-4.2541146"
            },
            {
                "entity": "MHA (Methodist Homes for the Aged)",
                "confidence_score": "-8.809649"
            }
        ]
    },
    {
        "mention": "greenpeace india",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Greenpeace India",
                "confidence_score": "11.641913"
            },
            {
                "entity": "Greenpeace",
                "confidence_score": "-1.5232947"
            },
            {
                "entity": "Greenpeace USA",
                "confidence_score": "-7.1717873"
            }
        ]
    },
    {
        "mention": "mha",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Master of Health Administration",
                "confidence_score": "-2.552178"
            },
            {
                "entity": "Rajya Sabha",
                "confidence_score": "-4.2541146"
            },
            {
                "entity": "MHA (Methodist Homes for the Aged)",
                "confidence_score": "-8.809649"
            }
        ]
    },
    {
        "mention": "ngo",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Non-governmental organization",
                "confidence_score": "7.5985384"
            },
            {
                "entity": "International non-governmental organization",
                "confidence_score": "-2.4419703"
            },
            {
                "entity": "Nonprofit organization",
                "confidence_score": "-5.26467"
            }
        ]
    },
    {
        "mention": "idbi",
        "type": "ORG",
        "predictions": [
            {
                "entity": "IDBI Bank",
                "confidence_score": "1.6861428"
            },
            {
                "entity": "Union Bank of India",
                "confidence_score": "-9.187858"
            },
            {
                "entity": "Insolvency and Bankruptcy Board of India",
                "confidence_score": "-10.422206"
            }
        ]
    },
    {
        "mention": "icici",
        "type": "ORG",
        "predictions": [
            {
                "entity": "ICICI Bank",
                "confidence_score": "5.071098"
            },
            {
                "entity": "International Chamber of Commerce",
                "confidence_score": "-7.6122136"
            },
            {
                "entity": "International Clearing Union",
                "confidence_score": "-9.183238"
            }
        ]
    },
    {
        "mention": "bank",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Bank",
                "confidence_score": "4.9896016"
            },
            {
                "entity": "Reserve Bank of India",
                "confidence_score": "-3.9144347"
            },
            {
                "entity": "Bank of India",
                "confidence_score": "-4.29546"
            }
        ]
    },
    {
        "mention": "ngo",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Non-governmental organization",
                "confidence_score": "7.5985384"
            },
            {
                "entity": "International non-governmental organization",
                "confidence_score": "-2.4419703"
            },
            {
                "entity": "Nonprofit organization",
                "confidence_score": "-5.26467"
            }
        ]
    },
    {
        "mention": "mha",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Master of Health Administration",
                "confidence_score": "-2.552178"
            },
            {
                "entity": "Rajya Sabha",
                "confidence_score": "-4.2541146"
            },
            {
                "entity": "MHA (Methodist Homes for the Aged)",
                "confidence_score": "-8.809649"
            }
        ]
    },
    {
        "mention": "ngo",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Non-governmental organization",
                "confidence_score": "7.5985384"
            },
            {
                "entity": "International non-governmental organization",
                "confidence_score": "-2.4419703"
            },
            {
                "entity": "Nonprofit organization",
                "confidence_score": "-5.26467"
            }
        ]
    },
    {
        "mention": "ngo",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Non-governmental organization",
                "confidence_score": "7.5985384"
            },
            {
                "entity": "International non-governmental organization",
                "confidence_score": "-2.4419703"
            },
            {
                "entity": "Nonprofit organization",
                "confidence_score": "-5.26467"
            }
        ]
    },
    {
        "mention": "euros",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Euro",
                "confidence_score": "9.414692"
            },
            {
                "entity": "Eurozone",
                "confidence_score": "-0.64442474"
            },
            {
                "entity": "European Union",
                "confidence_score": "-1.7834802"
            }
        ]
    },
    {
        "mention": "stichting greenpeace council",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Greenpeace",
                "confidence_score": "-3.4288359"
            },
            {
                "entity": "ICLEI",
                "confidence_score": "-10.60101"
            },
            {
                "entity": "Water Supply and Sanitation Collaborative Council",
                "confidence_score": "-11.518219"
            }
        ]
    },
    {
        "mention": "greenpeace india",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Greenpeace India",
                "confidence_score": "11.641913"
            },
            {
                "entity": "Greenpeace",
                "confidence_score": "-1.5232947"
            },
            {
                "entity": "Greenpeace USA",
                "confidence_score": "-7.1717873"
            }
        ]
    },
    {
        "mention": "david cameron",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "David Cameron",
                "confidence_score": "11.115639"
            },
            {
                "entity": "David Cameron (rower)",
                "confidence_score": "-10.001212"
            },
            {
                "entity": "Christian Cameron",
                "confidence_score": "-11.292661"
            }
        ]
    },
    {
        "mention": "cameron",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "David Cameron",
                "confidence_score": "0.045297895"
            },
            {
                "entity": "James Cameron",
                "confidence_score": "-4.4155126"
            },
            {
                "entity": "Cameron, Fife",
                "confidence_score": "-10.010605"
            }
        ]
    },
    {
        "mention": "un",
        "type": "ORG",
        "predictions": [
            {
                "entity": "United Nations",
                "confidence_score": "7.8057218"
            },
            {
                "entity": "United Nations General Assembly",
                "confidence_score": "-3.8032172"
            },
            {
                "entity": "Unrepresented Nations and Peoples Organization",
                "confidence_score": "-7.051578"
            }
        ]
    },
    {
        "mention": "europe",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Europe",
                "confidence_score": "3.7219584"
            },
            {
                "entity": "European Union",
                "confidence_score": "-1.9110476"
            },
            {
                "entity": "Council of Europe",
                "confidence_score": "-3.662894"
            }
        ]
    },
    {
        "mention": "eu",
        "type": "ORG",
        "predictions": [
            {
                "entity": "European Union",
                "confidence_score": "8.22565"
            },
            {
                "entity": "Euro",
                "confidence_score": "-1.3719231"
            },
            {
                "entity": "Council of Europe",
                "confidence_score": "-3.778574"
            }
        ]
    },
    {
        "mention": "europe",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Europe",
                "confidence_score": "3.7219584"
            },
            {
                "entity": "European Union",
                "confidence_score": "-1.9110476"
            },
            {
                "entity": "Council of Europe",
                "confidence_score": "-3.662894"
            }
        ]
    },
    {
        "mention": "mahesh sharma",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mahesh Sharma",
                "confidence_score": "10.007316"
            },
            {
                "entity": "Mahesh Joshi",
                "confidence_score": "-11.039451"
            },
            {
                "entity": "Mahesh Joshi (Madhya Pradesh politician)",
                "confidence_score": "-12.039932"
            }
        ]
    },
    {
        "mention": "pdt current affairs daily voice",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Daily Voice (South African newspaper)",
                "confidence_score": "-10.9162855"
            },
            {
                "entity": "Liberal Democrat Voice",
                "confidence_score": "-11.389053"
            },
            {
                "entity": "Dalit Voice",
                "confidence_score": "-11.728372"
            }
        ]
    },
    {
        "mention": "mahesh sharma",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mahesh Sharma",
                "confidence_score": "10.007316"
            },
            {
                "entity": "Mahesh Joshi",
                "confidence_score": "-11.039451"
            },
            {
                "entity": "Mahesh Joshi (Madhya Pradesh politician)",
                "confidence_score": "-12.039932"
            }
        ]
    },
    {
        "mention": "afs interviewer",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Inter Press Service",
                "confidence_score": "-11.325374"
            },
            {
                "entity": "Associated Press Service",
                "confidence_score": "-11.666579"
            },
            {
                "entity": "All India Radio Monitoring Service",
                "confidence_score": "-11.790071"
            }
        ]
    },
    {
        "mention": "sindhu jain bhattacharya",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Basu Bhattacharya",
                "confidence_score": "-11.010752"
            },
            {
                "entity": "Abinash Chandra Bhattacharya",
                "confidence_score": "-11.403957"
            },
            {
                "entity": "Sanjay Bhattacharya",
                "confidence_score": "-11.781011"
            }
        ]
    },
    {
        "mention": "\u0930\u093e\u0937\u094d\u091f\u094d\u0930\u092e\u0902\u0921\u0932 \u0916\u0947\u0932",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Jhala Nath Khanal",
                "confidence_score": "-9.743552"
            },
            {
                "entity": "Khanal cabinet",
                "confidence_score": "-11.335571"
            },
            {
                "entity": "Khandkhel",
                "confidence_score": "-11.445943"
            }
        ]
    },
    {
        "mention": "\u0930\u093e\u0937\u094d\u091f\u094d\u0930\u092e\u0902\u0921\u0932 \u0916\u0947\u0932",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Jhala Nath Khanal",
                "confidence_score": "-9.743552"
            },
            {
                "entity": "Khanal cabinet",
                "confidence_score": "-11.335571"
            },
            {
                "entity": "Khandkhel",
                "confidence_score": "-11.445943"
            }
        ]
    },
    {
        "mention": "\u0928\u093f\u0930\u094d\u0923\u092f \u092a\u0930",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Pandit Narendra Sharma",
                "confidence_score": "-11.24919"
            },
            {
                "entity": "Niranjan Iyengar",
                "confidence_score": "-11.865438"
            },
            {
                "entity": "National Institute of Rehabilitation Training and Research",
                "confidence_score": "-11.91757"
            }
        ]
    },
    {
        "mention": "hdp",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Peoples' Democratic Party (Turkey)",
                "confidence_score": "4.72046"
            },
            {
                "entity": "Democratic Regions Party",
                "confidence_score": "-6.1960216"
            },
            {
                "entity": "Free Democratic Party (Northern Cyprus)",
                "confidence_score": "-10.064738"
            }
        ]
    },
    {
        "mention": "raymond mould",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Raymond Mould",
                "confidence_score": "12.062735"
            },
            {
                "entity": "Philip Mould",
                "confidence_score": "-9.620217"
            },
            {
                "entity": "Raymond Floyd",
                "confidence_score": "-11.772363"
            }
        ]
    },
    {
        "mention": "the grand national",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Grand National",
                "confidence_score": "-1.3877977"
            },
            {
                "entity": "Grand National (greyhounds)",
                "confidence_score": "-5.1724935"
            },
            {
                "entity": "Irish Grand National",
                "confidence_score": "-10.400422"
            }
        ]
    },
    {
        "mention": "champion chase",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Champion Hurdle",
                "confidence_score": "-4.991761"
            },
            {
                "entity": "Queen Mother Champion Chase",
                "confidence_score": "-5.113881"
            },
            {
                "entity": "JNwine.com Champion Chase",
                "confidence_score": "-6.9157424"
            }
        ]
    },
    {
        "mention": "george vi chase",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "King George VI Chase",
                "confidence_score": "-1.6328759"
            },
            {
                "entity": "Cheltenham Chase",
                "confidence_score": "-9.101781"
            },
            {
                "entity": "Ascot Chase",
                "confidence_score": "-9.964832"
            }
        ]
    },
    {
        "mention": "mould",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Philip Mould",
                "confidence_score": "-2.245812"
            },
            {
                "entity": "Bob Mould",
                "confidence_score": "-7.934108"
            },
            {
                "entity": "Jon Mould",
                "confidence_score": "-8.926639"
            }
        ]
    },
    {
        "mention": "sprinter sacre",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sprinter Sacre",
                "confidence_score": "3.6181815"
            },
            {
                "entity": "French Renewal",
                "confidence_score": "-10.837056"
            },
            {
                "entity": "Fr\u00e9d\u00e9ric Belaubre",
                "confidence_score": "-11.516375"
            }
        ]
    },
    {
        "mention": "charter party",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Charter Party",
                "confidence_score": "5.193508"
            },
            {
                "entity": "Charter Party (horse)",
                "confidence_score": "-3.5406468"
            },
            {
                "entity": "Charter 88",
                "confidence_score": "-8.877359"
            }
        ]
    },
    {
        "mention": "barton bank",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Barton Bank",
                "confidence_score": "1.6099434"
            },
            {
                "entity": "Doe Bank",
                "confidence_score": "-11.610986"
            },
            {
                "entity": "Dean Bank",
                "confidence_score": "-12.1787"
            }
        ]
    },
    {
        "mention": "pillar properties",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Capita Property and Infrastructure",
                "confidence_score": "-10.482668"
            },
            {
                "entity": "Landmark (company)",
                "confidence_score": "-11.176497"
            },
            {
                "entity": "Element 14 (company)",
                "confidence_score": "-12.045583"
            }
        ]
    },
    {
        "mention": "mould",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Philip Mould",
                "confidence_score": "-2.245812"
            },
            {
                "entity": "Bob Mould",
                "confidence_score": "-7.934108"
            },
            {
                "entity": "Jon Mould",
                "confidence_score": "-8.926639"
            }
        ]
    },
    {
        "mention": "guiting power",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Guiting Power",
                "confidence_score": "8.496997"
            },
            {
                "entity": "Upcycling",
                "confidence_score": "-11.02222"
            },
            {
                "entity": "Aquamarine Power",
                "confidence_score": "-11.658291"
            }
        ]
    },
    {
        "mention": "gloucestershire",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Gloucestershire",
                "confidence_score": "6.0391774"
            },
            {
                "entity": "University of Gloucestershire",
                "confidence_score": "-5.831635"
            },
            {
                "entity": "South Gloucestershire",
                "confidence_score": "-8.389405"
            }
        ]
    },
    {
        "mention": "david nicholson",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "David Nicholson (journalist)",
                "confidence_score": "1.4030061"
            },
            {
                "entity": "David Nicholson (British politician)",
                "confidence_score": "0.764122"
            },
            {
                "entity": "David Nicholson (horse racing)",
                "confidence_score": "-7.605444"
            }
        ]
    },
    {
        "mention": "charter party",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Charter Party",
                "confidence_score": "5.193508"
            },
            {
                "entity": "Charter Party (horse)",
                "confidence_score": "-3.5406468"
            },
            {
                "entity": "Charter 88",
                "confidence_score": "-8.877359"
            }
        ]
    },
    {
        "mention": "colin",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Colin Phipps",
                "confidence_score": "-7.427876"
            },
            {
                "entity": "Colin Matthews (businessman)",
                "confidence_score": "-7.8862023"
            },
            {
                "entity": "Colin Smith (British rower)",
                "confidence_score": "-10.450628"
            }
        ]
    },
    {
        "mention": "clare smith",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Clare Calbraith",
                "confidence_score": "-11.647167"
            },
            {
                "entity": "Clare Greenwood",
                "confidence_score": "-11.7904215"
            },
            {
                "entity": "Clare Smyth",
                "confidence_score": "-12.002308"
            }
        ]
    },
    {
        "mention": "barton bank",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Barton Bank",
                "confidence_score": "1.6099434"
            },
            {
                "entity": "Doe Bank",
                "confidence_score": "-11.610986"
            },
            {
                "entity": "Dean Bank",
                "confidence_score": "-12.1787"
            }
        ]
    },
    {
        "mention": "george vi chase",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "King George VI Chase",
                "confidence_score": "-1.6328759"
            },
            {
                "entity": "Cheltenham Chase",
                "confidence_score": "-9.101781"
            },
            {
                "entity": "Ascot Chase",
                "confidence_score": "-9.964832"
            }
        ]
    },
    {
        "mention": "nigel twiston-davies",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Nigel Twiston-Davies",
                "confidence_score": "11.401046"
            },
            {
                "entity": "Sam Twiston-Davies",
                "confidence_score": "-3.430216"
            },
            {
                "entity": "Twist Magic",
                "confidence_score": "-10.475499"
            }
        ]
    },
    {
        "mention": "mould",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Philip Mould",
                "confidence_score": "-2.245812"
            },
            {
                "entity": "Bob Mould",
                "confidence_score": "-7.934108"
            },
            {
                "entity": "Jon Mould",
                "confidence_score": "-8.926639"
            }
        ]
    },
    {
        "mention": "twiston-davies",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Nigel Twiston-Davies",
                "confidence_score": "1.4381024"
            },
            {
                "entity": "Twiston",
                "confidence_score": "-8.886714"
            },
            {
                "entity": "Twist Magic",
                "confidence_score": "-11.145813"
            }
        ]
    },
    {
        "mention": "guiting power",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Guiting Power",
                "confidence_score": "8.496997"
            },
            {
                "entity": "Upcycling",
                "confidence_score": "-11.02222"
            },
            {
                "entity": "Aquamarine Power",
                "confidence_score": "-11.658291"
            }
        ]
    },
    {
        "mention": "twiston-davies",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Nigel Twiston-Davies",
                "confidence_score": "1.4381024"
            },
            {
                "entity": "Twiston",
                "confidence_score": "-8.886714"
            },
            {
                "entity": "Twist Magic",
                "confidence_score": "-11.145813"
            }
        ]
    },
    {
        "mention": "raymond",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Raymond",
                "confidence_score": "-5.1941285"
            },
            {
                "entity": "Christian Raymond",
                "confidence_score": "-10.151926"
            },
            {
                "entity": "Raymond Martin (cyclist)",
                "confidence_score": "-11.60287"
            }
        ]
    },
    {
        "mention": "tipping tim",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Tip Tipping",
                "confidence_score": "-9.992383"
            },
            {
                "entity": "Tipit",
                "confidence_score": "-10.863245"
            },
            {
                "entity": "Tipping Point (game show)",
                "confidence_score": "-11.157781"
            }
        ]
    },
    {
        "mention": "sprinter sacre",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Sprinter Sacre",
                "confidence_score": "8.483996"
            },
            {
                "entity": "Serpentine Running Club",
                "confidence_score": "-10.663048"
            },
            {
                "entity": "Cartier Champion Sprinter",
                "confidence_score": "-10.790881"
            }
        ]
    },
    {
        "mention": "caroline",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Caroline Pierce",
                "confidence_score": "-11.365865"
            },
            {
                "entity": "Caroline Pratt",
                "confidence_score": "-11.393584"
            },
            {
                "entity": "Caroline Buchanan",
                "confidence_score": "-11.701561"
            }
        ]
    },
    {
        "mention": "the queen mother champion chase",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Queen Mother Champion Chase",
                "confidence_score": "9.97849"
            },
            {
                "entity": "2015 Queen Mother Champion Chase",
                "confidence_score": "-7.3754067"
            },
            {
                "entity": "Kingmaker Novices' Chase",
                "confidence_score": "-11.336513"
            }
        ]
    },
    {
        "mention": "the timeform organisation",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Timeform",
                "confidence_score": "-1.4261535"
            },
            {
                "entity": "Cycling Time Trials",
                "confidence_score": "-11.565618"
            },
            {
                "entity": "The Track & Field Organisation",
                "confidence_score": "-11.824578"
            }
        ]
    },
    {
        "mention": "national hunt",
        "type": "ORG",
        "predictions": [
            {
                "entity": "National Hunt racing",
                "confidence_score": "6.0748844"
            },
            {
                "entity": "National Trust for Places of Historic Interest or Natural Beauty",
                "confidence_score": "-8.410076"
            },
            {
                "entity": "The Royal Caledonian Hunt",
                "confidence_score": "-10.40365"
            }
        ]
    },
    {
        "mention": "flyingbolt",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Flyingbolt",
                "confidence_score": "9.404755"
            },
            {
                "entity": "Flyingbolt Novice Chase",
                "confidence_score": "-2.9916608"
            },
            {
                "entity": "Flying Eleven",
                "confidence_score": "-12.062841"
            }
        ]
    },
    {
        "mention": "mn",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Minnesota",
                "confidence_score": "0.008507854"
            },
            {
                "entity": "Minneapolis",
                "confidence_score": "-3.562429"
            },
            {
                "entity": "MN",
                "confidence_score": "-11.657489"
            }
        ]
    },
    {
        "mention": "the goodman group",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Goodman Group",
                "confidence_score": "-7.1211195"
            },
            {
                "entity": "Goodman Global",
                "confidence_score": "-8.808243"
            },
            {
                "entity": "Goodman Gallery",
                "confidence_score": "-12.027571"
            }
        ]
    },
    {
        "mention": "jennifer ballantine",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jennifer Crystal Foley",
                "confidence_score": "-11.160103"
            },
            {
                "entity": "Patrick J. Ballantine",
                "confidence_score": "-11.926982"
            },
            {
                "entity": "Sara Ballantyne",
                "confidence_score": "-12.352129"
            }
        ]
    },
    {
        "mention": "able palms home health of minneapolis",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Jonathan, Minnesota",
                "confidence_score": "-10.24767"
            },
            {
                "entity": "List of people from Minneapolis",
                "confidence_score": "-10.871371"
            },
            {
                "entity": "Edina Realty",
                "confidence_score": "-11.611315"
            }
        ]
    },
    {
        "mention": "ballantine",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Ballantine, Montana",
                "confidence_score": "-9.58007"
            },
            {
                "entity": "Patrick J. Ballantine",
                "confidence_score": "-10.125121"
            },
            {
                "entity": "Ballentine, South Carolina",
                "confidence_score": "-11.96888"
            }
        ]
    },
    {
        "mention": "craig edinger",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Evan Edinger",
                "confidence_score": "-11.7964945"
            },
            {
                "entity": "Craig Miller (CEO)",
                "confidence_score": "-12.272685"
            },
            {
                "entity": "Jeff Edler",
                "confidence_score": "-12.428829"
            }
        ]
    },
    {
        "mention": "senior living & health care",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Elderly care",
                "confidence_score": "-8.198955"
            },
            {
                "entity": "Health and Social Care",
                "confidence_score": "-9.334951"
            },
            {
                "entity": "Health and Retirement Study",
                "confidence_score": "-10.780785"
            }
        ]
    },
    {
        "mention": "the goodman group",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Goodman Group",
                "confidence_score": "-7.1211195"
            },
            {
                "entity": "Goodman Global",
                "confidence_score": "-8.808243"
            },
            {
                "entity": "Goodman Gallery",
                "confidence_score": "-12.027571"
            }
        ]
    },
    {
        "mention": "jennifer",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jennifer Aniston",
                "confidence_score": "-9.237152"
            },
            {
                "entity": "Jennifer Lavoie",
                "confidence_score": "-10.980504"
            },
            {
                "entity": "Jennifer Jako",
                "confidence_score": "-11.82011"
            }
        ]
    },
    {
        "mention": "edinger",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Edward F. Edinger",
                "confidence_score": "-3.9674938"
            },
            {
                "entity": "Evan Edinger",
                "confidence_score": "-7.721009"
            },
            {
                "entity": "Edinger",
                "confidence_score": "-8.142102"
            }
        ]
    },
    {
        "mention": "ballantine",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Ballantine, Montana",
                "confidence_score": "-9.58007"
            },
            {
                "entity": "Patrick J. Ballantine",
                "confidence_score": "-10.125121"
            },
            {
                "entity": "Ballentine, South Carolina",
                "confidence_score": "-11.96888"
            }
        ]
    },
    {
        "mention": "recover health",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Recovery International",
                "confidence_score": "-8.6431875"
            },
            {
                "entity": "Steward Health Care System",
                "confidence_score": "-10.768306"
            },
            {
                "entity": "Renown Health",
                "confidence_score": "-11.781255"
            }
        ]
    },
    {
        "mention": "medicare",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Medicare (United States)",
                "confidence_score": "8.6975155"
            },
            {
                "entity": "Medicare (Canada)",
                "confidence_score": "-5.5937285"
            },
            {
                "entity": "Medicare",
                "confidence_score": "-6.107386"
            }
        ]
    },
    {
        "mention": "ballantine",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Ballantine, Montana",
                "confidence_score": "-9.58007"
            },
            {
                "entity": "Patrick J. Ballantine",
                "confidence_score": "-10.125121"
            },
            {
                "entity": "Ballentine, South Carolina",
                "confidence_score": "-11.96888"
            }
        ]
    },
    {
        "mention": "metropolitan state university",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Metropolitan State University",
                "confidence_score": "11.63488"
            },
            {
                "entity": "Minnesota State University, Mankato",
                "confidence_score": "-8.540732"
            },
            {
                "entity": "Southwest Minnesota State University",
                "confidence_score": "-10.424705"
            }
        ]
    },
    {
        "mention": "twin cities",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Twin cities",
                "confidence_score": "5.7361817"
            },
            {
                "entity": "Minneapolis\u2013Saint Paul",
                "confidence_score": "4.6527057"
            },
            {
                "entity": "Minneapolis",
                "confidence_score": "-1.4091647"
            }
        ]
    },
    {
        "mention": "community health accreditation partner",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Community Health Accreditation Program",
                "confidence_score": "-1.7553334"
            },
            {
                "entity": "Accreditation Commission for Health Care",
                "confidence_score": "-9.197819"
            },
            {
                "entity": "Community rating",
                "confidence_score": "-10.103427"
            }
        ]
    },
    {
        "mention": "able palms home & health care services",
        "type": "ORG",
        "predictions": [
            {
                "entity": "PSA Healthcare",
                "confidence_score": "-10.8723345"
            },
            {
                "entity": "HealthEast Care System",
                "confidence_score": "-11.137187"
            },
            {
                "entity": "Manatee Palms Youth Services",
                "confidence_score": "-12.147605"
            }
        ]
    },
    {
        "mention": "the goodman group",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Goodman Group",
                "confidence_score": "-7.1211195"
            },
            {
                "entity": "Goodman Global",
                "confidence_score": "-8.808243"
            },
            {
                "entity": "Goodman Gallery",
                "confidence_score": "-12.027571"
            }
        ]
    },
    {
        "mention": "the goodman group",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Goodman Group",
                "confidence_score": "-7.1211195"
            },
            {
                "entity": "Goodman Global",
                "confidence_score": "-8.808243"
            },
            {
                "entity": "Goodman Gallery",
                "confidence_score": "-12.027571"
            }
        ]
    },
    {
        "mention": "arch coal",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Arch Coal",
                "confidence_score": "9.177982"
            },
            {
                "entity": "Patriot Coal",
                "confidence_score": "-9.803976"
            },
            {
                "entity": "Enoco Coal Mine",
                "confidence_score": "-12.007482"
            }
        ]
    },
    {
        "mention": "aci",
        "type": "ORG",
        "predictions": [
            {
                "entity": "American Competitiveness Initiative",
                "confidence_score": "-0.014987714"
            },
            {
                "entity": "Airports Council International",
                "confidence_score": "-2.6019385"
            },
            {
                "entity": "American Competitiveness Institute",
                "confidence_score": "-3.7877538"
            }
        ]
    },
    {
        "mention": "aci",
        "type": "ORG",
        "predictions": [
            {
                "entity": "American Competitiveness Initiative",
                "confidence_score": "-0.014987714"
            },
            {
                "entity": "Airports Council International",
                "confidence_score": "-2.6019385"
            },
            {
                "entity": "American Competitiveness Institute",
                "confidence_score": "-3.7877538"
            }
        ]
    },
    {
        "mention": "free",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Free software",
                "confidence_score": "-3.2891412"
            },
            {
                "entity": "Liberty",
                "confidence_score": "-5.296305"
            },
            {
                "entity": "Freetown",
                "confidence_score": "-5.9360566"
            }
        ]
    },
    {
        "mention": "gazette",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Gazette",
                "confidence_score": "11.215565"
            },
            {
                "entity": "The Gazette of India",
                "confidence_score": "-0.9696276"
            },
            {
                "entity": "Government gazette",
                "confidence_score": "-2.4597008"
            }
        ]
    },
    {
        "mention": "carmen de miguel",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Carmen Ejogo",
                "confidence_score": "-11.072501"
            },
            {
                "entity": "Carmen Solomons",
                "confidence_score": "-11.959321"
            },
            {
                "entity": "Carmen Pereira",
                "confidence_score": "-12.306169"
            }
        ]
    },
    {
        "mention": "the university of alabama at birmingham",
        "type": "ORG",
        "predictions": [
            {
                "entity": "University of Alabama at Birmingham",
                "confidence_score": "4.5339465"
            },
            {
                "entity": "University of Alabama",
                "confidence_score": "-6.5175643"
            },
            {
                "entity": "University of Birmingham",
                "confidence_score": "-6.876284"
            }
        ]
    },
    {
        "mention": "de miguel",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Miguel de Miguel",
                "confidence_score": "-6.6036286"
            },
            {
                "entity": "Diego de Miguel",
                "confidence_score": "-11.013737"
            },
            {
                "entity": "De Bono",
                "confidence_score": "-11.076583"
            }
        ]
    },
    {
        "mention": "de miguel",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Miguel de Miguel",
                "confidence_score": "-6.6036286"
            },
            {
                "entity": "Diego de Miguel",
                "confidence_score": "-11.013737"
            },
            {
                "entity": "De Bono",
                "confidence_score": "-11.076583"
            }
        ]
    },
    {
        "mention": "bmi",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Body mass index",
                "confidence_score": "3.268693"
            },
            {
                "entity": "Birmingham and Midland Institute",
                "confidence_score": "-6.33865"
            },
            {
                "entity": "Bank Melli Iran",
                "confidence_score": "-7.8381524"
            }
        ]
    },
    {
        "mention": "de miguel",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Miguel de Miguel",
                "confidence_score": "-6.6036286"
            },
            {
                "entity": "Diego de Miguel",
                "confidence_score": "-11.013737"
            },
            {
                "entity": "De Bono",
                "confidence_score": "-11.076583"
            }
        ]
    },
    {
        "mention": "de miguel c",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Miguel de Miguel",
                "confidence_score": "-7.783598"
            },
            {
                "entity": "Sony Xperia C",
                "confidence_score": "-10.96512"
            },
            {
                "entity": "Nagai Go no Horror Gekijo: Mannequin",
                "confidence_score": "-11.409646"
            }
        ]
    },
    {
        "mention": "xu x",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Xu Yuhua",
                "confidence_score": "-11.412566"
            },
            {
                "entity": "Xu Chen",
                "confidence_score": "-11.678302"
            },
            {
                "entity": "Xuso Jones",
                "confidence_score": "-11.761205"
            }
        ]
    },
    {
        "mention": "pan y",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Pan Entertainment",
                "confidence_score": "-10.940717"
            },
            {
                "entity": "Javi Pa\u00f1os",
                "confidence_score": "-11.89249"
            },
            {
                "entity": "Pan (band)",
                "confidence_score": "-11.976176"
            }
        ]
    },
    {
        "mention": "teen",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Adolescence",
                "confidence_score": "-1.9932653"
            },
            {
                "entity": "Youth",
                "confidence_score": "-4.6531844"
            },
            {
                "entity": "2017 ISSF Junior World Cup",
                "confidence_score": "-12.217517"
            }
        ]
    },
    {
        "mention": "american heart association abstract",
        "type": "ORG",
        "predictions": [
            {
                "entity": "American Heart Association",
                "confidence_score": "-2.8352323"
            },
            {
                "entity": "American Heart Journal",
                "confidence_score": "-11.05169"
            },
            {
                "entity": "The American Heart Institute",
                "confidence_score": "-11.86385"
            }
        ]
    },
    {
        "mention": "cine kolly deivamagal - 26-09-2015",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Cin\u00e9+",
                "confidence_score": "-10.647243"
            },
            {
                "entity": "Cignal",
                "confidence_score": "-11.599072"
            },
            {
                "entity": "Cine+",
                "confidence_score": "-12.748718"
            }
        ]
    },
    {
        "mention": "sun tv deivamagal",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Deivamagal",
                "confidence_score": "1.929848"
            },
            {
                "entity": "Sun TV (India)",
                "confidence_score": "-0.77363366"
            },
            {
                "entity": "Sun TV Network",
                "confidence_score": "-4.6779943"
            }
        ]
    },
    {
        "mention": "sun tv serials",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Tamil television soap opera",
                "confidence_score": "-3.1146088"
            },
            {
                "entity": "Sun TV Network",
                "confidence_score": "-4.6152062"
            },
            {
                "entity": "List of programs broadcast by Star Vijay",
                "confidence_score": "-10.665736"
            }
        ]
    },
    {
        "mention": "youtube video deivamagal",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "YouTube",
                "confidence_score": "-4.0530386"
            },
            {
                "entity": "Google Video",
                "confidence_score": "-9.71354"
            },
            {
                "entity": "PewDiePie videography",
                "confidence_score": "-11.504464"
            }
        ]
    },
    {
        "mention": "vikatan tele vistas",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Vu Televisions",
                "confidence_score": "-10.618415"
            },
            {
                "entity": "Tele Picture archiving and communication system",
                "confidence_score": "-11.428846"
            },
            {
                "entity": "Video detection and ranging",
                "confidence_score": "-11.937547"
            }
        ]
    },
    {
        "mention": "thirumathi selvam",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Thirumathi Selvam",
                "confidence_score": "12.114499"
            },
            {
                "entity": "Thirumayam",
                "confidence_score": "-10.151469"
            },
            {
                "entity": "Thirumathi Palanisamy",
                "confidence_score": "-11.191888"
            }
        ]
    },
    {
        "mention": "thirumathi selvam",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Thirumathi Selvam",
                "confidence_score": "12.114499"
            },
            {
                "entity": "Thirumayam",
                "confidence_score": "-10.151469"
            },
            {
                "entity": "Thirumathi Palanisamy",
                "confidence_score": "-11.191888"
            }
        ]
    },
    {
        "mention": "sun",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sun Microsystems",
                "confidence_score": "-3.78396"
            },
            {
                "entity": "Sun TV Network",
                "confidence_score": "-4.2329564"
            },
            {
                "entity": "SUN Group",
                "confidence_score": "-4.9375844"
            }
        ]
    },
    {
        "mention": "post comments",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Comments section",
                "confidence_score": "-3.9577708"
            },
            {
                "entity": "Comment (computer programming)",
                "confidence_score": "-4.852222"
            },
            {
                "entity": "POST (HTTP)",
                "confidence_score": "-10.333184"
            }
        ]
    },
    {
        "mention": "los angeles professional employment forecast",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Los Angeles Business Journal",
                "confidence_score": "-10.098554"
            },
            {
                "entity": "Ten-year occupational employment projections",
                "confidence_score": "-10.636722"
            },
            {
                "entity": "Job guarantee",
                "confidence_score": "-10.82089"
            }
        ]
    },
    {
        "mention": "robert half indicates",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Robert Half International",
                "confidence_score": "-3.875004"
            },
            {
                "entity": "Robert Halfon",
                "confidence_score": "-9.7362795"
            },
            {
                "entity": "Rob Richie",
                "confidence_score": "-11.600632"
            }
        ]
    },
    {
        "mention": "freezing",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Freezing",
                "confidence_score": "6.6845775"
            },
            {
                "entity": "Frozen food",
                "confidence_score": "-1.5573266"
            },
            {
                "entity": "Freezing level",
                "confidence_score": "-5.03105"
            }
        ]
    },
    {
        "mention": "brandi britton",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Brandi Britton",
                "confidence_score": "10.034802"
            },
            {
                "entity": "Mark Britton",
                "confidence_score": "-11.491027"
            },
            {
                "entity": "Jim Britton",
                "confidence_score": "-11.552197"
            }
        ]
    },
    {
        "mention": "robert half",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Robert Half International",
                "confidence_score": "4.106703"
            },
            {
                "entity": "Half & Half",
                "confidence_score": "-10.146896"
            },
            {
                "entity": "Robert Halfon",
                "confidence_score": "-10.216338"
            }
        ]
    },
    {
        "mention": "robert half's",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Robert Half International",
                "confidence_score": "5.643333"
            },
            {
                "entity": "One half",
                "confidence_score": "-9.479556"
            },
            {
                "entity": "Half & Half",
                "confidence_score": "-10.1513195"
            }
        ]
    },
    {
        "mention": "the professional employment forecast",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Ten-year occupational employment projections",
                "confidence_score": "-10.643827"
            },
            {
                "entity": "Monster Employment Index",
                "confidence_score": "-11.176484"
            },
            {
                "entity": "Occupational Employment Statistics",
                "confidence_score": "-11.334029"
            }
        ]
    },
    {
        "mention": "compensation upgrades",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Golden boot compensation",
                "confidence_score": "-10.370938"
            },
            {
                "entity": "Job Creation and Worker Assistance Act of 2002",
                "confidence_score": "-10.692373"
            },
            {
                "entity": "Hiring Incentives to Restore Employment Act",
                "confidence_score": "-11.115085"
            }
        ]
    },
    {
        "mention": "the professional employment forecast",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Ten-year occupational employment projections",
                "confidence_score": "-10.643827"
            },
            {
                "entity": "Monster Employment Index",
                "confidence_score": "-11.176484"
            },
            {
                "entity": "Occupational Employment Statistics",
                "confidence_score": "-11.334029"
            }
        ]
    },
    {
        "mention": "robert half",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Robert Half International",
                "confidence_score": "4.106703"
            },
            {
                "entity": "Half & Half",
                "confidence_score": "-10.146896"
            },
            {
                "entity": "Robert Halfon",
                "confidence_score": "-10.216338"
            }
        ]
    },
    {
        "mention": "robert half",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Robert Half International",
                "confidence_score": "4.106703"
            },
            {
                "entity": "Half & Half",
                "confidence_score": "-10.146896"
            },
            {
                "entity": "Robert Halfon",
                "confidence_score": "-10.216338"
            }
        ]
    },
    {
        "mention": "roberthalf.com/los-angeles",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Los Angeles",
                "confidence_score": "-1.6272929"
            },
            {
                "entity": "Los Angeles (magazine)",
                "confidence_score": "-9.39667"
            },
            {
                "entity": "Los Angeles Media Fund",
                "confidence_score": "-11.470977"
            }
        ]
    },
    {
        "mention": "pr newswire",
        "type": "ORG",
        "predictions": [
            {
                "entity": "PR Newswire",
                "confidence_score": "12.28939"
            },
            {
                "entity": "U.S. Newswire",
                "confidence_score": "-4.3931184"
            },
            {
                "entity": "SatireWire",
                "confidence_score": "-10.547641"
            }
        ]
    },
    {
        "mention": "paul a. eisenstein",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Melvin A. Eisenberg",
                "confidence_score": "-11.520937"
            },
            {
                "entity": "Paul Achleitner",
                "confidence_score": "-11.736992"
            },
            {
                "entity": "Paul Oberschneider",
                "confidence_score": "-11.808063"
            }
        ]
    },
    {
        "mention": "cameroon",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Cameroon",
                "confidence_score": "0.70326823"
            },
            {
                "entity": "Vehicle registration plates of Cameroon",
                "confidence_score": "-10.599984"
            },
            {
                "entity": "Cameroon (disambiguation)",
                "confidence_score": "-12.948282"
            }
        ]
    },
    {
        "mention": "the international organization of motor vehicle manufacturers",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Organisation Internationale des Constructeurs d'Automobiles",
                "confidence_score": "-6.320154"
            },
            {
                "entity": "European Automobile Manufacturers Association",
                "confidence_score": "-10.5458555"
            },
            {
                "entity": "International Road Transport Union",
                "confidence_score": "-11.094427"
            }
        ]
    },
    {
        "mention": "oica",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Organisation Internationale des Constructeurs d'Automobiles",
                "confidence_score": "3.1770308"
            },
            {
                "entity": "Owner\u2013Operator Independent Drivers Association",
                "confidence_score": "-11.248149"
            },
            {
                "entity": "Organic Crop Improvement Association",
                "confidence_score": "-11.626438"
            }
        ]
    },
    {
        "mention": "africa",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Africa",
                "confidence_score": "4.098563"
            },
            {
                "entity": "Libya",
                "confidence_score": "-4.8353124"
            },
            {
                "entity": "Nigeria",
                "confidence_score": "-5.9973283"
            }
        ]
    },
    {
        "mention": "oica",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Organisation Internationale des Constructeurs d'Automobiles",
                "confidence_score": "3.1770308"
            },
            {
                "entity": "Owner\u2013Operator Independent Drivers Association",
                "confidence_score": "-11.248149"
            },
            {
                "entity": "Organic Crop Improvement Association",
                "confidence_score": "-11.626438"
            }
        ]
    },
    {
        "mention": "africa",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Africa",
                "confidence_score": "4.098563"
            },
            {
                "entity": "Libya",
                "confidence_score": "-4.8353124"
            },
            {
                "entity": "Nigeria",
                "confidence_score": "-5.9973283"
            }
        ]
    },
    {
        "mention": "americas",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Americas",
                "confidence_score": "5.7528462"
            },
            {
                "entity": "Americas (terminology)",
                "confidence_score": "-2.2610369"
            },
            {
                "entity": "FIBA Americas",
                "confidence_score": "-7.2188897"
            }
        ]
    },
    {
        "mention": "europe",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Europe",
                "confidence_score": "3.0246418"
            },
            {
                "entity": "European Union",
                "confidence_score": "-2.369415"
            },
            {
                "entity": "Africa",
                "confidence_score": "-4.2561126"
            }
        ]
    },
    {
        "mention": "asia",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Asia",
                "confidence_score": "3.6870348"
            },
            {
                "entity": "Asia-Pacific",
                "confidence_score": "-5.1116576"
            },
            {
                "entity": ".asia",
                "confidence_score": "-7.637167"
            }
        ]
    },
    {
        "mention": "oica",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Organisation Internationale des Constructeurs d'Automobiles",
                "confidence_score": "3.1770308"
            },
            {
                "entity": "Owner\u2013Operator Independent Drivers Association",
                "confidence_score": "-11.248149"
            },
            {
                "entity": "Organic Crop Improvement Association",
                "confidence_score": "-11.626438"
            }
        ]
    },
    {
        "mention": "matthias wissmann",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Matthias Wissmann",
                "confidence_score": "11.339347"
            },
            {
                "entity": "Wiesmann",
                "confidence_score": "-9.896157"
            },
            {
                "entity": "Tanner Tessmann",
                "confidence_score": "-11.903149"
            }
        ]
    },
    {
        "mention": "the german association of the automotive industry",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Verband der Automobilindustrie",
                "confidence_score": "5.5893908"
            },
            {
                "entity": "ADAC",
                "confidence_score": "-3.7372396"
            },
            {
                "entity": "European Automobile Manufacturers Association",
                "confidence_score": "-8.519704"
            }
        ]
    },
    {
        "mention": "frankfurt messe",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Messe Frankfurt",
                "confidence_score": "1.348766"
            },
            {
                "entity": "Frankfurt",
                "confidence_score": "-2.3083203"
            },
            {
                "entity": "Frankfurt Messe station",
                "confidence_score": "-6.7007165"
            }
        ]
    },
    {
        "mention": "renault/nissan",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Nissan",
                "confidence_score": "-2.7255714"
            },
            {
                "entity": "Renault",
                "confidence_score": "-3.1167414"
            },
            {
                "entity": "Renault\u2013Nissan\u2013Mitsubishi Alliance",
                "confidence_score": "-3.7134962"
            }
        ]
    },
    {
        "mention": "carlos ghosn",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Carlos Ghosn",
                "confidence_score": "10.845214"
            },
            {
                "entity": "Caroline Ghosn",
                "confidence_score": "-7.5380697"
            },
            {
                "entity": "Nicolas Ghosn",
                "confidence_score": "-12.238341"
            }
        ]
    },
    {
        "mention": "gm",
        "type": "ORG",
        "predictions": [
            {
                "entity": "General Motors",
                "confidence_score": "8.120402"
            },
            {
                "entity": "Microsoft",
                "confidence_score": "-6.8251414"
            },
            {
                "entity": "General Motors Canada",
                "confidence_score": "-8.129058"
            }
        ]
    },
    {
        "mention": "justice department",
        "type": "ORG",
        "predictions": [
            {
                "entity": "United States Department of Justice",
                "confidence_score": "5.0132995"
            },
            {
                "entity": "United States Department of Defense",
                "confidence_score": "-5.8902106"
            },
            {
                "entity": "United States Department of State",
                "confidence_score": "-7.387035"
            }
        ]
    },
    {
        "mention": "oica",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Organisation Internationale des Constructeurs d'Automobiles",
                "confidence_score": "3.1770308"
            },
            {
                "entity": "Owner\u2013Operator Independent Drivers Association",
                "confidence_score": "-11.248149"
            },
            {
                "entity": "Organic Crop Improvement Association",
                "confidence_score": "-11.626438"
            }
        ]
    },
    {
        "mention": "oica",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Organisation Internationale des Constructeurs d'Automobiles",
                "confidence_score": "3.1770308"
            },
            {
                "entity": "Owner\u2013Operator Independent Drivers Association",
                "confidence_score": "-11.248149"
            },
            {
                "entity": "Organic Crop Improvement Association",
                "confidence_score": "-11.626438"
            }
        ]
    },
    {
        "mention": "daimler",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Daimler AG",
                "confidence_score": "4.9833813"
            },
            {
                "entity": "Daimler Motoren Gesellschaft",
                "confidence_score": "-2.496652"
            },
            {
                "entity": "Daimler Company",
                "confidence_score": "-3.6441295"
            }
        ]
    },
    {
        "mention": "africa",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Africa",
                "confidence_score": "4.098563"
            },
            {
                "entity": "Libya",
                "confidence_score": "-4.8353124"
            },
            {
                "entity": "Nigeria",
                "confidence_score": "-5.9973283"
            }
        ]
    },
    {
        "mention": "europe",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Europe",
                "confidence_score": "3.0246418"
            },
            {
                "entity": "European Union",
                "confidence_score": "-2.369415"
            },
            {
                "entity": "Africa",
                "confidence_score": "-4.2561126"
            }
        ]
    },
    {
        "mention": "americas",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Americas",
                "confidence_score": "5.7528462"
            },
            {
                "entity": "Americas (terminology)",
                "confidence_score": "-2.2610369"
            },
            {
                "entity": "FIBA Americas",
                "confidence_score": "-7.2188897"
            }
        ]
    },
    {
        "mention": "europe",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Europe",
                "confidence_score": "3.0246418"
            },
            {
                "entity": "European Union",
                "confidence_score": "-2.369415"
            },
            {
                "entity": "Africa",
                "confidence_score": "-4.2561126"
            }
        ]
    },
    {
        "mention": "honda",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Honda",
                "confidence_score": "9.132179"
            },
            {
                "entity": "Honda Accord",
                "confidence_score": "-4.2652316"
            },
            {
                "entity": "American Honda Motor Company",
                "confidence_score": "-6.4955707"
            }
        ]
    },
    {
        "mention": "click here",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Nissan (disambiguation)",
                "confidence_score": "-11.386483"
            },
            {
                "entity": "Honda CBR250",
                "confidence_score": "-12.88758"
            },
            {
                "entity": "Renault GTA",
                "confidence_score": "-12.906182"
            }
        ]
    },
    {
        "mention": "renault/nissan\u2019s",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Nissan",
                "confidence_score": "-1.1531711"
            },
            {
                "entity": "Renault\u2013Nissan\u2013Mitsubishi Alliance",
                "confidence_score": "-3.4423015"
            },
            {
                "entity": "Nissan 200SX",
                "confidence_score": "-11.387139"
            }
        ]
    },
    {
        "mention": "africa",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Africa",
                "confidence_score": "4.098563"
            },
            {
                "entity": "Libya",
                "confidence_score": "-4.8353124"
            },
            {
                "entity": "Nigeria",
                "confidence_score": "-5.9973283"
            }
        ]
    },
    {
        "mention": "oica",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Organisation Internationale des Constructeurs d'Automobiles",
                "confidence_score": "3.1770308"
            },
            {
                "entity": "Owner\u2013Operator Independent Drivers Association",
                "confidence_score": "-11.248149"
            },
            {
                "entity": "Organic Crop Improvement Association",
                "confidence_score": "-11.626438"
            }
        ]
    },
    {
        "mention": "oica",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Organisation Internationale des Constructeurs d'Automobiles",
                "confidence_score": "3.1770308"
            },
            {
                "entity": "Owner\u2013Operator Independent Drivers Association",
                "confidence_score": "-11.248149"
            },
            {
                "entity": "Organic Crop Improvement Association",
                "confidence_score": "-11.626438"
            }
        ]
    },
    {
        "mention": "paul a. eisenstein",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Paul A. Schneider",
                "confidence_score": "-11.430123"
            },
            {
                "entity": "Pablo Eisenberg",
                "confidence_score": "-12.089656"
            },
            {
                "entity": "Paul Eibeler",
                "confidence_score": "-12.232968"
            }
        ]
    },
    {
        "mention": "automobiles",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Automobile (magazine)",
                "confidence_score": "-4.559293"
            },
            {
                "entity": "2014 Dubai 24 Hour",
                "confidence_score": "-12.264899"
            },
            {
                "entity": "Automobile Association",
                "confidence_score": "-12.647675"
            }
        ]
    },
    {
        "mention": "nissan",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Nissan",
                "confidence_score": "4.688967"
            },
            {
                "entity": "Nissan Group",
                "confidence_score": "-6.311645"
            },
            {
                "entity": "Nissan Chemical Corporation",
                "confidence_score": "-10.663416"
            }
        ]
    },
    {
        "mention": "joey wang",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Joey Wong",
                "confidence_score": "1.3677337"
            },
            {
                "entity": "Joey Wat",
                "confidence_score": "-10.83706"
            },
            {
                "entity": "Jade Wang",
                "confidence_score": "-11.0838785"
            }
        ]
    },
    {
        "mention": "nissan",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Nissan",
                "confidence_score": "4.688967"
            },
            {
                "entity": "Nissan Group",
                "confidence_score": "-6.311645"
            },
            {
                "entity": "Nissan Chemical Corporation",
                "confidence_score": "-10.663416"
            }
        ]
    },
    {
        "mention": "nissan",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Nissan",
                "confidence_score": "4.688967"
            },
            {
                "entity": "Nissan Group",
                "confidence_score": "-6.311645"
            },
            {
                "entity": "Nissan Chemical Corporation",
                "confidence_score": "-10.663416"
            }
        ]
    },
    {
        "mention": "lanniao",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Nissan Lannia",
                "confidence_score": "-4.159739"
            },
            {
                "entity": "Lancia",
                "confidence_score": "-6.5898585"
            },
            {
                "entity": "Lannoo",
                "confidence_score": "-7.1668572"
            }
        ]
    },
    {
        "mention": "the nissan china design center (ndc",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Nintendo System Development",
                "confidence_score": "-11.417943"
            },
            {
                "entity": "Toyota National College of Technology",
                "confidence_score": "-11.779648"
            },
            {
                "entity": "Thailand Creative & Design Center",
                "confidence_score": "-12.364513"
            }
        ]
    },
    {
        "mention": "the nissan global design center",
        "type": "ORG",
        "predictions": [
            {
                "entity": "International Technical Development Center",
                "confidence_score": "-10.580587"
            },
            {
                "entity": "Global Design Effort",
                "confidence_score": "-11.431853"
            },
            {
                "entity": "Digital Design Studio",
                "confidence_score": "-12.41994"
            }
        ]
    },
    {
        "mention": "ngdc",
        "type": "ORG",
        "predictions": [
            {
                "entity": "National Commission for Denotified, Nomadic and Semi-Nomadic Tribes",
                "confidence_score": "-11.838946"
            },
            {
                "entity": "National Center for Digital Certification",
                "confidence_score": "-12.013958"
            },
            {
                "entity": "Nigerian Tourism Development Corporation",
                "confidence_score": "-12.403844"
            }
        ]
    },
    {
        "mention": "dongfeng-nissan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Dongfeng Motor Co., Ltd.",
                "confidence_score": "-4.691514"
            },
            {
                "entity": "Dongfeng Motor Corporation",
                "confidence_score": "-6.270923"
            },
            {
                "entity": "Dongfeng Peugeot-Citro\u00ebn",
                "confidence_score": "-9.524495"
            }
        ]
    },
    {
        "mention": "nissan",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Nissan",
                "confidence_score": "4.688967"
            },
            {
                "entity": "Nissan Group",
                "confidence_score": "-6.311645"
            },
            {
                "entity": "Nissan Chemical Corporation",
                "confidence_score": "-10.663416"
            }
        ]
    },
    {
        "mention": "fuzzy",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Fuzzy classification",
                "confidence_score": "-7.445805"
            },
            {
                "entity": "Fuzzy agent",
                "confidence_score": "-8.880872"
            },
            {
                "entity": "Fuzzy Logix",
                "confidence_score": "-9.443469"
            }
        ]
    },
    {
        "mention": "nissan",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Nissan",
                "confidence_score": "4.688967"
            },
            {
                "entity": "Nissan Group",
                "confidence_score": "-6.311645"
            },
            {
                "entity": "Nissan Chemical Corporation",
                "confidence_score": "-10.663416"
            }
        ]
    },
    {
        "mention": "loading",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Loader (computing)",
                "confidence_score": "-7.0049486"
            },
            {
                "entity": "Load (computing)",
                "confidence_score": "-7.2788634"
            },
            {
                "entity": "Loading screen",
                "confidence_score": "-8.180894"
            }
        ]
    },
    {
        "mention": "dongfeng-nissan nissan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Nissan",
                "confidence_score": "-3.6953557"
            },
            {
                "entity": "Dongfeng Motor Co., Ltd.",
                "confidence_score": "-5.858568"
            },
            {
                "entity": "Dongfeng Peugeot-Citro\u00ebn",
                "confidence_score": "-8.437636"
            }
        ]
    },
    {
        "mention": "ritz",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Ritz Tower",
                "confidence_score": "-9.453444"
            },
            {
                "entity": "Ritz-Carlton, Riyadh",
                "confidence_score": "-9.922773"
            },
            {
                "entity": "Hotel Ritz, Madrid",
                "confidence_score": "-10.581843"
            }
        ]
    },
    {
        "mention": "the ritz-carlton \u2018s",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Ritz-Carlton Hotel Company",
                "confidence_score": "-3.5609815"
            },
            {
                "entity": "The Ritz-Carlton, Washington, D.C.",
                "confidence_score": "-5.7847757"
            },
            {
                "entity": "Ritz-Carlton, Riyadh",
                "confidence_score": "-8.837169"
            }
        ]
    },
    {
        "mention": "the mexican embassy",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Embassy of Mexico, Paris",
                "confidence_score": "-9.635977"
            },
            {
                "entity": "Embassy of the United States, Mexico City",
                "confidence_score": "-9.975193"
            },
            {
                "entity": "Embassy of Mexico, Ottawa",
                "confidence_score": "-10.635633"
            }
        ]
    },
    {
        "mention": "the chile en nogada",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Chile",
                "confidence_score": "-4.134796"
            },
            {
                "entity": "Chiles en nogada",
                "confidence_score": "-7.6637974"
            },
            {
                "entity": "Chilean Antarctic Territory",
                "confidence_score": "-9.617642"
            }
        ]
    },
    {
        "mention": "nogada",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Negada",
                "confidence_score": "-11.620681"
            },
            {
                "entity": "Noja",
                "confidence_score": "-11.838738"
            },
            {
                "entity": "Nogaye Lo",
                "confidence_score": "-12.190002"
            }
        ]
    },
    {
        "mention": "jorge carlos gonz\u00e1les",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Rodolfo Gonzales",
                "confidence_score": "-11.635093"
            },
            {
                "entity": "Jose Gonzales-Gonzales",
                "confidence_score": "-11.721436"
            },
            {
                "entity": "Carlos Gonzales (Peruvian footballer)",
                "confidence_score": "-12.435639"
            }
        ]
    },
    {
        "mention": "ritz",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Ritz Tower",
                "confidence_score": "-9.453444"
            },
            {
                "entity": "Ritz-Carlton, Riyadh",
                "confidence_score": "-9.922773"
            },
            {
                "entity": "Hotel Ritz, Madrid",
                "confidence_score": "-10.581843"
            }
        ]
    },
    {
        "mention": "nett",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Net (economics)",
                "confidence_score": "4.3096323"
            },
            {
                "entity": "Net register tonnage",
                "confidence_score": "-2.2065575"
            },
            {
                "entity": "Net tonnage",
                "confidence_score": "-6.4103622"
            }
        ]
    },
    {
        "mention": "ritz",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Ritz Tower",
                "confidence_score": "-9.453444"
            },
            {
                "entity": "Ritz-Carlton, Riyadh",
                "confidence_score": "-9.922773"
            },
            {
                "entity": "Hotel Ritz, Madrid",
                "confidence_score": "-10.581843"
            }
        ]
    },
    {
        "mention": "jibby & co.",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Robinsons & Co.",
                "confidence_score": "-10.038628"
            },
            {
                "entity": "JYJ",
                "confidence_score": "-11.469079"
            },
            {
                "entity": "Jiban Bima Corporation",
                "confidence_score": "-11.826677"
            }
        ]
    },
    {
        "mention": "juicy",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Sweetness",
                "confidence_score": "-7.4782825"
            },
            {
                "entity": "Juicy Couture",
                "confidence_score": "-8.619092"
            },
            {
                "entity": "Juicy",
                "confidence_score": "-11.459817"
            }
        ]
    },
    {
        "mention": "meaty",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Meaty",
                "confidence_score": "-10.120429"
            },
            {
                "entity": "Meat magazine",
                "confidence_score": "-10.797367"
            },
            {
                "entity": "Meat Katie",
                "confidence_score": "-11.755659"
            }
        ]
    },
    {
        "mention": "the jibby & co.",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Robinsons & Co.",
                "confidence_score": "-9.274943"
            },
            {
                "entity": "Shanghai Jiushi Group",
                "confidence_score": "-11.5455675"
            },
            {
                "entity": "Jiffy (convenience store)",
                "confidence_score": "-11.843783"
            }
        ]
    },
    {
        "mention": "empire shopping gallery",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Empire City Mall",
                "confidence_score": "-9.864724"
            },
            {
                "entity": "Starhill Gallery",
                "confidence_score": "-10.152752"
            },
            {
                "entity": "\u00c6ON Bukit Indah Shopping Centre",
                "confidence_score": "-11.565492"
            }
        ]
    },
    {
        "mention": "jason chan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jason Chan Chi-san",
                "confidence_score": "4.2236667"
            },
            {
                "entity": "Jason Chan (singer)",
                "confidence_score": "-1.4546729"
            },
            {
                "entity": "Jason Lo",
                "confidence_score": "-11.752342"
            }
        ]
    },
    {
        "mention": "cheesy",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Cheetos",
                "confidence_score": "-8.672715"
            },
            {
                "entity": "Cheesy",
                "confidence_score": "-11.186189"
            },
            {
                "entity": "Kiko-chan's Smile",
                "confidence_score": "-11.644924"
            }
        ]
    },
    {
        "mention": "meaty",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Meaty",
                "confidence_score": "-10.120429"
            },
            {
                "entity": "Meat magazine",
                "confidence_score": "-10.797367"
            },
            {
                "entity": "Meat Katie",
                "confidence_score": "-11.755659"
            }
        ]
    },
    {
        "mention": "al dente",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Al dente",
                "confidence_score": "5.1736054"
            },
            {
                "entity": "Al Dente (film)",
                "confidence_score": "-3.0616586"
            },
            {
                "entity": "Esm\u00e9e Denters",
                "confidence_score": "-11.083591"
            }
        ]
    },
    {
        "mention": "publika\u2019s white box",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Bali Post Media Group",
                "confidence_score": "-11.758748"
            },
            {
                "entity": "Publika TV",
                "confidence_score": "-11.82215"
            },
            {
                "entity": "Publika FM",
                "confidence_score": "-12.114303"
            }
        ]
    },
    {
        "mention": "kl",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kuala Lumpur",
                "confidence_score": "6.567741"
            },
            {
                "entity": "Kuala Lumpur Tower",
                "confidence_score": "-4.515897"
            },
            {
                "entity": "Kuala Lumpur City Centre",
                "confidence_score": "-9.591512"
            }
        ]
    },
    {
        "mention": "ritz",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Ritz Tower",
                "confidence_score": "-9.453444"
            },
            {
                "entity": "Ritz-Carlton, Riyadh",
                "confidence_score": "-9.922773"
            },
            {
                "entity": "Hotel Ritz, Madrid",
                "confidence_score": "-10.581843"
            }
        ]
    },
    {
        "mention": "kl editor",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Contributing editor",
                "confidence_score": "-8.451207"
            },
            {
                "entity": "Copy editing",
                "confidence_score": "-8.517801"
            },
            {
                "entity": "City editor",
                "confidence_score": "-9.798249"
            }
        ]
    },
    {
        "mention": "dee may",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Deems May",
                "confidence_score": "-9.396798"
            },
            {
                "entity": "Katie May",
                "confidence_score": "-9.953279"
            },
            {
                "entity": "Sarah May",
                "confidence_score": "-11.283624"
            }
        ]
    },
    {
        "mention": "mia wasikowska",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mia Wasikowska",
                "confidence_score": "12.496398"
            },
            {
                "entity": "Mia Kirshner",
                "confidence_score": "-10.904408"
            },
            {
                "entity": "Mia Murano",
                "confidence_score": "-11.625081"
            }
        ]
    },
    {
        "mention": "alice",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Alice (Alice's Adventures in Wonderland)",
                "confidence_score": "-5.663775"
            },
            {
                "entity": "Alice (Resident Evil)",
                "confidence_score": "-7.6751227"
            },
            {
                "entity": "Alice Fulks",
                "confidence_score": "-10.098265"
            }
        ]
    },
    {
        "mention": "crimson peak",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Crimson Peak",
                "confidence_score": "12.409198"
            },
            {
                "entity": "Diamond Peak (ski area)",
                "confidence_score": "-11.954026"
            },
            {
                "entity": "Patterson Peak",
                "confidence_score": "-13.042562"
            }
        ]
    },
    {
        "mention": "edith cushing",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Catherine Chisholm Cushing",
                "confidence_score": "-10.852707"
            },
            {
                "entity": "Aric Cushing",
                "confidence_score": "-11.111829"
            },
            {
                "entity": "Nick Cushing",
                "confidence_score": "-11.802096"
            }
        ]
    },
    {
        "mention": "mia wasikowska",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mia Wasikowska",
                "confidence_score": "12.496398"
            },
            {
                "entity": "Mia Kirshner",
                "confidence_score": "-10.904408"
            },
            {
                "entity": "Mia Murano",
                "confidence_score": "-11.625081"
            }
        ]
    },
    {
        "mention": "tim burton",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Tim Burton",
                "confidence_score": "12.258313"
            },
            {
                "entity": "Tim Burton (musician)",
                "confidence_score": "-5.8413815"
            },
            {
                "entity": "Tim Sharp",
                "confidence_score": "-12.01359"
            }
        ]
    },
    {
        "mention": "edith cushing",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Catherine Chisholm Cushing",
                "confidence_score": "-10.852707"
            },
            {
                "entity": "Aric Cushing",
                "confidence_score": "-11.111829"
            },
            {
                "entity": "Nick Cushing",
                "confidence_score": "-11.802096"
            }
        ]
    },
    {
        "mention": "edith",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Edith",
                "confidence_score": "0.99857825"
            },
            {
                "entity": "Edith Wharton",
                "confidence_score": "-6.2298093"
            },
            {
                "entity": "Edith Roosevelt",
                "confidence_score": "-9.000337"
            }
        ]
    },
    {
        "mention": "edith",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Edith",
                "confidence_score": "0.99857825"
            },
            {
                "entity": "Edith Wharton",
                "confidence_score": "-6.2298093"
            },
            {
                "entity": "Edith Roosevelt",
                "confidence_score": "-9.000337"
            }
        ]
    },
    {
        "mention": "guillermo del toro",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Guillermo del Toro",
                "confidence_score": "11.722655"
            },
            {
                "entity": "Del Toro (surname)",
                "confidence_score": "-10.581647"
            },
            {
                "entity": "Del Toro",
                "confidence_score": "-12.388013"
            }
        ]
    },
    {
        "mention": "frankenstein",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Frankenstein",
                "confidence_score": "6.1125445"
            },
            {
                "entity": "Frankenstein's monster",
                "confidence_score": "1.3760395"
            },
            {
                "entity": "Frankenstein (Death Race)",
                "confidence_score": "-3.9146419"
            }
        ]
    },
    {
        "mention": "edith",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Edith",
                "confidence_score": "0.99857825"
            },
            {
                "entity": "Edith Wharton",
                "confidence_score": "-6.2298093"
            },
            {
                "entity": "Edith Roosevelt",
                "confidence_score": "-9.000337"
            }
        ]
    },
    {
        "mention": "edith",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Edith",
                "confidence_score": "3.8860838"
            },
            {
                "entity": "Edith Wharton",
                "confidence_score": "-5.28323"
            },
            {
                "entity": "Edith Roosevelt",
                "confidence_score": "-8.221506"
            }
        ]
    },
    {
        "mention": "callum greene",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Sonia Greene",
                "confidence_score": "-10.847867"
            },
            {
                "entity": "Sean Greene",
                "confidence_score": "-11.528211"
            },
            {
                "entity": "Callum Cooke",
                "confidence_score": "-11.749079"
            }
        ]
    },
    {
        "mention": "edith",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Edith",
                "confidence_score": "3.8860838"
            },
            {
                "entity": "Edith Wharton",
                "confidence_score": "-5.28323"
            },
            {
                "entity": "Edith Roosevelt",
                "confidence_score": "-8.221506"
            }
        ]
    },
    {
        "mention": "mia",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mia Wasikowska",
                "confidence_score": "0.42582998"
            },
            {
                "entity": "Mia Farrow",
                "confidence_score": "-3.4473245"
            },
            {
                "entity": "Mia Wallace",
                "confidence_score": "-10.072122"
            }
        ]
    },
    {
        "mention": "guillermo del toro",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Guillermo del Toro",
                "confidence_score": "11.722655"
            },
            {
                "entity": "Del Toro (surname)",
                "confidence_score": "-10.581647"
            },
            {
                "entity": "Del Toro",
                "confidence_score": "-12.388013"
            }
        ]
    },
    {
        "mention": "universal pictures",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Universal Pictures",
                "confidence_score": "10.809438"
            },
            {
                "entity": "Universal Pictures Home Entertainment",
                "confidence_score": "-6.088585"
            },
            {
                "entity": "Universal Animation Studios",
                "confidence_score": "-9.212479"
            }
        ]
    },
    {
        "mention": "mia wasikowska",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mia Wasikowska",
                "confidence_score": "12.496398"
            },
            {
                "entity": "Mia Kirshner",
                "confidence_score": "-10.904408"
            },
            {
                "entity": "Mia Murano",
                "confidence_score": "-11.625081"
            }
        ]
    },
    {
        "mention": "alice",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Alice (Alice's Adventures in Wonderland)",
                "confidence_score": "-5.663775"
            },
            {
                "entity": "Alice (Resident Evil)",
                "confidence_score": "-7.6751227"
            },
            {
                "entity": "Alice Fulks",
                "confidence_score": "-10.098265"
            }
        ]
    },
    {
        "mention": "academy award\u00ae",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Academy Awards",
                "confidence_score": "6.936744"
            },
            {
                "entity": "Academy Honorary Award",
                "confidence_score": "-7.4301753"
            },
            {
                "entity": "Academy Award for Best Adapted Screenplay",
                "confidence_score": "-8.451261"
            }
        ]
    },
    {
        "mention": "jessica chastain",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jessica Chastain",
                "confidence_score": "11.668368"
            },
            {
                "entity": "Jessica Chastain on screen and stage",
                "confidence_score": "-3.7395797"
            },
            {
                "entity": "Jessica Chaffin",
                "confidence_score": "-10.921571"
            }
        ]
    },
    {
        "mention": "interstellar",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Interstellar (film)",
                "confidence_score": "5.6974053"
            },
            {
                "entity": "Interstellar (band)",
                "confidence_score": "-7.619841"
            },
            {
                "entity": "InterWorld",
                "confidence_score": "-9.779013"
            }
        ]
    },
    {
        "mention": "tom hiddleston",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Tom Hiddleston",
                "confidence_score": "11.254449"
            },
            {
                "entity": "Tom Stourton",
                "confidence_score": "-10.272843"
            },
            {
                "entity": "Tom Lotherington",
                "confidence_score": "-11.402518"
            }
        ]
    },
    {
        "mention": "charlie hunnam",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Charlie Hunnam",
                "confidence_score": "8.1086855"
            },
            {
                "entity": "Charlie Huston",
                "confidence_score": "-10.491383"
            },
            {
                "entity": "Charlie Huddy",
                "confidence_score": "-11.375828"
            }
        ]
    },
    {
        "mention": "crimson peak",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Crimson Peak",
                "confidence_score": "12.409198"
            },
            {
                "entity": "Diamond Peak (ski area)",
                "confidence_score": "-11.954026"
            },
            {
                "entity": "Patterson Peak",
                "confidence_score": "-13.042562"
            }
        ]
    },
    {
        "mention": "crimson peak",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Crimson Peak",
                "confidence_score": "12.409198"
            },
            {
                "entity": "Diamond Peak (ski area)",
                "confidence_score": "-11.954026"
            },
            {
                "entity": "Patterson Peak",
                "confidence_score": "-13.042562"
            }
        ]
    },
    {
        "mention": "united international pictures",
        "type": "ORG",
        "predictions": [
            {
                "entity": "United International Pictures",
                "confidence_score": "9.586531"
            },
            {
                "entity": "United Artists",
                "confidence_score": "-6.8787265"
            },
            {
                "entity": "United States Pictures",
                "confidence_score": "-10.623298"
            }
        ]
    },
    {
        "mention": "columbia pictures",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Columbia Pictures",
                "confidence_score": "11.634309"
            },
            {
                "entity": "Sony Pictures",
                "confidence_score": "-5.239906"
            },
            {
                "entity": "Columbia Records",
                "confidence_score": "-7.7462115"
            }
        ]
    },
    {
        "mention": "steel dynamics, inc.",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Steel Dynamics",
                "confidence_score": "10.771548"
            },
            {
                "entity": "Steel Connect",
                "confidence_score": "-9.523234"
            },
            {
                "entity": "Hughes Dynamics",
                "confidence_score": "-10.975592"
            }
        ]
    },
    {
        "mention": "thomson reuters",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Thomson Reuters",
                "confidence_score": "9.638311"
            },
            {
                "entity": "Reuters",
                "confidence_score": "-0.45496362"
            },
            {
                "entity": "Thomson Corporation",
                "confidence_score": "-4.6108785"
            }
        ]
    },
    {
        "mention": "ratings network.com",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Nielsen Media Research",
                "confidence_score": "-5.9906774"
            },
            {
                "entity": "RatingsIntel",
                "confidence_score": "-9.997657"
            },
            {
                "entity": "Salesforce.com",
                "confidence_score": "-11.292388"
            }
        ]
    },
    {
        "mention": "steel dynamics, inc.",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Steel Dynamics",
                "confidence_score": "10.771548"
            },
            {
                "entity": "Steel Connect",
                "confidence_score": "-9.523234"
            },
            {
                "entity": "Hughes Dynamics",
                "confidence_score": "-10.975592"
            }
        ]
    },
    {
        "mention": "steel dynamics, inc.",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Steel Dynamics",
                "confidence_score": "10.771548"
            },
            {
                "entity": "Steel Connect",
                "confidence_score": "-9.523234"
            },
            {
                "entity": "Hughes Dynamics",
                "confidence_score": "-10.975592"
            }
        ]
    },
    {
        "mention": "steel dynamics",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Steel Dynamics",
                "confidence_score": "10.97265"
            },
            {
                "entity": "Microsoft Dynamics",
                "confidence_score": "-10.607012"
            },
            {
                "entity": "Data Dynamics",
                "confidence_score": "-11.980631"
            }
        ]
    },
    {
        "mention": "thomson reuters",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Thomson Reuters",
                "confidence_score": "9.638311"
            },
            {
                "entity": "Reuters",
                "confidence_score": "-0.45496362"
            },
            {
                "entity": "Thomson Corporation",
                "confidence_score": "-4.6108785"
            }
        ]
    },
    {
        "mention": "steel dynamics, inc.",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Steel Dynamics",
                "confidence_score": "10.771548"
            },
            {
                "entity": "Steel Connect",
                "confidence_score": "-9.523234"
            },
            {
                "entity": "Hughes Dynamics",
                "confidence_score": "-10.975592"
            }
        ]
    },
    {
        "mention": "bank of america",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Bank of America",
                "confidence_score": "10.363298"
            },
            {
                "entity": "Bank of America Canada",
                "confidence_score": "-9.669135"
            },
            {
                "entity": "Bank of Canada",
                "confidence_score": "-10.202437"
            }
        ]
    },
    {
        "mention": "steel dynamics",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Steel Dynamics",
                "confidence_score": "10.97265"
            },
            {
                "entity": "Microsoft Dynamics",
                "confidence_score": "-10.607012"
            },
            {
                "entity": "Data Dynamics",
                "confidence_score": "-11.980631"
            }
        ]
    },
    {
        "mention": "thestreet",
        "type": "ORG",
        "predictions": [
            {
                "entity": "TheStreet.com",
                "confidence_score": "6.1778307"
            },
            {
                "entity": "Teachstreet",
                "confidence_score": "-11.761054"
            },
            {
                "entity": "T. R. Abhilashi Memorial Institute of Engineering and Technology",
                "confidence_score": "-12.351509"
            }
        ]
    },
    {
        "mention": "steel dynamics",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Steel Dynamics",
                "confidence_score": "10.97265"
            },
            {
                "entity": "Microsoft Dynamics",
                "confidence_score": "-10.607012"
            },
            {
                "entity": "Data Dynamics",
                "confidence_score": "-11.980631"
            }
        ]
    },
    {
        "mention": "jefferies group",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Jefferies Group",
                "confidence_score": "10.725219"
            },
            {
                "entity": "Jefferies Financial Group",
                "confidence_score": "-3.9308202"
            },
            {
                "entity": "Jeff Taylor (entrepreneur)",
                "confidence_score": "-12.148281"
            }
        ]
    },
    {
        "mention": "steel dynamics",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Steel Dynamics",
                "confidence_score": "10.97265"
            },
            {
                "entity": "Microsoft Dynamics",
                "confidence_score": "-10.607012"
            },
            {
                "entity": "Data Dynamics",
                "confidence_score": "-11.980631"
            }
        ]
    },
    {
        "mention": "zacks cut steel dynamics",
        "type": "ORG",
        "predictions": [
            {
                "entity": "S&Z Metalworks",
                "confidence_score": "-10.891122"
            },
            {
                "entity": "JSW Ispat Steel",
                "confidence_score": "-11.54094"
            },
            {
                "entity": "Zacks Place",
                "confidence_score": "-12.117669"
            }
        ]
    },
    {
        "mention": "macquarie",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Macquarie Group",
                "confidence_score": "-3.2178836"
            },
            {
                "entity": "Division of Macquarie",
                "confidence_score": "-7.35669"
            },
            {
                "entity": "Macquarie, Australian Capital Territory",
                "confidence_score": "-7.63287"
            }
        ]
    },
    {
        "mention": "steel dynamics",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Steel Dynamics",
                "confidence_score": "10.97265"
            },
            {
                "entity": "Microsoft Dynamics",
                "confidence_score": "-10.607012"
            },
            {
                "entity": "Data Dynamics",
                "confidence_score": "-11.980631"
            }
        ]
    },
    {
        "mention": "steel dynamics",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Steel Dynamics",
                "confidence_score": "10.97265"
            },
            {
                "entity": "Microsoft Dynamics",
                "confidence_score": "-10.607012"
            },
            {
                "entity": "Data Dynamics",
                "confidence_score": "-11.980631"
            }
        ]
    },
    {
        "mention": "buy",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Buy\u2013sell agreement",
                "confidence_score": "-6.094321"
            },
            {
                "entity": "Buy-write",
                "confidence_score": "-10.413722"
            },
            {
                "entity": "BuyBlue.org",
                "confidence_score": "-10.726434"
            }
        ]
    },
    {
        "mention": "john c. bates",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "John C. Bates",
                "confidence_score": "9.03237"
            },
            {
                "entity": "John D. Bates",
                "confidence_score": "-6.3821397"
            },
            {
                "entity": "John Bates (basketball)",
                "confidence_score": "-10.758024"
            }
        ]
    },
    {
        "mention": "steel dynamics",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Steel Dynamics",
                "confidence_score": "10.97265"
            },
            {
                "entity": "Microsoft Dynamics",
                "confidence_score": "-10.607012"
            },
            {
                "entity": "Data Dynamics",
                "confidence_score": "-11.980631"
            }
        ]
    },
    {
        "mention": "the securities & exchange commission",
        "type": "ORG",
        "predictions": [
            {
                "entity": "U.S. Securities and Exchange Commission",
                "confidence_score": "9.095315"
            },
            {
                "entity": "Securities and Exchange Commission (Philippines)",
                "confidence_score": "-0.75468516"
            },
            {
                "entity": "Securities commission",
                "confidence_score": "-5.1703134"
            }
        ]
    },
    {
        "mention": "sec",
        "type": "ORG",
        "predictions": [
            {
                "entity": "U.S. Securities and Exchange Commission",
                "confidence_score": "11.189518"
            },
            {
                "entity": "Securities and Exchange Commission (Philippines)",
                "confidence_score": "3.3183484"
            },
            {
                "entity": "Federal Communications Commission",
                "confidence_score": "-5.5948358"
            }
        ]
    },
    {
        "mention": "steel dynamics, inc.",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Steel Dynamics",
                "confidence_score": "10.771548"
            },
            {
                "entity": "Steel Connect",
                "confidence_score": "-9.523234"
            },
            {
                "entity": "Hughes Dynamics",
                "confidence_score": "-10.975592"
            }
        ]
    },
    {
        "mention": "company",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Company",
                "confidence_score": "9.122138"
            },
            {
                "entity": "Railway company",
                "confidence_score": "-9.171106"
            },
            {
                "entity": "Company (band)",
                "confidence_score": "-9.874451"
            }
        ]
    },
    {
        "mention": "galvalume",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Leonard Green & Partners",
                "confidence_score": "-11.019694"
            },
            {
                "entity": "Galvanize (software company)",
                "confidence_score": "-11.780145"
            },
            {
                "entity": "Villaume",
                "confidence_score": "-12.879235"
            }
        ]
    },
    {
        "mention": "galfan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Galcon",
                "confidence_score": "-11.121753"
            },
            {
                "entity": "Galson Laboratories",
                "confidence_score": "-11.801229"
            },
            {
                "entity": "Gen4 Energy",
                "confidence_score": "-11.810819"
            }
        ]
    },
    {
        "mention": "nsw",
        "type": "ORG",
        "predictions": [
            {
                "entity": "New South Wales",
                "confidence_score": "4.151885"
            },
            {
                "entity": "Government of New South Wales",
                "confidence_score": "-5.101606"
            },
            {
                "entity": "Cricket NSW",
                "confidence_score": "-8.525115"
            }
        ]
    },
    {
        "mention": "warners bay",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Warners Bay, New South Wales",
                "confidence_score": "0.68160397"
            },
            {
                "entity": "Warners Bay High School",
                "confidence_score": "-7.5000434"
            },
            {
                "entity": "Neutral Bay, New South Wales",
                "confidence_score": "-9.364075"
            }
        ]
    },
    {
        "mention": "nsw police",
        "type": "ORG",
        "predictions": [
            {
                "entity": "New South Wales Police Force",
                "confidence_score": "6.0501738"
            },
            {
                "entity": "South Wales Police",
                "confidence_score": "-8.049785"
            },
            {
                "entity": "New Jersey State Police",
                "confidence_score": "-10.894244"
            }
        ]
    },
    {
        "mention": "postmans lane",
        "type": "ORG",
        "predictions": [
            {
                "entity": "General Holmes Drive",
                "confidence_score": "-10.702782"
            },
            {
                "entity": "The Horsley Drive",
                "confidence_score": "-11.599172"
            },
            {
                "entity": "Postmans Ridge, Queensland",
                "confidence_score": "-12.257697"
            }
        ]
    },
    {
        "mention": "the professional standards command",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Standards Australia",
                "confidence_score": "-8.421635"
            },
            {
                "entity": "Professional certification",
                "confidence_score": "-8.621453"
            },
            {
                "entity": "Professional certification (business)",
                "confidence_score": "-9.270954"
            }
        ]
    },
    {
        "mention": "australia post",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australia Post",
                "confidence_score": "11.573691"
            },
            {
                "entity": "Australia",
                "confidence_score": "-2.9743903"
            },
            {
                "entity": "Australian Associated Press",
                "confidence_score": "-8.732738"
            }
        ]
    },
    {
        "mention": "cheryl taylor",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Cheryl Taylor",
                "confidence_score": "9.581741"
            },
            {
                "entity": "Lenore Taylor",
                "confidence_score": "-8.987637"
            },
            {
                "entity": "Cory Taylor",
                "confidence_score": "-9.894231"
            }
        ]
    },
    {
        "mention": "the newcastle herald",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Newcastle Herald",
                "confidence_score": "10.175796"
            },
            {
                "entity": "The Sydney Morning Herald",
                "confidence_score": "-7.7066045"
            },
            {
                "entity": "The Dickson Herald",
                "confidence_score": "-11.656241"
            }
        ]
    },
    {
        "mention": "dawson\u2019s oysters",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Bentley's Oyster Bar and Grill",
                "confidence_score": "-11.092195"
            },
            {
                "entity": "St Matthias, Bristol",
                "confidence_score": "-11.651017"
            },
            {
                "entity": "Portland Creek, Newfoundland and Labrador",
                "confidence_score": "-12.385438"
            }
        ]
    },
    {
        "mention": "king street",
        "type": "ORG",
        "predictions": [
            {
                "entity": "King Street, Sydney",
                "confidence_score": "-3.0228975"
            },
            {
                "entity": "King Street, Manchester",
                "confidence_score": "-10.287187"
            },
            {
                "entity": "King Street, St James's",
                "confidence_score": "-11.125939"
            }
        ]
    },
    {
        "mention": "newcastle herald",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Newcastle Herald",
                "confidence_score": "9.92756"
            },
            {
                "entity": "The Sydney Morning Herald",
                "confidence_score": "-7.128462"
            },
            {
                "entity": "The Dickson Herald",
                "confidence_score": "-11.758367"
            }
        ]
    },
    {
        "mention": "the islamic state",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Islamic State of Iraq and the Levant",
                "confidence_score": "-4.269014"
            },
            {
                "entity": "Islamic state",
                "confidence_score": "-5.0754538"
            },
            {
                "entity": "Islamic State of Iraq",
                "confidence_score": "-8.971911"
            }
        ]
    },
    {
        "mention": "kashmir",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Jammu and Kashmir",
                "confidence_score": "-2.3491042"
            },
            {
                "entity": "Kashmir",
                "confidence_score": "-4.370001"
            },
            {
                "entity": "Kashmir Valley",
                "confidence_score": "-6.7483745"
            }
        ]
    },
    {
        "mention": "union home",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Union territory",
                "confidence_score": "-2.4101503"
            },
            {
                "entity": "Eurasian Home",
                "confidence_score": "-11.371952"
            },
            {
                "entity": "Union Place",
                "confidence_score": "-11.630657"
            }
        ]
    },
    {
        "mention": "rajnath singh",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Rajnath Singh",
                "confidence_score": "11.382765"
            },
            {
                "entity": "Rajnesh Singh",
                "confidence_score": "-11.979644"
            },
            {
                "entity": "Rajesh Singh",
                "confidence_score": "-12.259593"
            }
        ]
    },
    {
        "mention": "home",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Home",
                "confidence_score": "6.4426885"
            },
            {
                "entity": "Home (sports)",
                "confidence_score": "-4.0045476"
            },
            {
                "entity": "Home Guard (India)",
                "confidence_score": "-9.0039015"
            }
        ]
    },
    {
        "mention": "ani",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Asian News International",
                "confidence_score": "5.5771437"
            },
            {
                "entity": "Autism Network International",
                "confidence_score": "-0.75915617"
            },
            {
                "entity": "Anthropological Survey of India",
                "confidence_score": "-9.779613"
            }
        ]
    },
    {
        "mention": "malir's rifa-i-aam society",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Pakistan",
                "confidence_score": "-2.2003558"
            },
            {
                "entity": "Somalis in Pakistan",
                "confidence_score": "-11.2595625"
            },
            {
                "entity": "Maliha Ali Asghar Khan",
                "confidence_score": "-11.787587"
            }
        ]
    },
    {
        "mention": "farhan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Farhan",
                "confidence_score": "0.2850706"
            },
            {
                "entity": "Farhan Saeed",
                "confidence_score": "-3.8176184"
            },
            {
                "entity": "Farhan Khan (actor)",
                "confidence_score": "-8.490853"
            }
        ]
    },
    {
        "mention": "pehlwan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Pehlwani",
                "confidence_score": "-1.5431557"
            },
            {
                "entity": "Pehlwan Goth",
                "confidence_score": "-2.354359"
            },
            {
                "entity": "Hyderabadi Pahalwan",
                "confidence_score": "-7.5133014"
            }
        ]
    },
    {
        "mention": "farhan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Farhan",
                "confidence_score": "0.2850706"
            },
            {
                "entity": "Farhan Saeed",
                "confidence_score": "-3.8176184"
            },
            {
                "entity": "Farhan Khan (actor)",
                "confidence_score": "-8.490853"
            }
        ]
    },
    {
        "mention": "rangers",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Pakistan Rangers",
                "confidence_score": "-0.046780016"
            },
            {
                "entity": "Rangers F.C.",
                "confidence_score": "-5.3806634"
            },
            {
                "entity": "Pakistan Tehreek-e-Insaf",
                "confidence_score": "-9.4207945"
            }
        ]
    },
    {
        "mention": "the gold coast",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Gold Coast, Queensland",
                "confidence_score": "7.8153176"
            },
            {
                "entity": "Gold Coast Suns",
                "confidence_score": "-5.0646243"
            },
            {
                "entity": "Gold Coast Titans",
                "confidence_score": "-5.4521875"
            }
        ]
    },
    {
        "mention": "owen eagles",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Owen Stephens",
                "confidence_score": "-9.763269"
            },
            {
                "entity": "Owen Trout",
                "confidence_score": "-11.297881"
            },
            {
                "entity": "Albion Park-Oak Flats Eagles",
                "confidence_score": "-11.441521"
            }
        ]
    },
    {
        "mention": "ski express nz/japan",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Kenichi Yamamoto (skier)",
                "confidence_score": "-8.078839"
            },
            {
                "entity": "Holiday Inn Express",
                "confidence_score": "-10.538987"
            },
            {
                "entity": "Kiwi Travel International Airlines",
                "confidence_score": "-10.751096"
            }
        ]
    },
    {
        "mention": "kirra holidays",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "KILROY International",
                "confidence_score": "-11.138485"
            },
            {
                "entity": "Kiwi Travel International Airlines",
                "confidence_score": "-11.437175"
            },
            {
                "entity": "Travel Adventures",
                "confidence_score": "-11.641141"
            }
        ]
    },
    {
        "mention": "toni portmann",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Toni Ko",
                "confidence_score": "-9.871561"
            },
            {
                "entity": "Christopher Portman, 10th Viscount Portman",
                "confidence_score": "-11.554369"
            },
            {
                "entity": "Franz Portmann",
                "confidence_score": "-11.79573"
            }
        ]
    },
    {
        "mention": "dhisco inc.",
        "type": "ORG",
        "predictions": [
            {
                "entity": "DSG International Ltd.",
                "confidence_score": "-11.146859"
            },
            {
                "entity": "Chemische Industrie Uithoorn",
                "confidence_score": "-11.551819"
            },
            {
                "entity": "Citicorp Railmark Inc. (Citirail)",
                "confidence_score": "-11.965083"
            }
        ]
    },
    {
        "mention": "karishma tulsidas",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Karishma Naina Sharma",
                "confidence_score": "-10.62347"
            },
            {
                "entity": "Karishma Tanna",
                "confidence_score": "-10.828245"
            },
            {
                "entity": "Karishma Sharma",
                "confidence_score": "-11.256057"
            }
        ]
    },
    {
        "mention": "jewelfest singapore",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Jewel Changi Airport",
                "confidence_score": "-9.446652"
            },
            {
                "entity": "Jewellery Design and Management International School",
                "confidence_score": "-10.13782"
            },
            {
                "entity": "World Jewellery Confederation",
                "confidence_score": "-10.164753"
            }
        ]
    },
    {
        "mention": "the singapore jewellery & gem fair",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Singapore Jewellery Design Award",
                "confidence_score": "-9.138576"
            },
            {
                "entity": "Jewellery Design and Management International School",
                "confidence_score": "-10.058896"
            },
            {
                "entity": "Hong Kong Brands and Products Expo",
                "confidence_score": "-11.716703"
            }
        ]
    },
    {
        "mention": "the singapore international jewelry expo",
        "type": "ORG",
        "predictions": [
            {
                "entity": "World Jewellery Confederation",
                "confidence_score": "-9.491719"
            },
            {
                "entity": "Singapore Jewellery Design Award",
                "confidence_score": "-9.633056"
            },
            {
                "entity": "Jewellery Design and Management International School",
                "confidence_score": "-10.488417"
            }
        ]
    },
    {
        "mention": "singapore tatler jewels & time",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Straits Times",
                "confidence_score": "-8.812173"
            },
            {
                "entity": "Singapore Standard Time",
                "confidence_score": "-10.116788"
            },
            {
                "entity": "Straits Times Index",
                "confidence_score": "-10.563089"
            }
        ]
    },
    {
        "mention": "michael koh",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Michael Kum",
                "confidence_score": "-8.994506"
            },
            {
                "entity": "Michael Kors",
                "confidence_score": "-10.809246"
            },
            {
                "entity": "Michael Hui",
                "confidence_score": "-10.841018"
            }
        ]
    },
    {
        "mention": "caratell",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kimi ni Aitaku Naru Kara",
                "confidence_score": "-11.640695"
            },
            {
                "entity": "Kabaddi at the 2018 Asian Games \u2013 Men",
                "confidence_score": "-12.042856"
            },
            {
                "entity": "Paragliding at the 2018 Asian Games \u2013 Women's individual accuracy",
                "confidence_score": "-12.176884"
            }
        ]
    },
    {
        "mention": "europe",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Europe",
                "confidence_score": "4.8904595"
            },
            {
                "entity": "Europe (band)",
                "confidence_score": "-8.405712"
            },
            {
                "entity": "2017\u201318 Biathlon World Cup \u2013 Nation Women",
                "confidence_score": "-12.102094"
            }
        ]
    },
    {
        "mention": "belinda chua",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Paige Chua",
                "confidence_score": "-9.954881"
            },
            {
                "entity": "Jamie Chua",
                "confidence_score": "-10.29943"
            },
            {
                "entity": "Belinda Lee Xin Yu",
                "confidence_score": "-11.180914"
            }
        ]
    },
    {
        "mention": "flower diamond",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Erica Diamond",
                "confidence_score": "-11.400855"
            },
            {
                "entity": "Anna Mercedes Morris",
                "confidence_score": "-11.504778"
            },
            {
                "entity": "RD (girl group)",
                "confidence_score": "-12.059081"
            }
        ]
    },
    {
        "mention": "belinda",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Belinda",
                "confidence_score": "-5.327825"
            },
            {
                "entity": "Belinda Peregr\u00edn",
                "confidence_score": "-10.0350685"
            },
            {
                "entity": "Belinda Chapple",
                "confidence_score": "-10.390681"
            }
        ]
    },
    {
        "mention": "ritu aurora",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Aurora",
                "confidence_score": "-7.579847"
            },
            {
                "entity": "Amitabh Aurora",
                "confidence_score": "-11.89929"
            },
            {
                "entity": "Rilana Erades",
                "confidence_score": "-11.990048"
            }
        ]
    },
    {
        "mention": "jaipore jewels",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jewel BEM Hunter Lime",
                "confidence_score": "-9.781619"
            },
            {
                "entity": "Jaimie Fuller",
                "confidence_score": "-11.525582"
            },
            {
                "entity": "2012 PSA World Series Finals",
                "confidence_score": "-11.893926"
            }
        ]
    },
    {
        "mention": "jaipore jewels",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jewel BEM Hunter Lime",
                "confidence_score": "-9.781619"
            },
            {
                "entity": "Jaimie Fuller",
                "confidence_score": "-11.525582"
            },
            {
                "entity": "2012 PSA World Series Finals",
                "confidence_score": "-11.893926"
            }
        ]
    },
    {
        "mention": "jaipore",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Chandrapore",
                "confidence_score": "-8.100399"
            },
            {
                "entity": "Jaitly",
                "confidence_score": "-11.211272"
            },
            {
                "entity": "Jaignes",
                "confidence_score": "-11.449147"
            }
        ]
    },
    {
        "mention": "the patron jeweller",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Pearl Source",
                "confidence_score": "-11.3287"
            },
            {
                "entity": "The Entertainer (discount publisher)",
                "confidence_score": "-11.606862"
            },
            {
                "entity": "Painting and Patronage",
                "confidence_score": "-12.189455"
            }
        ]
    },
    {
        "mention": "the islamic fashion festival",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Islamic fashion",
                "confidence_score": "-8.134654"
            },
            {
                "entity": "Asia Fashion Exchange",
                "confidence_score": "-10.600425"
            },
            {
                "entity": "Dhaka Fashion Week",
                "confidence_score": "-12.148165"
            }
        ]
    },
    {
        "mention": "goh shuet-li",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Shueh-li Ong",
                "confidence_score": "-10.876456"
            },
            {
                "entity": "Goh V Shem",
                "confidence_score": "-11.590011"
            },
            {
                "entity": "Goh Sze Fei",
                "confidence_score": "-12.295375"
            }
        ]
    },
    {
        "mention": "la putri",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "La (Javanese)",
                "confidence_score": "-11.232955"
            },
            {
                "entity": "Vania Larissa",
                "confidence_score": "-11.868726"
            },
            {
                "entity": "Putri Ayu Silaen",
                "confidence_score": "-11.989604"
            }
        ]
    },
    {
        "mention": "marilyn tan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Vincent Tan",
                "confidence_score": "-9.297695"
            },
            {
                "entity": "Ling Tan",
                "confidence_score": "-9.334699"
            },
            {
                "entity": "Josephine Tan",
                "confidence_score": "-10.734073"
            }
        ]
    },
    {
        "mention": "marilyn tan jewellery",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jewellery Design and Management International School",
                "confidence_score": "-9.426469"
            },
            {
                "entity": "On Cheong Jewellery",
                "confidence_score": "-9.688912"
            },
            {
                "entity": "Straits Chinese Jewellery Museum",
                "confidence_score": "-10.263551"
            }
        ]
    },
    {
        "mention": "marilyn",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Marilyn Monroe",
                "confidence_score": "-5.549219"
            },
            {
                "entity": "Marilyn Miller",
                "confidence_score": "-9.76688"
            },
            {
                "entity": "Marilyn Villamayor",
                "confidence_score": "-10.732512"
            }
        ]
    },
    {
        "mention": "sara taseer",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Taman Tasik Prima",
                "confidence_score": "-10.51924"
            },
            {
                "entity": "Salmaan Taseer",
                "confidence_score": "-10.613047"
            },
            {
                "entity": "Taseer Badar",
                "confidence_score": "-11.267087"
            }
        ]
    },
    {
        "mention": "sara taseer",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Taman Tasik Prima",
                "confidence_score": "-10.51924"
            },
            {
                "entity": "Salmaan Taseer",
                "confidence_score": "-10.613047"
            },
            {
                "entity": "Taseer Badar",
                "confidence_score": "-11.267087"
            }
        ]
    },
    {
        "mention": "sara taseer",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Taman Tasik Prima",
                "confidence_score": "-10.51924"
            },
            {
                "entity": "Salmaan Taseer",
                "confidence_score": "-10.613047"
            },
            {
                "entity": "Taseer Badar",
                "confidence_score": "-11.267087"
            }
        ]
    },
    {
        "mention": "simone ng",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Simone Singh",
                "confidence_score": "-11.119492"
            },
            {
                "entity": "Simone Egeriis",
                "confidence_score": "-11.126487"
            },
            {
                "entity": "Simone Forbes",
                "confidence_score": "-11.193332"
            }
        ]
    },
    {
        "mention": "nayantara dhillon",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Navneet Kaur Dhillon",
                "confidence_score": "-7.8974285"
            },
            {
                "entity": "Poonam Dhillon",
                "confidence_score": "-8.014462"
            },
            {
                "entity": "Navjeet Dhillon",
                "confidence_score": "-9.854204"
            }
        ]
    },
    {
        "mention": "michelle buxani",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Michelle Brooke-Marciniak",
                "confidence_score": "-11.462352"
            },
            {
                "entity": "Michelle Lee (actress)",
                "confidence_score": "-11.80247"
            },
            {
                "entity": "Michelle DuBarry",
                "confidence_score": "-11.863744"
            }
        ]
    },
    {
        "mention": "temptations",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "The Temptations",
                "confidence_score": "-1.466572"
            },
            {
                "entity": "Temptation",
                "confidence_score": "-4.1608887"
            },
            {
                "entity": "Soapstars",
                "confidence_score": "-9.83251"
            }
        ]
    },
    {
        "mention": "vihari poddar",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Abhishek Poddar",
                "confidence_score": "-8.586196"
            },
            {
                "entity": "Vidhya Mohan",
                "confidence_score": "-11.620339"
            },
            {
                "entity": "List of Rosy Business characters",
                "confidence_score": "-11.957194"
            }
        ]
    },
    {
        "mention": "vihari jewels",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jewels (mixed martial arts)",
                "confidence_score": "-10.95498"
            },
            {
                "entity": "Heartsdales",
                "confidence_score": "-11.722877"
            },
            {
                "entity": "Run the Jewels",
                "confidence_score": "-11.767944"
            }
        ]
    },
    {
        "mention": "gia",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Gia Johnson",
                "confidence_score": "-5.600068"
            },
            {
                "entity": "Gia (name)",
                "confidence_score": "-10.0890465"
            },
            {
                "entity": "Gia Gunn",
                "confidence_score": "-10.697341"
            }
        ]
    },
    {
        "mention": "jessie foo",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jessie Cave",
                "confidence_score": "-10.932743"
            },
            {
                "entity": "Jessie Ware",
                "confidence_score": "-11.295608"
            },
            {
                "entity": "Jessie Zhou",
                "confidence_score": "-11.372216"
            }
        ]
    },
    {
        "mention": "yuli inc \u2013 fine jewellery",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Xiaomi",
                "confidence_score": "-8.274638"
            },
            {
                "entity": "Iconiq",
                "confidence_score": "-11.227624"
            },
            {
                "entity": "Yumi",
                "confidence_score": "-11.621428"
            }
        ]
    },
    {
        "mention": "jessie",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jessie Chung",
                "confidence_score": "-8.458671"
            },
            {
                "entity": "Jessie J",
                "confidence_score": "-9.063708"
            },
            {
                "entity": "Jessie Ware",
                "confidence_score": "-10.382716"
            }
        ]
    },
    {
        "mention": "jewellery",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jewellery",
                "confidence_score": "8.436512"
            },
            {
                "entity": "Jewellery design",
                "confidence_score": "-4.8921857"
            },
            {
                "entity": "Jewelry (group)",
                "confidence_score": "-10.586587"
            }
        ]
    },
    {
        "mention": "sara taseer",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Taman Tasik Prima",
                "confidence_score": "-10.51924"
            },
            {
                "entity": "Salmaan Taseer",
                "confidence_score": "-10.613047"
            },
            {
                "entity": "Taseer Badar",
                "confidence_score": "-11.267087"
            }
        ]
    },
    {
        "mention": "jessie foo",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jessie Cave",
                "confidence_score": "-10.932743"
            },
            {
                "entity": "Jessie Ware",
                "confidence_score": "-11.295608"
            },
            {
                "entity": "Jessie Zhou",
                "confidence_score": "-11.372216"
            }
        ]
    },
    {
        "mention": "belinda chua",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Paige Chua",
                "confidence_score": "-9.954881"
            },
            {
                "entity": "Jamie Chua",
                "confidence_score": "-10.29943"
            },
            {
                "entity": "Belinda Lee Xin Yu",
                "confidence_score": "-11.180914"
            }
        ]
    },
    {
        "mention": "michael koh",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Michael Kum",
                "confidence_score": "-8.994506"
            },
            {
                "entity": "Michael Kors",
                "confidence_score": "-10.809246"
            },
            {
                "entity": "Michael Hui",
                "confidence_score": "-10.841018"
            }
        ]
    },
    {
        "mention": "nayantara dhillon",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Navneet Kaur Dhillon",
                "confidence_score": "-7.8974285"
            },
            {
                "entity": "Poonam Dhillon",
                "confidence_score": "-8.014462"
            },
            {
                "entity": "Navjeet Dhillon",
                "confidence_score": "-9.854204"
            }
        ]
    },
    {
        "mention": "michelle buxani",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Michelle Brooke-Marciniak",
                "confidence_score": "-11.462352"
            },
            {
                "entity": "Michelle Lee (actress)",
                "confidence_score": "-11.80247"
            },
            {
                "entity": "Michelle DuBarry",
                "confidence_score": "-11.863744"
            }
        ]
    },
    {
        "mention": "friends of singapore tatler",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Malaysian Insider",
                "confidence_score": "-10.598672"
            },
            {
                "entity": "List of National University of Singapore people",
                "confidence_score": "-11.617703"
            },
            {
                "entity": "Citibank International Personal Bank Singapore",
                "confidence_score": "-11.621086"
            }
        ]
    },
    {
        "mention": "vihari poddar",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Abhishek Poddar",
                "confidence_score": "-8.586196"
            },
            {
                "entity": "Vidhya Mohan",
                "confidence_score": "-11.620339"
            },
            {
                "entity": "List of Rosy Business characters",
                "confidence_score": "-11.957194"
            }
        ]
    },
    {
        "mention": "marilyn tan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Vincent Tan",
                "confidence_score": "-9.297695"
            },
            {
                "entity": "Ling Tan",
                "confidence_score": "-9.334699"
            },
            {
                "entity": "Josephine Tan",
                "confidence_score": "-10.734073"
            }
        ]
    },
    {
        "mention": "ritu aurora",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Aurora",
                "confidence_score": "-7.579847"
            },
            {
                "entity": "Amitabh Aurora",
                "confidence_score": "-11.89929"
            },
            {
                "entity": "Rilana Erades",
                "confidence_score": "-11.990048"
            }
        ]
    },
    {
        "mention": "simone ng",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Simone Singh",
                "confidence_score": "-11.119492"
            },
            {
                "entity": "Simone Egeriis",
                "confidence_score": "-11.126487"
            },
            {
                "entity": "Simone Forbes",
                "confidence_score": "-11.193332"
            }
        ]
    },
    {
        "mention": "goh shuet-li",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Shueh-li Ong",
                "confidence_score": "-10.876456"
            },
            {
                "entity": "Goh V Shem",
                "confidence_score": "-11.590011"
            },
            {
                "entity": "Goh Sze Fei",
                "confidence_score": "-12.295375"
            }
        ]
    },
    {
        "mention": "flawless",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Flawless (dance troupe)",
                "confidence_score": "-7.6733947"
            },
            {
                "entity": "Flawless (Beyonc\u00e9 song)",
                "confidence_score": "-10.311395"
            },
            {
                "entity": "Flawless (The Ones song)",
                "confidence_score": "-10.448258"
            }
        ]
    },
    {
        "mention": "bulgari",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bulgari",
                "confidence_score": "8.936749"
            },
            {
                "entity": "Bulgars",
                "confidence_score": "-2.9057195"
            },
            {
                "entity": "Brigitta Bulgari",
                "confidence_score": "-6.6097717"
            }
        ]
    },
    {
        "mention": "detroit",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Detroit",
                "confidence_score": "9.890772"
            },
            {
                "entity": "Metro Detroit",
                "confidence_score": "-2.2366202"
            },
            {
                "entity": "Detroit (disambiguation)",
                "confidence_score": "-11.025088"
            }
        ]
    },
    {
        "mention": "atlantic",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Atlantic Ocean",
                "confidence_score": "4.698329"
            },
            {
                "entity": "Atlantic hurricane",
                "confidence_score": "-3.7984896"
            },
            {
                "entity": "The Atlantic",
                "confidence_score": "-4.974521"
            }
        ]
    },
    {
        "mention": "the east coast",
        "type": "LOC",
        "predictions": [
            {
                "entity": "East Coast of the United States",
                "confidence_score": "-2.0340664"
            },
            {
                "entity": "Gulf Coast of the United States",
                "confidence_score": "-8.368827"
            },
            {
                "entity": "Oregon Coast",
                "confidence_score": "-9.539765"
            }
        ]
    },
    {
        "mention": "the great lakes",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Great Lakes",
                "confidence_score": "1.6418152"
            },
            {
                "entity": "Great Lakes region",
                "confidence_score": "-4.706344"
            },
            {
                "entity": "Great Lakes Megalopolis",
                "confidence_score": "-6.8674603"
            }
        ]
    },
    {
        "mention": "lake huron",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Lake Huron",
                "confidence_score": "9.36548"
            },
            {
                "entity": "Lake Michigan\u2013Huron",
                "confidence_score": "-4.702829"
            },
            {
                "entity": "Lake Michigan",
                "confidence_score": "-5.1608253"
            }
        ]
    },
    {
        "mention": "showers",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Shower",
                "confidence_score": "-1.2876487"
            },
            {
                "entity": "Water polo at the 2011 World Aquatics Championships \u2013 Women's tournament",
                "confidence_score": "-11.43439"
            },
            {
                "entity": "Matches of Polish men's volleyball national team conducted by St\u00e9phane Antiga",
                "confidence_score": "-12.688642"
            }
        ]
    },
    {
        "mention": "desperately",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Desperately (Don Williams song)",
                "confidence_score": "-11.312842"
            },
            {
                "entity": "Desperately",
                "confidence_score": "-11.710676"
            },
            {
                "entity": "Desperately Wanting",
                "confidence_score": "-11.911909"
            }
        ]
    },
    {
        "mention": "lisa herbold",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Lisa Herbold",
                "confidence_score": "12.056472"
            },
            {
                "entity": "Patricia L. Herbold",
                "confidence_score": "-10.407132"
            },
            {
                "entity": "Lisa Bender",
                "confidence_score": "-10.837328"
            }
        ]
    },
    {
        "mention": "councilmember licata's office",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Nick Licata (politician)",
                "confidence_score": "-8.631442"
            },
            {
                "entity": "2016 Lipa local elections",
                "confidence_score": "-12.324182"
            },
            {
                "entity": "2019 Lipa local elections",
                "confidence_score": "-12.391976"
            }
        ]
    },
    {
        "mention": "councilmember sawant's office",
        "type": "ORG",
        "predictions": [
            {
                "entity": "List of Nunavut Legislative Assemblies",
                "confidence_score": "-11.901475"
            },
            {
                "entity": "R\u00e4tsepa, P\u00f5hja-P\u00e4rnumaa Parish",
                "confidence_score": "-12.386149"
            },
            {
                "entity": "2015 Gilgit Baltistan Assembly election",
                "confidence_score": "-12.415324"
            }
        ]
    },
    {
        "mention": "dana robinson slote",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "List of census-designated places in North Dakota",
                "confidence_score": "-11.51553"
            },
            {
                "entity": "Nosimo Balindlela",
                "confidence_score": "-12.208592"
            },
            {
                "entity": "Jane Censoria Cajes\u2013Yap",
                "confidence_score": "-12.480653"
            }
        ]
    },
    {
        "mention": "council  committee",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Committee",
                "confidence_score": "-1.3599354"
            },
            {
                "entity": "Council of the District of Columbia",
                "confidence_score": "-8.527604"
            },
            {
                "entity": "Planning committee",
                "confidence_score": "-8.720871"
            }
        ]
    },
    {
        "mention": "limiting rent increases",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Brooke Amendment",
                "confidence_score": "-10.612016"
            },
            {
                "entity": "Deferment rate",
                "confidence_score": "-10.727611"
            },
            {
                "entity": "Seattle's minimum wage ordinance",
                "confidence_score": "-11.256536"
            }
        ]
    },
    {
        "mention": "seattle - council's housing affordability, human services and economic  resiliency committee",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Seattle",
                "confidence_score": "-2.8440382"
            },
            {
                "entity": "Seattle City Council",
                "confidence_score": "-3.6341088"
            },
            {
                "entity": "Seattle Metropolitan Chamber of Commerce",
                "confidence_score": "-9.857411"
            }
        ]
    },
    {
        "mention": "councilmembers",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Councillor",
                "confidence_score": "-2.4868972"
            },
            {
                "entity": "List of members of the Council of Representatives of Iraq, 2005\u20132010",
                "confidence_score": "-11.702203"
            },
            {
                "entity": "Members of the Constituent Assembly of Namibia",
                "confidence_score": "-12.356446"
            }
        ]
    },
    {
        "mention": "kshama sawant",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kshama Sawant",
                "confidence_score": "12.244417"
            },
            {
                "entity": "Kassiane Asasumasu",
                "confidence_score": "-11.764058"
            },
            {
                "entity": "Maysara Abu Hamdiya",
                "confidence_score": "-12.496794"
            }
        ]
    },
    {
        "mention": "nick licata",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Nick Licata (politician)",
                "confidence_score": "11.648968"
            },
            {
                "entity": "Nick Licata (mobster)",
                "confidence_score": "-6.2948675"
            },
            {
                "entity": "Nick Liotta",
                "confidence_score": "-10.208274"
            }
        ]
    },
    {
        "mention": "washington state",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Washington (state)",
                "confidence_score": "5.5801234"
            },
            {
                "entity": "Washington State University",
                "confidence_score": "-3.665579"
            },
            {
                "entity": "Washington State",
                "confidence_score": "-10.781856"
            }
        ]
    },
    {
        "mention": "councilmembers  sawant",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kshama Sawant",
                "confidence_score": "-2.0385404"
            },
            {
                "entity": "Federal Way City Council",
                "confidence_score": "-9.416805"
            },
            {
                "entity": "Paragliding at the 2018 Asian Games \u2013 Women's individual accuracy",
                "confidence_score": "-12.36328"
            }
        ]
    },
    {
        "mention": "licata",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Nick Licata (politician)",
                "confidence_score": "5.988609"
            },
            {
                "entity": "Licata",
                "confidence_score": "0.3136353"
            },
            {
                "entity": "Tita Aida",
                "confidence_score": "-11.912841"
            }
        ]
    },
    {
        "mention": "council",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Council",
                "confidence_score": "4.339744"
            },
            {
                "entity": "Council of the Church",
                "confidence_score": "-5.837897"
            },
            {
                "entity": "Navajo Nation Council",
                "confidence_score": "-7.5006175"
            }
        ]
    },
    {
        "mention": "councilmembers licata",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Nick Licata (politician)",
                "confidence_score": "0.46282932"
            },
            {
                "entity": "Council of the District of Columbia Period 14",
                "confidence_score": "-11.752452"
            },
            {
                "entity": "Council of the District of Columbia Period 15",
                "confidence_score": "-11.7538595"
            }
        ]
    },
    {
        "mention": "full council",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Seattle City Council",
                "confidence_score": "-1.451632"
            },
            {
                "entity": "Chief Seattle Council",
                "confidence_score": "-8.199632"
            },
            {
                "entity": "Council of the Church",
                "confidence_score": "-8.758779"
            }
        ]
    },
    {
        "mention": "councilmembers  licata",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Nick Licata (politician)",
                "confidence_score": "0.43720028"
            },
            {
                "entity": "List of members of the Louisville Metro Council",
                "confidence_score": "-12.014221"
            },
            {
                "entity": "Members of the 5th National Assembly for Wales",
                "confidence_score": "-12.184952"
            }
        ]
    },
    {
        "mention": "the council's housing affordability, human services",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Seattle City Council",
                "confidence_score": "-4.708395"
            },
            {
                "entity": "Center for Council",
                "confidence_score": "-10.711799"
            },
            {
                "entity": "Council for Affordable Quality Healthcare",
                "confidence_score": "-11.20251"
            }
        ]
    },
    {
        "mention": "economic resiliency committee",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Community resilience",
                "confidence_score": "-9.28616"
            },
            {
                "entity": "Sustainable Seattle",
                "confidence_score": "-9.406857"
            },
            {
                "entity": "Institute for Local Self-Reliance",
                "confidence_score": "-10.592935"
            }
        ]
    },
    {
        "mention": "second floor",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Stairs",
                "confidence_score": "-7.641441"
            },
            {
                "entity": "Lower Queen Anne, Seattle",
                "confidence_score": "-9.8298235"
            },
            {
                "entity": "2+U",
                "confidence_score": "-10.788342"
            }
        ]
    },
    {
        "mention": "nick licata",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Nick Licata (politician)",
                "confidence_score": "11.648968"
            },
            {
                "entity": "Nick Licata (mobster)",
                "confidence_score": "-6.2948675"
            },
            {
                "entity": "Nick Liotta",
                "confidence_score": "-10.208274"
            }
        ]
    },
    {
        "mention": "the council\u2019s housing affordability, human services",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Housing and Community Development Act of 1992",
                "confidence_score": "-10.258757"
            },
            {
                "entity": "Housing and Community Development Act of 1987",
                "confidence_score": "-10.819142"
            },
            {
                "entity": "2010 New Orleans city council election",
                "confidence_score": "-11.906914"
            }
        ]
    },
    {
        "mention": "arya stark",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Arya Stark",
                "confidence_score": "10.690959"
            },
            {
                "entity": "Ned Stark",
                "confidence_score": "-4.608479"
            },
            {
                "entity": "Robb Stark",
                "confidence_score": "-6.157898"
            }
        ]
    },
    {
        "mention": "maisie williams",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Maisie Williams",
                "confidence_score": "12.085702"
            },
            {
                "entity": "Maizie Williams",
                "confidence_score": "-7.2570853"
            },
            {
                "entity": "Maiya Williams",
                "confidence_score": "-10.620336"
            }
        ]
    },
    {
        "mention": "youtube",
        "type": "ORG",
        "predictions": [
            {
                "entity": "YouTube",
                "confidence_score": "10.731659"
            },
            {
                "entity": "YouTuber",
                "confidence_score": "-6.747682"
            },
            {
                "entity": "YouTube (channel)",
                "confidence_score": "-7.647054"
            }
        ]
    },
    {
        "mention": "maisie",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Maisie Williams",
                "confidence_score": "8.9909"
            },
            {
                "entity": "Maisie",
                "confidence_score": "1.4028592"
            },
            {
                "entity": "Maisie Smith",
                "confidence_score": "-5.2384276"
            }
        ]
    },
    {
        "mention": "youtube",
        "type": "ORG",
        "predictions": [
            {
                "entity": "YouTube",
                "confidence_score": "10.731659"
            },
            {
                "entity": "YouTuber",
                "confidence_score": "-6.747682"
            },
            {
                "entity": "YouTube (channel)",
                "confidence_score": "-7.647054"
            }
        ]
    },
    {
        "mention": "youtube",
        "type": "ORG",
        "predictions": [
            {
                "entity": "YouTube",
                "confidence_score": "10.731659"
            },
            {
                "entity": "YouTuber",
                "confidence_score": "-6.747682"
            },
            {
                "entity": "YouTube (channel)",
                "confidence_score": "-7.647054"
            }
        ]
    },
    {
        "mention": "ms williams",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Venus Williams",
                "confidence_score": "-5.8795123"
            },
            {
                "entity": "Jessica Williams (actress)",
                "confidence_score": "-8.328695"
            },
            {
                "entity": "Jessica Williams",
                "confidence_score": "-11.973672"
            }
        ]
    },
    {
        "mention": "youtube",
        "type": "ORG",
        "predictions": [
            {
                "entity": "YouTube",
                "confidence_score": "10.731659"
            },
            {
                "entity": "YouTuber",
                "confidence_score": "-6.747682"
            },
            {
                "entity": "YouTube (channel)",
                "confidence_score": "-7.647054"
            }
        ]
    },
    {
        "mention": "maisie williams/youtube",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Maisie Williams",
                "confidence_score": "4.3699503"
            },
            {
                "entity": "Maizie Williams",
                "confidence_score": "-3.9681423"
            },
            {
                "entity": "Maisie Smith",
                "confidence_score": "-9.647899"
            }
        ]
    },
    {
        "mention": "sophie",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sophie, Countess of Wessex",
                "confidence_score": "-6.304443"
            },
            {
                "entity": "Sophie",
                "confidence_score": "-9.127511"
            },
            {
                "entity": "Sophie Simnett",
                "confidence_score": "-9.567245"
            }
        ]
    },
    {
        "mention": "maisie\u2019s",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Maisie Williams",
                "confidence_score": "-0.7427547"
            },
            {
                "entity": "Maisie",
                "confidence_score": "-1.1856271"
            },
            {
                "entity": "Maisie Smith",
                "confidence_score": "-9.251408"
            }
        ]
    },
    {
        "mention": "maisie williams",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Maisie Williams",
                "confidence_score": "12.085702"
            },
            {
                "entity": "Maizie Williams",
                "confidence_score": "-7.2570853"
            },
            {
                "entity": "Maiya Williams",
                "confidence_score": "-10.620336"
            }
        ]
    },
    {
        "mention": "@maisie_williams",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Maisie Williams",
                "confidence_score": "-0.8978305"
            },
            {
                "entity": "Kimi e...",
                "confidence_score": "-10.551426"
            },
            {
                "entity": "Ai Maeda (voice actress)",
                "confidence_score": "-11.595332"
            }
        ]
    },
    {
        "mention": "mughal",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mughal Empire",
                "confidence_score": "6.9409075"
            },
            {
                "entity": "Mughal emperors",
                "confidence_score": "-1.2927883"
            },
            {
                "entity": "Mughal architecture",
                "confidence_score": "-5.8385167"
            }
        ]
    },
    {
        "mention": "the delhi cantonment",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Delhi Cantonment",
                "confidence_score": "0.044629987"
            },
            {
                "entity": "Jalandhar Cantonment",
                "confidence_score": "-7.517198"
            },
            {
                "entity": "Taxila Cantonment",
                "confidence_score": "-10.380227"
            }
        ]
    },
    {
        "mention": "agra",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Agra",
                "confidence_score": "9.960967"
            },
            {
                "entity": "Agra district",
                "confidence_score": "-5.7101855"
            },
            {
                "entity": "Agra, Punjab",
                "confidence_score": "-7.250465"
            }
        ]
    },
    {
        "mention": "jantar mantar",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Jantar Mantar",
                "confidence_score": "9.296149"
            },
            {
                "entity": "Jantar Mantar, New Delhi",
                "confidence_score": "-0.5665424"
            },
            {
                "entity": "Jantar Mantar, Jaipur",
                "confidence_score": "-2.4672198"
            }
        ]
    },
    {
        "mention": "agra",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Agra",
                "confidence_score": "9.960967"
            },
            {
                "entity": "Agra district",
                "confidence_score": "-5.7101855"
            },
            {
                "entity": "Agra, Punjab",
                "confidence_score": "-7.250465"
            }
        ]
    },
    {
        "mention": "mahabharata",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Mahabharata",
                "confidence_score": "10.583041"
            },
            {
                "entity": "The Mahabharata (Narayan book)",
                "confidence_score": "-9.141131"
            },
            {
                "entity": "Mahabharata (disambiguation)",
                "confidence_score": "-10.984652"
            }
        ]
    },
    {
        "mention": "agra",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Agra",
                "confidence_score": "7.208212"
            },
            {
                "entity": "Agra Fort",
                "confidence_score": "-5.529025"
            },
            {
                "entity": "Agra, Punjab",
                "confidence_score": "-6.496818"
            }
        ]
    },
    {
        "mention": "ka mah",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mah",
                "confidence_score": "-8.022498"
            },
            {
                "entity": "Mahesh (name)",
                "confidence_score": "-10.276853"
            },
            {
                "entity": "Kahma",
                "confidence_score": "-11.768374"
            }
        ]
    },
    {
        "mention": "ka mah",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mah",
                "confidence_score": "-8.022498"
            },
            {
                "entity": "Mahesh (name)",
                "confidence_score": "-10.276853"
            },
            {
                "entity": "Kahma",
                "confidence_score": "-11.768374"
            }
        ]
    },
    {
        "mention": "jhandewalan metro station",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jhandewalan metro station",
                "confidence_score": "8.67785"
            },
            {
                "entity": "Jahangirpuri metro station",
                "confidence_score": "-11.044398"
            },
            {
                "entity": "Jhilmil metro station",
                "confidence_score": "-11.189336"
            }
        ]
    },
    {
        "mention": "shah tughlaq",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Firuz Shah Tughlaq",
                "confidence_score": "-0.82747626"
            },
            {
                "entity": "Tughlaq dynasty",
                "confidence_score": "-4.6367216"
            },
            {
                "entity": "Shah Jahan",
                "confidence_score": "-8.991059"
            }
        ]
    },
    {
        "mention": "mahal",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Mahal (palace)",
                "confidence_score": "6.642782"
            },
            {
                "entity": "Jahaz Mahal",
                "confidence_score": "-6.4280477"
            },
            {
                "entity": "Mirza Mahal",
                "confidence_score": "-9.393955"
            }
        ]
    },
    {
        "mention": "mahal",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Mahal (palace)",
                "confidence_score": "6.642782"
            },
            {
                "entity": "Jahaz Mahal",
                "confidence_score": "-6.4280477"
            },
            {
                "entity": "Mirza Mahal",
                "confidence_score": "-9.393955"
            }
        ]
    },
    {
        "mention": "ka mahal",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Taj Mahal",
                "confidence_score": "-8.960652"
            },
            {
                "entity": "Jahaz Mahal",
                "confidence_score": "-10.273468"
            },
            {
                "entity": "Hawa Mahal",
                "confidence_score": "-10.715512"
            }
        ]
    },
    {
        "mention": "mahal",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Mahal (palace)",
                "confidence_score": "6.642782"
            },
            {
                "entity": "Jahaz Mahal",
                "confidence_score": "-6.4280477"
            },
            {
                "entity": "Mirza Mahal",
                "confidence_score": "-9.393955"
            }
        ]
    },
    {
        "mention": "daily mail\u2019s",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Daily Mail",
                "confidence_score": "11.0001955"
            },
            {
                "entity": "Mail",
                "confidence_score": "-1.5973269"
            },
            {
                "entity": "The Daily Telegraph",
                "confidence_score": "-7.545195"
            }
        ]
    },
    {
        "mention": "macmillan",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Harold Macmillan",
                "confidence_score": "-4.0180745"
            },
            {
                "entity": "Macmillan Media",
                "confidence_score": "-4.5724106"
            },
            {
                "entity": "Macmillan Publishers",
                "confidence_score": "-4.957259"
            }
        ]
    },
    {
        "mention": "stanley crook",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Stanley Crook",
                "confidence_score": "7.563377"
            },
            {
                "entity": "Stanley Crooks",
                "confidence_score": "-8.675647"
            },
            {
                "entity": "Mackenzie Crook",
                "confidence_score": "-10.542045"
            }
        ]
    },
    {
        "mention": "charlene harker",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Baker Street Babes",
                "confidence_score": "-11.275212"
            },
            {
                "entity": "Charlene Carruthers",
                "confidence_score": "-11.516325"
            },
            {
                "entity": "Charline Joiner",
                "confidence_score": "-11.655975"
            }
        ]
    },
    {
        "mention": "beryl",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Beryl",
                "confidence_score": "4.0090036"
            },
            {
                "entity": "Beryl the Peril",
                "confidence_score": "0.88032657"
            },
            {
                "entity": "Hurricane Beryl",
                "confidence_score": "-7.003842"
            }
        ]
    },
    {
        "mention": "ellie harker",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ellie Darcey-Alden",
                "confidence_score": "-8.318237"
            },
            {
                "entity": "Ellie Harvie",
                "confidence_score": "-9.352502"
            },
            {
                "entity": "Ellie Haddington",
                "confidence_score": "-9.678484"
            }
        ]
    },
    {
        "mention": "harry weatherald",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Tim Weatherald",
                "confidence_score": "-10.089182"
            },
            {
                "entity": "Jake Weatherald",
                "confidence_score": "-10.182248"
            },
            {
                "entity": "Thomas Milton Weatherald",
                "confidence_score": "-11.648656"
            }
        ]
    },
    {
        "mention": "stanley crook primary school",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Stanley Crook",
                "confidence_score": "-7.3152614"
            },
            {
                "entity": "Stanley Clark School",
                "confidence_score": "-11.7428255"
            },
            {
                "entity": "Stanley Crooks",
                "confidence_score": "-12.044744"
            }
        ]
    },
    {
        "mention": "leon",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Leon Jackson",
                "confidence_score": "-7.4546633"
            },
            {
                "entity": "Leon Ford",
                "confidence_score": "-9.308982"
            },
            {
                "entity": "Leon Davies",
                "confidence_score": "-11.546831"
            }
        ]
    },
    {
        "mention": "luke mason",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Luke Lennox",
                "confidence_score": "-11.141234"
            },
            {
                "entity": "Luke Cain",
                "confidence_score": "-11.22554"
            },
            {
                "entity": "Luke Jackson (singer)",
                "confidence_score": "-11.906255"
            }
        ]
    },
    {
        "mention": "connie herron",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Connie Martinson",
                "confidence_score": "-11.014575"
            },
            {
                "entity": "Connie Smotek",
                "confidence_score": "-11.656192"
            },
            {
                "entity": "Mazola McKerson",
                "confidence_score": "-11.6571865"
            }
        ]
    },
    {
        "mention": "loading",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Loaded (band)",
                "confidence_score": "-7.4567914"
            },
            {
                "entity": "Loading screen",
                "confidence_score": "-9.334056"
            },
            {
                "entity": "2010 Royal International Horse Show",
                "confidence_score": "-11.975203"
            }
        ]
    },
    {
        "mention": "joanne laundy",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Joanne Salley",
                "confidence_score": "-11.110135"
            },
            {
                "entity": "Kathleen Ridder",
                "confidence_score": "-11.4136915"
            },
            {
                "entity": "Joanne Hogg",
                "confidence_score": "-12.099306"
            }
        ]
    },
    {
        "mention": "macmillan",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Macmillan Publishers",
                "confidence_score": "-4.069004"
            },
            {
                "entity": "Macmillan Media",
                "confidence_score": "-4.205039"
            },
            {
                "entity": "John MacMillan",
                "confidence_score": "-9.887166"
            }
        ]
    },
    {
        "mention": "billy row",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Billy Row",
                "confidence_score": "1.4891739"
            },
            {
                "entity": "Billy Bush",
                "confidence_score": "-11.029965"
            },
            {
                "entity": "R\u00e1ndall Row",
                "confidence_score": "-12.091285"
            }
        ]
    },
    {
        "mention": "julie proud",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Julie Berry",
                "confidence_score": "-11.103684"
            },
            {
                "entity": "Come Out with Pride",
                "confidence_score": "-11.233918"
            },
            {
                "entity": "Gary Proud",
                "confidence_score": "-11.92639"
            }
        ]
    },
    {
        "mention": "lisa brewster",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Lisa Joann Thompson",
                "confidence_score": "-11.009149"
            },
            {
                "entity": "Lisa Thompson (politician)",
                "confidence_score": "-11.46158"
            },
            {
                "entity": "Lisa Schultz",
                "confidence_score": "-11.531719"
            }
        ]
    },
    {
        "mention": "donna thompson",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Donna Simpson (internet celebrity)",
                "confidence_score": "-10.867162"
            },
            {
                "entity": "Donna Kellogg",
                "confidence_score": "-11.90902"
            },
            {
                "entity": "Donna Ritchie",
                "confidence_score": "-12.061093"
            }
        ]
    },
    {
        "mention": "leanne harker",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Leanne Battersby",
                "confidence_score": "-9.964695"
            },
            {
                "entity": "Leanne Hall",
                "confidence_score": "-11.004845"
            },
            {
                "entity": "Leanne Callaghan",
                "confidence_score": "-11.457825"
            }
        ]
    },
    {
        "mention": "amy howcroft",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Amy Willerton",
                "confidence_score": "-10.411445"
            },
            {
                "entity": "Amy Howorth",
                "confidence_score": "-11.5355015"
            },
            {
                "entity": "Amy Howden-Chapman",
                "confidence_score": "-11.911824"
            }
        ]
    },
    {
        "mention": "john thompson",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "John Thompson (sociologist)",
                "confidence_score": "-2.5995758"
            },
            {
                "entity": "John Thompson (boxer)",
                "confidence_score": "-7.48679"
            },
            {
                "entity": "John Thompson (footballer, born 1981)",
                "confidence_score": "-8.616793"
            }
        ]
    },
    {
        "mention": "jane herron",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jane Danson",
                "confidence_score": "-7.5946956"
            },
            {
                "entity": "Jane Aronson",
                "confidence_score": "-11.157407"
            },
            {
                "entity": "Jane Watson",
                "confidence_score": "-11.725734"
            }
        ]
    },
    {
        "mention": "baltimore orioles",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Baltimore Orioles",
                "confidence_score": "8.931406"
            },
            {
                "entity": "2004 Baltimore Orioles season",
                "confidence_score": "-6.6319423"
            },
            {
                "entity": "1981 Baltimore Orioles season",
                "confidence_score": "-8.844806"
            }
        ]
    },
    {
        "mention": "manny machado",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Manny Machado",
                "confidence_score": "9.632159"
            },
            {
                "entity": "Roger Machado (baseball)",
                "confidence_score": "-10.173802"
            },
            {
                "entity": "Robert Machado",
                "confidence_score": "-10.21009"
            }
        ]
    },
    {
        "mention": "washington nationals",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Washington Nationals",
                "confidence_score": "8.68574"
            },
            {
                "entity": "Washington Nationals (NA)",
                "confidence_score": "-4.452497"
            },
            {
                "entity": "2010 Washington Nationals season",
                "confidence_score": "-5.700689"
            }
        ]
    },
    {
        "mention": "jonathan papelbon",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jonathan Papelbon",
                "confidence_score": "9.717675"
            },
            {
                "entity": "Jonathan Albon",
                "confidence_score": "-12.016818"
            },
            {
                "entity": "Jonathan Paiement",
                "confidence_score": "-12.350949"
            }
        ]
    },
    {
        "mention": "max scherzer",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Max Scherzer",
                "confidence_score": "9.35327"
            },
            {
                "entity": "Max Scherr",
                "confidence_score": "-11.017763"
            },
            {
                "entity": "Max Scharping",
                "confidence_score": "-11.574894"
            }
        ]
    },
    {
        "mention": "papelbon",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jonathan Papelbon",
                "confidence_score": "-4.295975"
            },
            {
                "entity": "2011 PSA World Series Finals",
                "confidence_score": "-11.509012"
            },
            {
                "entity": "2012 TEAN International",
                "confidence_score": "-11.956001"
            }
        ]
    },
    {
        "mention": "papelbon",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jonathan Papelbon",
                "confidence_score": "-4.295975"
            },
            {
                "entity": "2011 PSA World Series Finals",
                "confidence_score": "-11.509012"
            },
            {
                "entity": "2012 TEAN International",
                "confidence_score": "-11.956001"
            }
        ]
    },
    {
        "mention": "reuters",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Reuters",
                "confidence_score": "9.007746"
            },
            {
                "entity": "Thomson Reuters",
                "confidence_score": "-3.8036516"
            },
            {
                "entity": "Reuters Group",
                "confidence_score": "-5.307069"
            }
        ]
    },
    {
        "mention": "the u.s. federal reserve",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Federal Reserve",
                "confidence_score": "0.58824587"
            },
            {
                "entity": "Federal Reserve Bank",
                "confidence_score": "-5.760767"
            },
            {
                "entity": "Federal Reserve Bank of New York",
                "confidence_score": "-8.100665"
            }
        ]
    },
    {
        "mention": "fed",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Federal Reserve",
                "confidence_score": "8.06001"
            },
            {
                "entity": "Fed model",
                "confidence_score": "-3.539654"
            },
            {
                "entity": "Federal Reserve Bank",
                "confidence_score": "-8.167174"
            }
        ]
    },
    {
        "mention": "csi300",
        "type": "ORG",
        "predictions": [
            {
                "entity": "ISO 9000",
                "confidence_score": "-9.439532"
            },
            {
                "entity": "ISO 14000",
                "confidence_score": "-9.904124"
            },
            {
                "entity": "CSI 300 Index",
                "confidence_score": "-10.714245"
            }
        ]
    },
    {
        "mention": "gmt",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Greenwich Mean Time",
                "confidence_score": "5.4173503"
            },
            {
                "entity": "GMT Records",
                "confidence_score": "-9.528771"
            },
            {
                "entity": "Graduate Assessment Test",
                "confidence_score": "-12.494129"
            }
        ]
    },
    {
        "mention": "the shanghai composite index",
        "type": "ORG",
        "predictions": [
            {
                "entity": "SSE Composite Index",
                "confidence_score": "-4.061337"
            },
            {
                "entity": "SZSE Composite Index",
                "confidence_score": "-9.119904"
            },
            {
                "entity": "Hang Seng Global Composite Index",
                "confidence_score": "-11.127813"
            }
        ]
    },
    {
        "mention": "hang seng",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Hang Seng Bank",
                "confidence_score": "3.9861526"
            },
            {
                "entity": "Hong Kong",
                "confidence_score": "-3.9766734"
            },
            {
                "entity": "Hangzhou",
                "confidence_score": "-6.519919"
            }
        ]
    },
    {
        "mention": "the china securities regulatory commission",
        "type": "ORG",
        "predictions": [
            {
                "entity": "China Securities Regulatory Commission",
                "confidence_score": "8.079721"
            },
            {
                "entity": "China Banking and Insurance Regulatory Commission",
                "confidence_score": "-7.7589817"
            },
            {
                "entity": "China Insurance Regulatory Commission",
                "confidence_score": "-8.852233"
            }
        ]
    },
    {
        "mention": "samuel shen",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "H.J. Shen",
                "confidence_score": "-10.385303"
            },
            {
                "entity": "Simon Shen",
                "confidence_score": "-10.883597"
            },
            {
                "entity": "Samuel Burke",
                "confidence_score": "-11.697881"
            }
        ]
    },
    {
        "mention": "pete sweeney",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Pete Sweeney",
                "confidence_score": "0.26639315"
            },
            {
                "entity": "Peter Sweeney",
                "confidence_score": "-7.2310147"
            },
            {
                "entity": "Pete Snyder",
                "confidence_score": "-11.793349"
            }
        ]
    },
    {
        "mention": "fenway",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Fenway Park",
                "confidence_score": "0.040342793"
            },
            {
                "entity": "Fenway (parkway)",
                "confidence_score": "-4.0177674"
            },
            {
                "entity": "Fenway\u2013Kenmore",
                "confidence_score": "-5.711334"
            }
        ]
    },
    {
        "mention": "red sox",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Boston Red Sox",
                "confidence_score": "0.58450556"
            },
            {
                "entity": "2009 Boston Red Sox season",
                "confidence_score": "-7.879129"
            },
            {
                "entity": "1967 Boston Red Sox season",
                "confidence_score": "-8.308279"
            }
        ]
    },
    {
        "mention": "harvard",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Harvard University",
                "confidence_score": "4.555863"
            },
            {
                "entity": "Harvard College",
                "confidence_score": "-3.013967"
            },
            {
                "entity": "Harvard, Massachusetts",
                "confidence_score": "-3.5128589"
            }
        ]
    },
    {
        "mention": "harvard yard",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Harvard Yard",
                "confidence_score": "11.096471"
            },
            {
                "entity": "Harvard Stadium",
                "confidence_score": "-8.000196"
            },
            {
                "entity": "Harvard Square",
                "confidence_score": "-8.02942"
            }
        ]
    },
    {
        "mention": "harvard",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Harvard University",
                "confidence_score": "4.555863"
            },
            {
                "entity": "Harvard College",
                "confidence_score": "-3.013967"
            },
            {
                "entity": "Harvard, Massachusetts",
                "confidence_score": "-3.5128589"
            }
        ]
    },
    {
        "mention": "ucf",
        "type": "ORG",
        "predictions": [
            {
                "entity": "UNCF",
                "confidence_score": "-9.81387"
            },
            {
                "entity": "University of South Florida",
                "confidence_score": "-9.840956"
            },
            {
                "entity": "University Press of Florida",
                "confidence_score": "-11.518891"
            }
        ]
    },
    {
        "mention": "curious george",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Curious George",
                "confidence_score": "0.87910205"
            },
            {
                "entity": "Curious George Brigade",
                "confidence_score": "-8.165929"
            },
            {
                "entity": "Curious George (film)",
                "confidence_score": "-9.09459"
            }
        ]
    },
    {
        "mention": "jillian",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jillian",
                "confidence_score": "-8.265801"
            },
            {
                "entity": "List of Sarah Lawrence College people",
                "confidence_score": "-11.872732"
            },
            {
                "entity": "List of Bennington College people",
                "confidence_score": "-12.060231"
            }
        ]
    },
    {
        "mention": "harvard",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Harvard University",
                "confidence_score": "4.555863"
            },
            {
                "entity": "Harvard College",
                "confidence_score": "-3.013967"
            },
            {
                "entity": "Harvard, Massachusetts",
                "confidence_score": "-3.5128589"
            }
        ]
    },
    {
        "mention": "red sox",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Boston Red Sox",
                "confidence_score": "0.58450556"
            },
            {
                "entity": "2009 Boston Red Sox season",
                "confidence_score": "-7.879129"
            },
            {
                "entity": "1967 Boston Red Sox season",
                "confidence_score": "-8.308279"
            }
        ]
    },
    {
        "mention": "red sox",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Boston Red Sox",
                "confidence_score": "0.58450556"
            },
            {
                "entity": "2009 Boston Red Sox season",
                "confidence_score": "-7.879129"
            },
            {
                "entity": "1967 Boston Red Sox season",
                "confidence_score": "-8.308279"
            }
        ]
    },
    {
        "mention": "fenway",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Fenway Park",
                "confidence_score": "0.040342793"
            },
            {
                "entity": "Fenway (parkway)",
                "confidence_score": "-4.0177674"
            },
            {
                "entity": "Fenway\u2013Kenmore",
                "confidence_score": "-5.711334"
            }
        ]
    },
    {
        "mention": "fenway",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Fenway Park",
                "confidence_score": "-1.6376158"
            },
            {
                "entity": "Fenway\u2013Kenmore",
                "confidence_score": "-6.3986726"
            },
            {
                "entity": "Fenway South",
                "confidence_score": "-7.519774"
            }
        ]
    },
    {
        "mention": "fenway",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Fenway Park",
                "confidence_score": "0.040342793"
            },
            {
                "entity": "Fenway (parkway)",
                "confidence_score": "-4.0177674"
            },
            {
                "entity": "Fenway\u2013Kenmore",
                "confidence_score": "-5.711334"
            }
        ]
    },
    {
        "mention": "cracker jack",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Cracker Jack",
                "confidence_score": "8.608442"
            },
            {
                "entity": "Cracker Jack, Pennsylvania",
                "confidence_score": "-9.676548"
            },
            {
                "entity": "Cracker (band)",
                "confidence_score": "-11.156343"
            }
        ]
    },
    {
        "mention": "take me",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Take Me (film)",
                "confidence_score": "-7.437864"
            },
            {
                "entity": "Take That",
                "confidence_score": "-9.970714"
            },
            {
                "entity": "Take On Me",
                "confidence_score": "-10.466869"
            }
        ]
    },
    {
        "mention": "ballgame",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ball game",
                "confidence_score": "3.4378371"
            },
            {
                "entity": "Ball J",
                "confidence_score": "-11.516437"
            },
            {
                "entity": "Matches of Polish men's volleyball national team conducted by St\u00e9phane Antiga",
                "confidence_score": "-12.696663"
            }
        ]
    },
    {
        "mention": "sweet caroline",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Sweet Caroline",
                "confidence_score": "-2.8346088"
            },
            {
                "entity": "Sweet Pussy Pauline",
                "confidence_score": "-11.748376"
            },
            {
                "entity": "Sweet Cobra",
                "confidence_score": "-11.914188"
            }
        ]
    },
    {
        "mention": "faneuil hall",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Faneuil Hall",
                "confidence_score": "2.4458926"
            },
            {
                "entity": "Swimming at the 2006 Commonwealth Games \u2013 Women's 50 metre EAD freestyle",
                "confidence_score": "-12.56511"
            },
            {
                "entity": "Swimming at the 2006 Commonwealth Games \u2013 Men's 50 metre EAD freestyle",
                "confidence_score": "-12.5687275"
            }
        ]
    },
    {
        "mention": "the north end",
        "type": "LOC",
        "predictions": [
            {
                "entity": "North End, Boston",
                "confidence_score": "-3.4153123"
            },
            {
                "entity": "North End, Halifax",
                "confidence_score": "-9.156511"
            },
            {
                "entity": "North End St. Catharines",
                "confidence_score": "-9.703259"
            }
        ]
    },
    {
        "mention": "north end",
        "type": "LOC",
        "predictions": [
            {
                "entity": "North End, Boston",
                "confidence_score": "6.2282286"
            },
            {
                "entity": "North End (band)",
                "confidence_score": "-3.5432014"
            },
            {
                "entity": "North End, Springfield, Massachusetts",
                "confidence_score": "-5.5503073"
            }
        ]
    },
    {
        "mention": "hubs",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Airline hub",
                "confidence_score": "-4.1495557"
            },
            {
                "entity": "Hub (network science)",
                "confidence_score": "-5.713201"
            },
            {
                "entity": "Hubs and nodes",
                "confidence_score": "-5.958527"
            }
        ]
    },
    {
        "mention": "mike",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "MIKE",
                "confidence_score": "-8.402079"
            },
            {
                "entity": "Mike Amoia",
                "confidence_score": "-10.299709"
            },
            {
                "entity": "Mike Mictlan",
                "confidence_score": "-11.583054"
            }
        ]
    },
    {
        "mention": "modern",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Modern history",
                "confidence_score": "-5.992842"
            },
            {
                "entity": "Modern English",
                "confidence_score": "-6.0394855"
            },
            {
                "entity": "Modern Men",
                "confidence_score": "-10.254603"
            }
        ]
    },
    {
        "mention": "mike",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "MIKE",
                "confidence_score": "-8.402079"
            },
            {
                "entity": "Mike Amoia",
                "confidence_score": "-10.299709"
            },
            {
                "entity": "Mike Mictlan",
                "confidence_score": "-11.583054"
            }
        ]
    },
    {
        "mention": "hubs",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Airline hub",
                "confidence_score": "-4.1495557"
            },
            {
                "entity": "Hub (network science)",
                "confidence_score": "-5.713201"
            },
            {
                "entity": "Hubs and nodes",
                "confidence_score": "-5.958527"
            }
        ]
    },
    {
        "mention": "oktoberfest",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Oktoberfest",
                "confidence_score": "11.591672"
            },
            {
                "entity": "Oktoberfest \u2013 La Crosse, Wisconsin",
                "confidence_score": "-6.8706307"
            },
            {
                "entity": "Oktoberfest tents",
                "confidence_score": "-7.3637652"
            }
        ]
    },
    {
        "mention": "windhoek",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Windhoek",
                "confidence_score": "11.715169"
            },
            {
                "entity": "Windhoek railway station",
                "confidence_score": "-7.8834343"
            },
            {
                "entity": "Windhoek West",
                "confidence_score": "-7.898023"
            }
        ]
    },
    {
        "mention": "namibia breweries",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Namibia Breweries Limited",
                "confidence_score": "2.8188846"
            },
            {
                "entity": "South African Breweries",
                "confidence_score": "-7.4977136"
            },
            {
                "entity": "Nigerian Breweries",
                "confidence_score": "-12.037639"
            }
        ]
    },
    {
        "mention": "oktoberfest",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Oktoberfest",
                "confidence_score": "11.591672"
            },
            {
                "entity": "Oktoberfest \u2013 La Crosse, Wisconsin",
                "confidence_score": "-6.8706307"
            },
            {
                "entity": "Oktoberfest tents",
                "confidence_score": "-7.3637652"
            }
        ]
    },
    {
        "mention": "skw",
        "type": "ORG",
        "predictions": [
            {
                "entity": "SKW",
                "confidence_score": "-10.299478"
            },
            {
                "entity": "Sportk\u00fcstenschifferschein",
                "confidence_score": "-10.949462"
            },
            {
                "entity": "Bobsleigh at the 2014 Winter Olympics",
                "confidence_score": "-11.870948"
            }
        ]
    },
    {
        "mention": "oktoberfest",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Oktoberfest",
                "confidence_score": "11.591672"
            },
            {
                "entity": "Oktoberfest \u2013 La Crosse, Wisconsin",
                "confidence_score": "-6.8706307"
            },
            {
                "entity": "Oktoberfest tents",
                "confidence_score": "-7.3637652"
            }
        ]
    },
    {
        "mention": "oktoberfest",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Oktoberfest",
                "confidence_score": "11.591672"
            },
            {
                "entity": "Oktoberfest \u2013 La Crosse, Wisconsin",
                "confidence_score": "-6.8706307"
            },
            {
                "entity": "Oktoberfest tents",
                "confidence_score": "-7.3637652"
            }
        ]
    },
    {
        "mention": "cathy mcgowan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Cathy McGowan (politician)",
                "confidence_score": "9.276536"
            },
            {
                "entity": "Cathy McGowan (presenter)",
                "confidence_score": "-0.5251003"
            },
            {
                "entity": "Angela McGowan",
                "confidence_score": "-7.83694"
            }
        ]
    },
    {
        "mention": "anzac",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Anzac Day",
                "confidence_score": "0.006327114"
            },
            {
                "entity": "ANZAC (acronym)",
                "confidence_score": "-7.6580634"
            },
            {
                "entity": "Anzacs (TV series)",
                "confidence_score": "-7.6852965"
            }
        ]
    },
    {
        "mention": "indi cathy mcgowan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Cathy McGowan (politician)",
                "confidence_score": "0.6377952"
            },
            {
                "entity": "Travis McGowan",
                "confidence_score": "-11.610791"
            },
            {
                "entity": "Cathy McGowan",
                "confidence_score": "-11.721185"
            }
        ]
    },
    {
        "mention": "spirit",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Spirit",
                "confidence_score": "4.9421077"
            },
            {
                "entity": "Spirit Radio Network",
                "confidence_score": "-1.8773068"
            },
            {
                "entity": "Spirit (media personality)",
                "confidence_score": "-9.936504"
            }
        ]
    },
    {
        "mention": "tony abbott",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Tony Abbott",
                "confidence_score": "12.264326"
            },
            {
                "entity": "Tony Abbott (author)",
                "confidence_score": "-4.1702037"
            },
            {
                "entity": "Tony Abbott (Canadian politician)",
                "confidence_score": "-8.431617"
            }
        ]
    },
    {
        "mention": "ms mcgowan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Cathy McGowan (politician)",
                "confidence_score": "-4.9800916"
            },
            {
                "entity": "Mark McGowan",
                "confidence_score": "-8.025987"
            },
            {
                "entity": "Maura McGowan",
                "confidence_score": "-8.283868"
            }
        ]
    },
    {
        "mention": "abbott",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Tony Abbott",
                "confidence_score": "-0.11773673"
            },
            {
                "entity": "Abbott Government",
                "confidence_score": "-4.764891"
            },
            {
                "entity": "Abbott Ministry",
                "confidence_score": "-6.536548"
            }
        ]
    },
    {
        "mention": "ms mcgowan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Cathy McGowan (politician)",
                "confidence_score": "-4.9800916"
            },
            {
                "entity": "Mark McGowan",
                "confidence_score": "-8.025987"
            },
            {
                "entity": "Maura McGowan",
                "confidence_score": "-8.283868"
            }
        ]
    },
    {
        "mention": "liberals",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Liberal Party of Canada",
                "confidence_score": "-5.036187"
            },
            {
                "entity": "Ontario Liberal Party",
                "confidence_score": "-7.9925056"
            },
            {
                "entity": "Nova Scotia Liberal Party",
                "confidence_score": "-10.624478"
            }
        ]
    },
    {
        "mention": "indi",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Division of Indi",
                "confidence_score": "4.2555566"
            },
            {
                "entity": "Indi Walker",
                "confidence_score": "-8.291438"
            },
            {
                "entity": "John Indi",
                "confidence_score": "-8.838471"
            }
        ]
    },
    {
        "mention": "sophie mirabella",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Sophie Mirabella",
                "confidence_score": "11.693441"
            },
            {
                "entity": "Mirabella",
                "confidence_score": "-9.110084"
            },
            {
                "entity": "Grace Mirabella",
                "confidence_score": "-10.683023"
            }
        ]
    },
    {
        "mention": "abbott",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Tony Abbott",
                "confidence_score": "-0.11773673"
            },
            {
                "entity": "Abbott Government",
                "confidence_score": "-4.764891"
            },
            {
                "entity": "Abbott Ministry",
                "confidence_score": "-6.536548"
            }
        ]
    },
    {
        "mention": "liberal party",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Liberal Party of Australia",
                "confidence_score": "5.5795703"
            },
            {
                "entity": "Liberal Party of Canada",
                "confidence_score": "-5.074076"
            },
            {
                "entity": "Ontario Liberal Party",
                "confidence_score": "-8.695327"
            }
        ]
    },
    {
        "mention": "ms mcgowan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Cathy McGowan (politician)",
                "confidence_score": "-4.9800916"
            },
            {
                "entity": "Mark McGowan",
                "confidence_score": "-8.025987"
            },
            {
                "entity": "Maura McGowan",
                "confidence_score": "-8.283868"
            }
        ]
    },
    {
        "mention": "liberal party",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Liberal Party of Australia",
                "confidence_score": "5.5795703"
            },
            {
                "entity": "Liberal Party of Canada",
                "confidence_score": "-5.074076"
            },
            {
                "entity": "Ontario Liberal Party",
                "confidence_score": "-8.695327"
            }
        ]
    },
    {
        "mention": "indi",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Division of Indi",
                "confidence_score": "-0.6704036"
            },
            {
                "entity": "Indi Walker",
                "confidence_score": "-8.705102"
            },
            {
                "entity": "Indah Putri Indriani",
                "confidence_score": "-12.393696"
            }
        ]
    },
    {
        "mention": "abbott",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Tony Abbott",
                "confidence_score": "-0.11773673"
            },
            {
                "entity": "Abbott Government",
                "confidence_score": "-4.764891"
            },
            {
                "entity": "Abbott Ministry",
                "confidence_score": "-6.536548"
            }
        ]
    },
    {
        "mention": "liberal",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Liberal Party of Canada",
                "confidence_score": "-6.08277"
            },
            {
                "entity": "Colombian Liberal Party",
                "confidence_score": "-10.075234"
            },
            {
                "entity": "Nova Scotia Liberal Party",
                "confidence_score": "-10.201619"
            }
        ]
    },
    {
        "mention": "mrs mirabella",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Sophie Mirabella",
                "confidence_score": "2.030365"
            },
            {
                "entity": "Mirabella",
                "confidence_score": "-6.2410946"
            },
            {
                "entity": "Grace Mirabella",
                "confidence_score": "-9.027606"
            }
        ]
    },
    {
        "mention": "lou lieberman",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Lou Lieberman",
                "confidence_score": "11.255984"
            },
            {
                "entity": "Joe Lieberman",
                "confidence_score": "-10.782419"
            },
            {
                "entity": "Lou Sullivan",
                "confidence_score": "-11.46239"
            }
        ]
    },
    {
        "mention": "brendan nelson",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Brendan Nelson",
                "confidence_score": "10.046449"
            },
            {
                "entity": "Brendan Cannon",
                "confidence_score": "-11.464131"
            },
            {
                "entity": "Brendan Grace",
                "confidence_score": "-11.933738"
            }
        ]
    },
    {
        "mention": "the australian war memorial",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian War Memorial",
                "confidence_score": "5.2633767"
            },
            {
                "entity": "National War Memorial (South Australia)",
                "confidence_score": "-7.6335855"
            },
            {
                "entity": "Australian\u2013American Memorial",
                "confidence_score": "-9.164122"
            }
        ]
    },
    {
        "mention": "spirit",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Spirit",
                "confidence_score": "4.9421077"
            },
            {
                "entity": "Spirit Radio Network",
                "confidence_score": "-1.8773068"
            },
            {
                "entity": "Spirit (media personality)",
                "confidence_score": "-9.936504"
            }
        ]
    },
    {
        "mention": "wodonga sports",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Wodonga",
                "confidence_score": "-4.5876737"
            },
            {
                "entity": "Wodonga Football Club",
                "confidence_score": "-8.266822"
            },
            {
                "entity": "Wodonga Raiders Football Club",
                "confidence_score": "-8.333758"
            }
        ]
    },
    {
        "mention": "leisure centre",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Leisure centre",
                "confidence_score": "5.74202"
            },
            {
                "entity": "Ryde Aquatic Leisure Centre",
                "confidence_score": "-10.717045"
            },
            {
                "entity": "Berry Sport and Recreation Centre",
                "confidence_score": "-11.227899"
            }
        ]
    },
    {
        "mention": "jenny o'connor",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jennifer O'Connor (netball)",
                "confidence_score": "-11.215902"
            },
            {
                "entity": "Jennifer O'Connor (musician)",
                "confidence_score": "-11.316007"
            },
            {
                "entity": "Elaine O'Connor",
                "confidence_score": "-11.772326"
            }
        ]
    },
    {
        "mention": "ms mcgowan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Cathy McGowan (politician)",
                "confidence_score": "-4.9800916"
            },
            {
                "entity": "Mark McGowan",
                "confidence_score": "-8.025987"
            },
            {
                "entity": "Maura McGowan",
                "confidence_score": "-8.283868"
            }
        ]
    },
    {
        "mention": "mrs mirabella",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Sophie Mirabella",
                "confidence_score": "2.030365"
            },
            {
                "entity": "Mirabella",
                "confidence_score": "-6.2410946"
            },
            {
                "entity": "Grace Mirabella",
                "confidence_score": "-9.027606"
            }
        ]
    },
    {
        "mention": "sophie",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sophie Lawrence",
                "confidence_score": "-7.599138"
            },
            {
                "entity": "Sophie Hopkins",
                "confidence_score": "-9.536452"
            },
            {
                "entity": "Sophie Ward",
                "confidence_score": "-9.969121"
            }
        ]
    },
    {
        "mention": "ms o'connor",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Frances O'Connor",
                "confidence_score": "-7.157365"
            },
            {
                "entity": "Natalie O'Connor",
                "confidence_score": "-8.655777"
            },
            {
                "entity": "Des O'Connor",
                "confidence_score": "-8.982987"
            }
        ]
    },
    {
        "mention": "cathy mcgowan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Cathy McGowan (politician)",
                "confidence_score": "9.276536"
            },
            {
                "entity": "Cathy McGowan (presenter)",
                "confidence_score": "-0.5251003"
            },
            {
                "entity": "Angela McGowan",
                "confidence_score": "-7.83694"
            }
        ]
    },
    {
        "mention": "abbott",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Tony Abbott",
                "confidence_score": "-0.11773673"
            },
            {
                "entity": "Abbott Government",
                "confidence_score": "-4.764891"
            },
            {
                "entity": "Abbott Ministry",
                "confidence_score": "-6.536548"
            }
        ]
    },
    {
        "mention": "farrer sussan ley",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sussan Ley",
                "confidence_score": "1.9322057"
            },
            {
                "entity": "Farrer Football League",
                "confidence_score": "-6.030201"
            },
            {
                "entity": "Harley Dickinson",
                "confidence_score": "-11.020953"
            }
        ]
    },
    {
        "mention": "wodonga",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Wodonga",
                "confidence_score": "8.673154"
            },
            {
                "entity": "City of Wodonga",
                "confidence_score": "-0.17178907"
            },
            {
                "entity": "Albury\u2013Wodonga",
                "confidence_score": "-4.027599"
            }
        ]
    },
    {
        "mention": "rod wangman",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Dennis Rodman",
                "confidence_score": "-10.591179"
            },
            {
                "entity": "Rodman Rockefeller",
                "confidence_score": "-11.107978"
            },
            {
                "entity": "Rodger Raderman",
                "confidence_score": "-11.415758"
            }
        ]
    },
    {
        "mention": "commonwealth bank",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Commonwealth Bank",
                "confidence_score": "12.094094"
            },
            {
                "entity": "Westpac",
                "confidence_score": "-7.093159"
            },
            {
                "entity": "Commonwealth Foundation",
                "confidence_score": "-9.983016"
            }
        ]
    },
    {
        "mention": "ian narev",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ian Narev",
                "confidence_score": "12.340396"
            },
            {
                "entity": "Narendra Jadhav",
                "confidence_score": "-11.675483"
            },
            {
                "entity": "Ian Herron",
                "confidence_score": "-11.990914"
            }
        ]
    },
    {
        "mention": "andrew penn",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Andrew Penn (cricketer)",
                "confidence_score": "-7.430454"
            },
            {
                "entity": "Andrew Penn (racewalker)",
                "confidence_score": "-10.297041"
            },
            {
                "entity": "Andrew Robb",
                "confidence_score": "-11.065112"
            }
        ]
    },
    {
        "mention": "abbott",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Tony Abbott",
                "confidence_score": "-0.11773673"
            },
            {
                "entity": "Abbott Government",
                "confidence_score": "-4.764891"
            },
            {
                "entity": "Abbott Ministry",
                "confidence_score": "-6.536548"
            }
        ]
    },
    {
        "mention": "tim fischer",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Tim Fischer",
                "confidence_score": "11.266134"
            },
            {
                "entity": "Tim Watts (politician)",
                "confidence_score": "-11.394667"
            },
            {
                "entity": "Tim Nielsen",
                "confidence_score": "-11.458278"
            }
        ]
    },
    {
        "mention": "andy fairweather",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Andy Fairweather Low",
                "confidence_score": "6.1999736"
            },
            {
                "entity": "Ian Fairweather",
                "confidence_score": "-10.594924"
            },
            {
                "entity": "Andrew Fairweather",
                "confidence_score": "-11.18331"
            }
        ]
    },
    {
        "mention": "welsh",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Welsh language",
                "confidence_score": "6.5837812"
            },
            {
                "entity": "Welsh people",
                "confidence_score": "0.8252626"
            },
            {
                "entity": "Wales",
                "confidence_score": "-0.36891365"
            }
        ]
    },
    {
        "mention": "hello suzy",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Hello (band)",
                "confidence_score": "-10.538263"
            },
            {
                "entity": "Hello Hooray",
                "confidence_score": "-11.3558855"
            },
            {
                "entity": "Hello Dave",
                "confidence_score": "-11.992561"
            }
        ]
    },
    {
        "mention": "paradise",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Paradise",
                "confidence_score": "1.9678035"
            },
            {
                "entity": "Paradise (band)",
                "confidence_score": "-8.242759"
            },
            {
                "entity": "Paradise (John Prine song)",
                "confidence_score": "-10.655071"
            }
        ]
    },
    {
        "mention": "andy",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Andy Revell",
                "confidence_score": "-7.235394"
            },
            {
                "entity": "Andy Powell",
                "confidence_score": "-7.453733"
            },
            {
                "entity": "Andy Ross",
                "confidence_score": "-8.458278"
            }
        ]
    },
    {
        "mention": "eric clapton",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Eric Clapton",
                "confidence_score": "12.659146"
            },
            {
                "entity": "Eric Johnson",
                "confidence_score": "-10.395478"
            },
            {
                "entity": "Eric Singer",
                "confidence_score": "-11.5342"
            }
        ]
    },
    {
        "mention": "george harrison",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "George Harrison",
                "confidence_score": "11.29125"
            },
            {
                "entity": "George Martin",
                "confidence_score": "-8.424743"
            },
            {
                "entity": "George Harrison (swimmer)",
                "confidence_score": "-10.980447"
            }
        ]
    },
    {
        "mention": "bob dylan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bob Dylan",
                "confidence_score": "11.524373"
            },
            {
                "entity": "Bob Marley",
                "confidence_score": "-10.518478"
            },
            {
                "entity": "Bob Rock",
                "confidence_score": "-11.414748"
            }
        ]
    },
    {
        "mention": "jimi hendrix",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jimi Hendrix",
                "confidence_score": "11.506654"
            },
            {
                "entity": "The Jimi Hendrix Experience",
                "confidence_score": "-4.728294"
            },
            {
                "entity": "Death of Jimi Hendrix",
                "confidence_score": "-5.741774"
            }
        ]
    },
    {
        "mention": "roger waters",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Roger Waters",
                "confidence_score": "11.70103"
            },
            {
                "entity": "Roger Troutman",
                "confidence_score": "-11.037559"
            },
            {
                "entity": "Richard Waters",
                "confidence_score": "-11.612699"
            }
        ]
    },
    {
        "mention": "joe cocker",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Joe Cocker",
                "confidence_score": "11.606954"
            },
            {
                "entity": "Joe Cocker!",
                "confidence_score": "-7.9671683"
            },
            {
                "entity": "Jarvis Cocker",
                "confidence_score": "-8.915912"
            }
        ]
    },
    {
        "mention": "steve winwood",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Steve Winwood",
                "confidence_score": "12.149531"
            },
            {
                "entity": "Steve Winwood (album)",
                "confidence_score": "-6.560865"
            },
            {
                "entity": "Winwood (album)",
                "confidence_score": "-8.47418"
            }
        ]
    },
    {
        "mention": "rockwell medical",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Rockwell Medical",
                "confidence_score": "11.957061"
            },
            {
                "entity": "Rockwell International",
                "confidence_score": "-6.751296"
            },
            {
                "entity": "Rockwell Automation",
                "confidence_score": "-9.0801935"
            }
        ]
    },
    {
        "mention": "trade-ideas",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Trade idea",
                "confidence_score": "7.7831807"
            },
            {
                "entity": "TradeIndia",
                "confidence_score": "-11.638684"
            },
            {
                "entity": "U.P. Town Center",
                "confidence_score": "-12.200831"
            }
        ]
    },
    {
        "mention": "rockwell medical",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Rockwell Medical",
                "confidence_score": "11.957061"
            },
            {
                "entity": "Rockwell International",
                "confidence_score": "-6.751296"
            },
            {
                "entity": "Rockwell Automation",
                "confidence_score": "-9.0801935"
            }
        ]
    },
    {
        "mention": "rmti",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Indian rupee",
                "confidence_score": "-5.285209"
            },
            {
                "entity": "Malaysian ringgit",
                "confidence_score": "-5.734074"
            },
            {
                "entity": "Nepalese rupee",
                "confidence_score": "-6.5607514"
            }
        ]
    },
    {
        "mention": "trade-ideas",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Trade idea",
                "confidence_score": "7.7831807"
            },
            {
                "entity": "TradeIndia",
                "confidence_score": "-11.638684"
            },
            {
                "entity": "U.P. Town Center",
                "confidence_score": "-12.200831"
            }
        ]
    },
    {
        "mention": "free",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Free trade",
                "confidence_score": "3.2144902"
            },
            {
                "entity": "Free market",
                "confidence_score": "-3.8779848"
            },
            {
                "entity": "Free software",
                "confidence_score": "-7.157618"
            }
        ]
    },
    {
        "mention": "rockwell medical, inc.",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Rockwell Medical",
                "confidence_score": "7.2292705"
            },
            {
                "entity": "Rockwell International",
                "confidence_score": "-7.5778193"
            },
            {
                "entity": "Rockwell Automation",
                "confidence_score": "-10.465592"
            }
        ]
    },
    {
        "mention": "rockwell medical",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Rockwell Medical",
                "confidence_score": "11.957061"
            },
            {
                "entity": "Rockwell International",
                "confidence_score": "-6.751296"
            },
            {
                "entity": "Rockwell Automation",
                "confidence_score": "-9.0801935"
            }
        ]
    },
    {
        "mention": "rockwell medical",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Rockwell Medical",
                "confidence_score": "11.957061"
            },
            {
                "entity": "Rockwell International",
                "confidence_score": "-6.751296"
            },
            {
                "entity": "Rockwell Automation",
                "confidence_score": "-9.0801935"
            }
        ]
    },
    {
        "mention": "rockwell medical",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Rockwell Medical",
                "confidence_score": "11.957061"
            },
            {
                "entity": "Rockwell International",
                "confidence_score": "-6.751296"
            },
            {
                "entity": "Rockwell Automation",
                "confidence_score": "-9.0801935"
            }
        ]
    },
    {
        "mention": "jim cramer's",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jim Cramer",
                "confidence_score": "-2.5926697"
            },
            {
                "entity": "William C. Cramer",
                "confidence_score": "-10.895045"
            },
            {
                "entity": "Kevin Cramer",
                "confidence_score": "-11.533919"
            }
        ]
    },
    {
        "mention": "rockwell medical",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Rockwell Medical",
                "confidence_score": "11.957061"
            },
            {
                "entity": "Rockwell International",
                "confidence_score": "-6.751296"
            },
            {
                "entity": "Rockwell Automation",
                "confidence_score": "-9.0801935"
            }
        ]
    },
    {
        "mention": "rockwell medical inc",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Rockwell Medical",
                "confidence_score": "9.856086"
            },
            {
                "entity": "Rockwell International",
                "confidence_score": "-7.8075733"
            },
            {
                "entity": "Rockwell Automation",
                "confidence_score": "-10.433992"
            }
        ]
    },
    {
        "mention": "rockwell medical inc",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Rockwell Medical",
                "confidence_score": "9.856086"
            },
            {
                "entity": "Rockwell International",
                "confidence_score": "-7.8075733"
            },
            {
                "entity": "Rockwell Automation",
                "confidence_score": "-10.433992"
            }
        ]
    },
    {
        "mention": "rockwell medical inc",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Rockwell Medical",
                "confidence_score": "9.856086"
            },
            {
                "entity": "Rockwell International",
                "confidence_score": "-7.8075733"
            },
            {
                "entity": "Rockwell Automation",
                "confidence_score": "-10.433992"
            }
        ]
    },
    {
        "mention": "rockwell medical ratings report",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Rockwell Medical",
                "confidence_score": "-8.667266"
            },
            {
                "entity": "Department of Defense Medical Examination Review Board",
                "confidence_score": "-11.605668"
            },
            {
                "entity": "Certified Medical Reimbursement Specialist",
                "confidence_score": "-11.878798"
            }
        ]
    },
    {
        "mention": "jim cramer's",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jim Cramer",
                "confidence_score": "-2.5926697"
            },
            {
                "entity": "William C. Cramer",
                "confidence_score": "-10.895045"
            },
            {
                "entity": "Kevin Cramer",
                "confidence_score": "-11.533919"
            }
        ]
    },
    {
        "mention": "jim cramer",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jim Cramer",
                "confidence_score": "12.371788"
            },
            {
                "entity": "Jim Forbes (journalist)",
                "confidence_score": "-11.51301"
            },
            {
                "entity": "Kevin Cramer",
                "confidence_score": "-11.804911"
            }
        ]
    },
    {
        "mention": "jack mohr",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Gordon \"Jack\" Mohr",
                "confidence_score": "-0.2828504"
            },
            {
                "entity": "Jack Mohr",
                "confidence_score": "-2.8242512"
            },
            {
                "entity": "Cole Mohr",
                "confidence_score": "-11.290423"
            }
        ]
    },
    {
        "mention": "large-cap",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Market capitalization",
                "confidence_score": "-5.15095"
            },
            {
                "entity": "Tax cap",
                "confidence_score": "-7.2972784"
            },
            {
                "entity": "Microcap stock",
                "confidence_score": "-10.087648"
            }
        ]
    },
    {
        "mention": "cramer    quant ratings",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Nielsen Media Research",
                "confidence_score": "-8.035184"
            },
            {
                "entity": "Market-Adjusted Performance Indicator",
                "confidence_score": "-11.450524"
            },
            {
                "entity": "Quantifind",
                "confidence_score": "-11.635441"
            }
        ]
    },
    {
        "mention": "david peltier",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "David Peltier",
                "confidence_score": "-1.7603106"
            },
            {
                "entity": "David Pelletier",
                "confidence_score": "-4.8570323"
            },
            {
                "entity": "David Pelletier (American figure skater)",
                "confidence_score": "-8.3565035"
            }
        ]
    },
    {
        "mention": "hold",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Buy and hold",
                "confidence_score": "-6.0823092"
            },
            {
                "entity": "Holding value",
                "confidence_score": "-7.04299"
            },
            {
                "entity": "Graduate",
                "confidence_score": "-9.621525"
            }
        ]
    },
    {
        "mention": "model portfolio  intra day trade",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mark to model",
                "confidence_score": "-9.335326"
            },
            {
                "entity": "Pattern day trader",
                "confidence_score": "-10.689525"
            },
            {
                "entity": "Target date fund",
                "confidence_score": "-10.712474"
            }
        ]
    },
    {
        "mention": "manufacturer",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Manufacturing",
                "confidence_score": "4.504576"
            },
            {
                "entity": "Aerospace manufacturer",
                "confidence_score": "-3.050848"
            },
            {
                "entity": "List of aircraft manufacturers",
                "confidence_score": "-5.333057"
            }
        ]
    },
    {
        "mention": "evektor-aerotechnik",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Evektor-Aerotechnik",
                "confidence_score": "8.827137"
            },
            {
                "entity": "Evektor EV-55 Outback",
                "confidence_score": "-9.104533"
            },
            {
                "entity": "Aeroprakt",
                "confidence_score": "-11.482926"
            }
        ]
    },
    {
        "mention": "lsa",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Light-sport aircraft",
                "confidence_score": "-0.034614634"
            },
            {
                "entity": "Lithuanian Space Association",
                "confidence_score": "-4.9406238"
            },
            {
                "entity": "Lufthansa",
                "confidence_score": "-7.2200055"
            }
        ]
    },
    {
        "mention": "evektor-aerotechnik",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Evektor-Aerotechnik",
                "confidence_score": "8.827137"
            },
            {
                "entity": "Evektor EV-55 Outback",
                "confidence_score": "-9.104533"
            },
            {
                "entity": "Aeroprakt",
                "confidence_score": "-11.482926"
            }
        ]
    },
    {
        "mention": "harmony lsa",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Civil Aviation Agency Slovenia",
                "confidence_score": "-11.409558"
            },
            {
                "entity": "Lithuanian Space Association",
                "confidence_score": "-11.624653"
            },
            {
                "entity": "Sapphire Aircraft Australia Sapphire LSA",
                "confidence_score": "-12.05624"
            }
        ]
    },
    {
        "mention": "usa faa",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Federal Aviation Administration",
                "confidence_score": "-0.32924747"
            },
            {
                "entity": "United States Anti-Doping Agency",
                "confidence_score": "-10.174729"
            },
            {
                "entity": "American Institute of Aeronautics and Astronautics",
                "confidence_score": "-10.8127985"
            }
        ]
    },
    {
        "mention": "the directorate general of civil aviation of chile",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Directorate General of Civil Aviation (Chile)",
                "confidence_score": "4.4458036"
            },
            {
                "entity": "Chile",
                "confidence_score": "-3.6560352"
            },
            {
                "entity": "Ministry of Transport and Telecommunications",
                "confidence_score": "-9.951182"
            }
        ]
    },
    {
        "mention": "harmony\revektor sportstar",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Evektor SportStar",
                "confidence_score": "-4.6328044"
            },
            {
                "entity": "Evektor-Aerotechnik",
                "confidence_score": "-11.383657"
            },
            {
                "entity": "Sport Copter Lightning",
                "confidence_score": "-11.999219"
            }
        ]
    },
    {
        "mention": "rtc aircraft",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Radio Technical Commission for Aeronautics",
                "confidence_score": "-9.259475"
            },
            {
                "entity": "Historical Aircraft Corporation",
                "confidence_score": "-11.194082"
            },
            {
                "entity": "ICON Aircraft",
                "confidence_score": "-11.325419"
            }
        ]
    },
    {
        "mention": "sportstar",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Evektor SportStar",
                "confidence_score": "9.221049"
            },
            {
                "entity": "Sportstar",
                "confidence_score": "3.9853246"
            },
            {
                "entity": "Sportsguitar",
                "confidence_score": "-10.435473"
            }
        ]
    },
    {
        "mention": "rtc",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Russian State Scientific Center for Robotics and Technical Cybernetics",
                "confidence_score": "0.9966831"
            },
            {
                "entity": "Radio Technical Commission for Aeronautics",
                "confidence_score": "-10.065144"
            },
            {
                "entity": "RTU FC",
                "confidence_score": "-11.441587"
            }
        ]
    },
    {
        "mention": "evektor-aerotechnik",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Evektor-Aerotechnik",
                "confidence_score": "8.827137"
            },
            {
                "entity": "Evektor EV-55 Outback",
                "confidence_score": "-9.104533"
            },
            {
                "entity": "Aeroprakt",
                "confidence_score": "-11.482926"
            }
        ]
    },
    {
        "mention": "harmony",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Harmony (ISS module)",
                "confidence_score": "-6.422423"
            },
            {
                "entity": "Harmony (band)",
                "confidence_score": "-9.62752"
            },
            {
                "entity": "Harmon Engineering Company",
                "confidence_score": "-11.307626"
            }
        ]
    },
    {
        "mention": "lsa",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Light-sport aircraft",
                "confidence_score": "-0.034614634"
            },
            {
                "entity": "Lithuanian Space Association",
                "confidence_score": "-4.9406238"
            },
            {
                "entity": "Lufthansa",
                "confidence_score": "-7.2200055"
            }
        ]
    },
    {
        "mention": "vortex",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Vortex (satellite)",
                "confidence_score": "-6.224005"
            },
            {
                "entity": "Vortex (software)",
                "confidence_score": "-8.846663"
            },
            {
                "entity": "Vortex Software",
                "confidence_score": "-9.762317"
            }
        ]
    },
    {
        "mention": "bwf - industrial felts for the aerospace industry",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Albany International",
                "confidence_score": "-7.4418178"
            },
            {
                "entity": "Orion Electric",
                "confidence_score": "-11.037095"
            },
            {
                "entity": "BWF",
                "confidence_score": "-12.067563"
            }
        ]
    },
    {
        "mention": "the bwf group",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Badminton World Federation",
                "confidence_score": "-8.058932"
            },
            {
                "entity": "Japan Railways Group",
                "confidence_score": "-10.5910015"
            },
            {
                "entity": "BWF",
                "confidence_score": "-12.288686"
            }
        ]
    },
    {
        "mention": "kuerzi avionics - avionics components and services",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Kuki Inc.",
                "confidence_score": "-10.405135"
            },
            {
                "entity": "Ikarus Kurir",
                "confidence_score": "-11.4756775"
            },
            {
                "entity": "Kur Kahriz",
                "confidence_score": "-11.716727"
            }
        ]
    },
    {
        "mention": "kuerzi avionics",
        "type": "ORG",
        "predictions": [
            {
                "entity": "JSC Kuznetsov",
                "confidence_score": "-11.151296"
            },
            {
                "entity": "Ikarus Kurir",
                "confidence_score": "-11.617747"
            },
            {
                "entity": "Tri-R Technologies",
                "confidence_score": "-11.656711"
            }
        ]
    },
    {
        "mention": "harmony",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Harmony (ISS module)",
                "confidence_score": "-6.422423"
            },
            {
                "entity": "Harmony (band)",
                "confidence_score": "-9.62752"
            },
            {
                "entity": "Harmon Engineering Company",
                "confidence_score": "-11.307626"
            }
        ]
    },
    {
        "mention": "ems",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Emergency medical services",
                "confidence_score": "6.5095267"
            },
            {
                "entity": "Instrument meteorological conditions",
                "confidence_score": "-8.04362"
            },
            {
                "entity": "Instrument landing system",
                "confidence_score": "-10.177106"
            }
        ]
    },
    {
        "mention": "vfr",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Visual flight rules",
                "confidence_score": "10.907287"
            },
            {
                "entity": "Night VFR",
                "confidence_score": "0.86028457"
            },
            {
                "entity": "Visual meteorological conditions",
                "confidence_score": "-2.1758988"
            }
        ]
    },
    {
        "mention": "propuls aes",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Rolls-Royce Exe",
                "confidence_score": "-11.276854"
            },
            {
                "entity": "Clapp's Cars Spyder Standard",
                "confidence_score": "-11.317895"
            },
            {
                "entity": "Splitterskyddad EnhetsPlattform",
                "confidence_score": "-11.5385275"
            }
        ]
    },
    {
        "mention": "the rotax 912",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Rotax 912",
                "confidence_score": "9.701241"
            },
            {
                "entity": "Rotax 914",
                "confidence_score": "-2.450577"
            },
            {
                "entity": "Rotax 915 iS",
                "confidence_score": "-10.922716"
            }
        ]
    },
    {
        "mention": "lsa",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Light-sport aircraft",
                "confidence_score": "-0.034614634"
            },
            {
                "entity": "Lithuanian Space Association",
                "confidence_score": "-4.9406238"
            },
            {
                "entity": "Lufthansa",
                "confidence_score": "-7.2200055"
            }
        ]
    },
    {
        "mention": "evektor-aerotechnik",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Evektor-Aerotechnik",
                "confidence_score": "8.827137"
            },
            {
                "entity": "Evektor EV-55 Outback",
                "confidence_score": "-9.104533"
            },
            {
                "entity": "Aeroprakt",
                "confidence_score": "-11.482926"
            }
        ]
    },
    {
        "mention": "harmony",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Harmony (ISS module)",
                "confidence_score": "-6.422423"
            },
            {
                "entity": "Harmony (band)",
                "confidence_score": "-9.62752"
            },
            {
                "entity": "Harmon Engineering Company",
                "confidence_score": "-11.307626"
            }
        ]
    },
    {
        "mention": "elizabeth armstrong moore",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Elizabeth Moore (historian)",
                "confidence_score": "-11.345449"
            },
            {
                "entity": "Elizabeth Oehlkers Wright",
                "confidence_score": "-12.0126915"
            },
            {
                "entity": "Elizabeth Anne McCauley",
                "confidence_score": "-12.268099"
            }
        ]
    },
    {
        "mention": "newser staff",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Newser",
                "confidence_score": "-3.7418249"
            },
            {
                "entity": "NewsRx",
                "confidence_score": "-10.255401"
            },
            {
                "entity": "NewsLeecher",
                "confidence_score": "-11.539338"
            }
        ]
    },
    {
        "mention": "the\u00a0 ",
        "type": "ORG",
        "predictions": [
            {
                "entity": "University of Colorado Boulder",
                "confidence_score": "-0.4627882"
            },
            {
                "entity": "University of Colorado",
                "confidence_score": "-2.6873162"
            },
            {
                "entity": "University of Colorado Denver",
                "confidence_score": "-6.123486"
            }
        ]
    },
    {
        "mention": "university of colorado",
        "type": "ORG",
        "predictions": [
            {
                "entity": "University of Colorado Boulder",
                "confidence_score": "9.624495"
            },
            {
                "entity": "University of Colorado",
                "confidence_score": "7.337195"
            },
            {
                "entity": "University of Colorado Colorado Springs",
                "confidence_score": "-7.3504305"
            }
        ]
    },
    {
        "mention": "science translational medicine",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Science Translational Medicine",
                "confidence_score": "6.06595"
            },
            {
                "entity": "Translational medicine",
                "confidence_score": "-2.4361577"
            },
            {
                "entity": "Translational research",
                "confidence_score": "-4.898734"
            }
        ]
    },
    {
        "mention": "npr",
        "type": "ORG",
        "predictions": [
            {
                "entity": "NPR",
                "confidence_score": "11.77745"
            },
            {
                "entity": "PBS",
                "confidence_score": "-5.2102942"
            },
            {
                "entity": "Association of Public Radio Stations",
                "confidence_score": "-7.979561"
            }
        ]
    },
    {
        "mention": "newser.com",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Newser",
                "confidence_score": "5.7257447"
            },
            {
                "entity": "NewsRx",
                "confidence_score": "-8.879338"
            },
            {
                "entity": "NewsON",
                "confidence_score": "-10.650759"
            }
        ]
    },
    {
        "mention": "frank",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Frank Endacott",
                "confidence_score": "-5.5879807"
            },
            {
                "entity": "Francis Maloney (rugby league)",
                "confidence_score": "-9.896699"
            },
            {
                "entity": "Francis Meli",
                "confidence_score": "-11.188002"
            }
        ]
    },
    {
        "mention": "nrl",
        "type": "ORG",
        "predictions": [
            {
                "entity": "National Rugby League",
                "confidence_score": "5.3634133"
            },
            {
                "entity": "Rugby league",
                "confidence_score": "-4.611774"
            },
            {
                "entity": "New South Wales Rugby League",
                "confidence_score": "-4.730779"
            }
        ]
    },
    {
        "mention": "the sydney roosters",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sydney Roosters",
                "confidence_score": "8.519383"
            },
            {
                "entity": "2010 Sydney Roosters season",
                "confidence_score": "-7.144399"
            },
            {
                "entity": "2003 Sydney Roosters season",
                "confidence_score": "-7.9537473"
            }
        ]
    },
    {
        "mention": "des hasler's",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Des Hasler",
                "confidence_score": "6.8061156"
            },
            {
                "entity": "2006 Bulldogs RLFC season",
                "confidence_score": "-10.517438"
            },
            {
                "entity": "2000 Bulldogs RLFC season",
                "confidence_score": "-10.877585"
            }
        ]
    },
    {
        "mention": "st george illawarra",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "St. George Illawarra Dragons",
                "confidence_score": "5.7417"
            },
            {
                "entity": "St. George Dragons",
                "confidence_score": "1.259391"
            },
            {
                "entity": "St George AFC",
                "confidence_score": "-8.548941"
            }
        ]
    },
    {
        "mention": "gareth widdop",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Gareth Widdop",
                "confidence_score": "12.057835"
            },
            {
                "entity": "Gareth Morton",
                "confidence_score": "-11.368827"
            },
            {
                "entity": "Gareth Frodsham",
                "confidence_score": "-11.475511"
            }
        ]
    },
    {
        "mention": "bronson harrison",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bronson Harrison",
                "confidence_score": "10.848612"
            },
            {
                "entity": "Bronson Matthews",
                "confidence_score": "-11.237391"
            },
            {
                "entity": "Bronson Murray",
                "confidence_score": "-11.75303"
            }
        ]
    },
    {
        "mention": "mal cochrane",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mal Cochrane",
                "confidence_score": "3.6471572"
            },
            {
                "entity": "Alun Cochrane",
                "confidence_score": "-10.514911"
            },
            {
                "entity": "Malcom Chalmers",
                "confidence_score": "-11.797835"
            }
        ]
    },
    {
        "mention": "sean garlick",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Sean Garlick",
                "confidence_score": "11.185458"
            },
            {
                "entity": "Stephen Garlick",
                "confidence_score": "-10.344502"
            },
            {
                "entity": "Raymond Garlick",
                "confidence_score": "-11.455039"
            }
        ]
    },
    {
        "mention": "the roosters at allianz stadium",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sydney Football Stadium",
                "confidence_score": "-3.7779624"
            },
            {
                "entity": "Stadium Australia",
                "confidence_score": "-8.999251"
            },
            {
                "entity": "Hindmarsh Stadium",
                "confidence_score": "-10.923217"
            }
        ]
    },
    {
        "mention": "widdop",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Gareth Widdop",
                "confidence_score": "0.22250405"
            },
            {
                "entity": "Widdop",
                "confidence_score": "-7.9022565"
            },
            {
                "entity": "Wivenhoe, Tasmania",
                "confidence_score": "-11.824872"
            }
        ]
    },
    {
        "mention": "bulldogs",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Canterbury-Bankstown Bulldogs",
                "confidence_score": "-3.9203165"
            },
            {
                "entity": "2009 Bulldogs RLFC season",
                "confidence_score": "-9.039708"
            },
            {
                "entity": "2000 Bulldogs RLFC season",
                "confidence_score": "-10.07378"
            }
        ]
    },
    {
        "mention": "nick ghabar",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Khabar, Shahr-e Babak",
                "confidence_score": "-11.81408"
            },
            {
                "entity": "Zoubin Ghahramani",
                "confidence_score": "-11.861615"
            },
            {
                "entity": "Nick Dzubnar",
                "confidence_score": "-11.881917"
            }
        ]
    },
    {
        "mention": "widdop",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Widdop",
                "confidence_score": "-5.602164"
            },
            {
                "entity": "Gareth Widdop",
                "confidence_score": "-6.9565325"
            },
            {
                "entity": "Matches of Polish men's volleyball national team conducted by St\u00e9phane Antiga",
                "confidence_score": "-12.744774"
            }
        ]
    },
    {
        "mention": "hasler with prtichard",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Des Hasler",
                "confidence_score": "-10.453496"
            },
            {
                "entity": "Nikol Hasler",
                "confidence_score": "-10.491104"
            },
            {
                "entity": "Chit Chat Von Loopin Stab",
                "confidence_score": "-11.676635"
            }
        ]
    },
    {
        "mention": "bulldogs",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Canterbury-Bankstown Bulldogs",
                "confidence_score": "-3.9203165"
            },
            {
                "entity": "2009 Bulldogs RLFC season",
                "confidence_score": "-9.039708"
            },
            {
                "entity": "2000 Bulldogs RLFC season",
                "confidence_score": "-10.07378"
            }
        ]
    },
    {
        "mention": "frankie",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Frankie Boyle",
                "confidence_score": "-9.016181"
            },
            {
                "entity": "Frankie (musician)",
                "confidence_score": "-10.001084"
            },
            {
                "entity": "Frankie Curry",
                "confidence_score": "-11.239305"
            }
        ]
    },
    {
        "mention": "james graham",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "James Graham (playwright)",
                "confidence_score": "0.37251776"
            },
            {
                "entity": "James Graham (singer)",
                "confidence_score": "-0.6197977"
            },
            {
                "entity": "James Graham (rugby league)",
                "confidence_score": "-7.428926"
            }
        ]
    },
    {
        "mention": "sony",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sony",
                "confidence_score": "10.840842"
            },
            {
                "entity": "Sony Corporation of America",
                "confidence_score": "-3.940872"
            },
            {
                "entity": "PlayStation",
                "confidence_score": "-5.7264104"
            }
        ]
    },
    {
        "mention": "sony",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sony",
                "confidence_score": "10.840842"
            },
            {
                "entity": "Sony Corporation of America",
                "confidence_score": "-3.940872"
            },
            {
                "entity": "PlayStation",
                "confidence_score": "-5.7264104"
            }
        ]
    },
    {
        "mention": "xperia blog",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Xperia Play",
                "confidence_score": "-10.125974"
            },
            {
                "entity": "Sony Ericsson Xperia pro",
                "confidence_score": "-11.099523"
            },
            {
                "entity": "Xperience Days",
                "confidence_score": "-11.957051"
            }
        ]
    },
    {
        "mention": "sony",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sony",
                "confidence_score": "10.840842"
            },
            {
                "entity": "Sony Corporation of America",
                "confidence_score": "-3.940872"
            },
            {
                "entity": "PlayStation",
                "confidence_score": "-5.7264104"
            }
        ]
    },
    {
        "mention": "sony",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sony",
                "confidence_score": "10.840842"
            },
            {
                "entity": "Sony Corporation of America",
                "confidence_score": "-3.940872"
            },
            {
                "entity": "PlayStation",
                "confidence_score": "-5.7264104"
            }
        ]
    },
    {
        "mention": "xperia z5 compact",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Sony Xperia Z5 Compact",
                "confidence_score": "7.580156"
            },
            {
                "entity": "Sony Xperia Z3 Compact",
                "confidence_score": "-3.2435815"
            },
            {
                "entity": "Sony Xperia Z3 Tablet Compact",
                "confidence_score": "-9.318572"
            }
        ]
    },
    {
        "mention": "sony",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sony",
                "confidence_score": "10.840842"
            },
            {
                "entity": "Sony Corporation of America",
                "confidence_score": "-3.940872"
            },
            {
                "entity": "PlayStation",
                "confidence_score": "-5.7264104"
            }
        ]
    },
    {
        "mention": "sony",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sony",
                "confidence_score": "10.840842"
            },
            {
                "entity": "Sony Corporation of America",
                "confidence_score": "-3.940872"
            },
            {
                "entity": "PlayStation",
                "confidence_score": "-5.7264104"
            }
        ]
    },
    {
        "mention": "sony",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sony",
                "confidence_score": "10.840842"
            },
            {
                "entity": "Sony Corporation of America",
                "confidence_score": "-3.940872"
            },
            {
                "entity": "PlayStation",
                "confidence_score": "-5.7264104"
            }
        ]
    },
    {
        "mention": "the xperia z1",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Sony Xperia Z1",
                "confidence_score": "5.4181767"
            },
            {
                "entity": "Sony Xperia XZ1",
                "confidence_score": "-9.96985"
            },
            {
                "entity": "Sony Xperia E1",
                "confidence_score": "-11.315503"
            }
        ]
    },
    {
        "mention": "sony",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sony",
                "confidence_score": "10.840842"
            },
            {
                "entity": "Sony Corporation of America",
                "confidence_score": "-3.940872"
            },
            {
                "entity": "PlayStation",
                "confidence_score": "-5.7264104"
            }
        ]
    },
    {
        "mention": "sony",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sony",
                "confidence_score": "10.840842"
            },
            {
                "entity": "Sony Corporation of America",
                "confidence_score": "-3.940872"
            },
            {
                "entity": "PlayStation",
                "confidence_score": "-5.7264104"
            }
        ]
    },
    {
        "mention": "sony",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sony",
                "confidence_score": "10.840842"
            },
            {
                "entity": "Sony Corporation of America",
                "confidence_score": "-3.940872"
            },
            {
                "entity": "PlayStation",
                "confidence_score": "-5.7264104"
            }
        ]
    },
    {
        "mention": "southern ca",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Southern California",
                "confidence_score": "-3.0381234"
            },
            {
                "entity": "Southern Pacific Transportation Company",
                "confidence_score": "-9.275487"
            },
            {
                "entity": "Southern California Gas Company",
                "confidence_score": "-9.555658"
            }
        ]
    },
    {
        "mention": "summer",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Summer",
                "confidence_score": "7.8512044"
            },
            {
                "entity": "June",
                "confidence_score": "-6.224282"
            },
            {
                "entity": "Graduate",
                "confidence_score": "-9.671268"
            }
        ]
    },
    {
        "mention": "the stylish housewife",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Today's Housewife",
                "confidence_score": "-9.949436"
            },
            {
                "entity": "Stylist (magazine)",
                "confidence_score": "-9.953347"
            },
            {
                "entity": "Stylista",
                "confidence_score": "-11.031051"
            }
        ]
    },
    {
        "mention": "loading inlinkz",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "LoadingReadyRun",
                "confidence_score": "-10.150963"
            },
            {
                "entity": "Linked Data Notifications",
                "confidence_score": "-11.0892725"
            },
            {
                "entity": "Loading Zones",
                "confidence_score": "-11.905949"
            }
        ]
    },
    {
        "mention": "the pleated poppy",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Poppy (entertainer)",
                "confidence_score": "-9.360101"
            },
            {
                "entity": "The Poppy Fields",
                "confidence_score": "-10.835494"
            },
            {
                "entity": "The Poppy Girls",
                "confidence_score": "-10.957667"
            }
        ]
    },
    {
        "mention": "mix & match fashion",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Matches Fashion",
                "confidence_score": "-7.451032"
            },
            {
                "entity": "Mixcloud",
                "confidence_score": "-10.810869"
            },
            {
                "entity": "Mixx",
                "confidence_score": "-11.315227"
            }
        ]
    },
    {
        "mention": "ali larijani",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ali Larijani",
                "confidence_score": "10.7804575"
            },
            {
                "entity": "Sadeq Larijani",
                "confidence_score": "-8.419525"
            },
            {
                "entity": "Ali Latifi",
                "confidence_score": "-11.27521"
            }
        ]
    },
    {
        "mention": "npr",
        "type": "ORG",
        "predictions": [
            {
                "entity": "NPR",
                "confidence_score": "8.644986"
            },
            {
                "entity": "All Things Considered",
                "confidence_score": "-6.1805573"
            },
            {
                "entity": "Association of Public Radio Stations",
                "confidence_score": "-9.088657"
            }
        ]
    },
    {
        "mention": "larijani",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Mohammad-Javad Larijani",
                "confidence_score": "5.8656707"
            },
            {
                "entity": "Ali Larijani",
                "confidence_score": "3.4093883"
            },
            {
                "entity": "Larijani",
                "confidence_score": "-3.8075123"
            }
        ]
    },
    {
        "mention": "npr",
        "type": "ORG",
        "predictions": [
            {
                "entity": "NPR",
                "confidence_score": "8.644986"
            },
            {
                "entity": "All Things Considered",
                "confidence_score": "-6.1805573"
            },
            {
                "entity": "Association of Public Radio Stations",
                "confidence_score": "-9.088657"
            }
        ]
    },
    {
        "mention": "larijani",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Mohammad-Javad Larijani",
                "confidence_score": "5.8656707"
            },
            {
                "entity": "Ali Larijani",
                "confidence_score": "3.4093883"
            },
            {
                "entity": "Larijani",
                "confidence_score": "-3.8075123"
            }
        ]
    },
    {
        "mention": "the u.s. congress",
        "type": "ORG",
        "predictions": [
            {
                "entity": "United States Congress",
                "confidence_score": "3.6461933"
            },
            {
                "entity": "Congress",
                "confidence_score": "-4.217595"
            },
            {
                "entity": "Congress of the Confederation",
                "confidence_score": "-7.625841"
            }
        ]
    },
    {
        "mention": "npr",
        "type": "ORG",
        "predictions": [
            {
                "entity": "NPR",
                "confidence_score": "8.644986"
            },
            {
                "entity": "All Things Considered",
                "confidence_score": "-6.1805573"
            },
            {
                "entity": "Association of Public Radio Stations",
                "confidence_score": "-9.088657"
            }
        ]
    },
    {
        "mention": "obama",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Barack Obama",
                "confidence_score": "6.637609"
            },
            {
                "entity": "Presidency of Barack Obama",
                "confidence_score": "-4.8801856"
            },
            {
                "entity": "Abraham Lincoln",
                "confidence_score": "-7.7160974"
            }
        ]
    },
    {
        "mention": "state",
        "type": "ORG",
        "predictions": [
            {
                "entity": "U.S. state",
                "confidence_score": "3.0444102"
            },
            {
                "entity": "United States Department of State",
                "confidence_score": "-2.7607617"
            },
            {
                "entity": "United States Secretary of State",
                "confidence_score": "-4.2258916"
            }
        ]
    },
    {
        "mention": "john kerry",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "John Kerry",
                "confidence_score": "10.6447115"
            },
            {
                "entity": "John McCain",
                "confidence_score": "-10.768056"
            },
            {
                "entity": "John R. Bolton",
                "confidence_score": "-11.150006"
            }
        ]
    },
    {
        "mention": "npr",
        "type": "ORG",
        "predictions": [
            {
                "entity": "NPR",
                "confidence_score": "8.644986"
            },
            {
                "entity": "All Things Considered",
                "confidence_score": "-6.1805573"
            },
            {
                "entity": "Association of Public Radio Stations",
                "confidence_score": "-9.088657"
            }
        ]
    },
    {
        "mention": "hezbollah",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Hezbollah",
                "confidence_score": "8.130948"
            },
            {
                "entity": "Hezbollah (Iran)",
                "confidence_score": "4.064799"
            },
            {
                "entity": "Hezbollah Organization",
                "confidence_score": "-4.5074363"
            }
        ]
    },
    {
        "mention": "larijani",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Mohammad-Javad Larijani",
                "confidence_score": "5.8656707"
            },
            {
                "entity": "Ali Larijani",
                "confidence_score": "3.4093883"
            },
            {
                "entity": "Larijani",
                "confidence_score": "-3.8075123"
            }
        ]
    },
    {
        "mention": "larijani",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Mohammad-Javad Larijani",
                "confidence_score": "5.8656707"
            },
            {
                "entity": "Ali Larijani",
                "confidence_score": "3.4093883"
            },
            {
                "entity": "Larijani",
                "confidence_score": "-3.8075123"
            }
        ]
    },
    {
        "mention": "hassan rouhani",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Hassan Rouhani",
                "confidence_score": "12.025309"
            },
            {
                "entity": "Fuad Rouhani",
                "confidence_score": "-10.638558"
            },
            {
                "entity": "Hassan Roudbarian",
                "confidence_score": "-11.937132"
            }
        ]
    },
    {
        "mention": "larijani",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Mohammad-Javad Larijani",
                "confidence_score": "5.8656707"
            },
            {
                "entity": "Ali Larijani",
                "confidence_score": "3.4093883"
            },
            {
                "entity": "Larijani",
                "confidence_score": "-3.8075123"
            }
        ]
    },
    {
        "mention": "larijani",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Mohammad-Javad Larijani",
                "confidence_score": "5.8656707"
            },
            {
                "entity": "Ali Larijani",
                "confidence_score": "3.4093883"
            },
            {
                "entity": "Larijani",
                "confidence_score": "-3.8075123"
            }
        ]
    },
    {
        "mention": "jason rezaian",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jason Rezaian",
                "confidence_score": "12.271817"
            },
            {
                "entity": "Jason Jorjani",
                "confidence_score": "-6.6562057"
            },
            {
                "entity": "Jason Kartalian",
                "confidence_score": "-12.112204"
            }
        ]
    },
    {
        "mention": "larijani",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Mohammad-Javad Larijani",
                "confidence_score": "5.8656707"
            },
            {
                "entity": "Ali Larijani",
                "confidence_score": "3.4093883"
            },
            {
                "entity": "Larijani",
                "confidence_score": "-3.8075123"
            }
        ]
    },
    {
        "mention": "saddam hussein",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Saddam Hussein",
                "confidence_score": "11.555086"
            },
            {
                "entity": "Iraq",
                "confidence_score": "-2.7303157"
            },
            {
                "entity": "Hussein of Jordan",
                "confidence_score": "-9.328941"
            }
        ]
    },
    {
        "mention": "ash carter",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ash Carter",
                "confidence_score": "10.3207"
            },
            {
                "entity": "Jimmy Carter",
                "confidence_score": "-8.545499"
            },
            {
                "entity": "Ash Avildsen",
                "confidence_score": "-10.846004"
            }
        ]
    },
    {
        "mention": "farsi",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Persian language",
                "confidence_score": "8.880752"
            },
            {
                "entity": "Western Persian",
                "confidence_score": "6.771075"
            },
            {
                "entity": "Fars Province",
                "confidence_score": "-8.190826"
            }
        ]
    },
    {
        "mention": "npr",
        "type": "ORG",
        "predictions": [
            {
                "entity": "NPR",
                "confidence_score": "8.644986"
            },
            {
                "entity": "All Things Considered",
                "confidence_score": "-6.1805573"
            },
            {
                "entity": "Association of Public Radio Stations",
                "confidence_score": "-9.088657"
            }
        ]
    },
    {
        "mention": "larijani",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Mohammad-Javad Larijani",
                "confidence_score": "5.8656707"
            },
            {
                "entity": "Ali Larijani",
                "confidence_score": "3.4093883"
            },
            {
                "entity": "Larijani",
                "confidence_score": "-3.8075123"
            }
        ]
    },
    {
        "mention": "larijani",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Mohammad-Javad Larijani",
                "confidence_score": "5.8656707"
            },
            {
                "entity": "Ali Larijani",
                "confidence_score": "3.4093883"
            },
            {
                "entity": "Larijani",
                "confidence_score": "-3.8075123"
            }
        ]
    },
    {
        "mention": "login",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Login",
                "confidence_score": "11.363369"
            },
            {
                "entity": "Login session",
                "confidence_score": "-0.9606112"
            },
            {
                "entity": "Login manager",
                "confidence_score": "-6.6410623"
            }
        ]
    },
    {
        "mention": "username & password required",
        "type": "ORG",
        "predictions": [
            {
                "entity": "User (computing)",
                "confidence_score": "-5.4775205"
            },
            {
                "entity": "User identifier",
                "confidence_score": "-6.23649"
            },
            {
                "entity": "Userscript",
                "confidence_score": "-9.673589"
            }
        ]
    },
    {
        "mention": "maruti suzuki india",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Maruti Suzuki",
                "confidence_score": "3.8024914"
            },
            {
                "entity": "Maruti 800",
                "confidence_score": "-5.286254"
            },
            {
                "entity": "Suzuki",
                "confidence_score": "-5.897353"
            }
        ]
    },
    {
        "mention": "gurgaon",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Gurgaon",
                "confidence_score": "10.069616"
            },
            {
                "entity": "Gurgaon district",
                "confidence_score": "-3.0816057"
            },
            {
                "entity": "New Gurgaon",
                "confidence_score": "-6.9231744"
            }
        ]
    },
    {
        "mention": "blue jeans network",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Blue Book Network",
                "confidence_score": "-11.08869"
            },
            {
                "entity": "BlueJeans Network",
                "confidence_score": "-11.3801565"
            },
            {
                "entity": "FM 876 Network",
                "confidence_score": "-11.694495"
            }
        ]
    },
    {
        "mention": "microsoft skype/lync",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Skype for Business Server",
                "confidence_score": "-0.31758332"
            },
            {
                "entity": "Skype",
                "confidence_score": "-5.3398347"
            },
            {
                "entity": "Skype for Business",
                "confidence_score": "-7.799041"
            }
        ]
    },
    {
        "mention": "zoom",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Zoom (company)",
                "confidence_score": "-2.1759977"
            },
            {
                "entity": "Zoom Video Communications",
                "confidence_score": "-2.8596191"
            },
            {
                "entity": "Zoom Telephonics",
                "confidence_score": "-3.4291096"
            }
        ]
    },
    {
        "mention": "blue jeans",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Jeans",
                "confidence_score": "2.282659"
            },
            {
                "entity": "Blue Jeans (Lana Del Rey song)",
                "confidence_score": "-9.666917"
            },
            {
                "entity": "BlueFace",
                "confidence_score": "-10.907931"
            }
        ]
    },
    {
        "mention": "lync, hangouts",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Google Hangouts",
                "confidence_score": "-9.4589815"
            },
            {
                "entity": "Locator/Identifier Separation Protocol",
                "confidence_score": "-11.251074"
            },
            {
                "entity": "Lyn-Z Adams Hawkins",
                "confidence_score": "-11.839108"
            }
        ]
    },
    {
        "mention": "polycom",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Polycom",
                "confidence_score": "11.761879"
            },
            {
                "entity": "Microcom",
                "confidence_score": "-10.909858"
            },
            {
                "entity": "Polymail",
                "confidence_score": "-11.902148"
            }
        ]
    },
    {
        "mention": "blue jimp",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Blue Byte",
                "confidence_score": "-10.849291"
            },
            {
                "entity": "BlueJ",
                "confidence_score": "-11.839496"
            },
            {
                "entity": "Jimm",
                "confidence_score": "-11.973129"
            }
        ]
    },
    {
        "mention": "blue jeans",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Jeans",
                "confidence_score": "2.282659"
            },
            {
                "entity": "Blue Jeans (Lana Del Rey song)",
                "confidence_score": "-9.666917"
            },
            {
                "entity": "BlueFace",
                "confidence_score": "-10.907931"
            }
        ]
    },
    {
        "mention": "facebook, gopro, pandora",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Facebook",
                "confidence_score": "-0.023043904"
            },
            {
                "entity": "Amazon (company)",
                "confidence_score": "-6.343079"
            },
            {
                "entity": "Facebook Platform",
                "confidence_score": "-9.273545"
            }
        ]
    },
    {
        "mention": "stanford university",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Stanford University",
                "confidence_score": "10.495671"
            },
            {
                "entity": "Stanford University Press",
                "confidence_score": "-6.6829185"
            },
            {
                "entity": "Stanford Law School",
                "confidence_score": "-6.8271976"
            }
        ]
    },
    {
        "mention": "blue jeans",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Jeans",
                "confidence_score": "2.282659"
            },
            {
                "entity": "Blue Jeans (Lana Del Rey song)",
                "confidence_score": "-9.666917"
            },
            {
                "entity": "BlueFace",
                "confidence_score": "-10.907931"
            }
        ]
    },
    {
        "mention": "primetime",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Primetime (American TV program)",
                "confidence_score": "-4.843532"
            },
            {
                "entity": "Primetime Race Group",
                "confidence_score": "-10.508782"
            },
            {
                "entity": "2012 PSA World Series Finals",
                "confidence_score": "-11.834396"
            }
        ]
    },
    {
        "mention": "new enterprise associates",
        "type": "ORG",
        "predictions": [
            {
                "entity": "New Enterprise Associates",
                "confidence_score": "9.0593815"
            },
            {
                "entity": "The New York Enterprise Report",
                "confidence_score": "-11.305951"
            },
            {
                "entity": "New Signature",
                "confidence_score": "-11.628841"
            }
        ]
    },
    {
        "mention": "blue jeans",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Jeans",
                "confidence_score": "2.282659"
            },
            {
                "entity": "Blue Jeans (Lana Del Rey song)",
                "confidence_score": "-9.666917"
            },
            {
                "entity": "BlueFace",
                "confidence_score": "-10.907931"
            }
        ]
    },
    {
        "mention": "accel partners",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Accel (venture capital firm)",
                "confidence_score": "9.500475"
            },
            {
                "entity": "Accel-KKR",
                "confidence_score": "-6.4480467"
            },
            {
                "entity": "ACC Capital Holdings",
                "confidence_score": "-10.721014"
            }
        ]
    },
    {
        "mention": "battery ventures",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Battery Ventures",
                "confidence_score": "12.420905"
            },
            {
                "entity": "Qualcomm Ventures",
                "confidence_score": "-11.184068"
            },
            {
                "entity": "Sierra Ventures",
                "confidence_score": "-11.1925535"
            }
        ]
    },
    {
        "mention": "glynn capital",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Alden Global Capital",
                "confidence_score": "-11.189538"
            },
            {
                "entity": "LRG Capital Funds",
                "confidence_score": "-11.250461"
            },
            {
                "entity": "Sageview Capital",
                "confidence_score": "-11.867641"
            }
        ]
    },
    {
        "mention": "norwest venture partners",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Norwest Venture Partners",
                "confidence_score": "12.630581"
            },
            {
                "entity": "Norwest Corporation",
                "confidence_score": "-9.164546"
            },
            {
                "entity": "Norvergence",
                "confidence_score": "-11.549666"
            }
        ]
    },
    {
        "mention": "quadrille capital",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Quadrangle Group",
                "confidence_score": "-9.868115"
            },
            {
                "entity": "Triangle Capital",
                "confidence_score": "-10.880435"
            },
            {
                "entity": "Quadrangle Development Corporation",
                "confidence_score": "-10.899531"
            }
        ]
    },
    {
        "mention": "new york yankees",
        "type": "ORG",
        "predictions": [
            {
                "entity": "New York Yankees",
                "confidence_score": "3.79796"
            },
            {
                "entity": "2017 New York Yankees season",
                "confidence_score": "-4.8271875"
            },
            {
                "entity": "2010 New York Yankees season",
                "confidence_score": "-4.887062"
            }
        ]
    },
    {
        "mention": "derek jeter",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Derek Jeter",
                "confidence_score": "12.209455"
            },
            {
                "entity": "Shawn Jeter",
                "confidence_score": "-11.026099"
            },
            {
                "entity": "Derek Milman",
                "confidence_score": "-11.789318"
            }
        ]
    },
    {
        "mention": "blue jeans",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Jeans",
                "confidence_score": "2.282659"
            },
            {
                "entity": "Blue Jeans (Lana Del Rey song)",
                "confidence_score": "-9.666917"
            },
            {
                "entity": "BlueFace",
                "confidence_score": "-10.907931"
            }
        ]
    },
    {
        "mention": "vbprofiles",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Bank of America",
                "confidence_score": "-7.613111"
            },
            {
                "entity": "Volvo Financial Services",
                "confidence_score": "-10.141739"
            },
            {
                "entity": "The Pepsi Bottling Group",
                "confidence_score": "-10.230353"
            }
        ]
    },
    {
        "mention": "antoinette jadaone",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Antoinette Jadaone",
                "confidence_score": "11.791148"
            },
            {
                "entity": "Antoinette Taus",
                "confidence_score": "-10.238744"
            },
            {
                "entity": "Tiara Jacquelina",
                "confidence_score": "-11.941598"
            }
        ]
    },
    {
        "mention": "jojo saguin",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "JoJo (singer)",
                "confidence_score": "-10.462535"
            },
            {
                "entity": "Jojo Acuin",
                "confidence_score": "-11.160273"
            },
            {
                "entity": "JoJo Siwa",
                "confidence_score": "-11.429673"
            }
        ]
    },
    {
        "mention": "james reid",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "James Reid (actor)",
                "confidence_score": "7.0040193"
            },
            {
                "entity": "James Reid (cyclist)",
                "confidence_score": "-5.855991"
            },
            {
                "entity": "James Reid (ice hockey)",
                "confidence_score": "-6.920304"
            }
        ]
    },
    {
        "mention": "nadine lustre",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Nadine Lustre",
                "confidence_score": "11.469524"
            },
            {
                "entity": "Nadine van der Velde",
                "confidence_score": "-10.815067"
            },
            {
                "entity": "Nadine Rose Mulkerrin",
                "confidence_score": "-10.997639"
            }
        ]
    },
    {
        "mention": "reid",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Reid",
                "confidence_score": "-2.6229036"
            },
            {
                "entity": "Alex Reid (fighter)",
                "confidence_score": "-8.508979"
            },
            {
                "entity": "List of The Voice (American TV series) contestants",
                "confidence_score": "-11.008271"
            }
        ]
    },
    {
        "mention": "lustre",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Nadine Lustre",
                "confidence_score": "9.068319"
            },
            {
                "entity": "Autodesk Lustre",
                "confidence_score": "-6.8713784"
            },
            {
                "entity": "Lustre (band)",
                "confidence_score": "-10.871447"
            }
        ]
    },
    {
        "mention": "ysabel ortega",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ysabel Ortega",
                "confidence_score": "10.637683"
            },
            {
                "entity": "Ysabella",
                "confidence_score": "-11.318709"
            },
            {
                "entity": "Jeannie Ortega",
                "confidence_score": "-11.343546"
            }
        ]
    },
    {
        "mention": "abs-cbn",
        "type": "ORG",
        "predictions": [
            {
                "entity": "ABS-CBN",
                "confidence_score": "11.421739"
            },
            {
                "entity": "ABS-CBN (TV network)",
                "confidence_score": "7.1703105"
            },
            {
                "entity": "CBS",
                "confidence_score": "-5.0610676"
            }
        ]
    },
    {
        "mention": "the filipino channel",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Filipino Channel",
                "confidence_score": "10.545868"
            },
            {
                "entity": "Truth Channel",
                "confidence_score": "-9.85997"
            },
            {
                "entity": "The Truth Channel",
                "confidence_score": "-10.177686"
            }
        ]
    },
    {
        "mention": "clark",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Clark",
                "confidence_score": "-1.728861"
            },
            {
                "entity": "The Clark Sisters",
                "confidence_score": "-9.661612"
            },
            {
                "entity": "Clark Beckham",
                "confidence_score": "-10.68189"
            }
        ]
    },
    {
        "mention": "clark medina",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Clark Middleton",
                "confidence_score": "-11.3064375"
            },
            {
                "entity": "Clark Kent (producer)",
                "confidence_score": "-11.603168"
            },
            {
                "entity": "Clark Beckham",
                "confidence_score": "-11.746989"
            }
        ]
    },
    {
        "mention": "leah olivar",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Roger Olivares",
                "confidence_score": "-10.071152"
            },
            {
                "entity": "Vanessa Olivarez",
                "confidence_score": "-10.710721"
            },
            {
                "entity": "Melissa Lozada-Oliva",
                "confidence_score": "-12.008957"
            }
        ]
    },
    {
        "mention": "filipina",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Philippines",
                "confidence_score": "-3.3341453"
            },
            {
                "entity": "Filipino Americans",
                "confidence_score": "-7.5425434"
            },
            {
                "entity": "Filipinas",
                "confidence_score": "-10.698062"
            }
        ]
    },
    {
        "mention": "leah",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Leah",
                "confidence_score": "-3.7191734"
            },
            {
                "entity": "Leah Moore",
                "confidence_score": "-10.624793"
            },
            {
                "entity": "Leah Clark",
                "confidence_score": "-11.25864"
            }
        ]
    },
    {
        "mention": "rona",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Rona, Inc.",
                "confidence_score": "-7.7967706"
            },
            {
                "entity": "North Rona",
                "confidence_score": "-7.8725495"
            },
            {
                "entity": "Ronya",
                "confidence_score": "-11.325486"
            }
        ]
    },
    {
        "mention": "leah",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Leah",
                "confidence_score": "-3.7191734"
            },
            {
                "entity": "Leah Moore",
                "confidence_score": "-10.624793"
            },
            {
                "entity": "Leah Clark",
                "confidence_score": "-11.25864"
            }
        ]
    },
    {
        "mention": "leah",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Leah",
                "confidence_score": "-3.7191734"
            },
            {
                "entity": "Leah Moore",
                "confidence_score": "-10.624793"
            },
            {
                "entity": "Leah Clark",
                "confidence_score": "-11.25864"
            }
        ]
    },
    {
        "mention": "sol",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Sol (musician)",
                "confidence_score": "-9.190414"
            },
            {
                "entity": "Sol Press",
                "confidence_score": "-10.409034"
            },
            {
                "entity": "Sol (band)",
                "confidence_score": "-11.190997"
            }
        ]
    },
    {
        "mention": "leah",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Leah",
                "confidence_score": "-3.7191734"
            },
            {
                "entity": "Leah Moore",
                "confidence_score": "-10.624793"
            },
            {
                "entity": "Leah Clark",
                "confidence_score": "-11.25864"
            }
        ]
    },
    {
        "mention": "jack",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jack (webcomic)",
                "confidence_score": "-8.818204"
            },
            {
                "entity": "Jack (band)",
                "confidence_score": "-8.981212"
            },
            {
                "entity": "Jack (Tekken)",
                "confidence_score": "-10.811569"
            }
        ]
    },
    {
        "mention": "jiggs",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jiggs (chimpanzee)",
                "confidence_score": "-5.848668"
            },
            {
                "entity": "Jiggs, Nevada",
                "confidence_score": "-7.1838903"
            },
            {
                "entity": "Jibbs",
                "confidence_score": "-10.891191"
            }
        ]
    },
    {
        "mention": "leah",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Leah",
                "confidence_score": "-3.7191734"
            },
            {
                "entity": "Leah Moore",
                "confidence_score": "-10.624793"
            },
            {
                "entity": "Leah Clark",
                "confidence_score": "-11.25864"
            }
        ]
    },
    {
        "mention": "leah",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Leah",
                "confidence_score": "-3.7191734"
            },
            {
                "entity": "Leah Moore",
                "confidence_score": "-10.624793"
            },
            {
                "entity": "Leah Clark",
                "confidence_score": "-11.25864"
            }
        ]
    },
    {
        "mention": "jack",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jack (webcomic)",
                "confidence_score": "-8.818204"
            },
            {
                "entity": "Jack (band)",
                "confidence_score": "-8.981212"
            },
            {
                "entity": "Jack (Tekken)",
                "confidence_score": "-10.811569"
            }
        ]
    },
    {
        "mention": "leah",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Leah",
                "confidence_score": "-3.7191734"
            },
            {
                "entity": "Leah Moore",
                "confidence_score": "-10.624793"
            },
            {
                "entity": "Leah Clark",
                "confidence_score": "-11.25864"
            }
        ]
    },
    {
        "mention": "jack",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jack (webcomic)",
                "confidence_score": "-8.818204"
            },
            {
                "entity": "Jack (band)",
                "confidence_score": "-8.981212"
            },
            {
                "entity": "Jack (Tekken)",
                "confidence_score": "-10.811569"
            }
        ]
    },
    {
        "mention": "clark",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Clark",
                "confidence_score": "-1.728861"
            },
            {
                "entity": "The Clark Sisters",
                "confidence_score": "-9.661612"
            },
            {
                "entity": "Clark Beckham",
                "confidence_score": "-10.68189"
            }
        ]
    },
    {
        "mention": "clark",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Clark",
                "confidence_score": "-1.728861"
            },
            {
                "entity": "The Clark Sisters",
                "confidence_score": "-9.661612"
            },
            {
                "entity": "Clark Beckham",
                "confidence_score": "-10.68189"
            }
        ]
    },
    {
        "mention": "ofelia",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Typhoon Ofelia",
                "confidence_score": "-6.3685555"
            },
            {
                "entity": "Ofelia Garcia",
                "confidence_score": "-7.900685"
            },
            {
                "entity": "Ofelia Hernandez",
                "confidence_score": "-8.016454"
            }
        ]
    },
    {
        "mention": "clark",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Clark",
                "confidence_score": "-1.728861"
            },
            {
                "entity": "The Clark Sisters",
                "confidence_score": "-9.661612"
            },
            {
                "entity": "Clark Beckham",
                "confidence_score": "-10.68189"
            }
        ]
    },
    {
        "mention": "clark",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Clark",
                "confidence_score": "-1.728861"
            },
            {
                "entity": "The Clark Sisters",
                "confidence_score": "-9.661612"
            },
            {
                "entity": "Clark Beckham",
                "confidence_score": "-10.68189"
            }
        ]
    },
    {
        "mention": "clark",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Clark",
                "confidence_score": "-1.728861"
            },
            {
                "entity": "The Clark Sisters",
                "confidence_score": "-9.661612"
            },
            {
                "entity": "Clark Beckham",
                "confidence_score": "-10.68189"
            }
        ]
    },
    {
        "mention": "clark",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Clark",
                "confidence_score": "-1.728861"
            },
            {
                "entity": "The Clark Sisters",
                "confidence_score": "-9.661612"
            },
            {
                "entity": "Clark Beckham",
                "confidence_score": "-10.68189"
            }
        ]
    },
    {
        "mention": "clark",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Clark",
                "confidence_score": "-1.728861"
            },
            {
                "entity": "The Clark Sisters",
                "confidence_score": "-9.661612"
            },
            {
                "entity": "Clark Beckham",
                "confidence_score": "-10.68189"
            }
        ]
    },
    {
        "mention": "jack",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jack (webcomic)",
                "confidence_score": "-8.818204"
            },
            {
                "entity": "Jack (band)",
                "confidence_score": "-8.981212"
            },
            {
                "entity": "Jack (Tekken)",
                "confidence_score": "-10.811569"
            }
        ]
    },
    {
        "mention": "leah",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Leah",
                "confidence_score": "-3.7191734"
            },
            {
                "entity": "Leah Moore",
                "confidence_score": "-10.624793"
            },
            {
                "entity": "Leah Clark",
                "confidence_score": "-11.25864"
            }
        ]
    },
    {
        "mention": "clark",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Clark",
                "confidence_score": "-1.728861"
            },
            {
                "entity": "The Clark Sisters",
                "confidence_score": "-9.661612"
            },
            {
                "entity": "Clark Beckham",
                "confidence_score": "-10.68189"
            }
        ]
    },
    {
        "mention": "leah",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Leah",
                "confidence_score": "-3.7191734"
            },
            {
                "entity": "Leah Moore",
                "confidence_score": "-10.624793"
            },
            {
                "entity": "Leah Clark",
                "confidence_score": "-11.25864"
            }
        ]
    },
    {
        "mention": "clark",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Clark",
                "confidence_score": "-1.728861"
            },
            {
                "entity": "The Clark Sisters",
                "confidence_score": "-9.661612"
            },
            {
                "entity": "Clark Beckham",
                "confidence_score": "-10.68189"
            }
        ]
    },
    {
        "mention": "the immigration and naturalization service",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Immigration and Naturalization Service",
                "confidence_score": "4.455292"
            },
            {
                "entity": "United States Citizenship and Immigration Services",
                "confidence_score": "-4.51466"
            },
            {
                "entity": "Irish Naturalisation and Immigration Service",
                "confidence_score": "-8.6830845"
            }
        ]
    },
    {
        "mention": "leah",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Leah",
                "confidence_score": "-3.7191734"
            },
            {
                "entity": "Leah Moore",
                "confidence_score": "-10.624793"
            },
            {
                "entity": "Leah Clark",
                "confidence_score": "-11.25864"
            }
        ]
    },
    {
        "mention": "clark",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Clark",
                "confidence_score": "-1.728861"
            },
            {
                "entity": "The Clark Sisters",
                "confidence_score": "-9.661612"
            },
            {
                "entity": "Clark Beckham",
                "confidence_score": "-10.68189"
            }
        ]
    },
    {
        "mention": "leah",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Leah",
                "confidence_score": "-3.7191734"
            },
            {
                "entity": "Leah Moore",
                "confidence_score": "-10.624793"
            },
            {
                "entity": "Leah Clark",
                "confidence_score": "-11.25864"
            }
        ]
    },
    {
        "mention": "clark",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Clark",
                "confidence_score": "-1.728861"
            },
            {
                "entity": "The Clark Sisters",
                "confidence_score": "-9.661612"
            },
            {
                "entity": "Clark Beckham",
                "confidence_score": "-10.68189"
            }
        ]
    },
    {
        "mention": "leah",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Leah",
                "confidence_score": "-3.7191734"
            },
            {
                "entity": "Leah Moore",
                "confidence_score": "-10.624793"
            },
            {
                "entity": "Leah Clark",
                "confidence_score": "-11.25864"
            }
        ]
    },
    {
        "mention": "clark",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Clark",
                "confidence_score": "-1.728861"
            },
            {
                "entity": "The Clark Sisters",
                "confidence_score": "-9.661612"
            },
            {
                "entity": "Clark Beckham",
                "confidence_score": "-10.68189"
            }
        ]
    },
    {
        "mention": "pinoy tambayan",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Pinoy Weekly",
                "confidence_score": "-8.183962"
            },
            {
                "entity": "Pinoy Idol",
                "confidence_score": "-8.888964"
            },
            {
                "entity": "Pinay Ballers League",
                "confidence_score": "-11.420327"
            }
        ]
    },
    {
        "mention": "tambayan replay",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Virtual replay",
                "confidence_score": "-10.510716"
            },
            {
                "entity": "2011 Tampere Open",
                "confidence_score": "-11.987852"
            },
            {
                "entity": "Replay (Tamta song)",
                "confidence_score": "-12.373687"
            }
        ]
    },
    {
        "mention": "gmt",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Greenwich Mean Time",
                "confidence_score": "9.210631"
            },
            {
                "entity": "GMT (TV programme)",
                "confidence_score": "-2.9180982"
            },
            {
                "entity": "Central European Time",
                "confidence_score": "-4.724177"
            }
        ]
    },
    {
        "mention": "harry burlton",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Harry Bickerton",
                "confidence_score": "-8.912656"
            },
            {
                "entity": "Harry Lewiston",
                "confidence_score": "-10.781564"
            },
            {
                "entity": "Harry Samuel Bickerton Brindley",
                "confidence_score": "-11.679765"
            }
        ]
    },
    {
        "mention": "zoe",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Zoe Williams",
                "confidence_score": "-6.901143"
            },
            {
                "entity": "Zoe French",
                "confidence_score": "-6.90927"
            },
            {
                "entity": "Z\u00f6e Salmon",
                "confidence_score": "-8.836568"
            }
        ]
    },
    {
        "mention": "soh mui mui",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Tan Chui Mui",
                "confidence_score": "-11.229165"
            },
            {
                "entity": "Ng Mui Wui",
                "confidence_score": "-11.330452"
            },
            {
                "entity": "Mu Zi",
                "confidence_score": "-11.840341"
            }
        ]
    },
    {
        "mention": "hi-lily",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Lily Ho Ngo Yee",
                "confidence_score": "-9.918379"
            },
            {
                "entity": "Hi-Hi",
                "confidence_score": "-11.220764"
            },
            {
                "entity": "Lily M",
                "confidence_score": "-12.631616"
            }
        ]
    },
    {
        "mention": "soh kau kau",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kao Chin Su-mei",
                "confidence_score": "-11.257447"
            },
            {
                "entity": "So Sau Wah",
                "confidence_score": "-11.439664"
            },
            {
                "entity": "John Kao",
                "confidence_score": "-11.659772"
            }
        ]
    },
    {
        "mention": "ching lai fong",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ching Cheong",
                "confidence_score": "-10.963683"
            },
            {
                "entity": "Mei Fong",
                "confidence_score": "-11.229758"
            },
            {
                "entity": "Lung Fong",
                "confidence_score": "-11.501388"
            }
        ]
    },
    {
        "mention": "soh mui mui's",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "MUI Group",
                "confidence_score": "-5.5312004"
            },
            {
                "entity": "Tan Chui Mui",
                "confidence_score": "-10.988402"
            },
            {
                "entity": "Mui Siu-ching",
                "confidence_score": "-11.552148"
            }
        ]
    },
    {
        "mention": "gm",
        "type": "ORG",
        "predictions": [
            {
                "entity": "General manager",
                "confidence_score": "5.0072765"
            },
            {
                "entity": "Grandmaster (chess)",
                "confidence_score": "-1.2808145"
            },
            {
                "entity": "Gamemaster",
                "confidence_score": "-3.09268"
            }
        ]
    },
    {
        "mention": "jm",
        "type": "ORG",
        "predictions": [
            {
                "entity": "JMJ Records",
                "confidence_score": "-9.975553"
            },
            {
                "entity": "Jamiatu Muslim Mindanao",
                "confidence_score": "-10.15587"
            },
            {
                "entity": "Johnson Stokes & Master",
                "confidence_score": "-10.703353"
            }
        ]
    },
    {
        "mention": "jms song bird",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "JHawk",
                "confidence_score": "-11.190145"
            },
            {
                "entity": "JMSN",
                "confidence_score": "-11.247088"
            },
            {
                "entity": "J-STARS",
                "confidence_score": "-12.2554865"
            }
        ]
    },
    {
        "mention": "circle",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Circle (band)",
                "confidence_score": "-10.755982"
            },
            {
                "entity": "Circle (jazz band)",
                "confidence_score": "-10.929894"
            },
            {
                "entity": "Circle II Circle",
                "confidence_score": "-11.399422"
            }
        ]
    },
    {
        "mention": "gm",
        "type": "ORG",
        "predictions": [
            {
                "entity": "General manager",
                "confidence_score": "5.0072765"
            },
            {
                "entity": "Grandmaster (chess)",
                "confidence_score": "-1.2808145"
            },
            {
                "entity": "Gamemaster",
                "confidence_score": "-3.09268"
            }
        ]
    },
    {
        "mention": "silent gorilla",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "The Silent Warrior",
                "confidence_score": "-10.633433"
            },
            {
                "entity": "Silent Servant",
                "confidence_score": "-10.718211"
            },
            {
                "entity": "Silent Circle",
                "confidence_score": "-11.286637"
            }
        ]
    },
    {
        "mention": "poker face",
        "type": "ORG",
        "predictions": [
            {
                "entity": "PokerFace",
                "confidence_score": "7.564985"
            },
            {
                "entity": "Poker Faces",
                "confidence_score": "-6.401858"
            },
            {
                "entity": "Poker face",
                "confidence_score": "-11.152776"
            }
        ]
    },
    {
        "mention": "silent gorilla",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "The Silent Warrior",
                "confidence_score": "-10.633433"
            },
            {
                "entity": "Silent Servant",
                "confidence_score": "-10.718211"
            },
            {
                "entity": "Silent Circle",
                "confidence_score": "-11.286637"
            }
        ]
    },
    {
        "mention": "warlord",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Warlord",
                "confidence_score": "4.802262"
            },
            {
                "entity": "Warlord (manhua)",
                "confidence_score": "-5.2037544"
            },
            {
                "entity": "Warlord (band)",
                "confidence_score": "-9.199963"
            }
        ]
    },
    {
        "mention": "messy",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Steve Huffman",
                "confidence_score": "-10.678642"
            },
            {
                "entity": "Andy Dinh",
                "confidence_score": "-10.9869375"
            },
            {
                "entity": "Messy",
                "confidence_score": "-12.057201"
            }
        ]
    },
    {
        "mention": "gm",
        "type": "ORG",
        "predictions": [
            {
                "entity": "General manager",
                "confidence_score": "5.0072765"
            },
            {
                "entity": "Grandmaster (chess)",
                "confidence_score": "-1.2808145"
            },
            {
                "entity": "Gamemaster",
                "confidence_score": "-3.09268"
            }
        ]
    },
    {
        "mention": "soh mui mui",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Tan Chui Mui",
                "confidence_score": "-11.229165"
            },
            {
                "entity": "Ng Mui Wui",
                "confidence_score": "-11.330452"
            },
            {
                "entity": "Mu Zi",
                "confidence_score": "-11.840341"
            }
        ]
    },
    {
        "mention": "mango",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Mango",
                "confidence_score": "3.5163772"
            },
            {
                "entity": "Mango (singer)",
                "confidence_score": "-10.348327"
            },
            {
                "entity": "Mango (group)",
                "confidence_score": "-11.340237"
            }
        ]
    },
    {
        "mention": "loud chick",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Football at the 2017 Southeast Asian Games \u2013 Women's team squads",
                "confidence_score": "-11.868988"
            },
            {
                "entity": "Paragliding at the 2018 Asian Games \u2013 Women's individual accuracy",
                "confidence_score": "-12.209268"
            },
            {
                "entity": "Boom chick",
                "confidence_score": "-12.988702"
            }
        ]
    },
    {
        "mention": "mango",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Mango",
                "confidence_score": "4.76069"
            },
            {
                "entity": "Mango (color)",
                "confidence_score": "-6.594252"
            },
            {
                "entity": "Mango (group)",
                "confidence_score": "-10.330002"
            }
        ]
    },
    {
        "mention": "soh miow miow",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ki Su U Ma I (Kiss Your Mind) / S.O.S. (Smile on Smile)",
                "confidence_score": "-10.563547"
            },
            {
                "entity": "Kih\u014d, Mie",
                "confidence_score": "-12.170186"
            },
            {
                "entity": "Shirow Miwa",
                "confidence_score": "-12.463885"
            }
        ]
    },
    {
        "mention": "messy",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Steve Huffman",
                "confidence_score": "-10.678642"
            },
            {
                "entity": "Andy Dinh",
                "confidence_score": "-10.9869375"
            },
            {
                "entity": "Messy",
                "confidence_score": "-12.057201"
            }
        ]
    },
    {
        "mention": "mango",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Mango",
                "confidence_score": "4.76069"
            },
            {
                "entity": "Mango (color)",
                "confidence_score": "-6.594252"
            },
            {
                "entity": "Mango (group)",
                "confidence_score": "-10.330002"
            }
        ]
    },
    {
        "mention": "jms",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "JMSN",
                "confidence_score": "-10.70329"
            },
            {
                "entity": "JLS",
                "confidence_score": "-11.227996"
            },
            {
                "entity": "JS Pictures",
                "confidence_score": "-11.521864"
            }
        ]
    },
    {
        "mention": "jms",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "JMSN",
                "confidence_score": "-10.70329"
            },
            {
                "entity": "JLS",
                "confidence_score": "-11.227996"
            },
            {
                "entity": "JS Pictures",
                "confidence_score": "-11.521864"
            }
        ]
    },
    {
        "mention": "michael",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Michael Jackson",
                "confidence_score": "-5.9904094"
            },
            {
                "entity": "Michael Michael",
                "confidence_score": "-7.5331197"
            },
            {
                "entity": "Michael Tao",
                "confidence_score": "-7.697785"
            }
        ]
    },
    {
        "mention": "charles tan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Charles Chong",
                "confidence_score": "-8.436384"
            },
            {
                "entity": "Charles Ng",
                "confidence_score": "-11.713767"
            },
            {
                "entity": "Rico Tan",
                "confidence_score": "-11.817546"
            }
        ]
    },
    {
        "mention": "tongkat",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Kitkat (comedian)",
                "confidence_score": "-10.508563"
            },
            {
                "entity": "Boomkat",
                "confidence_score": "-11.304426"
            },
            {
                "entity": "Sinosikat?",
                "confidence_score": "-11.796936"
            }
        ]
    },
    {
        "mention": "kirie kimono",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kiko Mizuhara",
                "confidence_score": "-11.392098"
            },
            {
                "entity": "Kim Ji-yeon",
                "confidence_score": "-11.862701"
            },
            {
                "entity": "Kei (singer)",
                "confidence_score": "-12.017795"
            }
        ]
    },
    {
        "mention": "chin",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Chin people",
                "confidence_score": "-0.3282439"
            },
            {
                "entity": "Japanese Chin",
                "confidence_score": "-6.8733068"
            },
            {
                "entity": "CHIN",
                "confidence_score": "-12.342367"
            }
        ]
    },
    {
        "mention": "gm",
        "type": "ORG",
        "predictions": [
            {
                "entity": "General manager",
                "confidence_score": "5.0072765"
            },
            {
                "entity": "Grandmaster (chess)",
                "confidence_score": "-1.2808145"
            },
            {
                "entity": "Gamemaster",
                "confidence_score": "-3.09268"
            }
        ]
    },
    {
        "mention": "styrofoam",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Styrofoam",
                "confidence_score": "8.773667"
            },
            {
                "entity": "Styrofoam (musician)",
                "confidence_score": "-3.585302"
            },
            {
                "entity": "Paragliding at the 2018 Asian Games \u2013 Women's individual accuracy",
                "confidence_score": "-12.301669"
            }
        ]
    },
    {
        "mention": "nrl",
        "type": "ORG",
        "predictions": [
            {
                "entity": "National Rugby League",
                "confidence_score": "7.383065"
            },
            {
                "entity": "2014 NRL season",
                "confidence_score": "-6.1423717"
            },
            {
                "entity": "2010 NRL season",
                "confidence_score": "-7.71929"
            }
        ]
    },
    {
        "mention": "nrl",
        "type": "ORG",
        "predictions": [
            {
                "entity": "National Rugby League",
                "confidence_score": "7.383065"
            },
            {
                "entity": "2014 NRL season",
                "confidence_score": "-6.1423717"
            },
            {
                "entity": "2010 NRL season",
                "confidence_score": "-7.71929"
            }
        ]
    },
    {
        "mention": "nrl",
        "type": "ORG",
        "predictions": [
            {
                "entity": "National Rugby League",
                "confidence_score": "7.383065"
            },
            {
                "entity": "2014 NRL season",
                "confidence_score": "-6.1423717"
            },
            {
                "entity": "2010 NRL season",
                "confidence_score": "-7.71929"
            }
        ]
    },
    {
        "mention": "greg inglis",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Greg Inglis",
                "confidence_score": "12.188951"
            },
            {
                "entity": "Greg Alexander",
                "confidence_score": "-10.22256"
            },
            {
                "entity": "Greg Smith (rugby league)",
                "confidence_score": "-11.0599575"
            }
        ]
    },
    {
        "mention": "rabbitohs",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "South Sydney Rabbitohs",
                "confidence_score": "5.8960776"
            },
            {
                "entity": "Sydney Roosters",
                "confidence_score": "1.6495299"
            },
            {
                "entity": "2003 South Sydney Rabbitohs season",
                "confidence_score": "-9.308834"
            }
        ]
    },
    {
        "mention": "brisbane",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Brisbane",
                "confidence_score": "7.9944043"
            },
            {
                "entity": "Brisbane Broncos",
                "confidence_score": "-2.2302477"
            },
            {
                "entity": "2018 Brisbane Broncos season",
                "confidence_score": "-8.196397"
            }
        ]
    },
    {
        "mention": "roosters",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Roosters",
                "confidence_score": "-0.4277949"
            },
            {
                "entity": "Roosters",
                "confidence_score": "-6.3489757"
            },
            {
                "entity": "Rooster (band)",
                "confidence_score": "-11.344303"
            }
        ]
    },
    {
        "mention": "michael maguire's",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Michael Maguire (rugby league)",
                "confidence_score": "4.5935664"
            },
            {
                "entity": "Michael Maguire (footballer)",
                "confidence_score": "-8.053923"
            },
            {
                "entity": "Michael Maguire (Gaelic footballer)",
                "confidence_score": "-11.683164"
            }
        ]
    },
    {
        "mention": "cronulla",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Cronulla-Sutherland Sharks",
                "confidence_score": "0.72116846"
            },
            {
                "entity": "Cronulla, New South Wales",
                "confidence_score": "-6.9870076"
            },
            {
                "entity": "1977 Cronulla-Sutherland Sharks season",
                "confidence_score": "-8.972044"
            }
        ]
    },
    {
        "mention": "dylan walker",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Dylan Walker",
                "confidence_score": "4.8928905"
            },
            {
                "entity": "Ben Walker",
                "confidence_score": "-10.224699"
            },
            {
                "entity": "Shane Walker (rugby league, born 1978)",
                "confidence_score": "-10.617158"
            }
        ]
    },
    {
        "mention": "alex [johnston]",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Alex Johnston (rugby league)",
                "confidence_score": "-6.1222034"
            },
            {
                "entity": "Alex Johnston",
                "confidence_score": "-7.8639436"
            },
            {
                "entity": "James Johnston (English musician)",
                "confidence_score": "-11.2509575"
            }
        ]
    },
    {
        "mention": "walker",
        "type": "ORG",
        "predictions": [
            {
                "entity": "2014 Dubai 24 Hour",
                "confidence_score": "-11.598541"
            },
            {
                "entity": "2012 PSA World Series Finals",
                "confidence_score": "-11.734194"
            },
            {
                "entity": "2011 PSA World Series Finals",
                "confidence_score": "-11.794743"
            }
        ]
    },
    {
        "mention": "adam reynolds",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Adam Reynolds",
                "confidence_score": "8.54242"
            },
            {
                "entity": "Josh Reynolds",
                "confidence_score": "-10.493801"
            },
            {
                "entity": "Adam Simpson",
                "confidence_score": "-11.233435"
            }
        ]
    },
    {
        "mention": "souths",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "South Sydney Rabbitohs",
                "confidence_score": "6.1558805"
            },
            {
                "entity": "Souths Rugby",
                "confidence_score": "-3.052418"
            },
            {
                "entity": "South (band)",
                "confidence_score": "-11.280638"
            }
        ]
    },
    {
        "mention": "rabbitohs",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "South Sydney Rabbitohs",
                "confidence_score": "5.8960776"
            },
            {
                "entity": "Sydney Roosters",
                "confidence_score": "1.6495299"
            },
            {
                "entity": "2003 South Sydney Rabbitohs season",
                "confidence_score": "-9.308834"
            }
        ]
    },
    {
        "mention": "george burgess",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "George Burgess (rugby league)",
                "confidence_score": "-0.4564182"
            },
            {
                "entity": "George Burgess (entrepreneur)",
                "confidence_score": "-5.7453704"
            },
            {
                "entity": "George Burgess (rugby union)",
                "confidence_score": "-9.799122"
            }
        ]
    },
    {
        "mention": "kane evans",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Kane Evans",
                "confidence_score": "1.5371611"
            },
            {
                "entity": "Kieran Evans",
                "confidence_score": "-11.294506"
            },
            {
                "entity": "Shane Evans",
                "confidence_score": "-11.871839"
            }
        ]
    },
    {
        "mention": "glenn stewart's",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Glenn Stewart",
                "confidence_score": "1.8319211"
            },
            {
                "entity": "Glenn Stewart (badminton)",
                "confidence_score": "-9.87662"
            },
            {
                "entity": "Glenn Lazarus",
                "confidence_score": "-11.121686"
            }
        ]
    },
    {
        "mention": "the catalans dragons",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Catalans Dragons",
                "confidence_score": "6.651554"
            },
            {
                "entity": "Dragons (rugby union)",
                "confidence_score": "-7.509596"
            },
            {
                "entity": "Barcelona Dragons",
                "confidence_score": "-7.513821"
            }
        ]
    },
    {
        "mention": "super league",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Super League",
                "confidence_score": "8.395976"
            },
            {
                "entity": "Rugby Super League (United States)",
                "confidence_score": "-6.9730697"
            },
            {
                "entity": "Super Rugby",
                "confidence_score": "-7.192388"
            }
        ]
    },
    {
        "mention": "burgess",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Tom Burgess (rugby league)",
                "confidence_score": "-4.5346656"
            },
            {
                "entity": "Luke Burgess (rugby league)",
                "confidence_score": "-5.3665237"
            },
            {
                "entity": "George Burgess (rugby league)",
                "confidence_score": "-5.911858"
            }
        ]
    },
    {
        "mention": "reynolds",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Josh Reynolds",
                "confidence_score": "-5.951422"
            },
            {
                "entity": "Ben Reynolds (Australian rugby league)",
                "confidence_score": "-7.874339"
            },
            {
                "entity": "Kirk Reynoldson",
                "confidence_score": "-9.576205"
            }
        ]
    },
    {
        "mention": "walker",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Willie Walker (rugby union)",
                "confidence_score": "-10.905626"
            },
            {
                "entity": "Eli Walker",
                "confidence_score": "-11.29544"
            },
            {
                "entity": "Walker Zimmerman",
                "confidence_score": "-11.718762"
            }
        ]
    },
    {
        "mention": "stewart",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Wayne Stewart",
                "confidence_score": "-9.795752"
            },
            {
                "entity": "Cam Stewart",
                "confidence_score": "-10.294998"
            },
            {
                "entity": "Jordan Stewart (speedway rider)",
                "confidence_score": "-10.533441"
            }
        ]
    },
    {
        "mention": "liz raftery",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Raftery",
                "confidence_score": "-9.703352"
            },
            {
                "entity": "Mary Raftery",
                "confidence_score": "-9.832445"
            },
            {
                "entity": "Rafter (band)",
                "confidence_score": "-12.345071"
            }
        ]
    },
    {
        "mention": "sam smith",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Sam Smith (singer)",
                "confidence_score": "-3.1793518"
            },
            {
                "entity": "Sam Smith (actor)",
                "confidence_score": "-5.7398524"
            },
            {
                "entity": "Sam Smith (artist)",
                "confidence_score": "-8.137803"
            }
        ]
    },
    {
        "mention": "james bond",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "James Bond",
                "confidence_score": "8.675059"
            },
            {
                "entity": "John James Bond",
                "confidence_score": "-10.807872"
            },
            {
                "entity": "Jess Bond",
                "confidence_score": "-11.417623"
            }
        ]
    },
    {
        "mention": "spectre",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "SPECTRE",
                "confidence_score": "7.600545"
            },
            {
                "entity": "Spectre (2015 film)",
                "confidence_score": "-1.5732436"
            },
            {
                "entity": "In/Spectre",
                "confidence_score": "-5.770476"
            }
        ]
    },
    {
        "mention": "itunes",
        "type": "ORG",
        "predictions": [
            {
                "entity": "ITunes",
                "confidence_score": "10.766058"
            },
            {
                "entity": "ITunes Store",
                "confidence_score": "1.2198566"
            },
            {
                "entity": "History of iTunes",
                "confidence_score": "-5.508824"
            }
        ]
    },
    {
        "mention": "bond",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Edward Bond",
                "confidence_score": "-4.6062846"
            },
            {
                "entity": "Bond (band)",
                "confidence_score": "-4.731076"
            },
            {
                "entity": "Nelson S. Bond",
                "confidence_score": "-5.7745123"
            }
        ]
    },
    {
        "mention": "michael g. wilson",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Michael G. Wilson",
                "confidence_score": "10.987083"
            },
            {
                "entity": "Michael Wilson (presenter)",
                "confidence_score": "-8.286804"
            },
            {
                "entity": "Michael J. Wilson",
                "confidence_score": "-9.844409"
            }
        ]
    },
    {
        "mention": "barbara broccoli",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Barbara Broccoli",
                "confidence_score": "11.989828"
            },
            {
                "entity": "Barbara S. Brogliatti",
                "confidence_score": "-11.42736"
            },
            {
                "entity": "Barbara Brousal",
                "confidence_score": "-11.9119425"
            }
        ]
    },
    {
        "mention": "the lonely hour",
        "type": "LOC",
        "predictions": [
            {
                "entity": "In the Lonely Hour",
                "confidence_score": "-4.666655"
            },
            {
                "entity": "The Lonely Island",
                "confidence_score": "-9.575347"
            },
            {
                "entity": "The Lonely Hours",
                "confidence_score": "-9.658789"
            }
        ]
    },
    {
        "mention": "adele",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Adele",
                "confidence_score": "9.025246"
            },
            {
                "entity": "Ad\u00e8le Anderson",
                "confidence_score": "-10.310531"
            },
            {
                "entity": "Adel\u00e9n",
                "confidence_score": "-10.80429"
            }
        ]
    },
    {
        "mention": "oscar",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Academy Awards",
                "confidence_score": "7.3065944"
            },
            {
                "entity": "Academy Award for Best Actor",
                "confidence_score": "-8.372924"
            },
            {
                "entity": "Academy Award for Best Actress",
                "confidence_score": "-8.551124"
            }
        ]
    },
    {
        "mention": "golden globe",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "69th Golden Globe Awards",
                "confidence_score": "-6.758411"
            },
            {
                "entity": "72nd Golden Globe Awards",
                "confidence_score": "-8.160646"
            },
            {
                "entity": "71st Golden Globe Awards",
                "confidence_score": "-8.391723"
            }
        ]
    },
    {
        "mention": "spectre",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "SPECTRE",
                "confidence_score": "7.600545"
            },
            {
                "entity": "Spectre (2015 film)",
                "confidence_score": "-1.5732436"
            },
            {
                "entity": "In/Spectre",
                "confidence_score": "-5.770476"
            }
        ]
    },
    {
        "mention": "reuters",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Reuters",
                "confidence_score": "9.683029"
            },
            {
                "entity": "Thomson Reuters",
                "confidence_score": "-4.8229494"
            },
            {
                "entity": "Reuters Group",
                "confidence_score": "-5.128231"
            }
        ]
    },
    {
        "mention": "halo trust",
        "type": "ORG",
        "predictions": [
            {
                "entity": "HALO Trust",
                "confidence_score": "6.8039308"
            },
            {
                "entity": "The HALO Corporation",
                "confidence_score": "-10.430485"
            },
            {
                "entity": "HALO Technology Holdings",
                "confidence_score": "-11.106395"
            }
        ]
    },
    {
        "mention": "mozambique",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Mozambique",
                "confidence_score": "4.787083"
            },
            {
                "entity": "Island of Mozambique",
                "confidence_score": "-6.0774693"
            },
            {
                "entity": "History of Mozambique",
                "confidence_score": "-8.389184"
            }
        ]
    },
    {
        "mention": "halo trust",
        "type": "ORG",
        "predictions": [
            {
                "entity": "HALO Trust",
                "confidence_score": "6.8039308"
            },
            {
                "entity": "The HALO Corporation",
                "confidence_score": "-10.430485"
            },
            {
                "entity": "HALO Technology Holdings",
                "confidence_score": "-11.106395"
            }
        ]
    },
    {
        "mention": "the indian ocean",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Indian Ocean",
                "confidence_score": "1.915093"
            },
            {
                "entity": "Sea",
                "confidence_score": "-5.9247346"
            },
            {
                "entity": "Southern Ocean",
                "confidence_score": "-6.143732"
            }
        ]
    },
    {
        "mention": "mozambique",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Mozambique",
                "confidence_score": "4.787083"
            },
            {
                "entity": "Island of Mozambique",
                "confidence_score": "-6.0774693"
            },
            {
                "entity": "History of Mozambique",
                "confidence_score": "-8.389184"
            }
        ]
    },
    {
        "mention": "ed stoddard",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "George D. Stoddard",
                "confidence_score": "-10.169612"
            },
            {
                "entity": "A. B. Stoddard",
                "confidence_score": "-10.240043"
            },
            {
                "entity": "Eugene C. Stoddard",
                "confidence_score": "-11.389059"
            }
        ]
    },
    {
        "mention": "joe brock",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Pete Brock",
                "confidence_score": "-10.515062"
            },
            {
                "entity": "Dave Brock",
                "confidence_score": "-10.628099"
            },
            {
                "entity": "Wayne Brock",
                "confidence_score": "-11.039817"
            }
        ]
    },
    {
        "mention": "george brandis",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "George Brandis",
                "confidence_score": "10.065862"
            },
            {
                "entity": "George Argyros",
                "confidence_score": "-11.371687"
            },
            {
                "entity": "George Caris",
                "confidence_score": "-11.984139"
            }
        ]
    },
    {
        "mention": "qc",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Queen's Counsel",
                "confidence_score": "5.5903077"
            },
            {
                "entity": "2012 PSA World Series Finals",
                "confidence_score": "-11.684432"
            },
            {
                "entity": "2014\u201315 Maltese First Division",
                "confidence_score": "-11.989502"
            }
        ]
    },
    {
        "mention": "brandis",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "George Brandis",
                "confidence_score": "-4.5354633"
            },
            {
                "entity": "Brandis, Germany",
                "confidence_score": "-7.9329515"
            },
            {
                "entity": "Brandis Raley-Ross",
                "confidence_score": "-11.277959"
            }
        ]
    },
    {
        "mention": "dubbo",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Dubbo",
                "confidence_score": "9.10349"
            },
            {
                "entity": "Electoral district of Dubbo",
                "confidence_score": "-5.7804046"
            },
            {
                "entity": "Dubbo railway station",
                "confidence_score": "-8.416384"
            }
        ]
    },
    {
        "mention": "nsw",
        "type": "ORG",
        "predictions": [
            {
                "entity": "New South Wales",
                "confidence_score": "8.552698"
            },
            {
                "entity": "Australia",
                "confidence_score": "-1.593492"
            },
            {
                "entity": "Wales",
                "confidence_score": "-5.9192986"
            }
        ]
    },
    {
        "mention": "mallee",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mallee (Victoria)",
                "confidence_score": "7.8310328"
            },
            {
                "entity": "Mallee (biogeographic region)",
                "confidence_score": "0.23362777"
            },
            {
                "entity": "Division of Mallee",
                "confidence_score": "-2.5306447"
            }
        ]
    },
    {
        "mention": "alice springs",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Alice Springs",
                "confidence_score": "11.385169"
            },
            {
                "entity": "Town of Alice Springs",
                "confidence_score": "-2.6224804"
            },
            {
                "entity": "Alice Springs Airport",
                "confidence_score": "-6.967815"
            }
        ]
    },
    {
        "mention": "the northern territory",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Northern Territory",
                "confidence_score": "8.391966"
            },
            {
                "entity": "Australian Capital Territory",
                "confidence_score": "-5.8897176"
            },
            {
                "entity": "History of the Northern Territory",
                "confidence_score": "-7.126779"
            }
        ]
    },
    {
        "mention": "kimberley",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kimberley (Western Australia)",
                "confidence_score": "3.6679995"
            },
            {
                "entity": "Kimberley, Northern Cape",
                "confidence_score": "-7.2783775"
            },
            {
                "entity": "Kimberley, British Columbia",
                "confidence_score": "-8.337994"
            }
        ]
    },
    {
        "mention": "brandis",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "George Brandis",
                "confidence_score": "-4.5354633"
            },
            {
                "entity": "Brandis, Germany",
                "confidence_score": "-7.9329515"
            },
            {
                "entity": "Brandis Raley-Ross",
                "confidence_score": "-11.277959"
            }
        ]
    },
    {
        "mention": "inner melbourne community legal's '",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Community Legal Centre",
                "confidence_score": "-8.219142"
            },
            {
                "entity": "Sisters of Charity Health Service, Melbourne",
                "confidence_score": "-11.043381"
            },
            {
                "entity": "2007 Melbourne CBD shootings",
                "confidence_score": "-11.173959"
            }
        ]
    },
    {
        "mention": "the royal women's hospital",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Royal Women's Hospital",
                "confidence_score": "4.0549836"
            },
            {
                "entity": "Royal Brisbane and Women's Hospital",
                "confidence_score": "-10.033116"
            },
            {
                "entity": "Royal Hospital for Women",
                "confidence_score": "-10.473869"
            }
        ]
    },
    {
        "mention": "brandis",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "George Brandis",
                "confidence_score": "-4.5354633"
            },
            {
                "entity": "Brandis, Germany",
                "confidence_score": "-7.9329515"
            },
            {
                "entity": "Brandis Raley-Ross",
                "confidence_score": "-11.277959"
            }
        ]
    },
    {
        "mention": "the productivity commission's",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Productivity Commission",
                "confidence_score": "1.3167756"
            },
            {
                "entity": "New Zealand Productivity Commission",
                "confidence_score": "-10.480937"
            },
            {
                "entity": "Australian Fair Pay Commission",
                "confidence_score": "-10.767423"
            }
        ]
    },
    {
        "mention": "the sydney morning herald",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Sydney Morning Herald",
                "confidence_score": "9.551835"
            },
            {
                "entity": "The Sun-Herald",
                "confidence_score": "-5.930471"
            },
            {
                "entity": "The Newcastle Herald",
                "confidence_score": "-10.513847"
            }
        ]
    },
    {
        "mention": "the connecticut coalition against domestic violence",
        "type": "ORG",
        "predictions": [
            {
                "entity": "National Coalition Against Domestic Violence",
                "confidence_score": "-9.462857"
            },
            {
                "entity": "New York State Coalition Against Domestic Violence",
                "confidence_score": "-10.599347"
            },
            {
                "entity": "Illinois Coalition Against Domestic Violence",
                "confidence_score": "-12.080613"
            }
        ]
    },
    {
        "mention": "arthur hapgood sr.",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Arthur Doe Jr.",
                "confidence_score": "-12.211156"
            },
            {
                "entity": "Arthur H. Vandenberg Jr.",
                "confidence_score": "-12.39268"
            },
            {
                "entity": "Arthur Langford Jr.",
                "confidence_score": "-12.568445"
            }
        ]
    },
    {
        "mention": "zaniyah calloway",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Lil Mama",
                "confidence_score": "-10.77662"
            },
            {
                "entity": "Honeyhoney",
                "confidence_score": "-11.755089"
            },
            {
                "entity": "Calloway",
                "confidence_score": "-12.567809"
            }
        ]
    },
    {
        "mention": "hapgood",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Hutchins Hapgood",
                "confidence_score": "-6.6143374"
            },
            {
                "entity": "Hapgood",
                "confidence_score": "-11.013846"
            },
            {
                "entity": "Hapgood House",
                "confidence_score": "-11.530589"
            }
        ]
    },
    {
        "mention": "marijuana",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Cannabis (drug)",
                "confidence_score": "1.556718"
            },
            {
                "entity": "Cannabis",
                "confidence_score": "-2.4198964"
            },
            {
                "entity": "Marijuana (film)",
                "confidence_score": "-11.190791"
            }
        ]
    },
    {
        "mention": "pcp",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Phencyclidine",
                "confidence_score": "3.990393"
            },
            {
                "entity": "Personal contract purchase",
                "confidence_score": "-6.322449"
            },
            {
                "entity": "Performance Co-Pilot",
                "confidence_score": "-6.5931764"
            }
        ]
    },
    {
        "mention": "gncc",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Grand National Curling Club",
                "confidence_score": "-4.1400294"
            },
            {
                "entity": "Greater New York Councils",
                "confidence_score": "-8.140412"
            },
            {
                "entity": "New York\u2013New Jersey Trail Conference",
                "confidence_score": "-11.179767"
            }
        ]
    },
    {
        "mention": "skyshot",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Skywriting",
                "confidence_score": "-9.857505"
            },
            {
                "entity": "Skyrunning",
                "confidence_score": "-10.446408"
            },
            {
                "entity": "Skytypers Air Show Team",
                "confidence_score": "-11.801428"
            }
        ]
    },
    {
        "mention": "michael swift",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Michael Swift (ice hockey)",
                "confidence_score": "-8.653503"
            },
            {
                "entity": "Michael Swift (American football)",
                "confidence_score": "-8.711123"
            },
            {
                "entity": "Michael Swift (rugby union)",
                "confidence_score": "-8.736312"
            }
        ]
    },
    {
        "mention": "kevin trantham",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Roy Trantham",
                "confidence_score": "-11.33381"
            },
            {
                "entity": "Kevin VanDam",
                "confidence_score": "-11.70449"
            },
            {
                "entity": "Evan Wickham",
                "confidence_score": "-12.170621"
            }
        ]
    },
    {
        "mention": "maxxis tires",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Max Records",
                "confidence_score": "-11.328092"
            },
            {
                "entity": "Maxxxwell Carlisle",
                "confidence_score": "-11.463344"
            },
            {
                "entity": "Max Creek",
                "confidence_score": "-11.818346"
            }
        ]
    },
    {
        "mention": "trantham",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Roy Trantham",
                "confidence_score": "-8.62604"
            },
            {
                "entity": "Esham",
                "confidence_score": "-10.627847"
            },
            {
                "entity": "Operator Generator",
                "confidence_score": "-11.908186"
            }
        ]
    },
    {
        "mention": "trantham",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Roy Trantham",
                "confidence_score": "-8.62604"
            },
            {
                "entity": "Esham",
                "confidence_score": "-10.627847"
            },
            {
                "entity": "Operator Generator",
                "confidence_score": "-11.908186"
            }
        ]
    },
    {
        "mention": "polaris scrambler 850xp",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australis-OSCAR 5",
                "confidence_score": "-11.665299"
            },
            {
                "entity": "Pearson Ensign",
                "confidence_score": "-11.668766"
            },
            {
                "entity": "Kosmos 850",
                "confidence_score": "-11.866054"
            }
        ]
    },
    {
        "mention": "trantham",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Roy Trantham",
                "confidence_score": "-8.62604"
            },
            {
                "entity": "Esham",
                "confidence_score": "-10.627847"
            },
            {
                "entity": "Operator Generator",
                "confidence_score": "-11.908186"
            }
        ]
    },
    {
        "mention": "don higbee",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Scott Donohue",
                "confidence_score": "-11.231226"
            },
            {
                "entity": "Don Higginbotham",
                "confidence_score": "-11.617112"
            },
            {
                "entity": "Don Hiney",
                "confidence_score": "-12.527684"
            }
        ]
    },
    {
        "mention": "graham widdicombe",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Angus Widdicombe",
                "confidence_score": "-11.699879"
            },
            {
                "entity": "Graham Symonds",
                "confidence_score": "-12.0539255"
            },
            {
                "entity": "Graham Windeatt",
                "confidence_score": "-12.1981125"
            }
        ]
    },
    {
        "mention": "polaris scrambler 850xp\u2019s",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Antarctic Ice Marathon & 100k ultra race",
                "confidence_score": "-11.818307"
            },
            {
                "entity": "Athletics at the 2003 Games of the Small States of Europe",
                "confidence_score": "-12.20882"
            },
            {
                "entity": "Biathlon at the 2015 Winter Universiade \u2013 Men's 10 km sprint",
                "confidence_score": "-12.420411"
            }
        ]
    },
    {
        "mention": "state",
        "type": "ORG",
        "predictions": [
            {
                "entity": "U.S. state",
                "confidence_score": "2.537277"
            },
            {
                "entity": "Athletics at the 2003 Games of the Small States of Europe",
                "confidence_score": "-12.160954"
            },
            {
                "entity": "Paragliding at the 2018 Asian Games \u2013 Women's individual accuracy",
                "confidence_score": "-12.510171"
            }
        ]
    },
    {
        "mention": "scrambler xp1000",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ski jumping at the 2018 Winter Olympics \u2013 Men's large hill team",
                "confidence_score": "-12.131634"
            },
            {
                "entity": "2014 UCI Mountain Bike & Trials World Championships \u2013 Men's junior cross-country",
                "confidence_score": "-12.445587"
            },
            {
                "entity": "2012 IAAF World Indoor Championships \u2013 Men's 4 \u00d7 400 metres relay",
                "confidence_score": "-12.451402"
            }
        ]
    },
    {
        "mention": "jeff wright",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jeff Wright (politician)",
                "confidence_score": "-8.894492"
            },
            {
                "entity": "Jeff Wright (defensive tackle)",
                "confidence_score": "-10.473279"
            },
            {
                "entity": "Jeff Wright (defensive back)",
                "confidence_score": "-10.696263"
            }
        ]
    },
    {
        "mention": "swift",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Swift Aircraft",
                "confidence_score": "-3.6940126"
            },
            {
                "entity": "Swift Engineering",
                "confidence_score": "-6.0308614"
            },
            {
                "entity": "Swift Transportation",
                "confidence_score": "-7.646995"
            }
        ]
    },
    {
        "mention": "polaris scrambler xp850",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "1:10 radio-controlled off-road buggy",
                "confidence_score": "-11.956845"
            },
            {
                "entity": "Ski jumping at the 2018 Winter Olympics \u2013 Men's large hill team",
                "confidence_score": "-12.092523"
            },
            {
                "entity": "2018 Pan Pacific Swimming Championships \u2013 Men's 10 kilometre open water",
                "confidence_score": "-12.636121"
            }
        ]
    },
    {
        "mention": "team uxc\u2019s",
        "type": "ORG",
        "predictions": [
            {
                "entity": "UXC",
                "confidence_score": "-6.770965"
            },
            {
                "entity": "Super Robot Wars UX",
                "confidence_score": "-9.354356"
            },
            {
                "entity": "KTM 450 EXC",
                "confidence_score": "-11.875648"
            }
        ]
    },
    {
        "mention": "swift",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Swift Aircraft",
                "confidence_score": "-3.6940126"
            },
            {
                "entity": "Swift Engineering",
                "confidence_score": "-6.0308614"
            },
            {
                "entity": "Swift Transportation",
                "confidence_score": "-7.646995"
            }
        ]
    },
    {
        "mention": "the customaxis shocks",
        "type": "ORG",
        "predictions": [
            {
                "entity": "List of Super Powers minicomics",
                "confidence_score": "-11.216742"
            },
            {
                "entity": "Starriors",
                "confidence_score": "-11.217558"
            },
            {
                "entity": "XMODS",
                "confidence_score": "-11.227241"
            }
        ]
    },
    {
        "mention": "scrambler",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Scrambler (comics)",
                "confidence_score": "3.3512099"
            },
            {
                "entity": "Triumph Scrambler",
                "confidence_score": "-7.521423"
            },
            {
                "entity": "Scramble (comics)",
                "confidence_score": "-9.538408"
            }
        ]
    },
    {
        "mention": "skyshot",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Skywriting",
                "confidence_score": "-9.857505"
            },
            {
                "entity": "Skyrunning",
                "confidence_score": "-10.446408"
            },
            {
                "entity": "Skytypers Air Show Team",
                "confidence_score": "-11.801428"
            }
        ]
    },
    {
        "mention": "maxxis",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Max Collins (musician)",
                "confidence_score": "-11.073919"
            },
            {
                "entity": "Max Schneider",
                "confidence_score": "-11.085154"
            },
            {
                "entity": "Max Raptor",
                "confidence_score": "-11.898175"
            }
        ]
    },
    {
        "mention": "the dp brakes",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Brakes",
                "confidence_score": "-8.312257"
            },
            {
                "entity": "DMX Krew",
                "confidence_score": "-11.430854"
            },
            {
                "entity": "The Raymond Brake",
                "confidence_score": "-12.253129"
            }
        ]
    },
    {
        "mention": "somerset",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Somerset",
                "confidence_score": "1.4524075"
            },
            {
                "entity": "Somerset County, New Jersey",
                "confidence_score": "-2.330531"
            },
            {
                "entity": "Somerset, New York",
                "confidence_score": "-2.9218085"
            }
        ]
    },
    {
        "mention": "polaris",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Polaris Industries",
                "confidence_score": "2.0648048"
            },
            {
                "entity": "Polaris (band)",
                "confidence_score": "-2.0040252"
            },
            {
                "entity": "Polaris Motor",
                "confidence_score": "-4.439551"
            }
        ]
    },
    {
        "mention": "driven powersports",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Power Challenge",
                "confidence_score": "-11.021035"
            },
            {
                "entity": "Driven To Perform",
                "confidence_score": "-11.23631"
            },
            {
                "entity": "DriveNow",
                "confidence_score": "-11.313512"
            }
        ]
    },
    {
        "mention": "maxxis tires",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Max Records",
                "confidence_score": "-11.328092"
            },
            {
                "entity": "Maxxxwell Carlisle",
                "confidence_score": "-11.463344"
            },
            {
                "entity": "Max Creek",
                "confidence_score": "-11.818346"
            }
        ]
    },
    {
        "mention": "fly racing",
        "type": "ORG",
        "predictions": [
            {
                "entity": "FlyQuest",
                "confidence_score": "-10.329416"
            },
            {
                "entity": "Fly Entertainment",
                "confidence_score": "-10.726794"
            },
            {
                "entity": "Fly Guys",
                "confidence_score": "-11.293313"
            }
        ]
    },
    {
        "mention": "evans coolant",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Evans Tuning",
                "confidence_score": "-10.281847"
            },
            {
                "entity": "Andy Evans (racing driver)",
                "confidence_score": "-11.028104"
            },
            {
                "entity": "Dave Evans (entrepreneur)",
                "confidence_score": "-11.48836"
            }
        ]
    },
    {
        "mention": "rox speed fx",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Roxor Games",
                "confidence_score": "-11.313931"
            },
            {
                "entity": "Rovio Entertainment",
                "confidence_score": "-11.395904"
            },
            {
                "entity": "Illinois Speed Press",
                "confidence_score": "-11.9681635"
            }
        ]
    },
    {
        "mention": "dirtworks, cool-it thermo tech",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Dirt 3",
                "confidence_score": "-11.3040495"
            },
            {
                "entity": "Lazy 8 Studios",
                "confidence_score": "-11.7677355"
            },
            {
                "entity": "Dirtnap Records",
                "confidence_score": "-12.027723"
            }
        ]
    },
    {
        "mention": "leatt",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Leatt-Brace",
                "confidence_score": "-7.5999446"
            },
            {
                "entity": "2014 UCI Mountain Bike & Trials World Championships \u2013 Women's under-23 cross-country",
                "confidence_score": "-12.2519655"
            },
            {
                "entity": "2014 UCI Mountain Bike & Trials World Championships \u2013 Men's under-23 cross-country",
                "confidence_score": "-12.352618"
            }
        ]
    },
    {
        "mention": "ricochet off road",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Evo Street Racers",
                "confidence_score": "-10.328282"
            },
            {
                "entity": "List of Grand Prix motorcycle racers: K",
                "confidence_score": "-12.204774"
            },
            {
                "entity": "List of Grand Prix motorcycle racers: J",
                "confidence_score": "-12.207643"
            }
        ]
    },
    {
        "mention": "spider",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Daniel Quirk",
                "confidence_score": "-5.498012"
            },
            {
                "entity": "Spider Riders",
                "confidence_score": "-8.207725"
            },
            {
                "entity": "Spider Miller",
                "confidence_score": "-10.654417"
            }
        ]
    },
    {
        "mention": "moto pro training",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "MotoAmerica",
                "confidence_score": "-9.352872"
            },
            {
                "entity": "Harley Race's Wrestling Academy",
                "confidence_score": "-10.939101"
            },
            {
                "entity": "Moto Rapido Racing",
                "confidence_score": "-11.60031"
            }
        ]
    },
    {
        "mention": "scott goggles",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Scott Sports",
                "confidence_score": "-10.074452"
            },
            {
                "entity": "Scotty James",
                "confidence_score": "-11.806987"
            },
            {
                "entity": "Scott Gow",
                "confidence_score": "-11.891634"
            }
        ]
    },
    {
        "mention": "k&n kilters, america\u2019s",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Johnson & Johnson",
                "confidence_score": "-7.301103"
            },
            {
                "entity": "K&N Engineering",
                "confidence_score": "-9.060027"
            },
            {
                "entity": "K2 Sports",
                "confidence_score": "-11.135401"
            }
        ]
    },
    {
        "mention": "rigid industries",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Bentley Systems",
                "confidence_score": "-11.13816"
            },
            {
                "entity": "Unison Industries",
                "confidence_score": "-11.348573"
            },
            {
                "entity": "Motion Industries",
                "confidence_score": "-11.351034"
            }
        ]
    },
    {
        "mention": "holz racing products",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Speedway Motorsports",
                "confidence_score": "-9.86752"
            },
            {
                "entity": "Holley Performance Products",
                "confidence_score": "-10.351837"
            },
            {
                "entity": "Hoosier Racing Tire",
                "confidence_score": "-10.490441"
            }
        ]
    },
    {
        "mention": "teixeira tech",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Tektronix",
                "confidence_score": "-10.9082365"
            },
            {
                "entity": "Tetra Tech",
                "confidence_score": "-11.1190605"
            },
            {
                "entity": "Tecnomatix",
                "confidence_score": "-11.191722"
            }
        ]
    },
    {
        "mention": "quad tec",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Quad (unit)",
                "confidence_score": "-8.097997"
            },
            {
                "entity": "Quad Electroacoustics",
                "confidence_score": "-10.930799"
            },
            {
                "entity": "Tesco Corporation",
                "confidence_score": "-10.952512"
            }
        ]
    },
    {
        "mention": "alco cleaner",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Alcoa",
                "confidence_score": "-8.507862"
            },
            {
                "entity": "Alcoa World Alumina and Chemicals",
                "confidence_score": "-10.743859"
            },
            {
                "entity": "Alcoa Power Generating Inc.",
                "confidence_score": "-11.43788"
            }
        ]
    },
    {
        "mention": "polaris power generators",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Polaris Industries",
                "confidence_score": "-8.722072"
            },
            {
                "entity": "Capstone Turbine",
                "confidence_score": "-11.368145"
            },
            {
                "entity": "Primus Power",
                "confidence_score": "-11.412137"
            }
        ]
    },
    {
        "mention": "omf wheels",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "OM Group",
                "confidence_score": "-10.415758"
            },
            {
                "entity": "Omsk Electromechanical Plant",
                "confidence_score": "-10.878589"
            },
            {
                "entity": "United Heavy Machinery",
                "confidence_score": "-11.10468"
            }
        ]
    },
    {
        "mention": "muzzy performance exhaust",
        "type": "ORG",
        "predictions": [
            {
                "entity": "HMF Engineering",
                "confidence_score": "-9.097611"
            },
            {
                "entity": "Musco Lighting",
                "confidence_score": "-11.207447"
            },
            {
                "entity": "Ventilation air methane thermal oxidizer",
                "confidence_score": "-12.171601"
            }
        ]
    },
    {
        "mention": "polaris engineered lubricants",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Polaris Industries",
                "confidence_score": "-9.29434"
            },
            {
                "entity": "Solutia",
                "confidence_score": "-10.909884"
            },
            {
                "entity": "Rayonier Advanced Materials",
                "confidence_score": "-11.654929"
            }
        ]
    },
    {
        "mention": "rcv",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Reactivity controlled compression ignition",
                "confidence_score": "-11.202815"
            },
            {
                "entity": "Riverside Manufacturing LLC",
                "confidence_score": "-11.597365"
            },
            {
                "entity": "RCF audio",
                "confidence_score": "-12.095437"
            }
        ]
    },
    {
        "mention": "airdam clutching, santor design company",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Jacobs Vehicle Systems",
                "confidence_score": "-11.418008"
            },
            {
                "entity": "Lindsay Manufacturing",
                "confidence_score": "-11.516088"
            },
            {
                "entity": "Arnott Air Suspension Products",
                "confidence_score": "-11.748534"
            }
        ]
    },
    {
        "mention": "quad tech",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Quad 4 engine",
                "confidence_score": "-10.777415"
            },
            {
                "entity": "Quad Electroacoustics",
                "confidence_score": "-11.489414"
            },
            {
                "entity": "Aventi Technology",
                "confidence_score": "-11.700445"
            }
        ]
    },
    {
        "mention": "alex trebek",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Alex Trebek",
                "confidence_score": "12.358435"
            },
            {
                "entity": "Alex Trevi\u00f1o",
                "confidence_score": "-11.763276"
            },
            {
                "entity": "Alex Treves",
                "confidence_score": "-11.780206"
            }
        ]
    },
    {
        "mention": "turd ferguson",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "List of characters in Call of Duty 6: Modern Warfare 2",
                "confidence_score": "-11.3215"
            },
            {
                "entity": "Mark D. Ferguson",
                "confidence_score": "-11.711086"
            },
            {
                "entity": "From Raxacoricofallapatorius with Love",
                "confidence_score": "-11.849004"
            }
        ]
    },
    {
        "mention": "mara montalbano",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Rodriguez, Rizal",
                "confidence_score": "-7.7385907"
            },
            {
                "entity": "Michael S. Montalbano",
                "confidence_score": "-10.561075"
            },
            {
                "entity": "James Montalbano",
                "confidence_score": "-10.900941"
            }
        ]
    },
    {
        "mention": "@maramontalbano",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Adam Montoya",
                "confidence_score": "-11.358433"
            },
            {
                "entity": "Ryan Allen Carrillo",
                "confidence_score": "-11.496192"
            },
            {
                "entity": "List of Million Dollar Listing Los Angeles episodes",
                "confidence_score": "-11.603438"
            }
        ]
    },
    {
        "mention": "katherine rees",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Katie Rees",
                "confidence_score": "2.9151437"
            },
            {
                "entity": "Katherine Brooks",
                "confidence_score": "-11.030238"
            },
            {
                "entity": "Teresa Rees",
                "confidence_score": "-11.185421"
            }
        ]
    },
    {
        "mention": "kevin \"dj kulprit\" sims",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kevin Federline",
                "confidence_score": "-10.7784605"
            },
            {
                "entity": "KVEG",
                "confidence_score": "-10.944877"
            },
            {
                "entity": "Kevin Kugler",
                "confidence_score": "-11.885346"
            }
        ]
    },
    {
        "mention": "miss nevada usa",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Miss Nevada USA",
                "confidence_score": "10.774248"
            },
            {
                "entity": "Miss Nevada Teen USA",
                "confidence_score": "-0.5382967"
            },
            {
                "entity": "Miss USA",
                "confidence_score": "-5.0258803"
            }
        ]
    },
    {
        "mention": "sims",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Sims",
                "confidence_score": "-0.3109095"
            },
            {
                "entity": "Title 36 of the United States Code",
                "confidence_score": "-11.21505"
            },
            {
                "entity": "Sims, Illinois",
                "confidence_score": "-11.698894"
            }
        ]
    },
    {
        "mention": "east pebble road",
        "type": "LOC",
        "predictions": [
            {
                "entity": "List of county routes in Elko County, Nevada",
                "confidence_score": "-11.608787"
            },
            {
                "entity": "Nevada State Route 80",
                "confidence_score": "-11.767531"
            },
            {
                "entity": "Nevada State Route 70",
                "confidence_score": "-11.802279"
            }
        ]
    },
    {
        "mention": "sims",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Sims",
                "confidence_score": "-0.3109095"
            },
            {
                "entity": "Title 36 of the United States Code",
                "confidence_score": "-11.21505"
            },
            {
                "entity": "Sims, Illinois",
                "confidence_score": "-11.698894"
            }
        ]
    },
    {
        "mention": "marijuana",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Cannabis (drug)",
                "confidence_score": "2.4334426"
            },
            {
                "entity": "Cannabis",
                "confidence_score": "-4.41742"
            },
            {
                "entity": "Marijuana (word)",
                "confidence_score": "-5.626587"
            }
        ]
    },
    {
        "mention": "sims",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "The Sims",
                "confidence_score": "2.7196126"
            },
            {
                "entity": "Diabetes",
                "confidence_score": "-6.147214"
            },
            {
                "entity": "Sims",
                "confidence_score": "-10.902659"
            }
        ]
    },
    {
        "mention": "sims",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "The Sims",
                "confidence_score": "2.7196126"
            },
            {
                "entity": "Diabetes",
                "confidence_score": "-6.147214"
            },
            {
                "entity": "Sims",
                "confidence_score": "-10.902659"
            }
        ]
    },
    {
        "mention": "sims",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Sims",
                "confidence_score": "-0.3109095"
            },
            {
                "entity": "Title 36 of the United States Code",
                "confidence_score": "-11.21505"
            },
            {
                "entity": "Sims, Illinois",
                "confidence_score": "-11.698894"
            }
        ]
    },
    {
        "mention": "xanax",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Alprazolam",
                "confidence_score": "0.63187575"
            },
            {
                "entity": "X-ray",
                "confidence_score": "-9.167794"
            },
            {
                "entity": "Xanax (disambiguation)",
                "confidence_score": "-12.027896"
            }
        ]
    },
    {
        "mention": "oxycontin",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Oxycodone",
                "confidence_score": "4.1961436"
            },
            {
                "entity": "Oxalic acid",
                "confidence_score": "-10.215274"
            },
            {
                "entity": "Oxycinchophen",
                "confidence_score": "-11.848606"
            }
        ]
    },
    {
        "mention": "rees",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Dave Rees",
                "confidence_score": "-7.468818"
            },
            {
                "entity": "William E. Rees",
                "confidence_score": "-9.049914"
            },
            {
                "entity": "Ennis Rees",
                "confidence_score": "-10.330256"
            }
        ]
    },
    {
        "mention": "marijuana",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Cannabis (drug)",
                "confidence_score": "2.4334426"
            },
            {
                "entity": "Cannabis",
                "confidence_score": "-4.41742"
            },
            {
                "entity": "Marijuana (word)",
                "confidence_score": "-5.626587"
            }
        ]
    },
    {
        "mention": "warren j. geller",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kenneth Geller",
                "confidence_score": "-8.536869"
            },
            {
                "entity": "Martin J. Geller",
                "confidence_score": "-9.076875"
            },
            {
                "entity": "Joe Geller",
                "confidence_score": "-10.880626"
            }
        ]
    },
    {
        "mention": "sims",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Sims",
                "confidence_score": "-0.3109095"
            },
            {
                "entity": "Title 36 of the United States Code",
                "confidence_score": "-11.21505"
            },
            {
                "entity": "Sims, Illinois",
                "confidence_score": "-11.698894"
            }
        ]
    },
    {
        "mention": "sims",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Sims",
                "confidence_score": "-0.3109095"
            },
            {
                "entity": "Title 36 of the United States Code",
                "confidence_score": "-11.21505"
            },
            {
                "entity": "Sims, Illinois",
                "confidence_score": "-11.698894"
            }
        ]
    },
    {
        "mention": "marijuana",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Cannabis (drug)",
                "confidence_score": "2.4334426"
            },
            {
                "entity": "Cannabis",
                "confidence_score": "-4.41742"
            },
            {
                "entity": "Marijuana (word)",
                "confidence_score": "-5.626587"
            }
        ]
    },
    {
        "mention": "sims",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "The Sims",
                "confidence_score": "2.7196126"
            },
            {
                "entity": "Diabetes",
                "confidence_score": "-6.147214"
            },
            {
                "entity": "Sims",
                "confidence_score": "-10.902659"
            }
        ]
    },
    {
        "mention": "sims",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Sims",
                "confidence_score": "-0.3109095"
            },
            {
                "entity": "Title 36 of the United States Code",
                "confidence_score": "-11.21505"
            },
            {
                "entity": "Sims, Illinois",
                "confidence_score": "-11.698894"
            }
        ]
    },
    {
        "mention": "sims",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "The Sims",
                "confidence_score": "2.7196126"
            },
            {
                "entity": "Diabetes",
                "confidence_score": "-6.147214"
            },
            {
                "entity": "Sims",
                "confidence_score": "-10.902659"
            }
        ]
    },
    {
        "mention": "the sixth circuit",
        "type": "ORG",
        "predictions": [
            {
                "entity": "United States Court of Appeals for the Sixth Circuit",
                "confidence_score": "-5.0467796"
            },
            {
                "entity": "Sixth Amendment to the United States Constitution",
                "confidence_score": "-9.962877"
            },
            {
                "entity": "Same-sex marriage in the Sixth Circuit",
                "confidence_score": "-10.741328"
            }
        ]
    },
    {
        "mention": "icp",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Interpol",
                "confidence_score": "-5.059827"
            },
            {
                "entity": "ICITAP",
                "confidence_score": "-6.1757197"
            },
            {
                "entity": "American College of Physicians",
                "confidence_score": "-10.083039"
            }
        ]
    },
    {
        "mention": "juggalos",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Juggalo",
                "confidence_score": "1.7451768"
            },
            {
                "entity": "Juggalo gangs",
                "confidence_score": "-4.5170054"
            },
            {
                "entity": "Juggalo Championship Wrestling",
                "confidence_score": "-9.311606"
            }
        ]
    },
    {
        "mention": "fbi",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Federal Bureau of Investigation",
                "confidence_score": "9.217677"
            },
            {
                "entity": "FBI Police",
                "confidence_score": "-4.8289266"
            },
            {
                "entity": "Drug Enforcement Administration",
                "confidence_score": "-6.7346997"
            }
        ]
    },
    {
        "mention": "juggalos",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Juggalo",
                "confidence_score": "1.9498007"
            },
            {
                "entity": "Juggalo gangs",
                "confidence_score": "-3.1521075"
            },
            {
                "entity": "Juggalo Championship Wrestling",
                "confidence_score": "-9.181735"
            }
        ]
    },
    {
        "mention": "2011 national gang threat assessment",
        "type": "ORG",
        "predictions": [
            {
                "entity": "National Gang Intelligence Center",
                "confidence_score": "-10.7383585"
            },
            {
                "entity": "National Gang Center",
                "confidence_score": "-10.90487"
            },
            {
                "entity": "National Crime Victimization Survey",
                "confidence_score": "-11.524386"
            }
        ]
    },
    {
        "mention": "nbc news",
        "type": "ORG",
        "predictions": [
            {
                "entity": "NBC News",
                "confidence_score": "10.77444"
            },
            {
                "entity": "NBC Nightly News",
                "confidence_score": "5.641852"
            },
            {
                "entity": "Yahoo! News",
                "confidence_score": "-9.560118"
            }
        ]
    },
    {
        "mention": "juggalos",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Juggalo",
                "confidence_score": "1.9498007"
            },
            {
                "entity": "Juggalo gangs",
                "confidence_score": "-3.1521075"
            },
            {
                "entity": "Juggalo Championship Wrestling",
                "confidence_score": "-9.181735"
            }
        ]
    },
    {
        "mention": "fbi",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Federal Bureau of Investigation",
                "confidence_score": "9.217677"
            },
            {
                "entity": "FBI Police",
                "confidence_score": "-4.8289266"
            },
            {
                "entity": "Drug Enforcement Administration",
                "confidence_score": "-6.7346997"
            }
        ]
    },
    {
        "mention": "violent j",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Violent J",
                "confidence_score": "11.863929"
            },
            {
                "entity": "Johnny \"J\"",
                "confidence_score": "-10.79045"
            },
            {
                "entity": "Spider-Man J",
                "confidence_score": "-10.803304"
            }
        ]
    },
    {
        "mention": "forab",
        "type": "ORG",
        "predictions": [
            {
                "entity": "List of living former members of the United States House of Representatives (D)",
                "confidence_score": "-12.532015"
            },
            {
                "entity": "List of living former members of the United States House of Representatives (T)",
                "confidence_score": "-12.578427"
            },
            {
                "entity": "List of living former members of the United States House of Representatives (N)",
                "confidence_score": "-12.627588"
            }
        ]
    },
    {
        "mention": "the sixth circuit",
        "type": "ORG",
        "predictions": [
            {
                "entity": "United States Court of Appeals for the Sixth Circuit",
                "confidence_score": "-2.1004183"
            },
            {
                "entity": "Sixth Amendment to the United States Constitution",
                "confidence_score": "-10.035118"
            },
            {
                "entity": "Same-sex marriage in the Sixth Circuit",
                "confidence_score": "-10.1331625"
            }
        ]
    },
    {
        "mention": "supreme court",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Supreme court",
                "confidence_score": "7.4486923"
            },
            {
                "entity": "Supreme Court of the United States",
                "confidence_score": "6.4784617"
            },
            {
                "entity": "New York Supreme Court",
                "confidence_score": "-6.373042"
            }
        ]
    },
    {
        "mention": "doj",
        "type": "ORG",
        "predictions": [
            {
                "entity": "United States Department of Justice",
                "confidence_score": "10.679116"
            },
            {
                "entity": "Oregon Department of Justice",
                "confidence_score": "0.8849738"
            },
            {
                "entity": "Direcci\u00f3n de Observaciones Judiciales",
                "confidence_score": "-3.5575686"
            }
        ]
    },
    {
        "mention": "juggalos",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Juggalo",
                "confidence_score": "1.9498007"
            },
            {
                "entity": "Juggalo gangs",
                "confidence_score": "-3.1521075"
            },
            {
                "entity": "Juggalo Championship Wrestling",
                "confidence_score": "-9.181735"
            }
        ]
    },
    {
        "mention": "fbi",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Federal Bureau of Investigation",
                "confidence_score": "9.217677"
            },
            {
                "entity": "FBI Police",
                "confidence_score": "-4.8289266"
            },
            {
                "entity": "Drug Enforcement Administration",
                "confidence_score": "-6.7346997"
            }
        ]
    },
    {
        "mention": "fbi",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Federal Bureau of Investigation",
                "confidence_score": "9.217677"
            },
            {
                "entity": "FBI Police",
                "confidence_score": "-4.8289266"
            },
            {
                "entity": "Drug Enforcement Administration",
                "confidence_score": "-6.7346997"
            }
        ]
    },
    {
        "mention": "juggalo",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Juggalo",
                "confidence_score": "6.203214"
            },
            {
                "entity": "Juggalo gangs",
                "confidence_score": "-2.8084056"
            },
            {
                "entity": "Juggalo Championship Wrestling",
                "confidence_score": "-7.546285"
            }
        ]
    },
    {
        "mention": "fbi",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Federal Bureau of Investigation",
                "confidence_score": "9.217677"
            },
            {
                "entity": "FBI Police",
                "confidence_score": "-4.8289266"
            },
            {
                "entity": "Drug Enforcement Administration",
                "confidence_score": "-6.7346997"
            }
        ]
    },
    {
        "mention": "army",
        "type": "ORG",
        "predictions": [
            {
                "entity": "United States Army",
                "confidence_score": "6.5852256"
            },
            {
                "entity": "Army",
                "confidence_score": "4.3819013"
            },
            {
                "entity": "Army of the United States",
                "confidence_score": "-1.7837559"
            }
        ]
    },
    {
        "mention": "juggalo",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Juggalo",
                "confidence_score": "6.203214"
            },
            {
                "entity": "Juggalo gangs",
                "confidence_score": "-2.8084056"
            },
            {
                "entity": "Juggalo Championship Wrestling",
                "confidence_score": "-7.546285"
            }
        ]
    },
    {
        "mention": "the sixth circuit",
        "type": "ORG",
        "predictions": [
            {
                "entity": "United States Court of Appeals for the Sixth Circuit",
                "confidence_score": "-5.0467796"
            },
            {
                "entity": "Sixth Amendment to the United States Constitution",
                "confidence_score": "-9.962877"
            },
            {
                "entity": "Same-sex marriage in the Sixth Circuit",
                "confidence_score": "-10.741328"
            }
        ]
    },
    {
        "mention": "forab",
        "type": "ORG",
        "predictions": [
            {
                "entity": "List of living former members of the United States House of Representatives (D)",
                "confidence_score": "-12.532015"
            },
            {
                "entity": "List of living former members of the United States House of Representatives (T)",
                "confidence_score": "-12.578427"
            },
            {
                "entity": "List of living former members of the United States House of Representatives (N)",
                "confidence_score": "-12.627588"
            }
        ]
    },
    {
        "mention": "icp",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Interpol",
                "confidence_score": "-5.059827"
            },
            {
                "entity": "ICITAP",
                "confidence_score": "-6.1757197"
            },
            {
                "entity": "American College of Physicians",
                "confidence_score": "-10.083039"
            }
        ]
    },
    {
        "mention": "juggalos",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Juggalo",
                "confidence_score": "1.9498007"
            },
            {
                "entity": "Juggalo gangs",
                "confidence_score": "-3.1521075"
            },
            {
                "entity": "Juggalo Championship Wrestling",
                "confidence_score": "-9.181735"
            }
        ]
    },
    {
        "mention": "violent j",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Violent J",
                "confidence_score": "11.863929"
            },
            {
                "entity": "Johnny \"J\"",
                "confidence_score": "-10.79045"
            },
            {
                "entity": "Spider-Man J",
                "confidence_score": "-10.803304"
            }
        ]
    },
    {
        "mention": "shaggy 2 dope",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Shaggy 2 Dope",
                "confidence_score": "12.713063"
            },
            {
                "entity": "Pure DOPE",
                "confidence_score": "-11.020109"
            },
            {
                "entity": "Kenny \"Dope\" Gonzalez",
                "confidence_score": "-11.634648"
            }
        ]
    },
    {
        "mention": "aclu",
        "type": "ORG",
        "predictions": [
            {
                "entity": "American Civil Liberties Union",
                "confidence_score": "6.4272265"
            },
            {
                "entity": "Anarchy Championship Wrestling",
                "confidence_score": "-10.664191"
            },
            {
                "entity": "ACW Tag Team Championship",
                "confidence_score": "-11.212325"
            }
        ]
    },
    {
        "mention": "juggalo",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Juggalo",
                "confidence_score": "6.203214"
            },
            {
                "entity": "Juggalo gangs",
                "confidence_score": "-2.8084056"
            },
            {
                "entity": "Juggalo Championship Wrestling",
                "confidence_score": "-7.546285"
            }
        ]
    },
    {
        "mention": "juggalos",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Juggalo",
                "confidence_score": "1.9498007"
            },
            {
                "entity": "Juggalo gangs",
                "confidence_score": "-3.1521075"
            },
            {
                "entity": "Juggalo Championship Wrestling",
                "confidence_score": "-9.181735"
            }
        ]
    },
    {
        "mention": "pioneer energy services",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Pioneer Natural Resources",
                "confidence_score": "-4.371561"
            },
            {
                "entity": "Pioneer Energy",
                "confidence_score": "-6.7368903"
            },
            {
                "entity": "CenterPoint Energy",
                "confidence_score": "-10.242797"
            }
        ]
    },
    {
        "mention": "the securities and exchange commission",
        "type": "ORG",
        "predictions": [
            {
                "entity": "U.S. Securities and Exchange Commission",
                "confidence_score": "3.6327417"
            },
            {
                "entity": "Securities commission",
                "confidence_score": "-5.6049166"
            },
            {
                "entity": "Commodity Futures Trading Commission",
                "confidence_score": "-8.439437"
            }
        ]
    },
    {
        "mention": "pioneer energy services",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Pioneer Natural Resources",
                "confidence_score": "-4.371561"
            },
            {
                "entity": "Pioneer Energy",
                "confidence_score": "-6.7368903"
            },
            {
                "entity": "CenterPoint Energy",
                "confidence_score": "-10.242797"
            }
        ]
    },
    {
        "mention": "pr newswire",
        "type": "ORG",
        "predictions": [
            {
                "entity": "PR Newswire",
                "confidence_score": "11.792455"
            },
            {
                "entity": "U.S. Newswire",
                "confidence_score": "-4.9524465"
            },
            {
                "entity": "Clearwire",
                "confidence_score": "-10.374602"
            }
        ]
    },
    {
        "mention": "pr newswire",
        "type": "ORG",
        "predictions": [
            {
                "entity": "PR Newswire",
                "confidence_score": "11.792455"
            },
            {
                "entity": "U.S. Newswire",
                "confidence_score": "-4.9524465"
            },
            {
                "entity": "Clearwire",
                "confidence_score": "-10.374602"
            }
        ]
    },
    {
        "mention": "aquib javed",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Aaqib Javed",
                "confidence_score": "-7.096915"
            },
            {
                "entity": "Aqib Javed",
                "confidence_score": "-8.656602"
            },
            {
                "entity": "Mohammad Javed",
                "confidence_score": "-9.36325"
            }
        ]
    },
    {
        "mention": "mohammad aamir",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Aamir Khan",
                "confidence_score": "0.44650868"
            },
            {
                "entity": "Aamir Liaquat Hussain",
                "confidence_score": "-9.596632"
            },
            {
                "entity": "Mohammad Aamer (cricketer, born 1995)",
                "confidence_score": "-12.145636"
            }
        ]
    },
    {
        "mention": "salman butt",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Salman Butt",
                "confidence_score": "8.982509"
            },
            {
                "entity": "Salman Khan",
                "confidence_score": "-8.716155"
            },
            {
                "entity": "Salman Ali",
                "confidence_score": "-11.559153"
            }
        ]
    },
    {
        "mention": "javed",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mohammad Javed",
                "confidence_score": "-5.6946874"
            },
            {
                "entity": "Javed Khan (actor)",
                "confidence_score": "-7.552927"
            },
            {
                "entity": "Javed Ali Khan",
                "confidence_score": "-10.668416"
            }
        ]
    },
    {
        "mention": "mohammad aamir",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Aamir Khan",
                "confidence_score": "0.44650868"
            },
            {
                "entity": "Aamir Liaquat Hussain",
                "confidence_score": "-9.596632"
            },
            {
                "entity": "Mohammad Aamer (cricketer, born 1995)",
                "confidence_score": "-12.145636"
            }
        ]
    },
    {
        "mention": "salman butt",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Salman Butt",
                "confidence_score": "8.982509"
            },
            {
                "entity": "Salman Khan",
                "confidence_score": "-8.716155"
            },
            {
                "entity": "Salman Ali",
                "confidence_score": "-11.559153"
            }
        ]
    },
    {
        "mention": "mohammad azharuddin",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mohammad Azharuddin",
                "confidence_score": "9.942309"
            },
            {
                "entity": "Azharullah",
                "confidence_score": "-9.80936"
            },
            {
                "entity": "Azhar Mahmood",
                "confidence_score": "-10.122569"
            }
        ]
    },
    {
        "mention": "saleem malik",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Saleem Malik",
                "confidence_score": "10.528774"
            },
            {
                "entity": "Shoaib Malik",
                "confidence_score": "-10.314174"
            },
            {
                "entity": "Sumit Malik",
                "confidence_score": "-12.119098"
            }
        ]
    },
    {
        "mention": "ipl",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Indian Premier League",
                "confidence_score": "1.0439676"
            },
            {
                "entity": "Pakistan Premier League",
                "confidence_score": "-9.055242"
            },
            {
                "entity": "Indian Cricket League",
                "confidence_score": "-10.098463"
            }
        ]
    },
    {
        "mention": "wwe",
        "type": "ORG",
        "predictions": [
            {
                "entity": "WWE",
                "confidence_score": "10.244563"
            },
            {
                "entity": "WWE Network",
                "confidence_score": "-4.679355"
            },
            {
                "entity": "WWE Championship",
                "confidence_score": "-7.1042356"
            }
        ]
    },
    {
        "mention": "aamir",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Aamir Khan",
                "confidence_score": "4.9611077"
            },
            {
                "entity": "Aamir Ali",
                "confidence_score": "-8.771077"
            },
            {
                "entity": "A.J. Styles",
                "confidence_score": "-9.560662"
            }
        ]
    },
    {
        "mention": "sachin tendulkar",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Sachin Tendulkar",
                "confidence_score": "10.014631"
            },
            {
                "entity": "Vijay Tendulkar",
                "confidence_score": "-10.599022"
            },
            {
                "entity": "Suresh Tendulkar",
                "confidence_score": "-11.013983"
            }
        ]
    },
    {
        "mention": "younis khan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Younis Khan",
                "confidence_score": "11.047756"
            },
            {
                "entity": "Younis Ahmed",
                "confidence_score": "-11.231347"
            },
            {
                "entity": "Yousaf Ali Khan",
                "confidence_score": "-11.671966"
            }
        ]
    },
    {
        "mention": "to bharat sunderesan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bharat Desai",
                "confidence_score": "-11.397184"
            },
            {
                "entity": "Bharat Arun",
                "confidence_score": "-11.645979"
            },
            {
                "entity": "Bharat Awasthy",
                "confidence_score": "-12.198295"
            }
        ]
    },
    {
        "mention": "leeds",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Leeds",
                "confidence_score": "7.7471566"
            },
            {
                "entity": "Leeds railway station",
                "confidence_score": "-3.4146326"
            },
            {
                "entity": "Leeds Rhinos",
                "confidence_score": "-4.189503"
            }
        ]
    },
    {
        "mention": "guiseley school",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Guiseley School",
                "confidence_score": "1.6135678"
            },
            {
                "entity": "Guiseley",
                "confidence_score": "-7.1593194"
            },
            {
                "entity": "Guiseley railway station",
                "confidence_score": "-10.261829"
            }
        ]
    },
    {
        "mention": "paul clayton",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Paul Clayton (actor)",
                "confidence_score": "-5.56202"
            },
            {
                "entity": "Paul Clayton (folksinger)",
                "confidence_score": "-8.833602"
            },
            {
                "entity": "Paul Clayton (footballer)",
                "confidence_score": "-9.438412"
            }
        ]
    },
    {
        "mention": "the red cross",
        "type": "ORG",
        "predictions": [
            {
                "entity": "International Red Cross and Red Crescent Movement",
                "confidence_score": "-1.5281973"
            },
            {
                "entity": "International Committee of the Red Cross",
                "confidence_score": "-5.6485333"
            },
            {
                "entity": "International Federation of Red Cross and Red Crescent Societies",
                "confidence_score": "-5.9238834"
            }
        ]
    },
    {
        "mention": "twitter",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Twitter",
                "confidence_score": "9.86045"
            },
            {
                "entity": "SMS",
                "confidence_score": "-7.260535"
            },
            {
                "entity": "Twitter diplomacy",
                "confidence_score": "-9.852778"
            }
        ]
    },
    {
        "mention": "all south group",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Allman Brothers Band",
                "confidence_score": "-11.1800995"
            },
            {
                "entity": "The All-American Rejects",
                "confidence_score": "-11.577991"
            },
            {
                "entity": "All Sports Band",
                "confidence_score": "-11.658354"
            }
        ]
    },
    {
        "mention": "adam whitman",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Adam Yauch",
                "confidence_score": "-11.558134"
            },
            {
                "entity": "John Whitman",
                "confidence_score": "-11.707204"
            },
            {
                "entity": "James Whitman",
                "confidence_score": "-11.801407"
            }
        ]
    },
    {
        "mention": "sparta live",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Sparta (band)",
                "confidence_score": "-8.174508"
            },
            {
                "entity": "Kilimanjaro Live",
                "confidence_score": "-11.829827"
            },
            {
                "entity": "Gorillaz Live",
                "confidence_score": "-11.994027"
            }
        ]
    },
    {
        "mention": "dhanna resort",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Dubai Parks and Resorts",
                "confidence_score": "-8.102038"
            },
            {
                "entity": "Dubailand",
                "confidence_score": "-9.33172"
            },
            {
                "entity": "Walt Disney World Swan and Dolphin Resort",
                "confidence_score": "-11.414078"
            }
        ]
    },
    {
        "mention": "lafayette gourmet food festival",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Gourmet",
                "confidence_score": "-6.428368"
            },
            {
                "entity": "Lafayette (restaurant)",
                "confidence_score": "-10.738457"
            },
            {
                "entity": "World Gourmet Summit",
                "confidence_score": "-11.973723"
            }
        ]
    },
    {
        "mention": "underwater zoo",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Marine mammal park",
                "confidence_score": "-7.877282"
            },
            {
                "entity": "Oceanarium",
                "confidence_score": "-7.916664"
            },
            {
                "entity": "Anglesey Sea Zoo",
                "confidence_score": "-11.38964"
            }
        ]
    },
    {
        "mention": "dulce ortega",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Delfy de Ortega",
                "confidence_score": "-11.483848"
            },
            {
                "entity": "Pedro Ortega",
                "confidence_score": "-12.1006975"
            },
            {
                "entity": "Jos\u00e9 Ortega",
                "confidence_score": "-13.038783"
            }
        ]
    },
    {
        "mention": "officers club",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Military officers' club",
                "confidence_score": "-2.5156572"
            },
            {
                "entity": "The Officers Club",
                "confidence_score": "-6.702206"
            },
            {
                "entity": "Officers Club",
                "confidence_score": "-10.348919"
            }
        ]
    },
    {
        "mention": "mea",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Minimum en route altitude",
                "confidence_score": "-3.4562898"
            },
            {
                "entity": "American Hotel and Lodging Association",
                "confidence_score": "-9.973548"
            },
            {
                "entity": "Motor & Equipment Manufacturers Association",
                "confidence_score": "-10.190371"
            }
        ]
    },
    {
        "mention": "intercontinental\u00ae hotels & resorts redefines luxury",
        "type": "ORG",
        "predictions": [
            {
                "entity": "InterContinental",
                "confidence_score": "2.4810255"
            },
            {
                "entity": "InterContinental Hotels Group",
                "confidence_score": "-4.9494843"
            },
            {
                "entity": "Waldorf Astoria Hotels & Resorts",
                "confidence_score": "-8.501395"
            }
        ]
    },
    {
        "mention": "club intercontinental",
        "type": "ORG",
        "predictions": [
            {
                "entity": "InterContinental",
                "confidence_score": "-6.314486"
            },
            {
                "entity": "ClubCorp",
                "confidence_score": "-10.333324"
            },
            {
                "entity": "Club International",
                "confidence_score": "-11.264144"
            }
        ]
    },
    {
        "mention": "turnout",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Voter turnout",
                "confidence_score": "-1.7916319"
            },
            {
                "entity": "Attendance",
                "confidence_score": "-2.3267543"
            },
            {
                "entity": "Turnout",
                "confidence_score": "-11.773126"
            }
        ]
    },
    {
        "mention": "yas mall",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Yas Island",
                "confidence_score": "-10.15635"
            },
            {
                "entity": "Yas",
                "confidence_score": "-10.688764"
            },
            {
                "entity": "Yaesu Chikagai",
                "confidence_score": "-11.269027"
            }
        ]
    },
    {
        "mention": "al jimi mall",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Al Jimi Mall",
                "confidence_score": "10.139295"
            },
            {
                "entity": "Jimi Hendrix Park",
                "confidence_score": "-11.454613"
            },
            {
                "entity": "Ray's MTB Indoor Park",
                "confidence_score": "-12.096895"
            }
        ]
    },
    {
        "mention": "mercedes",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Mercedes-Benz",
                "confidence_score": "-2.388546"
            },
            {
                "entity": "Mercedes Jones",
                "confidence_score": "-8.991301"
            },
            {
                "entity": "Mercedes (singer)",
                "confidence_score": "-9.338724"
            }
        ]
    },
    {
        "mention": "waldorf astoria",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Waldorf Astoria Hotels & Resorts",
                "confidence_score": "-1.6795684"
            },
            {
                "entity": "Waldorf Astoria New York",
                "confidence_score": "-1.7785724"
            },
            {
                "entity": "Waldorf\u2013Astoria (1893\u20131929)",
                "confidence_score": "-5.067922"
            }
        ]
    },
    {
        "mention": "jason atherton",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jason Atherton",
                "confidence_score": "9.923142"
            },
            {
                "entity": "Jason Gould",
                "confidence_score": "-11.141628"
            },
            {
                "entity": "Jason Solomons",
                "confidence_score": "-11.527774"
            }
        ]
    },
    {
        "mention": "dhanna resort",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Dubai Parks and Resorts",
                "confidence_score": "-8.102038"
            },
            {
                "entity": "Dubailand",
                "confidence_score": "-9.33172"
            },
            {
                "entity": "Walt Disney World Swan and Dolphin Resort",
                "confidence_score": "-11.414078"
            }
        ]
    },
    {
        "mention": "bushman",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bushman (reggae singer)",
                "confidence_score": "-10.298779"
            },
            {
                "entity": "Yevgeni Bushmanov",
                "confidence_score": "-11.46525"
            },
            {
                "entity": "Bushman",
                "confidence_score": "-11.733359"
            }
        ]
    },
    {
        "mention": "momcinchanz show",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Tech Toyz",
                "confidence_score": "-11.891408"
            },
            {
                "entity": "List of Pakistan Idol episodes",
                "confidence_score": "-11.927607"
            },
            {
                "entity": "Moms (talk show)",
                "confidence_score": "-12.395232"
            }
        ]
    },
    {
        "mention": "healthcare",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Health care",
                "confidence_score": "7.8033137"
            },
            {
                "entity": "Healthcare industry",
                "confidence_score": "-1.706749"
            },
            {
                "entity": "Healthcare in India",
                "confidence_score": "-5.7922616"
            }
        ]
    },
    {
        "mention": "raghu das",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Raghubar Das",
                "confidence_score": "8.536983"
            },
            {
                "entity": "Raghu Ram",
                "confidence_score": "-10.980072"
            },
            {
                "entity": "Raghuvansh Prasad Singh",
                "confidence_score": "-11.172228"
            }
        ]
    },
    {
        "mention": "rick diamond/getty",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Getty family",
                "confidence_score": "-7.638806"
            },
            {
                "entity": "Charlie Getty",
                "confidence_score": "-11.558099"
            },
            {
                "entity": "Armstrong & Getty",
                "confidence_score": "-11.626846"
            }
        ]
    },
    {
        "mention": "eric",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jessie James Decker",
                "confidence_score": "7.4153786"
            },
            {
                "entity": "Eric Decker",
                "confidence_score": "-6.016327"
            },
            {
                "entity": "Daniel Decker",
                "confidence_score": "-11.696915"
            }
        ]
    },
    {
        "mention": "jessie james decker \u2018s",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jessie James Decker",
                "confidence_score": "11.179743"
            },
            {
                "entity": "Natalie Decker",
                "confidence_score": "-11.536051"
            },
            {
                "entity": "Claire Decker",
                "confidence_score": "-11.705059"
            }
        ]
    },
    {
        "mention": "vivianne rose",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Arden Rose",
                "confidence_score": "-11.2634"
            },
            {
                "entity": "Debi Rose",
                "confidence_score": "-11.288483"
            },
            {
                "entity": "Vivianne Collins",
                "confidence_score": "-11.468726"
            }
        ]
    },
    {
        "mention": "vivi",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Vivi (magazine)",
                "confidence_score": "-4.594382"
            },
            {
                "entity": "ViVi (singer)",
                "confidence_score": "-9.203469"
            },
            {
                "entity": "Vi",
                "confidence_score": "-10.160596"
            }
        ]
    },
    {
        "mention": "james decker",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jessie James Decker",
                "confidence_score": "3.1656404"
            },
            {
                "entity": "Daniel Decker",
                "confidence_score": "-11.165511"
            },
            {
                "entity": "Claire Decker",
                "confidence_score": "-11.523219"
            }
        ]
    },
    {
        "mention": "vivianne",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Vivianne Crowley",
                "confidence_score": "-4.522098"
            },
            {
                "entity": "Vivianne Collins",
                "confidence_score": "-8.789145"
            },
            {
                "entity": "Vivian Girls",
                "confidence_score": "-10.889866"
            }
        ]
    },
    {
        "mention": "jessie james decker",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jessie James Decker",
                "confidence_score": "11.4922085"
            },
            {
                "entity": "Natalie Decker",
                "confidence_score": "-11.168967"
            },
            {
                "entity": "Claire Decker",
                "confidence_score": "-11.551244"
            }
        ]
    },
    {
        "mention": "@jessiejamesdecker",
        "type": "ORG",
        "predictions": [
            {
                "entity": "American Ninja Warrior (season 11)",
                "confidence_score": "-11.792045"
            },
            {
                "entity": "Timeline for September following the September 11 attacks",
                "confidence_score": "-11.912368"
            },
            {
                "entity": "Timeline for October following the September 11 attacks",
                "confidence_score": "-12.177433"
            }
        ]
    },
    {
        "mention": "james decker",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jessie James Decker",
                "confidence_score": "3.1656404"
            },
            {
                "entity": "Daniel Decker",
                "confidence_score": "-11.165511"
            },
            {
                "entity": "Claire Decker",
                "confidence_score": "-11.523219"
            }
        ]
    },
    {
        "mention": "james decker",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jessie James Decker",
                "confidence_score": "3.1656404"
            },
            {
                "entity": "Daniel Decker",
                "confidence_score": "-11.165511"
            },
            {
                "entity": "Claire Decker",
                "confidence_score": "-11.523219"
            }
        ]
    },
    {
        "mention": "vivianne",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Vivianne Crowley",
                "confidence_score": "-4.522098"
            },
            {
                "entity": "Vivianne Collins",
                "confidence_score": "-8.789145"
            },
            {
                "entity": "Vivian Girls",
                "confidence_score": "-10.889866"
            }
        ]
    },
    {
        "mention": "vivi",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Vivi (magazine)",
                "confidence_score": "-0.7756751"
            },
            {
                "entity": "ViVi (singer)",
                "confidence_score": "-6.7917585"
            },
            {
                "entity": "Vi",
                "confidence_score": "-9.870178"
            }
        ]
    },
    {
        "mention": "eric decker",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Eric Decker",
                "confidence_score": "5.6499677"
            },
            {
                "entity": "Eric Decker (Australian politician)",
                "confidence_score": "-6.658856"
            },
            {
                "entity": "Eric Christian Olsen",
                "confidence_score": "-10.554417"
            }
        ]
    },
    {
        "mention": "james decker",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jessie James Decker",
                "confidence_score": "3.1656404"
            },
            {
                "entity": "Daniel Decker",
                "confidence_score": "-11.165511"
            },
            {
                "entity": "Claire Decker",
                "confidence_score": "-11.523219"
            }
        ]
    },
    {
        "mention": "anya leon",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Anya Jenkins",
                "confidence_score": "-9.634597"
            },
            {
                "entity": "Anya Taylor-Joy",
                "confidence_score": "-11.253058"
            },
            {
                "entity": "Anya Rozova",
                "confidence_score": "-11.508821"
            }
        ]
    },
    {
        "mention": "prweb",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Population Reference Bureau",
                "confidence_score": "-7.6139197"
            },
            {
                "entity": "Ministry of Social and Family Development",
                "confidence_score": "-10.921858"
            },
            {
                "entity": "Poverty Reduction Strategy Paper",
                "confidence_score": "-11.391945"
            }
        ]
    },
    {
        "mention": "the american cancer society",
        "type": "ORG",
        "predictions": [
            {
                "entity": "American Cancer Society",
                "confidence_score": "10.193181"
            },
            {
                "entity": "American Association for Cancer Research",
                "confidence_score": "-5.5837398"
            },
            {
                "entity": "American Nuclear Society",
                "confidence_score": "-11.183601"
            }
        ]
    },
    {
        "mention": "michael lazar",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Michael Lazarou",
                "confidence_score": "-8.945752"
            },
            {
                "entity": "Andrew Lazar",
                "confidence_score": "-10.816024"
            },
            {
                "entity": "Jonathan Lazar",
                "confidence_score": "-12.008439"
            }
        ]
    },
    {
        "mention": "hifu",
        "type": "ORG",
        "predictions": [
            {
                "entity": "High-intensity focused ultrasound",
                "confidence_score": "-0.70003664"
            },
            {
                "entity": "Hi5",
                "confidence_score": "-11.028163"
            },
            {
                "entity": "Hibutsu",
                "confidence_score": "-11.625176"
            }
        ]
    },
    {
        "mention": "hifu",
        "type": "ORG",
        "predictions": [
            {
                "entity": "High-intensity focused ultrasound",
                "confidence_score": "-0.70003664"
            },
            {
                "entity": "Hi5",
                "confidence_score": "-11.028163"
            },
            {
                "entity": "Hibutsu",
                "confidence_score": "-11.625176"
            }
        ]
    },
    {
        "mention": "psa",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Photographic Society of America",
                "confidence_score": "0.73007566"
            },
            {
                "entity": "Professional Squash Association",
                "confidence_score": "-1.737004"
            },
            {
                "entity": "Phone-paid Services Authority",
                "confidence_score": "-3.0740108"
            }
        ]
    },
    {
        "mention": "hifu",
        "type": "ORG",
        "predictions": [
            {
                "entity": "High-intensity focused ultrasound",
                "confidence_score": "-0.70003664"
            },
            {
                "entity": "Hi5",
                "confidence_score": "-11.028163"
            },
            {
                "entity": "Hibutsu",
                "confidence_score": "-11.625176"
            }
        ]
    },
    {
        "mention": "fda",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Food and Drug Administration",
                "confidence_score": "9.530396"
            },
            {
                "entity": "Drug Enforcement Administration",
                "confidence_score": "-5.8648577"
            },
            {
                "entity": "Federal Communications Commission",
                "confidence_score": "-8.16484"
            }
        ]
    },
    {
        "mention": "hifu",
        "type": "ORG",
        "predictions": [
            {
                "entity": "High-intensity focused ultrasound",
                "confidence_score": "-0.70003664"
            },
            {
                "entity": "Hi5",
                "confidence_score": "-11.028163"
            },
            {
                "entity": "Hibutsu",
                "confidence_score": "-11.625176"
            }
        ]
    },
    {
        "mention": "lazar",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Seymour Lazar",
                "confidence_score": "-8.251321"
            },
            {
                "entity": "Zachary Lazar",
                "confidence_score": "-8.306617"
            },
            {
                "entity": "Lazar",
                "confidence_score": "-12.117059"
            }
        ]
    },
    {
        "mention": "michael lazar",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Michael Lazarou",
                "confidence_score": "-8.945752"
            },
            {
                "entity": "Andrew Lazar",
                "confidence_score": "-10.816024"
            },
            {
                "entity": "Jonathan Lazar",
                "confidence_score": "-12.008439"
            }
        ]
    },
    {
        "mention": "santa rosa",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Santa Rosa, California",
                "confidence_score": "10.077419"
            },
            {
                "entity": "Santa Rosa, Arizona",
                "confidence_score": "-8.374631"
            },
            {
                "entity": "Santa Rosa, La Pampa",
                "confidence_score": "-9.854208"
            }
        ]
    },
    {
        "mention": "hazmat",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Hazmat suit",
                "confidence_score": "-6.1803145"
            },
            {
                "entity": "Hazmat Modine",
                "confidence_score": "-8.720078"
            },
            {
                "entity": "HazMat (film)",
                "confidence_score": "-9.29053"
            }
        ]
    },
    {
        "mention": "sauvignon blanc",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sauvignon blanc",
                "confidence_score": "9.249314"
            },
            {
                "entity": "Cabernet Sauvignon",
                "confidence_score": "-4.8373604"
            },
            {
                "entity": "Sauvignon",
                "confidence_score": "-12.433902"
            }
        ]
    },
    {
        "mention": "craigy",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Craigkelly transmitting station",
                "confidence_score": "-6.8451667"
            },
            {
                "entity": "Craig Low",
                "confidence_score": "-10.021306"
            },
            {
                "entity": "Craigie",
                "confidence_score": "-12.762584"
            }
        ]
    },
    {
        "mention": "mills & boon",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Mills & Boon",
                "confidence_score": "10.739887"
            },
            {
                "entity": "Mills & Reeve",
                "confidence_score": "-9.458449"
            },
            {
                "entity": "Mills and Ware",
                "confidence_score": "-11.060039"
            }
        ]
    },
    {
        "mention": "jack reacher",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jack Reacher",
                "confidence_score": "8.516049"
            },
            {
                "entity": "Jack Reacher (film)",
                "confidence_score": "-1.6884911"
            },
            {
                "entity": "Jack Nicholson",
                "confidence_score": "-8.055649"
            }
        ]
    },
    {
        "mention": "wiine",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Wii",
                "confidence_score": "-5.0936704"
            },
            {
                "entity": "Pilates",
                "confidence_score": "-6.102322"
            },
            {
                "entity": "Insulin",
                "confidence_score": "-7.0668364"
            }
        ]
    },
    {
        "mention": "insanity challenge",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Insanity Fight",
                "confidence_score": "-11.027833"
            },
            {
                "entity": "Osaka Kawasaki Rehabilitation University",
                "confidence_score": "-11.463712"
            },
            {
                "entity": "Aichi Medical College for Physical and Occupational Therapy",
                "confidence_score": "-11.633431"
            }
        ]
    },
    {
        "mention": "street fighter",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Street Fighter",
                "confidence_score": "3.974343"
            },
            {
                "entity": "Street Fighter (video game)",
                "confidence_score": "-2.9350097"
            },
            {
                "entity": "Street Fighter EX",
                "confidence_score": "-8.652843"
            }
        ]
    },
    {
        "mention": "mummy",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mummy",
                "confidence_score": "4.888062"
            },
            {
                "entity": "Mummy (monster)",
                "confidence_score": "-8.822412"
            },
            {
                "entity": "Mummy (Dungeons & Dragons)",
                "confidence_score": "-9.246108"
            }
        ]
    },
    {
        "mention": "netflix",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Netflix",
                "confidence_score": "11.158345"
            },
            {
                "entity": "Valve Corporation",
                "confidence_score": "-8.630753"
            },
            {
                "entity": "Netflix Prize",
                "confidence_score": "-8.984571"
            }
        ]
    },
    {
        "mention": "captain america",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Captain America",
                "confidence_score": "-0.21371786"
            },
            {
                "entity": "Captain America (1990 film)",
                "confidence_score": "-5.8872023"
            },
            {
                "entity": "Captain America (comic book)",
                "confidence_score": "-8.294366"
            }
        ]
    },
    {
        "mention": "dublin",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Dublin",
                "confidence_score": "7.3117948"
            },
            {
                "entity": "Dublin, Texas",
                "confidence_score": "-7.899513"
            },
            {
                "entity": "Dublin (disambiguation)",
                "confidence_score": "-11.792133"
            }
        ]
    },
    {
        "mention": "research and markets",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Research and development",
                "confidence_score": "-7.857689"
            },
            {
                "entity": "Medical Associates for Research and Communication",
                "confidence_score": "-11.593366"
            },
            {
                "entity": "Research Ideas and Outcomes",
                "confidence_score": "-11.6242075"
            }
        ]
    },
    {
        "mention": "the global molecular diagnostics",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Molecular diagnostics",
                "confidence_score": "-5.3375115"
            },
            {
                "entity": "National Institute of General Medical Sciences",
                "confidence_score": "-10.106475"
            },
            {
                "entity": "JN-International Medical Corporation",
                "confidence_score": "-11.091339"
            }
        ]
    },
    {
        "mention": "the securities exchange commission",
        "type": "ORG",
        "predictions": [
            {
                "entity": "U.S. Securities and Exchange Commission",
                "confidence_score": "2.205327"
            },
            {
                "entity": "Securities commission",
                "confidence_score": "-6.224046"
            },
            {
                "entity": "Commodity Futures Trading Commission",
                "confidence_score": "-8.727707"
            }
        ]
    },
    {
        "mention": "sec",
        "type": "ORG",
        "predictions": [
            {
                "entity": "U.S. Securities and Exchange Commission",
                "confidence_score": "8.640548"
            },
            {
                "entity": "Securities and Exchange Commission (Philippines)",
                "confidence_score": "-1.9099634"
            },
            {
                "entity": "Securities and Exchange Commission (Nigeria)",
                "confidence_score": "-2.4969196"
            }
        ]
    },
    {
        "mention": "media contact:",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Media relations",
                "confidence_score": "-7.3452377"
            },
            {
                "entity": "Media consultant",
                "confidence_score": "-7.892909"
            },
            {
                "entity": "Media contacts database",
                "confidence_score": "-8.547567"
            }
        ]
    },
    {
        "mention": "laura wood",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Laura Woods",
                "confidence_score": "0.2341895"
            },
            {
                "entity": "Laura Newbold Wood",
                "confidence_score": "-5.389903"
            },
            {
                "entity": "Laura Miller (writer)",
                "confidence_score": "-11.351572"
            }
        ]
    },
    {
        "mention": "pr newswire",
        "type": "ORG",
        "predictions": [
            {
                "entity": "PR Newswire",
                "confidence_score": "11.889104"
            },
            {
                "entity": "U.S. Newswire",
                "confidence_score": "-5.2292175"
            },
            {
                "entity": "SatireWire",
                "confidence_score": "-10.555501"
            }
        ]
    },
    {
        "mention": "the thomas edison",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Thomas Edison",
                "confidence_score": "-1.5791097"
            },
            {
                "entity": "The Edison",
                "confidence_score": "-4.9428983"
            },
            {
                "entity": "Thomas A. Edison, Inc.",
                "confidence_score": "-6.282781"
            }
        ]
    },
    {
        "mention": "henry ford",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Henry Ford",
                "confidence_score": "7.3060913"
            },
            {
                "entity": "The Henry Ford",
                "confidence_score": "4.0254426"
            },
            {
                "entity": "Henry P. Ford",
                "confidence_score": "-11.31442"
            }
        ]
    },
    {
        "mention": "silicon valley",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Silicon Valley",
                "confidence_score": "7.6953807"
            },
            {
                "entity": "Simi Valley, California",
                "confidence_score": "-8.752507"
            },
            {
                "entity": "Silicon Valley (TV series)",
                "confidence_score": "-8.854701"
            }
        ]
    },
    {
        "mention": "robert noyce",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Robert Noyce",
                "confidence_score": "12.682679"
            },
            {
                "entity": "Phillip Noyce",
                "confidence_score": "-10.192939"
            },
            {
                "entity": "Mark Noyce",
                "confidence_score": "-11.131219"
            }
        ]
    },
    {
        "mention": "silicon valley",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Silicon Valley",
                "confidence_score": "7.6953807"
            },
            {
                "entity": "Simi Valley, California",
                "confidence_score": "-8.752507"
            },
            {
                "entity": "Silicon Valley (TV series)",
                "confidence_score": "-8.854701"
            }
        ]
    },
    {
        "mention": "leslie berlin",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Leslie Berlin",
                "confidence_score": "12.722016"
            },
            {
                "entity": "Leslie Hoffman",
                "confidence_score": "-11.853341"
            },
            {
                "entity": "Leslie Feinberg",
                "confidence_score": "-11.860128"
            }
        ]
    },
    {
        "mention": "the silicon valley archives",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Hoover Institution Library and Archives",
                "confidence_score": "-6.5420394"
            },
            {
                "entity": "Silicon Valley Historical Association",
                "confidence_score": "-9.01363"
            },
            {
                "entity": "University of Southern California Libraries",
                "confidence_score": "-10.636604"
            }
        ]
    },
    {
        "mention": "stanford university",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Stanford University",
                "confidence_score": "11.576615"
            },
            {
                "entity": "Stanford University Press",
                "confidence_score": "-5.9635024"
            },
            {
                "entity": "Stanford Law School",
                "confidence_score": "-6.2537546"
            }
        ]
    },
    {
        "mention": "jeff frick",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Stephen Frick",
                "confidence_score": "-10.256689"
            },
            {
                "entity": "Jeff Spock",
                "confidence_score": "-11.542432"
            },
            {
                "entity": "Jeff Bates (technologist)",
                "confidence_score": "-11.89877"
            }
        ]
    },
    {
        "mention": "intel",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Intel",
                "confidence_score": "11.424021"
            },
            {
                "entity": "Intel Capital",
                "confidence_score": "-7.5767293"
            },
            {
                "entity": "Intel Atom",
                "confidence_score": "-9.848425"
            }
        ]
    },
    {
        "mention": "robert noyce",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Robert Noyce",
                "confidence_score": "12.682679"
            },
            {
                "entity": "Phillip Noyce",
                "confidence_score": "-10.192939"
            },
            {
                "entity": "Mark Noyce",
                "confidence_score": "-11.131219"
            }
        ]
    },
    {
        "mention": "the invention of silicon valley",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Silicon Valley",
                "confidence_score": "-5.2629905"
            },
            {
                "entity": "Invention of the integrated circuit",
                "confidence_score": "-10.098175"
            },
            {
                "entity": "Pirates of Silicon Valley",
                "confidence_score": "-10.671358"
            }
        ]
    },
    {
        "mention": "noyce",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Robert Noyce",
                "confidence_score": "7.0931554"
            },
            {
                "entity": "IEEE Robert N. Noyce Medal",
                "confidence_score": "-5.9793973"
            },
            {
                "entity": "The NOCO Company",
                "confidence_score": "-9.323169"
            }
        ]
    },
    {
        "mention": "northern california",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Northern California",
                "confidence_score": "9.65029"
            },
            {
                "entity": "Southern California",
                "confidence_score": "-6.483355"
            },
            {
                "entity": "Northern California Power Agency",
                "confidence_score": "-9.885758"
            }
        ]
    },
    {
        "mention": "fairchild semiconductor international, inc.",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Fairchild Semiconductor",
                "confidence_score": "9.180472"
            },
            {
                "entity": "Fairchild Systems",
                "confidence_score": "-8.065933"
            },
            {
                "entity": "Fairchild Industries",
                "confidence_score": "-9.100502"
            }
        ]
    },
    {
        "mention": "intel corp.",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Intel",
                "confidence_score": "10.321364"
            },
            {
                "entity": "Intel Capital",
                "confidence_score": "-8.120007"
            },
            {
                "entity": "China Electronics Corporation",
                "confidence_score": "-11.846009"
            }
        ]
    },
    {
        "mention": "noyce",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Robert Noyce",
                "confidence_score": "7.0931554"
            },
            {
                "entity": "IEEE Robert N. Noyce Medal",
                "confidence_score": "-5.9793973"
            },
            {
                "entity": "The NOCO Company",
                "confidence_score": "-9.323169"
            }
        ]
    },
    {
        "mention": "silicon valley",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Silicon Valley",
                "confidence_score": "7.6953807"
            },
            {
                "entity": "Simi Valley, California",
                "confidence_score": "-8.752507"
            },
            {
                "entity": "Silicon Valley (TV series)",
                "confidence_score": "-8.854701"
            }
        ]
    },
    {
        "mention": "noyce",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Robert Noyce",
                "confidence_score": "7.0931554"
            },
            {
                "entity": "IEEE Robert N. Noyce Medal",
                "confidence_score": "-5.9793973"
            },
            {
                "entity": "The NOCO Company",
                "confidence_score": "-9.323169"
            }
        ]
    },
    {
        "mention": "noyce",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Robert Noyce",
                "confidence_score": "7.0931554"
            },
            {
                "entity": "IEEE Robert N. Noyce Medal",
                "confidence_score": "-5.9793973"
            },
            {
                "entity": "The NOCO Company",
                "confidence_score": "-9.323169"
            }
        ]
    },
    {
        "mention": "noyce",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Robert Noyce",
                "confidence_score": "7.0931554"
            },
            {
                "entity": "IEEE Robert N. Noyce Medal",
                "confidence_score": "-5.9793973"
            },
            {
                "entity": "The NOCO Company",
                "confidence_score": "-9.323169"
            }
        ]
    },
    {
        "mention": "silicon valley",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Silicon Valley",
                "confidence_score": "7.6953807"
            },
            {
                "entity": "Simi Valley, California",
                "confidence_score": "-8.752507"
            },
            {
                "entity": "Silicon Valley (TV series)",
                "confidence_score": "-8.854701"
            }
        ]
    },
    {
        "mention": "noyce",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Robert Noyce",
                "confidence_score": "7.0931554"
            },
            {
                "entity": "IEEE Robert N. Noyce Medal",
                "confidence_score": "-5.9793973"
            },
            {
                "entity": "The NOCO Company",
                "confidence_score": "-9.323169"
            }
        ]
    },
    {
        "mention": "noyce",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Robert Noyce",
                "confidence_score": "7.0931554"
            },
            {
                "entity": "IEEE Robert N. Noyce Medal",
                "confidence_score": "-5.9793973"
            },
            {
                "entity": "The NOCO Company",
                "confidence_score": "-9.323169"
            }
        ]
    },
    {
        "mention": "silicon valley",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Silicon Valley",
                "confidence_score": "7.6953807"
            },
            {
                "entity": "Simi Valley, California",
                "confidence_score": "-8.752507"
            },
            {
                "entity": "Silicon Valley (TV series)",
                "confidence_score": "-8.854701"
            }
        ]
    },
    {
        "mention": "silicon valley",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Silicon Valley",
                "confidence_score": "7.6953807"
            },
            {
                "entity": "Simi Valley, California",
                "confidence_score": "-8.752507"
            },
            {
                "entity": "Silicon Valley (TV series)",
                "confidence_score": "-8.854701"
            }
        ]
    },
    {
        "mention": "cloud",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Cloud",
                "confidence_score": "3.8923333"
            },
            {
                "entity": "Cloud computing",
                "confidence_score": "-1.5095904"
            },
            {
                "entity": "Cloud albedo",
                "confidence_score": "-8.983817"
            }
        ]
    },
    {
        "mention": "silicon valley",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Silicon Valley",
                "confidence_score": "7.6953807"
            },
            {
                "entity": "Simi Valley, California",
                "confidence_score": "-8.752507"
            },
            {
                "entity": "Silicon Valley (TV series)",
                "confidence_score": "-8.854701"
            }
        ]
    },
    {
        "mention": "silicon valley",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Silicon Valley",
                "confidence_score": "7.6953807"
            },
            {
                "entity": "Simi Valley, California",
                "confidence_score": "-8.752507"
            },
            {
                "entity": "Silicon Valley (TV series)",
                "confidence_score": "-8.854701"
            }
        ]
    },
    {
        "mention": "marlene den bleyker marlene",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Blokwatch",
                "confidence_score": "-11.48195"
            },
            {
                "entity": "12secondstv",
                "confidence_score": "-12.05319"
            },
            {
                "entity": "Marlene",
                "confidence_score": "-12.201461"
            }
        ]
    },
    {
        "mention": "siliconangle media's",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Digital media",
                "confidence_score": "-9.741878"
            },
            {
                "entity": "Funke Digital TV",
                "confidence_score": "-11.480588"
            },
            {
                "entity": "Mingle Media TV Network",
                "confidence_score": "-11.686318"
            }
        ]
    },
    {
        "mention": "marlene den bleyker",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Walter De Brouwer",
                "confidence_score": "-11.514417"
            },
            {
                "entity": "Evan Blass",
                "confidence_score": "-11.813035"
            },
            {
                "entity": "Alfie Dennen",
                "confidence_score": "-11.839919"
            }
        ]
    },
    {
        "mention": "apple",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Apple Inc.",
                "confidence_score": "4.0746527"
            },
            {
                "entity": "Apple",
                "confidence_score": "-0.4233754"
            },
            {
                "entity": "Apple (band)",
                "confidence_score": "-8.850406"
            }
        ]
    },
    {
        "mention": "the siliconangle newsletter",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Newsletter",
                "confidence_score": "0.29155672"
            },
            {
                "entity": "Silicon Republic",
                "confidence_score": "-11.946187"
            },
            {
                "entity": "ACBF Newsletter",
                "confidence_score": "-12.613145"
            }
        ]
    },
    {
        "mention": "kristofer martin",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kristofer Martin",
                "confidence_score": "11.6727295"
            },
            {
                "entity": "Kristel de Catalina",
                "confidence_score": "-11.825611"
            },
            {
                "entity": "Kristel Moreno",
                "confidence_score": "-11.84768"
            }
        ]
    },
    {
        "mention": "joyce ching",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Joyce Ching",
                "confidence_score": "11.941098"
            },
            {
                "entity": "Joyce Jimenez",
                "confidence_score": "-11.27615"
            },
            {
                "entity": "Joyce Chu",
                "confidence_score": "-11.795945"
            }
        ]
    },
    {
        "mention": "mickey ferriols",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mickey Ferriols",
                "confidence_score": "12.517222"
            },
            {
                "entity": "John Ferriols",
                "confidence_score": "-9.407824"
            },
            {
                "entity": "Mickey Facchinello",
                "confidence_score": "-11.550006"
            }
        ]
    },
    {
        "mention": "kailan ba tama ang mali",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kailan Ba Tama ang Mali?",
                "confidence_score": "5.7923036"
            },
            {
                "entity": "Wow Mali",
                "confidence_score": "-10.26815"
            },
            {
                "entity": "Malich",
                "confidence_score": "-12.802333"
            }
        ]
    },
    {
        "mention": "gma pinoy tv",
        "type": "ORG",
        "predictions": [
            {
                "entity": "GMA Pinoy TV",
                "confidence_score": "11.675157"
            },
            {
                "entity": "GMA Network",
                "confidence_score": "-2.9622688"
            },
            {
                "entity": "GMA Network (company)",
                "confidence_score": "-6.593367"
            }
        ]
    },
    {
        "mention": "tambayan replay",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Replay (sports)",
                "confidence_score": "-7.8537087"
            },
            {
                "entity": "Virtual replay",
                "confidence_score": "-10.79959"
            },
            {
                "entity": "Tomie: Replay",
                "confidence_score": "-11.20707"
            }
        ]
    },
    {
        "mention": "the executive remuneration working group",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Efficiency and Reform Group",
                "confidence_score": "-11.074065"
            },
            {
                "entity": "Companies' Remuneration Reports Bill",
                "confidence_score": "-11.102667"
            },
            {
                "entity": "Joint Committee on Tax Law Rewrite Bills",
                "confidence_score": "-11.6270275"
            }
        ]
    },
    {
        "mention": "the investment association",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Investment Association",
                "confidence_score": "3.7977734"
            },
            {
                "entity": "Investment Management Association",
                "confidence_score": "1.9148675"
            },
            {
                "entity": "Investment Adviser Association",
                "confidence_score": "-8.307127"
            }
        ]
    },
    {
        "mention": "daniel godfrey",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Dan Godfrey",
                "confidence_score": "-1.7300365"
            },
            {
                "entity": "Daniel Strong Godfrey",
                "confidence_score": "-6.565685"
            },
            {
                "entity": "Daniel Pinto (financier)",
                "confidence_score": "-11.125046"
            }
        ]
    },
    {
        "mention": "the investment association",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Investment Association",
                "confidence_score": "3.7977734"
            },
            {
                "entity": "Investment Management Association",
                "confidence_score": "1.9148675"
            },
            {
                "entity": "Investment Adviser Association",
                "confidence_score": "-8.307127"
            }
        ]
    },
    {
        "mention": "nigel wilson",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Nigel Wilson (businessman)",
                "confidence_score": "12.405133"
            },
            {
                "entity": "Nigel Wilson",
                "confidence_score": "0.90378475"
            },
            {
                "entity": "Nigel Guy Wilson",
                "confidence_score": "-5.646127"
            }
        ]
    },
    {
        "mention": "legal & general",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Legal & General",
                "confidence_score": "12.504683"
            },
            {
                "entity": "Ernst & Young",
                "confidence_score": "-8.374655"
            },
            {
                "entity": "Law Debenture",
                "confidence_score": "-9.554503"
            }
        ]
    },
    {
        "mention": "ftse",
        "type": "ORG",
        "predictions": [
            {
                "entity": "FTSE Group",
                "confidence_score": "-0.10674975"
            },
            {
                "entity": "FTSE 100 Index",
                "confidence_score": "-1.1494522"
            },
            {
                "entity": "FTSE",
                "confidence_score": "-10.67216"
            }
        ]
    },
    {
        "mention": "david tyler",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "David Tyler (businessman)",
                "confidence_score": "9.496988"
            },
            {
                "entity": "David Tyler (producer)",
                "confidence_score": "-5.5582957"
            },
            {
                "entity": "David Tyler",
                "confidence_score": "-11.240688"
            }
        ]
    },
    {
        "mention": "sainsbury",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sainsbury's",
                "confidence_score": "5.6088643"
            },
            {
                "entity": "Sainsbury family",
                "confidence_score": "1.4537688"
            },
            {
                "entity": "John Sainsbury, Baron Sainsbury of Preston Candover",
                "confidence_score": "-3.4738038"
            }
        ]
    },
    {
        "mention": "burberry",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Burberry",
                "confidence_score": "10.763467"
            },
            {
                "entity": "Ebuyer",
                "confidence_score": "-10.964683"
            },
            {
                "entity": "Bilberry",
                "confidence_score": "-11.051904"
            }
        ]
    },
    {
        "mention": "helena morrissey",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Helena Morrissey",
                "confidence_score": "12.4708185"
            },
            {
                "entity": "Helena Shovelton",
                "confidence_score": "-11.337715"
            },
            {
                "entity": "James Morrissey",
                "confidence_score": "-11.3933115"
            }
        ]
    },
    {
        "mention": "bny mellon's",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Bank of New York Mellon",
                "confidence_score": "2.5827045"
            },
            {
                "entity": "Mellon Financial",
                "confidence_score": "-8.361842"
            },
            {
                "entity": "Carnegie Mellon University",
                "confidence_score": "-8.401823"
            }
        ]
    },
    {
        "mention": "newton investment management the investment association's",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Newton Investment Management",
                "confidence_score": "-0.90514743"
            },
            {
                "entity": "Institute of Asset Management",
                "confidence_score": "-11.122203"
            },
            {
                "entity": "Cambridge Investment Research",
                "confidence_score": "-11.31495"
            }
        ]
    },
    {
        "mention": "edi truell",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "ICG Enterprise Trust",
                "confidence_score": "-10.817267"
            },
            {
                "entity": "Institute of Chartered Accountants in England and Wales",
                "confidence_score": "-11.055156"
            },
            {
                "entity": "Institute of Chartered Accountants of Scotland",
                "confidence_score": "-11.718247"
            }
        ]
    },
    {
        "mention": "boris johnson",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Boris Johnson",
                "confidence_score": "11.597151"
            },
            {
                "entity": "Stanley Johnson (writer)",
                "confidence_score": "-8.037985"
            },
            {
                "entity": "Damian Johnson (broadcaster)",
                "confidence_score": "-11.740574"
            }
        ]
    },
    {
        "mention": "russell king",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Russell King (fraudster)",
                "confidence_score": "1.6295776"
            },
            {
                "entity": "Russell King (politician)",
                "confidence_score": "-9.670633"
            },
            {
                "entity": "Russell King",
                "confidence_score": "-9.79378"
            }
        ]
    },
    {
        "mention": "aggreko and spectris",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Spectris",
                "confidence_score": "-1.3166323"
            },
            {
                "entity": "Welsh, Carson, Anderson & Stowe",
                "confidence_score": "-11.109861"
            },
            {
                "entity": "Cr\u00e9dit Agricole Corporate and Investment Bank",
                "confidence_score": "-11.279719"
            }
        ]
    },
    {
        "mention": "ftse",
        "type": "ORG",
        "predictions": [
            {
                "entity": "FTSE Group",
                "confidence_score": "-0.10674975"
            },
            {
                "entity": "FTSE 100 Index",
                "confidence_score": "-1.1494522"
            },
            {
                "entity": "FTSE",
                "confidence_score": "-10.67216"
            }
        ]
    },
    {
        "mention": "ftse",
        "type": "ORG",
        "predictions": [
            {
                "entity": "FTSE Group",
                "confidence_score": "-0.10674975"
            },
            {
                "entity": "FTSE 100 Index",
                "confidence_score": "-1.1494522"
            },
            {
                "entity": "FTSE",
                "confidence_score": "-10.67216"
            }
        ]
    },
    {
        "mention": "self-improvement month",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Self-help",
                "confidence_score": "-4.678049"
            },
            {
                "entity": "Mental Health Awareness Month",
                "confidence_score": "-9.375929"
            },
            {
                "entity": "Self-enhancement",
                "confidence_score": "-9.728357"
            }
        ]
    },
    {
        "mention": "ny",
        "type": "ORG",
        "predictions": [
            {
                "entity": "New York City",
                "confidence_score": "1.142284"
            },
            {
                "entity": "New York (state)",
                "confidence_score": "-2.6529887"
            },
            {
                "entity": "New York",
                "confidence_score": "-11.903819"
            }
        ]
    },
    {
        "mention": "ny",
        "type": "ORG",
        "predictions": [
            {
                "entity": "New York City",
                "confidence_score": "1.142284"
            },
            {
                "entity": "New York (state)",
                "confidence_score": "-2.6529887"
            },
            {
                "entity": "New York",
                "confidence_score": "-11.903819"
            }
        ]
    },
    {
        "mention": "prweb",
        "type": "ORG",
        "predictions": [
            {
                "entity": "PR Newswire",
                "confidence_score": "-9.323785"
            },
            {
                "entity": "Classified Ventures",
                "confidence_score": "-11.721824"
            },
            {
                "entity": "SP-29 (Brazil)",
                "confidence_score": "-12.64419"
            }
        ]
    },
    {
        "mention": "self-improvement month",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Self-help",
                "confidence_score": "-4.678049"
            },
            {
                "entity": "Mental Health Awareness Month",
                "confidence_score": "-9.375929"
            },
            {
                "entity": "Self-enhancement",
                "confidence_score": "-9.728357"
            }
        ]
    },
    {
        "mention": "john lanzetta",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "John de Lancie",
                "confidence_score": "-10.304204"
            },
            {
                "entity": "John La Gatta",
                "confidence_score": "-11.185038"
            },
            {
                "entity": "John Coletta",
                "confidence_score": "-11.228605"
            }
        ]
    },
    {
        "mention": "chris chondrogiannis",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Giannis Chondrogiannis",
                "confidence_score": "-11.004365"
            },
            {
                "entity": "Christos Chomenidis",
                "confidence_score": "-11.863265"
            },
            {
                "entity": "Dimitrios Chondrokoukis",
                "confidence_score": "-11.931107"
            }
        ]
    },
    {
        "mention": "alyssa gursky",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Yuri Gurski",
                "confidence_score": "-11.014455"
            },
            {
                "entity": "Annabelle Gurwitch",
                "confidence_score": "-11.054784"
            },
            {
                "entity": "Alyssa Miller",
                "confidence_score": "-11.183948"
            }
        ]
    },
    {
        "mention": "matthew neary",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Matthew Causey",
                "confidence_score": "-10.513783"
            },
            {
                "entity": "Matthew Moy",
                "confidence_score": "-11.221783"
            },
            {
                "entity": "John Neary",
                "confidence_score": "-11.336711"
            }
        ]
    },
    {
        "mention": "ny",
        "type": "ORG",
        "predictions": [
            {
                "entity": "New York City",
                "confidence_score": "1.142284"
            },
            {
                "entity": "New York (state)",
                "confidence_score": "-2.6529887"
            },
            {
                "entity": "New York",
                "confidence_score": "-11.903819"
            }
        ]
    },
    {
        "mention": "lanzetta",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "James J. Lanzetta",
                "confidence_score": "-3.904218"
            },
            {
                "entity": "Leo Lanzetta",
                "confidence_score": "-6.4190917"
            },
            {
                "entity": "The Lanzetta Brothers",
                "confidence_score": "-8.503539"
            }
        ]
    },
    {
        "mention": "chondrogiannis",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Chondroyannos",
                "confidence_score": "-0.65183055"
            },
            {
                "entity": "Giannis Chondrogiannis",
                "confidence_score": "-6.3861647"
            },
            {
                "entity": "Dimitrios Chondrokoukis",
                "confidence_score": "-11.976914"
            }
        ]
    },
    {
        "mention": "gursky",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Samuel Gursky",
                "confidence_score": "-2.0000727"
            },
            {
                "entity": "Andreas Gursky",
                "confidence_score": "-5.2138023"
            },
            {
                "entity": "Yuri Gurski",
                "confidence_score": "-7.269423"
            }
        ]
    },
    {
        "mention": "neary",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Shay Neary",
                "confidence_score": "-6.900327"
            },
            {
                "entity": "J. Peter Neary",
                "confidence_score": "-8.053518"
            },
            {
                "entity": "John Neary",
                "confidence_score": "-8.956969"
            }
        ]
    },
    {
        "mention": "ny",
        "type": "ORG",
        "predictions": [
            {
                "entity": "New York City",
                "confidence_score": "1.142284"
            },
            {
                "entity": "New York (state)",
                "confidence_score": "-2.6529887"
            },
            {
                "entity": "New York",
                "confidence_score": "-11.903819"
            }
        ]
    },
    {
        "mention": "ny",
        "type": "ORG",
        "predictions": [
            {
                "entity": "New York City",
                "confidence_score": "1.142284"
            },
            {
                "entity": "New York (state)",
                "confidence_score": "-2.6529887"
            },
            {
                "entity": "New York",
                "confidence_score": "-11.903819"
            }
        ]
    },
    {
        "mention": "drs",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Physician",
                "confidence_score": "-1.9132121"
            },
            {
                "entity": "Doctor of Science",
                "confidence_score": "-8.66151"
            },
            {
                "entity": "Doctor of Social Work",
                "confidence_score": "-9.117049"
            }
        ]
    },
    {
        "mention": "lanzetta",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "James J. Lanzetta",
                "confidence_score": "-3.904218"
            },
            {
                "entity": "Leo Lanzetta",
                "confidence_score": "-6.4190917"
            },
            {
                "entity": "The Lanzetta Brothers",
                "confidence_score": "-8.503539"
            }
        ]
    },
    {
        "mention": "chondrogiannis",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Chondroyannos",
                "confidence_score": "-0.65183055"
            },
            {
                "entity": "Giannis Chondrogiannis",
                "confidence_score": "-6.3861647"
            },
            {
                "entity": "Dimitrios Chondrokoukis",
                "confidence_score": "-11.976914"
            }
        ]
    },
    {
        "mention": "gursky",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Samuel Gursky",
                "confidence_score": "-2.0000727"
            },
            {
                "entity": "Andreas Gursky",
                "confidence_score": "-5.2138023"
            },
            {
                "entity": "Yuri Gurski",
                "confidence_score": "-7.269423"
            }
        ]
    },
    {
        "mention": "ny",
        "type": "ORG",
        "predictions": [
            {
                "entity": "New York City",
                "confidence_score": "1.142284"
            },
            {
                "entity": "New York (state)",
                "confidence_score": "-2.6529887"
            },
            {
                "entity": "New York",
                "confidence_score": "-11.903819"
            }
        ]
    },
    {
        "mention": "john lanzetta",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "John de Lancie",
                "confidence_score": "-10.304204"
            },
            {
                "entity": "John La Gatta",
                "confidence_score": "-11.185038"
            },
            {
                "entity": "John Coletta",
                "confidence_score": "-11.228605"
            }
        ]
    },
    {
        "mention": "chris chondrogiannis",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Giannis Chondrogiannis",
                "confidence_score": "-11.004365"
            },
            {
                "entity": "Christos Chomenidis",
                "confidence_score": "-11.863265"
            },
            {
                "entity": "Dimitrios Chondrokoukis",
                "confidence_score": "-11.931107"
            }
        ]
    },
    {
        "mention": "alyssa gursky",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Yuri Gurski",
                "confidence_score": "-11.014455"
            },
            {
                "entity": "Annabelle Gurwitch",
                "confidence_score": "-11.054784"
            },
            {
                "entity": "Alyssa Miller",
                "confidence_score": "-11.183948"
            }
        ]
    },
    {
        "mention": "matthew neary",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Matthew Causey",
                "confidence_score": "-10.513783"
            },
            {
                "entity": "Matthew Moy",
                "confidence_score": "-11.221783"
            },
            {
                "entity": "John Neary",
                "confidence_score": "-11.336711"
            }
        ]
    },
    {
        "mention": "prweb",
        "type": "ORG",
        "predictions": [
            {
                "entity": "PR Newswire",
                "confidence_score": "-10.400137"
            },
            {
                "entity": "Pacific Research Laboratories",
                "confidence_score": "-11.057162"
            },
            {
                "entity": "SP-29 (Brazil)",
                "confidence_score": "-12.533347"
            }
        ]
    },
    {
        "mention": "tampa bay",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Tampa Bay",
                "confidence_score": "7.647836"
            },
            {
                "entity": "Tampa, Florida",
                "confidence_score": "-4.5728154"
            },
            {
                "entity": "Tampa Bay Rays",
                "confidence_score": "-6.093323"
            }
        ]
    },
    {
        "mention": "smoke",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Smoke",
                "confidence_score": "7.133982"
            },
            {
                "entity": "Tobacco smoke",
                "confidence_score": "-3.2859309"
            },
            {
                "entity": "Smoke (American band)",
                "confidence_score": "-11.326229"
            }
        ]
    },
    {
        "mention": "poutinerie",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Smoke's Poutinerie",
                "confidence_score": "-5.135813"
            },
            {
                "entity": "Poutine",
                "confidence_score": "-9.567143"
            },
            {
                "entity": "P\u00f6yry",
                "confidence_score": "-11.083997"
            }
        ]
    },
    {
        "mention": "smoke",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Smoke",
                "confidence_score": "7.133982"
            },
            {
                "entity": "Tobacco smoke",
                "confidence_score": "-3.2859309"
            },
            {
                "entity": "Smoke (American band)",
                "confidence_score": "-11.326229"
            }
        ]
    },
    {
        "mention": "poutinerie",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Smoke's Poutinerie",
                "confidence_score": "-2.056957"
            },
            {
                "entity": "Poutine",
                "confidence_score": "-6.5779386"
            },
            {
                "entity": "\u00c9mile Poirier",
                "confidence_score": "-11.885435"
            }
        ]
    },
    {
        "mention": "amalie arena",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Amalie Arena",
                "confidence_score": "6.3789687"
            },
            {
                "entity": "AMSOIL Arena",
                "confidence_score": "-11.399834"
            },
            {
                "entity": "Odense Isstadion",
                "confidence_score": "-11.975073"
            }
        ]
    },
    {
        "mention": "ryan smolkin",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ryan Flinn (ice hockey)",
                "confidence_score": "-12.153287"
            },
            {
                "entity": "Ryan Smillie",
                "confidence_score": "-12.277028"
            },
            {
                "entity": "Ryan Mulhern",
                "confidence_score": "-12.401957"
            }
        ]
    },
    {
        "mention": "non-league",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Non-League football",
                "confidence_score": "-3.3578646"
            },
            {
                "entity": "Minor League Baseball",
                "confidence_score": "-5.8877444"
            },
            {
                "entity": "Independent baseball league",
                "confidence_score": "-9.316045"
            }
        ]
    },
    {
        "mention": "cornwall",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Cornwall",
                "confidence_score": "0.06247021"
            },
            {
                "entity": "Cornwall, Connecticut",
                "confidence_score": "-3.990637"
            },
            {
                "entity": "Cornwall, New York",
                "confidence_score": "-4.9768653"
            }
        ]
    },
    {
        "mention": "monroe-woodbury",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Monroe County, New York",
                "confidence_score": "-8.621379"
            },
            {
                "entity": "Monroe County, Florida",
                "confidence_score": "-9.506399"
            },
            {
                "entity": "Monroe County, Tennessee",
                "confidence_score": "-10.650791"
            }
        ]
    },
    {
        "mention": "nash",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Nash County, North Carolina",
                "confidence_score": "-7.8910613"
            },
            {
                "entity": "Nash, Texas",
                "confidence_score": "-8.207181"
            },
            {
                "entity": "Nash, Oklahoma",
                "confidence_score": "-9.66731"
            }
        ]
    },
    {
        "mention": "erica bachiller",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Erica Bougard",
                "confidence_score": "-11.63207"
            },
            {
                "entity": "Erica Allar",
                "confidence_score": "-11.9245205"
            },
            {
                "entity": "Erica Clevenger",
                "confidence_score": "-12.145377"
            }
        ]
    },
    {
        "mention": "mackenzie tuttle",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bill Tuttle (swimmer)",
                "confidence_score": "-11.621792"
            },
            {
                "entity": "Reggie Crist",
                "confidence_score": "-11.667431"
            },
            {
                "entity": "Mackenzie Caquatto",
                "confidence_score": "-12.42677"
            }
        ]
    },
    {
        "mention": "allison dubick",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Allison Jones (athlete)",
                "confidence_score": "-11.291356"
            },
            {
                "entity": "Allie Kiick",
                "confidence_score": "-11.971826"
            },
            {
                "entity": "Carmen Marton",
                "confidence_score": "-12.037533"
            }
        ]
    },
    {
        "mention": "samantha reichman",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Samantha Futerman",
                "confidence_score": "-11.352302"
            },
            {
                "entity": "Alyssa Spellman",
                "confidence_score": "-11.452722"
            },
            {
                "entity": "Alyssa Beckerman",
                "confidence_score": "-11.546927"
            }
        ]
    },
    {
        "mention": "mackenzie tuttle",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bill Tuttle (swimmer)",
                "confidence_score": "-11.621792"
            },
            {
                "entity": "Reggie Crist",
                "confidence_score": "-11.667431"
            },
            {
                "entity": "Mackenzie Caquatto",
                "confidence_score": "-12.42677"
            }
        ]
    },
    {
        "mention": "lauren beneat",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Candice Bennatt",
                "confidence_score": "-11.597295"
            },
            {
                "entity": "Laurenne Ross",
                "confidence_score": "-11.623813"
            },
            {
                "entity": "Jessica Benites",
                "confidence_score": "-11.82593"
            }
        ]
    },
    {
        "mention": "ashley ayala",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ashley Greene",
                "confidence_score": "-11.2443695"
            },
            {
                "entity": "Ashley Ortega",
                "confidence_score": "-11.770499"
            },
            {
                "entity": "Ashley Robinson",
                "confidence_score": "-11.89666"
            }
        ]
    },
    {
        "mention": "monroe-woodbury",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Monroe County, New York",
                "confidence_score": "-8.621379"
            },
            {
                "entity": "Monroe County, Florida",
                "confidence_score": "-9.506399"
            },
            {
                "entity": "Monroe County, Tennessee",
                "confidence_score": "-10.650791"
            }
        ]
    },
    {
        "mention": "bachiller, beneat",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Chris Bachelder",
                "confidence_score": "-11.86659"
            },
            {
                "entity": "2016\u201317 Biathlon World Cup \u2013 Nation Men",
                "confidence_score": "-12.1328535"
            },
            {
                "entity": "2016 US Open \u2013 Men's Singles Qualifying",
                "confidence_score": "-12.339508"
            }
        ]
    },
    {
        "mention": "christman, tuttle",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Christ Church USA",
                "confidence_score": "-10.486736"
            },
            {
                "entity": "Theodore Christman",
                "confidence_score": "-11.15556"
            },
            {
                "entity": "Christman",
                "confidence_score": "-12.471601"
            }
        ]
    },
    {
        "mention": "jessica ebner",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jessica Jerome",
                "confidence_score": "-11.609344"
            },
            {
                "entity": "Jessica Fridrich",
                "confidence_score": "-11.63212"
            },
            {
                "entity": "Jessica Oyelowo",
                "confidence_score": "-12.286273"
            }
        ]
    },
    {
        "mention": "cornwall",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Cornwall",
                "confidence_score": "0.06247021"
            },
            {
                "entity": "Cornwall, Connecticut",
                "confidence_score": "-3.990637"
            },
            {
                "entity": "Cornwall, New York",
                "confidence_score": "-4.9768653"
            }
        ]
    },
    {
        "mention": "leishman",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Thomas Leishman",
                "confidence_score": "-6.654786"
            },
            {
                "entity": "Leishmania",
                "confidence_score": "-8.245823"
            },
            {
                "entity": "Leishman",
                "confidence_score": "-8.966611"
            }
        ]
    },
    {
        "mention": "cornwall",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Cornwall",
                "confidence_score": "0.06247021"
            },
            {
                "entity": "Cornwall, Connecticut",
                "confidence_score": "-3.990637"
            },
            {
                "entity": "Cornwall, New York",
                "confidence_score": "-4.9768653"
            }
        ]
    },
    {
        "mention": "bush",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "George W. Bush",
                "confidence_score": "-4.3197527"
            },
            {
                "entity": "George H. W. Bush",
                "confidence_score": "-4.6453295"
            },
            {
                "entity": "George P. Bush",
                "confidence_score": "-8.875043"
            }
        ]
    },
    {
        "mention": "lapinski",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Lisa Lapinski",
                "confidence_score": "-7.2525544"
            },
            {
                "entity": "2012 PSA World Series Finals",
                "confidence_score": "-11.674033"
            },
            {
                "entity": "Matches of Polish men's volleyball national team conducted by St\u00e9phane Antiga",
                "confidence_score": "-12.60867"
            }
        ]
    },
    {
        "mention": "pfeifer",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Scott Pfeifer",
                "confidence_score": "-6.657193"
            },
            {
                "entity": "John R. Pfeifer",
                "confidence_score": "-6.762395"
            },
            {
                "entity": "Melanie Pfeifer",
                "confidence_score": "-8.608552"
            }
        ]
    },
    {
        "mention": "alexander",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Alexander Hill (rower)",
                "confidence_score": "-11.330859"
            },
            {
                "entity": "Alexander Wolf",
                "confidence_score": "-11.411745"
            },
            {
                "entity": "Alexander M\u00fcller (skeleton racer)",
                "confidence_score": "-11.559477"
            }
        ]
    },
    {
        "mention": "corinne debeauvernet",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Jason De Carteret",
                "confidence_score": "-11.632197"
            },
            {
                "entity": "Corinne Debaine-Francfort",
                "confidence_score": "-11.711181"
            },
            {
                "entity": "Corinne Paliard",
                "confidence_score": "-12.116683"
            }
        ]
    },
    {
        "mention": "pb",
        "type": "ORG",
        "predictions": [
            {
                "entity": "P",
                "confidence_score": "-10.254656"
            },
            {
                "entity": "List of reporting marks: P",
                "confidence_score": "-12.217887"
            },
            {
                "entity": "Matches of Polish men's volleyball national team conducted by St\u00e9phane Antiga",
                "confidence_score": "-12.809165"
            }
        ]
    },
    {
        "mention": "rachel shih",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Rachel Komisarz",
                "confidence_score": "-11.299937"
            },
            {
                "entity": "Rachel Freier",
                "confidence_score": "-11.732764"
            },
            {
                "entity": "Rachel Lam",
                "confidence_score": "-11.919049"
            }
        ]
    },
    {
        "mention": "nicole lapinski",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Lisa Lapinski",
                "confidence_score": "-10.767877"
            },
            {
                "entity": "Nicole Lapin",
                "confidence_score": "-10.883456"
            },
            {
                "entity": "Tara Lipinski",
                "confidence_score": "-11.219933"
            }
        ]
    },
    {
        "mention": "pb",
        "type": "ORG",
        "predictions": [
            {
                "entity": "P",
                "confidence_score": "-10.254656"
            },
            {
                "entity": "List of reporting marks: P",
                "confidence_score": "-12.217887"
            },
            {
                "entity": "Matches of Polish men's volleyball national team conducted by St\u00e9phane Antiga",
                "confidence_score": "-12.809165"
            }
        ]
    },
    {
        "mention": "laurel parker-chan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "2865 Laurel",
                "confidence_score": "-11.932832"
            },
            {
                "entity": "Philip Chan",
                "confidence_score": "-13.075999"
            },
            {
                "entity": "Ray Chan",
                "confidence_score": "-13.112369"
            }
        ]
    },
    {
        "mention": "pb",
        "type": "ORG",
        "predictions": [
            {
                "entity": "P",
                "confidence_score": "-10.254656"
            },
            {
                "entity": "List of reporting marks: P",
                "confidence_score": "-12.217887"
            },
            {
                "entity": "Matches of Polish men's volleyball national team conducted by St\u00e9phane Antiga",
                "confidence_score": "-12.809165"
            }
        ]
    },
    {
        "mention": "pb",
        "type": "ORG",
        "predictions": [
            {
                "entity": "P",
                "confidence_score": "-10.254656"
            },
            {
                "entity": "List of reporting marks: P",
                "confidence_score": "-12.217887"
            },
            {
                "entity": "Matches of Polish men's volleyball national team conducted by St\u00e9phane Antiga",
                "confidence_score": "-12.809165"
            }
        ]
    },
    {
        "mention": "samantha hufcut",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Samantha Jo",
                "confidence_score": "-11.668341"
            },
            {
                "entity": "Samantha Foyle",
                "confidence_score": "-11.861024"
            },
            {
                "entity": "Samantha Baggett",
                "confidence_score": "-12.095691"
            }
        ]
    },
    {
        "mention": "pb",
        "type": "ORG",
        "predictions": [
            {
                "entity": "P",
                "confidence_score": "-10.254656"
            },
            {
                "entity": "List of reporting marks: P",
                "confidence_score": "-12.217887"
            },
            {
                "entity": "Matches of Polish men's volleyball national team conducted by St\u00e9phane Antiga",
                "confidence_score": "-12.809165"
            }
        ]
    },
    {
        "mention": "jamie alexander",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Scott Alexander",
                "confidence_score": "-11.429856"
            },
            {
                "entity": "Constantine Alexander",
                "confidence_score": "-12.101272"
            },
            {
                "entity": "Lee Alexander",
                "confidence_score": "-13.069893"
            }
        ]
    },
    {
        "mention": "pb",
        "type": "ORG",
        "predictions": [
            {
                "entity": "P",
                "confidence_score": "-10.254656"
            },
            {
                "entity": "List of reporting marks: P",
                "confidence_score": "-12.217887"
            },
            {
                "entity": "Matches of Polish men's volleyball national team conducted by St\u00e9phane Antiga",
                "confidence_score": "-12.809165"
            }
        ]
    },
    {
        "mention": "rachel shih",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Rachel Komisarz",
                "confidence_score": "-11.299937"
            },
            {
                "entity": "Rachel Freier",
                "confidence_score": "-11.732764"
            },
            {
                "entity": "Rachel Lam",
                "confidence_score": "-11.919049"
            }
        ]
    },
    {
        "mention": "korianne moore",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Carissa Moore",
                "confidence_score": "-11.223605"
            },
            {
                "entity": "Kristie Moore",
                "confidence_score": "-11.278461"
            },
            {
                "entity": "Mari\u00ebtte Koehorst",
                "confidence_score": "-12.68774"
            }
        ]
    },
    {
        "mention": "caceres, caceres, caceres, caceres",
        "type": "ORG",
        "predictions": [
            {
                "entity": "C\u00e1ceres, Spain",
                "confidence_score": "-5.076136"
            },
            {
                "entity": "Province of C\u00e1ceres",
                "confidence_score": "-6.3269353"
            },
            {
                "entity": "C\u00e1ceres, Antioquia",
                "confidence_score": "-8.255482"
            }
        ]
    },
    {
        "mention": "dexter",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Dexter, Maine",
                "confidence_score": "-6.3326216"
            },
            {
                "entity": "Dexter, New York",
                "confidence_score": "-6.455734"
            },
            {
                "entity": "Dexter, Illinois",
                "confidence_score": "-9.611636"
            }
        ]
    },
    {
        "mention": "martin",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Martin, Tennessee",
                "confidence_score": "-4.9932456"
            },
            {
                "entity": "Martin County, Florida",
                "confidence_score": "-6.452383"
            },
            {
                "entity": "Martin County, North Carolina",
                "confidence_score": "-7.516536"
            }
        ]
    },
    {
        "mention": "wilson",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Wilson County, Tennessee",
                "confidence_score": "-6.240399"
            },
            {
                "entity": "Wilson, North Carolina",
                "confidence_score": "-7.5464306"
            },
            {
                "entity": "Wilson, Texas",
                "confidence_score": "-8.418337"
            }
        ]
    },
    {
        "mention": "thompson",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Thompson (electoral district)",
                "confidence_score": "-7.4668307"
            },
            {
                "entity": "John Thompson (boxer)",
                "confidence_score": "-11.002513"
            },
            {
                "entity": "Antoine Thompson",
                "confidence_score": "-11.105662"
            }
        ]
    },
    {
        "mention": "jillian donlon",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Roger Donlon",
                "confidence_score": "-11.569699"
            },
            {
                "entity": "Jillian Saulnier",
                "confidence_score": "-11.624831"
            },
            {
                "entity": "Billy Donlon",
                "confidence_score": "-11.905844"
            }
        ]
    },
    {
        "mention": "julia laquidara",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Laquidara",
                "confidence_score": "-11.994839"
            },
            {
                "entity": "Lakia Aisha Barber",
                "confidence_score": "-12.019631"
            },
            {
                "entity": "Giuseppe Laguidara",
                "confidence_score": "-12.155067"
            }
        ]
    },
    {
        "mention": "erin mcgoldrick",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ryan McGoldrick",
                "confidence_score": "-10.974271"
            },
            {
                "entity": "Sean McGoldrick",
                "confidence_score": "-11.360036"
            },
            {
                "entity": "Erin McKenney",
                "confidence_score": "-11.366087"
            }
        ]
    },
    {
        "mention": "dana schmidt",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Dana Vollmer",
                "confidence_score": "-11.212751"
            },
            {
                "entity": "Dana Plato",
                "confidence_score": "-11.219087"
            },
            {
                "entity": "Dana Rettke",
                "confidence_score": "-11.973202"
            }
        ]
    },
    {
        "mention": "kaitlyn smolar",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Eugeniusz Smolar",
                "confidence_score": "-11.401817"
            },
            {
                "entity": "Kaitlyn Jones",
                "confidence_score": "-11.931025"
            },
            {
                "entity": "Kaitlynne Postel",
                "confidence_score": "-12.109621"
            }
        ]
    },
    {
        "mention": "molly miller",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Molly Miller",
                "confidence_score": "0.7731775"
            },
            {
                "entity": "Molly Culver",
                "confidence_score": "-11.278398"
            },
            {
                "entity": "Molly Wood",
                "confidence_score": "-11.404301"
            }
        ]
    },
    {
        "mention": "jordan thompson",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jordan Thompson (American football)",
                "confidence_score": "-8.184998"
            },
            {
                "entity": "Jordan Thompson (volleyball)",
                "confidence_score": "-8.280993"
            },
            {
                "entity": "Jordan Thompson (tennis)",
                "confidence_score": "-8.326518"
            }
        ]
    },
    {
        "mention": "magee",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Samantha Magee",
                "confidence_score": "-8.47456"
            },
            {
                "entity": "Magee, New York",
                "confidence_score": "-8.769428"
            },
            {
                "entity": "Lee Magee",
                "confidence_score": "-9.036416"
            }
        ]
    },
    {
        "mention": "davis",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Josh Davis (swimmer)",
                "confidence_score": "-10.500339"
            },
            {
                "entity": "Davis Steven",
                "confidence_score": "-11.579198"
            },
            {
                "entity": "D\u0101vis",
                "confidence_score": "-12.212327"
            }
        ]
    },
    {
        "mention": "scelta",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Liselotta Lennartsson",
                "confidence_score": "-11.690816"
            },
            {
                "entity": "Ritva Lemettinen",
                "confidence_score": "-11.862844"
            },
            {
                "entity": "Team SCA",
                "confidence_score": "-11.896632"
            }
        ]
    },
    {
        "mention": "mickayla dexter",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Sally Dexter",
                "confidence_score": "-10.85406"
            },
            {
                "entity": "Kim Dexter",
                "confidence_score": "-11.378102"
            },
            {
                "entity": "Viper (wrestler)",
                "confidence_score": "-11.427379"
            }
        ]
    },
    {
        "mention": "alyssa martin",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Alyssa Miller",
                "confidence_score": "-11.371221"
            },
            {
                "entity": "Alyssa Anderson",
                "confidence_score": "-11.44469"
            },
            {
                "entity": "Alyssa Leonard",
                "confidence_score": "-11.815791"
            }
        ]
    },
    {
        "mention": "littell",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Franklin Littell",
                "confidence_score": "-5.129702"
            },
            {
                "entity": "Littell",
                "confidence_score": "-7.8707776"
            },
            {
                "entity": "Robert E. Littell",
                "confidence_score": "-8.811658"
            }
        ]
    },
    {
        "mention": "flannery",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Flannery",
                "confidence_score": "7.5777273"
            },
            {
                "entity": "Jerry Flannery",
                "confidence_score": "-7.3536067"
            },
            {
                "entity": "John Flannery (golfer)",
                "confidence_score": "-9.7058115"
            }
        ]
    },
    {
        "mention": "kingston 93",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Volleyball at the 2010 Central American and Caribbean Games",
                "confidence_score": "-12.00921"
            },
            {
                "entity": "List of England national rugby union team results 2000\u201309",
                "confidence_score": "-12.282325"
            },
            {
                "entity": "Matches of Polish men's volleyball national team conducted by St\u00e9phane Antiga",
                "confidence_score": "-12.699129"
            }
        ]
    },
    {
        "mention": "rhinebeck",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Rhinebeck (town), New York",
                "confidence_score": "-4.6632495"
            },
            {
                "entity": "Rhinebeck (village), New York",
                "confidence_score": "-5.3310857"
            },
            {
                "entity": "Rhinebeck",
                "confidence_score": "-11.087489"
            }
        ]
    },
    {
        "mention": "vitale",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Jonathan Vitale",
                "confidence_score": "-9.491218"
            },
            {
                "entity": "John Vitale",
                "confidence_score": "-10.463062"
            },
            {
                "entity": "Danny Vitale",
                "confidence_score": "-10.511175"
            }
        ]
    },
    {
        "mention": "sistare, behrens, dahl",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sandris B\u0113rzi\u0146\u0161",
                "confidence_score": "-11.992166"
            },
            {
                "entity": "Jonas Leandersson",
                "confidence_score": "-12.020834"
            },
            {
                "entity": "J\u0101nis B\u0113rzi\u0146\u0161 (biathlete)",
                "confidence_score": "-12.390237"
            }
        ]
    },
    {
        "mention": "taylor robey",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Taylor Spivey",
                "confidence_score": "-11.465317"
            },
            {
                "entity": "Taylor Caby",
                "confidence_score": "-11.580527"
            },
            {
                "entity": "Taylor Bagley",
                "confidence_score": "-11.961347"
            }
        ]
    },
    {
        "mention": "miya vitale",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ami Vitale",
                "confidence_score": "-10.703291"
            },
            {
                "entity": "Amy Vitale",
                "confidence_score": "-10.926652"
            },
            {
                "entity": "Natallia Mikhnevich",
                "confidence_score": "-12.122735"
            }
        ]
    },
    {
        "mention": "kelly manley",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jake Manley",
                "confidence_score": "-11.340362"
            },
            {
                "entity": "Alyssa Manley",
                "confidence_score": "-11.627921"
            },
            {
                "entity": "Alvin Manley",
                "confidence_score": "-11.647689"
            }
        ]
    },
    {
        "mention": "brianna stinemire",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "St. Francis Xavier women's ice hockey",
                "confidence_score": "-11.303053"
            },
            {
                "entity": "Brianna Van Buren",
                "confidence_score": "-11.843126"
            },
            {
                "entity": "2005 Wismilak International",
                "confidence_score": "-11.990783"
            }
        ]
    },
    {
        "mention": "helen wunderlich",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Renner Wunderlich",
                "confidence_score": "-11.234546"
            },
            {
                "entity": "Janis Mars Wunderlich",
                "confidence_score": "-11.30878"
            },
            {
                "entity": "Magdalena Wunderlich",
                "confidence_score": "-11.692592"
            }
        ]
    },
    {
        "mention": "erin leirey",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Aliz\u00e9e Brien",
                "confidence_score": "-11.943175"
            },
            {
                "entity": "Erin Pearl",
                "confidence_score": "-12.064458"
            },
            {
                "entity": "Erin Gallagher",
                "confidence_score": "-12.288123"
            }
        ]
    },
    {
        "mention": "miya vitale",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ami Vitale",
                "confidence_score": "-10.703291"
            },
            {
                "entity": "Amy Vitale",
                "confidence_score": "-10.926652"
            },
            {
                "entity": "Natallia Mikhnevich",
                "confidence_score": "-12.122735"
            }
        ]
    },
    {
        "mention": "kingston",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Kingston, Jamaica",
                "confidence_score": "8.18341"
            },
            {
                "entity": "Kingston, Ontario",
                "confidence_score": "-0.2605768"
            },
            {
                "entity": "Kingston, New York",
                "confidence_score": "-6.002463"
            }
        ]
    },
    {
        "mention": "leirey",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Water polo at the 2011 World Aquatics Championships \u2013 Men's team rosters",
                "confidence_score": "-11.918557"
            },
            {
                "entity": "Radiance / Chi ni Kaeru (On the Earth)",
                "confidence_score": "-12.032064"
            },
            {
                "entity": "Eurytides callias",
                "confidence_score": "-12.806521"
            }
        ]
    },
    {
        "mention": "mitchell",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mitchell County, North Carolina",
                "confidence_score": "-7.5610466"
            },
            {
                "entity": "D. J. Mitchell",
                "confidence_score": "-11.048248"
            },
            {
                "entity": "Mitchell Gale",
                "confidence_score": "-11.665242"
            }
        ]
    },
    {
        "mention": "kathryn sistare",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kathryn Deane",
                "confidence_score": "-11.492135"
            },
            {
                "entity": "Kathryn Simmonds",
                "confidence_score": "-11.50648"
            },
            {
                "entity": "Kathryn Minshew",
                "confidence_score": "-11.869072"
            }
        ]
    },
    {
        "mention": "rhinebeck 0-1",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Rhinebeck (town), New York",
                "confidence_score": "-8.878014"
            },
            {
                "entity": "German exonyms (Lidzbark Warmi\u0144ski County)",
                "confidence_score": "-12.182052"
            },
            {
                "entity": "Matches of Polish men's volleyball national team conducted by St\u00e9phane Antiga",
                "confidence_score": "-12.723323"
            }
        ]
    },
    {
        "mention": "minisink valley 112",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Minisink",
                "confidence_score": "-5.292115"
            },
            {
                "entity": "Minisink, New York",
                "confidence_score": "-9.926866"
            },
            {
                "entity": "Minisink Ford, New York",
                "confidence_score": "-11.827094"
            }
        ]
    },
    {
        "mention": "grace murray",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Grace Murray",
                "confidence_score": "-4.2593913"
            },
            {
                "entity": "Grace Weber",
                "confidence_score": "-12.023028"
            },
            {
                "entity": "Grace Griffith",
                "confidence_score": "-12.120787"
            }
        ]
    },
    {
        "mention": "abigail gibbons",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Abigail Hopper Gibbons",
                "confidence_score": "-4.7268777"
            },
            {
                "entity": "Abigail Savage",
                "confidence_score": "-11.080864"
            },
            {
                "entity": "Abigail Guthrie",
                "confidence_score": "-11.939286"
            }
        ]
    },
    {
        "mention": "jennifer borek",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jennifer Ketcham",
                "confidence_score": "-11.305047"
            },
            {
                "entity": "Jennifer Rowe",
                "confidence_score": "-11.583733"
            },
            {
                "entity": "Scott Borek",
                "confidence_score": "-12.180284"
            }
        ]
    },
    {
        "mention": "erin clifford",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Brendan Clifford",
                "confidence_score": "-11.475499"
            },
            {
                "entity": "Erin Walter",
                "confidence_score": "-11.714674"
            },
            {
                "entity": "Erinn Walters",
                "confidence_score": "-11.797816"
            }
        ]
    },
    {
        "mention": "abigail gibbons",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Abigail Hopper Gibbons",
                "confidence_score": "-4.7268777"
            },
            {
                "entity": "Abigail Savage",
                "confidence_score": "-11.080864"
            },
            {
                "entity": "Abigail Guthrie",
                "confidence_score": "-11.939286"
            }
        ]
    },
    {
        "mention": "melina blezy",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Melina Almodovar",
                "confidence_score": "-11.212278"
            },
            {
                "entity": "List of The Clique characters",
                "confidence_score": "-11.519792"
            },
            {
                "entity": "Melina Le\u00f3n",
                "confidence_score": "-11.525133"
            }
        ]
    },
    {
        "mention": "madeline callahan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Madeline Weeks",
                "confidence_score": "-10.990237"
            },
            {
                "entity": "Madeline Harper",
                "confidence_score": "-11.238919"
            },
            {
                "entity": "19741 Callahan",
                "confidence_score": "-11.816297"
            }
        ]
    },
    {
        "mention": "scanlon",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Scanlon, Minnesota",
                "confidence_score": "-7.4168653"
            },
            {
                "entity": "Bill Scanlon",
                "confidence_score": "-9.031262"
            },
            {
                "entity": "Scanlon",
                "confidence_score": "-12.013302"
            }
        ]
    },
    {
        "mention": "losauro",
        "type": "ORG",
        "predictions": [
            {
                "entity": "2017\u201318 Biathlon World Cup \u2013 Nation Men",
                "confidence_score": "-11.837165"
            },
            {
                "entity": "2017\u201318 Biathlon World Cup \u2013 Nation Women",
                "confidence_score": "-11.983282"
            },
            {
                "entity": "Paragliding at the 2018 Asian Games \u2013 Women's individual accuracy",
                "confidence_score": "-12.187364"
            }
        ]
    },
    {
        "mention": "jennifer borek",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jennifer Ketcham",
                "confidence_score": "-11.305047"
            },
            {
                "entity": "Jennifer Rowe",
                "confidence_score": "-11.583733"
            },
            {
                "entity": "Scott Borek",
                "confidence_score": "-12.180284"
            }
        ]
    },
    {
        "mention": "scanlon",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Scanlon, Minnesota",
                "confidence_score": "-7.4168653"
            },
            {
                "entity": "Bill Scanlon",
                "confidence_score": "-9.031262"
            },
            {
                "entity": "Scanlon",
                "confidence_score": "-12.013302"
            }
        ]
    },
    {
        "mention": "middletown 1-2",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Middletown, Orange County, New York",
                "confidence_score": "-8.173306"
            },
            {
                "entity": "Middletown High School (New York)",
                "confidence_score": "-10.779649"
            },
            {
                "entity": "Middletown (New Rochelle)",
                "confidence_score": "-11.117455"
            }
        ]
    },
    {
        "mention": "newburgh free academy",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Newburgh Free Academy",
                "confidence_score": "10.947212"
            },
            {
                "entity": "Norwich Free Academy",
                "confidence_score": "-10.542211"
            },
            {
                "entity": "New Haven Academy",
                "confidence_score": "-11.832493"
            }
        ]
    },
    {
        "mention": "roosevelt",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Theodore Roosevelt",
                "confidence_score": "-0.7063169"
            },
            {
                "entity": "Franklin D. Roosevelt",
                "confidence_score": "-0.93159556"
            },
            {
                "entity": "Roosevelt family",
                "confidence_score": "-6.177557"
            }
        ]
    },
    {
        "mention": "emma reining",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Eloise Wellings",
                "confidence_score": "-11.348949"
            },
            {
                "entity": "Emma Polusny",
                "confidence_score": "-12.182616"
            },
            {
                "entity": "Emma Cumming",
                "confidence_score": "-12.350956"
            }
        ]
    },
    {
        "mention": "jamie forbes",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "DJ Forbes",
                "confidence_score": "-9.272473"
            },
            {
                "entity": "Jamie Ryan Scott Brooks",
                "confidence_score": "-11.832215"
            },
            {
                "entity": "Lee Jamieson",
                "confidence_score": "-11.862876"
            }
        ]
    },
    {
        "mention": "nfa",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Namibia Football Association",
                "confidence_score": "-8.300901"
            },
            {
                "entity": "National Women's Football Association",
                "confidence_score": "-10.895158"
            },
            {
                "entity": "National Intercollegiate Women's Fencing Association",
                "confidence_score": "-11.261217"
            }
        ]
    },
    {
        "mention": "ashley patierno",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ashley Tesoro",
                "confidence_score": "-11.493422"
            },
            {
                "entity": "Ashley House (TV presenter)",
                "confidence_score": "-11.633511"
            },
            {
                "entity": "Amy Palmiero-Winters",
                "confidence_score": "-11.7736435"
            }
        ]
    },
    {
        "mention": "fdr",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Athletics at the 2016 Summer Olympics \u2013 Women's 200 metres",
                "confidence_score": "-11.99917"
            },
            {
                "entity": "2012 PSA World Series Finals",
                "confidence_score": "-12.05205"
            },
            {
                "entity": "Paragliding at the 2018 Asian Games \u2013 Women's individual accuracy",
                "confidence_score": "-12.492154"
            }
        ]
    },
    {
        "mention": "emma reining",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Eloise Wellings",
                "confidence_score": "-11.348949"
            },
            {
                "entity": "Emma Polusny",
                "confidence_score": "-12.182616"
            },
            {
                "entity": "Emma Cumming",
                "confidence_score": "-12.350956"
            }
        ]
    },
    {
        "mention": "mercedes dewitt",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mercedes Gonzalez",
                "confidence_score": "-11.6119"
            },
            {
                "entity": "Mercedes Nicoll",
                "confidence_score": "-12.153664"
            },
            {
                "entity": "Mercedes P\u00e9rez",
                "confidence_score": "-12.266618"
            }
        ]
    },
    {
        "mention": "nfa",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Namibia Football Association",
                "confidence_score": "-8.300901"
            },
            {
                "entity": "National Women's Football Association",
                "confidence_score": "-10.895158"
            },
            {
                "entity": "National Intercollegiate Women's Fencing Association",
                "confidence_score": "-11.261217"
            }
        ]
    },
    {
        "mention": "kaylee debenedictus",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Lisa DeBenedictis",
                "confidence_score": "-11.376528"
            },
            {
                "entity": "23131 Debenedictis",
                "confidence_score": "-11.659012"
            },
            {
                "entity": "Sybilla Dekker",
                "confidence_score": "-11.822026"
            }
        ]
    },
    {
        "mention": "nfa",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Namibia Football Association",
                "confidence_score": "-8.300901"
            },
            {
                "entity": "National Women's Football Association",
                "confidence_score": "-10.895158"
            },
            {
                "entity": "National Intercollegiate Women's Fencing Association",
                "confidence_score": "-11.261217"
            }
        ]
    },
    {
        "mention": "roosevelt",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Theodore Roosevelt",
                "confidence_score": "-0.7063169"
            },
            {
                "entity": "Franklin D. Roosevelt",
                "confidence_score": "-0.93159556"
            },
            {
                "entity": "Roosevelt family",
                "confidence_score": "-6.177557"
            }
        ]
    },
    {
        "mention": "sydney reede",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Danielle Drady-Harte",
                "confidence_score": "-11.871711"
            },
            {
                "entity": "Desiree Reed-Francois",
                "confidence_score": "-12.025798"
            },
            {
                "entity": "Florian Becke",
                "confidence_score": "-12.082028"
            }
        ]
    },
    {
        "mention": "amber diehl",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Edith Diehl",
                "confidence_score": "-10.898139"
            },
            {
                "entity": "Ernie Diehl",
                "confidence_score": "-11.753831"
            },
            {
                "entity": "George Diehl",
                "confidence_score": "-11.889114"
            }
        ]
    },
    {
        "mention": "fdr",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Athletics at the 2016 Summer Olympics \u2013 Women's 200 metres",
                "confidence_score": "-11.99917"
            },
            {
                "entity": "2012 PSA World Series Finals",
                "confidence_score": "-12.05205"
            },
            {
                "entity": "Paragliding at the 2018 Asian Games \u2013 Women's individual accuracy",
                "confidence_score": "-12.492154"
            }
        ]
    },
    {
        "mention": "roosevelt",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Theodore Roosevelt",
                "confidence_score": "-0.7063169"
            },
            {
                "entity": "Franklin D. Roosevelt",
                "confidence_score": "-0.93159556"
            },
            {
                "entity": "Roosevelt family",
                "confidence_score": "-6.177557"
            }
        ]
    },
    {
        "mention": "mahoney, giesler",
        "type": "ORG",
        "predictions": [
            {
                "entity": "James E. Mahoney",
                "confidence_score": "-10.010175"
            },
            {
                "entity": "J. Daniel Mahoney",
                "confidence_score": "-10.85868"
            },
            {
                "entity": "John Mahoney (Ohio politician)",
                "confidence_score": "-11.785099"
            }
        ]
    },
    {
        "mention": "colaizzo,",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Anthony Colaizzo",
                "confidence_score": "-7.8121324"
            },
            {
                "entity": "Kevin Lenzo",
                "confidence_score": "-11.407506"
            },
            {
                "entity": "Matt Apuzzo",
                "confidence_score": "-11.782017"
            }
        ]
    },
    {
        "mention": "dosio",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Dosolo",
                "confidence_score": "-11.105489"
            },
            {
                "entity": "Doster",
                "confidence_score": "-12.950224"
            },
            {
                "entity": "Dosso",
                "confidence_score": "-12.973463"
            }
        ]
    },
    {
        "mention": "newburgh free academy",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Newburgh Free Academy",
                "confidence_score": "10.947212"
            },
            {
                "entity": "Norwich Free Academy",
                "confidence_score": "-10.542211"
            },
            {
                "entity": "New Haven Academy",
                "confidence_score": "-11.832493"
            }
        ]
    },
    {
        "mention": "roosevelt",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Theodore Roosevelt",
                "confidence_score": "-0.7063169"
            },
            {
                "entity": "Franklin D. Roosevelt",
                "confidence_score": "-0.93159556"
            },
            {
                "entity": "Roosevelt family",
                "confidence_score": "-6.177557"
            }
        ]
    },
    {
        "mention": "valley central 101",
        "type": "LOC",
        "predictions": [
            {
                "entity": "List of New York area codes",
                "confidence_score": "-10.60297"
            },
            {
                "entity": "Lower Northfield, Nova Scotia",
                "confidence_score": "-12.319566"
            },
            {
                "entity": "N-Formylmethionine (data page)",
                "confidence_score": "-12.610504"
            }
        ]
    },
    {
        "mention": "marlboro",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Marlboro Township, New Jersey",
                "confidence_score": "-7.3392353"
            },
            {
                "entity": "Marlboro County, South Carolina",
                "confidence_score": "-8.040996"
            },
            {
                "entity": "Marlboro, New York",
                "confidence_score": "-8.578139"
            }
        ]
    },
    {
        "mention": "marlboro",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Marlboro Township, New Jersey",
                "confidence_score": "-7.3392353"
            },
            {
                "entity": "Marlboro County, South Carolina",
                "confidence_score": "-8.040996"
            },
            {
                "entity": "Marlboro, New York",
                "confidence_score": "-8.578139"
            }
        ]
    },
    {
        "mention": "callo",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Callabo",
                "confidence_score": "-11.269776"
            },
            {
                "entity": "Rob Calloway",
                "confidence_score": "-11.617122"
            },
            {
                "entity": "Calle Brown",
                "confidence_score": "-11.9989"
            }
        ]
    },
    {
        "mention": "flynn",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Flynn",
                "confidence_score": "-1.6803088"
            },
            {
                "entity": "Flynn (PaaS)",
                "confidence_score": "-10.030812"
            },
            {
                "entity": "Flynn, Texas",
                "confidence_score": "-10.3690195"
            }
        ]
    },
    {
        "mention": "benkert",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Joseph Benkert",
                "confidence_score": "-6.5621977"
            },
            {
                "entity": "Kurt Benkert",
                "confidence_score": "-9.788466"
            },
            {
                "entity": "Benkert",
                "confidence_score": "-10.3453"
            }
        ]
    },
    {
        "mention": "demassio",
        "type": "ORG",
        "predictions": [
            {
                "entity": "DeMeco Ryans",
                "confidence_score": "-11.537514"
            },
            {
                "entity": "Harry A. DeMaso",
                "confidence_score": "-11.57184"
            },
            {
                "entity": "Demelza Reveley",
                "confidence_score": "-11.730565"
            }
        ]
    },
    {
        "mention": "daniela lusuriello",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Daniela Sruoga",
                "confidence_score": "-12.154161"
            },
            {
                "entity": "Daniela Porcelli",
                "confidence_score": "-12.17199"
            },
            {
                "entity": "Daniela Fusar Poli",
                "confidence_score": "-12.490572"
            }
        ]
    },
    {
        "mention": "vc",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Paragliding at the 2018 Asian Games \u2013 Women's individual accuracy",
                "confidence_score": "-11.93056"
            },
            {
                "entity": "Swimming at the 2006 Commonwealth Games \u2013 Women's 100 metre EAD freestyle",
                "confidence_score": "-12.478692"
            },
            {
                "entity": "Swimming at the 2006 Commonwealth Games \u2013 Women's 50 metre EAD freestyle",
                "confidence_score": "-12.507053"
            }
        ]
    },
    {
        "mention": "erin carmody",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Erin Carmody",
                "confidence_score": "4.873707"
            },
            {
                "entity": "Anson Carmody",
                "confidence_score": "-11.8346815"
            },
            {
                "entity": "Carmody",
                "confidence_score": "-11.857648"
            }
        ]
    },
    {
        "mention": "vc",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Paragliding at the 2018 Asian Games \u2013 Women's individual accuracy",
                "confidence_score": "-11.93056"
            },
            {
                "entity": "Swimming at the 2006 Commonwealth Games \u2013 Women's 100 metre EAD freestyle",
                "confidence_score": "-12.478692"
            },
            {
                "entity": "Swimming at the 2006 Commonwealth Games \u2013 Women's 50 metre EAD freestyle",
                "confidence_score": "-12.507053"
            }
        ]
    },
    {
        "mention": "emily boldrin",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Emily Bold",
                "confidence_score": "-9.389087"
            },
            {
                "entity": "Michele Boldrin",
                "confidence_score": "-11.181996"
            },
            {
                "entity": "Mikaela Shiffrin",
                "confidence_score": "-11.435277"
            }
        ]
    },
    {
        "mention": "vc",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Paragliding at the 2018 Asian Games \u2013 Women's individual accuracy",
                "confidence_score": "-11.3718"
            },
            {
                "entity": "Volleyball at the 2010 Central American and Caribbean Games",
                "confidence_score": "-11.400844"
            },
            {
                "entity": "Swimming at the 2006 Commonwealth Games \u2013 Women's 100 metre EAD freestyle",
                "confidence_score": "-11.941209"
            }
        ]
    },
    {
        "mention": "daniela lusuriello",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Daniela Sruoga",
                "confidence_score": "-12.154161"
            },
            {
                "entity": "Daniela Porcelli",
                "confidence_score": "-12.17199"
            },
            {
                "entity": "Daniela Fusar Poli",
                "confidence_score": "-12.490572"
            }
        ]
    },
    {
        "mention": "vc",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Paragliding at the 2018 Asian Games \u2013 Women's individual accuracy",
                "confidence_score": "-11.3718"
            },
            {
                "entity": "Volleyball at the 2010 Central American and Caribbean Games",
                "confidence_score": "-11.400844"
            },
            {
                "entity": "Swimming at the 2006 Commonwealth Games \u2013 Women's 100 metre EAD freestyle",
                "confidence_score": "-11.941209"
            }
        ]
    },
    {
        "mention": "erin carmody",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Erin Carmody",
                "confidence_score": "4.873707"
            },
            {
                "entity": "Anson Carmody",
                "confidence_score": "-11.8346815"
            },
            {
                "entity": "Carmody",
                "confidence_score": "-11.857648"
            }
        ]
    },
    {
        "mention": "vc",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Paragliding at the 2018 Asian Games \u2013 Women's individual accuracy",
                "confidence_score": "-11.3718"
            },
            {
                "entity": "Volleyball at the 2010 Central American and Caribbean Games",
                "confidence_score": "-11.400844"
            },
            {
                "entity": "Swimming at the 2006 Commonwealth Games \u2013 Women's 100 metre EAD freestyle",
                "confidence_score": "-11.941209"
            }
        ]
    },
    {
        "mention": "antonia demassio",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Nancy Dell'Olio",
                "confidence_score": "-11.596481"
            },
            {
                "entity": "Nancy Patricia Guti\u00e9rrez",
                "confidence_score": "-11.766889"
            },
            {
                "entity": "Deanna Demuzio",
                "confidence_score": "-12.183866"
            }
        ]
    },
    {
        "mention": "garcia",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Garc\u00eda Municipality",
                "confidence_score": "-10.249774"
            },
            {
                "entity": "Garc\u00eda, Nuevo Le\u00f3n",
                "confidence_score": "-10.677299"
            },
            {
                "entity": "Garc\u00eda",
                "confidence_score": "-12.67892"
            }
        ]
    },
    {
        "mention": "carmody",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Anson Carmody",
                "confidence_score": "-6.5828695"
            },
            {
                "entity": "Erin Carmody",
                "confidence_score": "-7.5903225"
            },
            {
                "entity": "Bill Carmody",
                "confidence_score": "-8.978158"
            }
        ]
    },
    {
        "mention": "emily boldrin",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Emily Bold",
                "confidence_score": "-9.389087"
            },
            {
                "entity": "Michele Boldrin",
                "confidence_score": "-11.181996"
            },
            {
                "entity": "Mikaela Shiffrin",
                "confidence_score": "-11.435277"
            }
        ]
    },
    {
        "mention": "vc",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Paragliding at the 2018 Asian Games \u2013 Women's individual accuracy",
                "confidence_score": "-11.3718"
            },
            {
                "entity": "Volleyball at the 2010 Central American and Caribbean Games",
                "confidence_score": "-11.400844"
            },
            {
                "entity": "Swimming at the 2006 Commonwealth Games \u2013 Women's 100 metre EAD freestyle",
                "confidence_score": "-11.941209"
            }
        ]
    },
    {
        "mention": "allison armistead",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Allison Schmitt",
                "confidence_score": "-11.507933"
            },
            {
                "entity": "Allison Ahlfeldt",
                "confidence_score": "-12.137777"
            },
            {
                "entity": "Allison George",
                "confidence_score": "-12.145497"
            }
        ]
    },
    {
        "mention": "vc",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Paragliding at the 2018 Asian Games \u2013 Women's individual accuracy",
                "confidence_score": "-11.93056"
            },
            {
                "entity": "Swimming at the 2006 Commonwealth Games \u2013 Women's 100 metre EAD freestyle",
                "confidence_score": "-12.478692"
            },
            {
                "entity": "Swimming at the 2006 Commonwealth Games \u2013 Women's 50 metre EAD freestyle",
                "confidence_score": "-12.507053"
            }
        ]
    },
    {
        "mention": "marlboro",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Marlboro Township, New Jersey",
                "confidence_score": "-7.3392353"
            },
            {
                "entity": "Marlboro County, South Carolina",
                "confidence_score": "-8.040996"
            },
            {
                "entity": "Marlboro, New York",
                "confidence_score": "-8.578139"
            }
        ]
    },
    {
        "mention": "callo",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Callabo",
                "confidence_score": "-11.269776"
            },
            {
                "entity": "Rob Calloway",
                "confidence_score": "-11.617122"
            },
            {
                "entity": "Calle Brown",
                "confidence_score": "-11.9989"
            }
        ]
    },
    {
        "mention": "camuso",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Caluso",
                "confidence_score": "-10.479674"
            },
            {
                "entity": "Cammoro",
                "confidence_score": "-10.969311"
            },
            {
                "entity": "Camino, Piedmont",
                "confidence_score": "-11.363803"
            }
        ]
    },
    {
        "mention": "demassio",
        "type": "ORG",
        "predictions": [
            {
                "entity": "DeMeco Ryans",
                "confidence_score": "-11.537514"
            },
            {
                "entity": "Harry A. DeMaso",
                "confidence_score": "-11.57184"
            },
            {
                "entity": "Demelza Reveley",
                "confidence_score": "-11.730565"
            }
        ]
    },
    {
        "mention": "benkert",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Joseph Benkert",
                "confidence_score": "-6.5621977"
            },
            {
                "entity": "Kurt Benkert",
                "confidence_score": "-9.788466"
            },
            {
                "entity": "Benkert",
                "confidence_score": "-10.3453"
            }
        ]
    },
    {
        "mention": "marlboro",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Marlboro Township, New Jersey",
                "confidence_score": "-7.3392353"
            },
            {
                "entity": "Marlboro County, South Carolina",
                "confidence_score": "-8.040996"
            },
            {
                "entity": "Marlboro, New York",
                "confidence_score": "-8.578139"
            }
        ]
    },
    {
        "mention": "jason day",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jason Day",
                "confidence_score": "11.941165"
            },
            {
                "entity": "Jason Day (fighter)",
                "confidence_score": "-6.49592"
            },
            {
                "entity": "Jason Day (rower)",
                "confidence_score": "-9.964982"
            }
        ]
    },
    {
        "mention": "rory mcilroy",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Rory McIlroy",
                "confidence_score": "11.389794"
            },
            {
                "entity": "Rory McIlroy PGA Tour",
                "confidence_score": "2.430814"
            },
            {
                "entity": "Drake McElroy",
                "confidence_score": "-11.8630495"
            }
        ]
    },
    {
        "mention": "the bmw championship",
        "type": "ORG",
        "predictions": [
            {
                "entity": "BMW Championship (PGA Tour)",
                "confidence_score": "0.024830054"
            },
            {
                "entity": "BMW PGA Championship",
                "confidence_score": "-5.0464377"
            },
            {
                "entity": "2018 BMW PGA Championship",
                "confidence_score": "-8.955152"
            }
        ]
    },
    {
        "mention": "east lake",
        "type": "LOC",
        "predictions": [
            {
                "entity": "East Lake Golf Club",
                "confidence_score": "-3.0913558"
            },
            {
                "entity": "East Lake, Atlanta",
                "confidence_score": "-6.2831025"
            },
            {
                "entity": "East Lake station",
                "confidence_score": "-10.386762"
            }
        ]
    },
    {
        "mention": "jordan spieth",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jordan Spieth",
                "confidence_score": "10.039963"
            },
            {
                "entity": "Jordan Schmaltz",
                "confidence_score": "-11.823285"
            },
            {
                "entity": "Jordan Schafer",
                "confidence_score": "-11.875056"
            }
        ]
    },
    {
        "mention": "rickie fowler",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Rickie Fowler",
                "confidence_score": "10.169225"
            },
            {
                "entity": "Kyle Fowler",
                "confidence_score": "-11.067958"
            },
            {
                "entity": "Dustin Fowler",
                "confidence_score": "-11.730964"
            }
        ]
    },
    {
        "mention": "henrik stenson",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Henrik Stenson",
                "confidence_score": "5.985124"
            },
            {
                "entity": "Henrik Larsson",
                "confidence_score": "-11.475375"
            },
            {
                "entity": "Henrik Gustavsson",
                "confidence_score": "-11.963934"
            }
        ]
    },
    {
        "mention": "bubba watson",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bubba Watson",
                "confidence_score": "10.985373"
            },
            {
                "entity": "Bob Watson",
                "confidence_score": "-10.456147"
            },
            {
                "entity": "Mark Watson (baseball)",
                "confidence_score": "-11.611037"
            }
        ]
    },
    {
        "mention": "stenson",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Henrik Stenson",
                "confidence_score": "2.7928336"
            },
            {
                "entity": "2012 PSA World Series Finals",
                "confidence_score": "-11.788213"
            },
            {
                "entity": "2010 Alessandria Challenger",
                "confidence_score": "-12.244683"
            }
        ]
    },
    {
        "mention": "ryder cup",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Ryder Cup",
                "confidence_score": "10.158486"
            },
            {
                "entity": "2008 Ryder Cup",
                "confidence_score": "-6.463095"
            },
            {
                "entity": "1983 Ryder Cup",
                "confidence_score": "-8.070945"
            }
        ]
    },
    {
        "mention": "jb holmes",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "J. B. Holmes",
                "confidence_score": "-9.298498"
            },
            {
                "entity": "John Holmes (actor)",
                "confidence_score": "-10.080251"
            },
            {
                "entity": "John Robert Holmes",
                "confidence_score": "-11.802656"
            }
        ]
    },
    {
        "mention": "spieth",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Jordan Spieth",
                "confidence_score": "-5.241715"
            },
            {
                "entity": "Spieth",
                "confidence_score": "-5.615644"
            },
            {
                "entity": "Fritz Seyferth",
                "confidence_score": "-11.705932"
            }
        ]
    },
    {
        "mention": "mcilroy",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Greg McElroy",
                "confidence_score": "-8.939454"
            },
            {
                "entity": "Sammy McIlroy",
                "confidence_score": "-9.070657"
            },
            {
                "entity": "Charles McIlroy",
                "confidence_score": "-9.35769"
            }
        ]
    },
    {
        "mention": "mcilroy",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Greg McElroy",
                "confidence_score": "-8.209774"
            },
            {
                "entity": "Sammy McIlroy",
                "confidence_score": "-8.422946"
            },
            {
                "entity": "Jimmy McIlroy",
                "confidence_score": "-8.885985"
            }
        ]
    },
    {
        "mention": "cv",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "National symbols of Croatia",
                "confidence_score": "-11.440146"
            },
            {
                "entity": "CP football",
                "confidence_score": "-11.52393"
            },
            {
                "entity": "Matches of Polish men's volleyball national team conducted by St\u00e9phane Antiga",
                "confidence_score": "-12.806836"
            }
        ]
    },
    {
        "mention": "billy horschel",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Billy Horschel",
                "confidence_score": "6.675257"
            },
            {
                "entity": "Billy Houliston",
                "confidence_score": "-11.850418"
            },
            {
                "entity": "Billy Kershaw",
                "confidence_score": "-12.184087"
            }
        ]
    },
    {
        "mention": "cheder chabad",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Cheder",
                "confidence_score": "-0.84406376"
            },
            {
                "entity": "Chabad",
                "confidence_score": "-1.5600905"
            },
            {
                "entity": "Cheder sheini",
                "confidence_score": "-6.693649"
            }
        ]
    },
    {
        "mention": "monsey",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Monsey, New York",
                "confidence_score": "0.4643005"
            },
            {
                "entity": "Ohr Somayach, Monsey",
                "confidence_score": "0.066042505"
            },
            {
                "entity": "Monsey, Benin",
                "confidence_score": "-7.2587476"
            }
        ]
    },
    {
        "mention": "torah",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Torah",
                "confidence_score": "8.946289"
            },
            {
                "entity": "Judaism",
                "confidence_score": "-2.1932926"
            },
            {
                "entity": "Torah Judaism",
                "confidence_score": "-2.223769"
            }
        ]
    },
    {
        "mention": "monsey",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Monsey, New York",
                "confidence_score": "0.4643005"
            },
            {
                "entity": "Ohr Somayach, Monsey",
                "confidence_score": "0.066042505"
            },
            {
                "entity": "Monsey, Benin",
                "confidence_score": "-7.2587476"
            }
        ]
    },
    {
        "mention": "abundance",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Flourishing",
                "confidence_score": "-8.196713"
            },
            {
                "entity": "Abundance (ecology)",
                "confidence_score": "-8.759666"
            },
            {
                "entity": "Abundance",
                "confidence_score": "-12.01004"
            }
        ]
    },
    {
        "mention": "parsha perspective",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Paroksha",
                "confidence_score": "-8.532312"
            },
            {
                "entity": "Pardes (Jewish exegesis)",
                "confidence_score": "-8.909293"
            },
            {
                "entity": "Parvarrish \u2013 Kuchh Khattee Kuchh Meethi",
                "confidence_score": "-11.168834"
            }
        ]
    },
    {
        "mention": "binyomin philipson",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Binyomin Adler",
                "confidence_score": "-9.807174"
            },
            {
                "entity": "Binyomin Paler",
                "confidence_score": "-11.014165"
            },
            {
                "entity": "Binjamin Wilkomirski",
                "confidence_score": "-11.560482"
            }
        ]
    },
    {
        "mention": "ellen",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ellen DeGeneres",
                "confidence_score": "-5.1869545"
            },
            {
                "entity": "Ellen Sue Miller",
                "confidence_score": "-10.035867"
            },
            {
                "entity": "Ellen Willis",
                "confidence_score": "-10.262172"
            }
        ]
    },
    {
        "mention": "judaism",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Judaism",
                "confidence_score": "6.8220835"
            },
            {
                "entity": "Jews",
                "confidence_score": "-4.76475"
            },
            {
                "entity": "Jewish culture",
                "confidence_score": "-7.6676197"
            }
        ]
    },
    {
        "mention": "judaism",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Judaism",
                "confidence_score": "6.8220835"
            },
            {
                "entity": "Jews",
                "confidence_score": "-4.76475"
            },
            {
                "entity": "Jewish culture",
                "confidence_score": "-7.6676197"
            }
        ]
    },
    {
        "mention": "torah",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Torah",
                "confidence_score": "8.946289"
            },
            {
                "entity": "Judaism",
                "confidence_score": "-2.1932926"
            },
            {
                "entity": "Torah Judaism",
                "confidence_score": "-2.223769"
            }
        ]
    },
    {
        "mention": "the land of\u00a0israel",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Land of Israel",
                "confidence_score": "4.692332"
            },
            {
                "entity": "Israel",
                "confidence_score": "-1.7469211"
            },
            {
                "entity": "History of Israel",
                "confidence_score": "-3.8692417"
            }
        ]
    },
    {
        "mention": "torah",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Torah",
                "confidence_score": "8.946289"
            },
            {
                "entity": "Judaism",
                "confidence_score": "-2.1932926"
            },
            {
                "entity": "Torah Judaism",
                "confidence_score": "-2.223769"
            }
        ]
    },
    {
        "mention": "vayelech",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Vayelech",
                "confidence_score": "11.061332"
            },
            {
                "entity": "Halakha",
                "confidence_score": "-3.9726608"
            },
            {
                "entity": "Torah reading",
                "confidence_score": "-6.3657146"
            }
        ]
    },
    {
        "mention": "devorim",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Devarim (parsha)",
                "confidence_score": "-1.3112812"
            },
            {
                "entity": "Devarim",
                "confidence_score": "-7.4052434"
            },
            {
                "entity": "Dvora, Israel",
                "confidence_score": "-10.990291"
            }
        ]
    },
    {
        "mention": "mitzvah",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Mitzvah",
                "confidence_score": "8.151362"
            },
            {
                "entity": "Judaism",
                "confidence_score": "-5.1479926"
            },
            {
                "entity": "Bar and bat mitzvah",
                "confidence_score": "-7.343223"
            }
        ]
    },
    {
        "mention": "moses",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Moses",
                "confidence_score": "9.2273"
            },
            {
                "entity": "Law of Moses",
                "confidence_score": "-7.2025633"
            },
            {
                "entity": "Aaron",
                "confidence_score": "-7.804746"
            }
        ]
    },
    {
        "mention": "shemittah",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Shmita",
                "confidence_score": "-3.483737"
            },
            {
                "entity": "Shabbaton",
                "confidence_score": "-6.6045055"
            },
            {
                "entity": "Shemini Atzeret",
                "confidence_score": "-9.449801"
            }
        ]
    },
    {
        "mention": "temple",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Temple",
                "confidence_score": "6.2415586"
            },
            {
                "entity": "Temple (Latter Day Saints)",
                "confidence_score": "-4.95863"
            },
            {
                "entity": "Temple (LDS Church)",
                "confidence_score": "-6.020338"
            }
        ]
    },
    {
        "mention": "torah",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Torah",
                "confidence_score": "8.946289"
            },
            {
                "entity": "Judaism",
                "confidence_score": "-2.1932926"
            },
            {
                "entity": "Torah Judaism",
                "confidence_score": "-2.223769"
            }
        ]
    },
    {
        "mention": "judaism",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Judaism",
                "confidence_score": "6.8220835"
            },
            {
                "entity": "Jews",
                "confidence_score": "-4.76475"
            },
            {
                "entity": "Jewish culture",
                "confidence_score": "-7.6676197"
            }
        ]
    },
    {
        "mention": "sinai",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sinai Peninsula",
                "confidence_score": "-1.5670881"
            },
            {
                "entity": "Mount Sinai",
                "confidence_score": "-6.1581283"
            },
            {
                "entity": "Sinaia",
                "confidence_score": "-9.737592"
            }
        ]
    },
    {
        "mention": "torah",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Torah",
                "confidence_score": "8.946289"
            },
            {
                "entity": "Judaism",
                "confidence_score": "-2.1932926"
            },
            {
                "entity": "Torah Judaism",
                "confidence_score": "-2.223769"
            }
        ]
    },
    {
        "mention": "torah",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Torah",
                "confidence_score": "8.946289"
            },
            {
                "entity": "Judaism",
                "confidence_score": "-2.1932926"
            },
            {
                "entity": "Torah Judaism",
                "confidence_score": "-2.223769"
            }
        ]
    },
    {
        "mention": "sinai",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sinai Peninsula",
                "confidence_score": "-1.5670881"
            },
            {
                "entity": "Mount Sinai",
                "confidence_score": "-6.1581283"
            },
            {
                "entity": "Sinaia",
                "confidence_score": "-9.737592"
            }
        ]
    },
    {
        "mention": "torah",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Torah",
                "confidence_score": "8.946289"
            },
            {
                "entity": "Judaism",
                "confidence_score": "-2.1932926"
            },
            {
                "entity": "Torah Judaism",
                "confidence_score": "-2.223769"
            }
        ]
    },
    {
        "mention": "temple",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Temple",
                "confidence_score": "8.245494"
            },
            {
                "entity": "Temple (Latter Day Saints)",
                "confidence_score": "0.060873926"
            },
            {
                "entity": "Temple (LDS Church)",
                "confidence_score": "-1.0192904"
            }
        ]
    },
    {
        "mention": "moshe",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Moses",
                "confidence_score": "-3.526408"
            },
            {
                "entity": "Moshe Schneersohn",
                "confidence_score": "-9.5615635"
            },
            {
                "entity": "Moshe Zvi of Savran",
                "confidence_score": "-10.812758"
            }
        ]
    },
    {
        "mention": "torah",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Torah",
                "confidence_score": "8.946289"
            },
            {
                "entity": "Judaism",
                "confidence_score": "-2.1932926"
            },
            {
                "entity": "Torah Judaism",
                "confidence_score": "-2.223769"
            }
        ]
    },
    {
        "mention": "sinai",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sinai Peninsula",
                "confidence_score": "-1.5670881"
            },
            {
                "entity": "Mount Sinai",
                "confidence_score": "-6.1581283"
            },
            {
                "entity": "Sinaia",
                "confidence_score": "-9.737592"
            }
        ]
    },
    {
        "mention": "sinai",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sinai Peninsula",
                "confidence_score": "-1.5670881"
            },
            {
                "entity": "Mount Sinai",
                "confidence_score": "-6.1581283"
            },
            {
                "entity": "Sinaia",
                "confidence_score": "-9.737592"
            }
        ]
    },
    {
        "mention": "chabad.org",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Chabad.org",
                "confidence_score": "12.246767"
            },
            {
                "entity": "Chabad",
                "confidence_score": "-0.796167"
            },
            {
                "entity": "Agudas Chasidei Chabad",
                "confidence_score": "-8.648515"
            }
        ]
    },
    {
        "mention": "feldman",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Yael Feldman",
                "confidence_score": "-6.658401"
            },
            {
                "entity": "Aharon Feldman",
                "confidence_score": "-8.534206"
            },
            {
                "entity": "Feldman",
                "confidence_score": "-8.929284"
            }
        ]
    },
    {
        "mention": "shabbos",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Shabbat",
                "confidence_score": "9.610666"
            },
            {
                "entity": "Shavuot",
                "confidence_score": "-2.1703236"
            },
            {
                "entity": "Shabbaton",
                "confidence_score": "-8.635697"
            }
        ]
    },
    {
        "mention": "the gold king mine",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Newmont Goldcorp",
                "confidence_score": "-8.816589"
            },
            {
                "entity": "Gold mining",
                "confidence_score": "-8.894177"
            },
            {
                "entity": "Gold Country",
                "confidence_score": "-10.176772"
            }
        ]
    },
    {
        "mention": "david weindorf",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "David Weinberger",
                "confidence_score": "-10.766471"
            },
            {
                "entity": "David Weisberg",
                "confidence_score": "-11.81852"
            },
            {
                "entity": "Dave Weinstein",
                "confidence_score": "-11.882699"
            }
        ]
    },
    {
        "mention": "texas tech university",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Texas Tech University",
                "confidence_score": "11.029931"
            },
            {
                "entity": "Michigan Technological University",
                "confidence_score": "-10.921628"
            },
            {
                "entity": "Tennessee Technological University",
                "confidence_score": "-11.393224"
            }
        ]
    },
    {
        "mention": "kevin lombard",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "John Lombard",
                "confidence_score": "-11.706773"
            },
            {
                "entity": "Jim Lombard",
                "confidence_score": "-11.83933"
            },
            {
                "entity": "Robert Lombard",
                "confidence_score": "-11.860044"
            }
        ]
    },
    {
        "mention": "new mexico state university",
        "type": "ORG",
        "predictions": [
            {
                "entity": "New Mexico State University",
                "confidence_score": "9.9945755"
            },
            {
                "entity": "University of New Mexico",
                "confidence_score": "-4.827562"
            },
            {
                "entity": "New Mexico Highlands University",
                "confidence_score": "-9.477769"
            }
        ]
    },
    {
        "mention": "the animas river",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Animas River (Mexico)",
                "confidence_score": "-1.7553355"
            },
            {
                "entity": "Animas River",
                "confidence_score": "-3.83599"
            },
            {
                "entity": "Animas Creek",
                "confidence_score": "-10.406238"
            }
        ]
    },
    {
        "mention": "weindorf",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Wei\u00dfendorf",
                "confidence_score": "-7.3497753"
            },
            {
                "entity": "United States v. Weitzenhoff",
                "confidence_score": "-11.162848"
            },
            {
                "entity": "James D. Weinrich",
                "confidence_score": "-11.783945"
            }
        ]
    },
    {
        "mention": "weindorf",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Wei\u00dfendorf",
                "confidence_score": "-7.3497753"
            },
            {
                "entity": "United States v. Weitzenhoff",
                "confidence_score": "-11.162848"
            },
            {
                "entity": "James D. Weinrich",
                "confidence_score": "-11.783945"
            }
        ]
    },
    {
        "mention": "animas",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Animas River",
                "confidence_score": "-7.026639"
            },
            {
                "entity": "Animas Mountains",
                "confidence_score": "-7.6782813"
            },
            {
                "entity": "Animas, New Mexico",
                "confidence_score": "-8.150809"
            }
        ]
    },
    {
        "mention": "the environmental protection agency",
        "type": "ORG",
        "predictions": [
            {
                "entity": "United States Environmental Protection Agency",
                "confidence_score": "7.655483"
            },
            {
                "entity": "Environment Agency",
                "confidence_score": "-9.659313"
            },
            {
                "entity": "Ohio Environmental Protection Agency",
                "confidence_score": "-10.570741"
            }
        ]
    },
    {
        "mention": "epa",
        "type": "ORG",
        "predictions": [
            {
                "entity": "United States Environmental Protection Agency",
                "confidence_score": "10.967144"
            },
            {
                "entity": "Drug Enforcement Administration",
                "confidence_score": "-7.5155416"
            },
            {
                "entity": "Ohio Environmental Protection Agency",
                "confidence_score": "-10.629864"
            }
        ]
    },
    {
        "mention": "the navajo nation",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Navajo Nation",
                "confidence_score": "7.488562"
            },
            {
                "entity": "Navajo",
                "confidence_score": "2.0658877"
            },
            {
                "entity": "Cherokee Nation",
                "confidence_score": "-10.051681"
            }
        ]
    },
    {
        "mention": "epa",
        "type": "ORG",
        "predictions": [
            {
                "entity": "United States Environmental Protection Agency",
                "confidence_score": "10.967144"
            },
            {
                "entity": "Drug Enforcement Administration",
                "confidence_score": "-7.5155416"
            },
            {
                "entity": "Ohio Environmental Protection Agency",
                "confidence_score": "-10.629864"
            }
        ]
    },
    {
        "mention": "weindorf and lombard",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Delta and Dawn",
                "confidence_score": "-11.402548"
            },
            {
                "entity": "Warner Norcross & Judd",
                "confidence_score": "-11.675132"
            },
            {
                "entity": "Dale L. Graybill",
                "confidence_score": "-12.29922"
            }
        ]
    },
    {
        "mention": "epa",
        "type": "ORG",
        "predictions": [
            {
                "entity": "United States Environmental Protection Agency",
                "confidence_score": "10.967144"
            },
            {
                "entity": "Drug Enforcement Administration",
                "confidence_score": "-7.5155416"
            },
            {
                "entity": "Ohio Environmental Protection Agency",
                "confidence_score": "-10.629864"
            }
        ]
    },
    {
        "mention": "laura allen",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Laura Allen",
                "confidence_score": "1.1854126"
            },
            {
                "entity": "Laura Dern",
                "confidence_score": "-10.786336"
            },
            {
                "entity": "Laura Moser",
                "confidence_score": "-11.846199"
            }
        ]
    },
    {
        "mention": "epa",
        "type": "ORG",
        "predictions": [
            {
                "entity": "United States Environmental Protection Agency",
                "confidence_score": "10.967144"
            },
            {
                "entity": "Drug Enforcement Administration",
                "confidence_score": "-7.5155416"
            },
            {
                "entity": "Ohio Environmental Protection Agency",
                "confidence_score": "-10.629864"
            }
        ]
    },
    {
        "mention": "the natural resources conservation service",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Natural Resources Conservation Service",
                "confidence_score": "8.929482"
            },
            {
                "entity": "National Park Service",
                "confidence_score": "-5.7618523"
            },
            {
                "entity": "United States Fish and Wildlife Service",
                "confidence_score": "-6.648952"
            }
        ]
    },
    {
        "mention": "the u.s. department of agriculture",
        "type": "ORG",
        "predictions": [
            {
                "entity": "United States Department of Agriculture",
                "confidence_score": "6.2610035"
            },
            {
                "entity": "United States Secretary of Agriculture",
                "confidence_score": "-3.7229621"
            },
            {
                "entity": "Pennsylvania Department of Agriculture",
                "confidence_score": "-10.323705"
            }
        ]
    },
    {
        "mention": "the nmsu agricultural science center",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Texas Tech University College of Agricultural Sciences & Natural Resources",
                "confidence_score": "-11.344987"
            },
            {
                "entity": "Colorado State University College of Agricultural Sciences",
                "confidence_score": "-11.38518"
            },
            {
                "entity": "Agriculture Innovation Center",
                "confidence_score": "-11.590854"
            }
        ]
    },
    {
        "mention": "san juan river",
        "type": "LOC",
        "predictions": [
            {
                "entity": "San Juan River (Colorado River tributary)",
                "confidence_score": "1.7076807"
            },
            {
                "entity": "San Juan Creek",
                "confidence_score": "-1.7842884"
            },
            {
                "entity": "San Juan River (Tamaulipas)",
                "confidence_score": "-7.063937"
            }
        ]
    },
    {
        "mention": "russell begaye",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Russell Begaye",
                "confidence_score": "12.496789"
            },
            {
                "entity": "Kelsey Begaye",
                "confidence_score": "-9.073139"
            },
            {
                "entity": "Enei Begaye",
                "confidence_score": "-10.150754"
            }
        ]
    },
    {
        "mention": "the navajo nation",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Navajo Nation",
                "confidence_score": "7.488562"
            },
            {
                "entity": "Navajo",
                "confidence_score": "2.0658877"
            },
            {
                "entity": "Cherokee Nation",
                "confidence_score": "-10.051681"
            }
        ]
    },
    {
        "mention": "begaye",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Russell Begaye",
                "confidence_score": "4.304874"
            },
            {
                "entity": "Kelsey Begaye",
                "confidence_score": "2.4302273"
            },
            {
                "entity": "Ray Begaye",
                "confidence_score": "-3.351105"
            }
        ]
    },
    {
        "mention": "u.s. house",
        "type": "ORG",
        "predictions": [
            {
                "entity": "United States House of Representatives",
                "confidence_score": "0.6208067"
            },
            {
                "entity": "United States Congress",
                "confidence_score": "-1.6465702"
            },
            {
                "entity": "United States",
                "confidence_score": "-1.8239053"
            }
        ]
    },
    {
        "mention": "epa",
        "type": "ORG",
        "predictions": [
            {
                "entity": "United States Environmental Protection Agency",
                "confidence_score": "10.967144"
            },
            {
                "entity": "Drug Enforcement Administration",
                "confidence_score": "-7.5155416"
            },
            {
                "entity": "Ohio Environmental Protection Agency",
                "confidence_score": "-10.629864"
            }
        ]
    },
    {
        "mention": "mark salley",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jerry Salley",
                "confidence_score": "-11.239314"
            },
            {
                "entity": "Nate Salley",
                "confidence_score": "-11.958098"
            },
            {
                "entity": "Mark Kersey",
                "confidence_score": "-12.288937"
            }
        ]
    },
    {
        "mention": "the new mexico environment department",
        "type": "ORG",
        "predictions": [
            {
                "entity": "New Mexico Environment Department",
                "confidence_score": "6.416036"
            },
            {
                "entity": "Wyoming Department of Environmental Quality",
                "confidence_score": "-11.15256"
            },
            {
                "entity": "Oklahoma Department of Environmental Quality",
                "confidence_score": "-11.786294"
            }
        ]
    },
    {
        "mention": "weindorf and lombard's",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Gable and Lombard",
                "confidence_score": "-10.788187"
            },
            {
                "entity": "Delta and Dawn",
                "confidence_score": "-11.419509"
            },
            {
                "entity": "Sarah and Karen Josephson",
                "confidence_score": "-11.7640295"
            }
        ]
    },
    {
        "mention": "allison majure",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Allison Strong",
                "confidence_score": "-11.114906"
            },
            {
                "entity": "Harlan Majure",
                "confidence_score": "-11.710187"
            },
            {
                "entity": "Allison Vulgamore",
                "confidence_score": "-12.189859"
            }
        ]
    },
    {
        "mention": "follow dan elliott",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "2014 Dubai 24 Hour",
                "confidence_score": "-11.71274"
            },
            {
                "entity": "Paragliding at the 2018 Asian Games \u2013 Women's individual accuracy",
                "confidence_score": "-12.451995"
            },
            {
                "entity": "Trails in Ithaca, New York",
                "confidence_score": "-12.625157"
            }
        ]
    },
    {
        "mention": "santander",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Banco Santander",
                "confidence_score": "3.9781563"
            },
            {
                "entity": "Santander UK",
                "confidence_score": "1.0583375"
            },
            {
                "entity": "Santander, Spain",
                "confidence_score": "-3.8471515"
            }
        ]
    },
    {
        "mention": "tesco bank",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Tesco Bank",
                "confidence_score": "9.60124"
            },
            {
                "entity": "Tesco",
                "confidence_score": "-4.62406"
            },
            {
                "entity": "Citibank",
                "confidence_score": "-9.109246"
            }
        ]
    },
    {
        "mention": "tesco bank",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Tesco Bank",
                "confidence_score": "9.60124"
            },
            {
                "entity": "Tesco",
                "confidence_score": "-4.62406"
            },
            {
                "entity": "Citibank",
                "confidence_score": "-9.109246"
            }
        ]
    },
    {
        "mention": "tesco bank current account",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Current account",
                "confidence_score": "-3.4769037"
            },
            {
                "entity": "Transaction account",
                "confidence_score": "-3.987747"
            },
            {
                "entity": "Tesco Bank",
                "confidence_score": "-4.3625903"
            }
        ]
    },
    {
        "mention": "clubcard",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Tesco Clubcard",
                "confidence_score": "9.392434"
            },
            {
                "entity": "Barclaycard",
                "confidence_score": "-6.956921"
            },
            {
                "entity": "CitizenCard",
                "confidence_score": "-10.43956"
            }
        ]
    },
    {
        "mention": "david mccreadle",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "David McCorkell",
                "confidence_score": "-10.86278"
            },
            {
                "entity": "David McCreery",
                "confidence_score": "-11.195897"
            },
            {
                "entity": "David Abercrombie",
                "confidence_score": "-13.017941"
            }
        ]
    },
    {
        "mention": "tesco bank",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Tesco Bank",
                "confidence_score": "9.60124"
            },
            {
                "entity": "Tesco",
                "confidence_score": "-4.62406"
            },
            {
                "entity": "Citibank",
                "confidence_score": "-9.109246"
            }
        ]
    },
    {
        "mention": "santander",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Banco Santander",
                "confidence_score": "3.9781563"
            },
            {
                "entity": "Santander UK",
                "confidence_score": "1.0583375"
            },
            {
                "entity": "Santander, Spain",
                "confidence_score": "-3.8471515"
            }
        ]
    },
    {
        "mention": "santander's",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Banco Santander",
                "confidence_score": "0.8478794"
            },
            {
                "entity": "Santander UK",
                "confidence_score": "-4.6573253"
            },
            {
                "entity": "Santander, Spain",
                "confidence_score": "-6.394996"
            }
        ]
    },
    {
        "mention": "tesco",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Tesco",
                "confidence_score": "11.188067"
            },
            {
                "entity": "Tesco Bank",
                "confidence_score": "0.14866371"
            },
            {
                "entity": "Tesco Corporation",
                "confidence_score": "-6.141289"
            }
        ]
    },
    {
        "mention": "nationwide flexdirect",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Flexible Support Fund",
                "confidence_score": "-11.492518"
            },
            {
                "entity": "Bus Service Operators Grant",
                "confidence_score": "-11.497483"
            },
            {
                "entity": "National Trip End Model",
                "confidence_score": "-11.712904"
            }
        ]
    },
    {
        "mention": "tsb",
        "type": "ORG",
        "predictions": [
            {
                "entity": "TSB Bank (United Kingdom)",
                "confidence_score": "-2.0135615"
            },
            {
                "entity": "TSB",
                "confidence_score": "-2.1037471"
            },
            {
                "entity": "HSBC",
                "confidence_score": "-6.5393834"
            }
        ]
    },
    {
        "mention": "tsb",
        "type": "ORG",
        "predictions": [
            {
                "entity": "TSB Bank (United Kingdom)",
                "confidence_score": "-2.0135615"
            },
            {
                "entity": "TSB",
                "confidence_score": "-2.1037471"
            },
            {
                "entity": "HSBC",
                "confidence_score": "-6.5393834"
            }
        ]
    },
    {
        "mention": "lloyds bank",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Lloyds Bank",
                "confidence_score": "9.460269"
            },
            {
                "entity": "Lloyds Banking Group",
                "confidence_score": "-0.042259373"
            },
            {
                "entity": "Lloyds Bank International",
                "confidence_score": "-5.9129653"
            }
        ]
    },
    {
        "mention": "bank of scotland's",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Bank of Scotland",
                "confidence_score": "9.490842"
            },
            {
                "entity": "Royal Bank of Scotland Group",
                "confidence_score": "-2.078891"
            },
            {
                "entity": "Royal Bank of Scotland",
                "confidence_score": "-2.7185166"
            }
        ]
    },
    {
        "mention": "vantage",
        "type": "ORG",
        "predictions": [
            {
                "entity": "VantageScore",
                "confidence_score": "-7.7273316"
            },
            {
                "entity": "Valuation (finance)",
                "confidence_score": "-8.2954645"
            },
            {
                "entity": "Vantage (cigarette)",
                "confidence_score": "-10.166728"
            }
        ]
    },
    {
        "mention": "halifax",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Halifax (bank)",
                "confidence_score": "3.4195416"
            },
            {
                "entity": "Halifax, Nova Scotia",
                "confidence_score": "-1.2426672"
            },
            {
                "entity": "Halifax, West Yorkshire",
                "confidence_score": "-3.211214"
            }
        ]
    },
    {
        "mention": "santander's",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Banco Santander",
                "confidence_score": "0.8478794"
            },
            {
                "entity": "Santander UK",
                "confidence_score": "-4.6573253"
            },
            {
                "entity": "Santander, Spain",
                "confidence_score": "-6.394996"
            }
        ]
    },
    {
        "mention": "fda",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Food and Drug Administration",
                "confidence_score": "9.777801"
            },
            {
                "entity": "Office of Global Regulatory Operations and Policy",
                "confidence_score": "-6.0382915"
            },
            {
                "entity": "Food and Drug Administration (Myanmar)",
                "confidence_score": "-6.060632"
            }
        ]
    },
    {
        "mention": "the world health organization",
        "type": "ORG",
        "predictions": [
            {
                "entity": "World Health Organization",
                "confidence_score": "5.650107"
            },
            {
                "entity": "World Health Assembly",
                "confidence_score": "-6.1673913"
            },
            {
                "entity": "Establishment of the World Health Organization",
                "confidence_score": "-8.554202"
            }
        ]
    },
    {
        "mention": "the state of the world's",
        "type": "ORG",
        "predictions": [
            {
                "entity": "State of the World (book series)",
                "confidence_score": "-8.3590765"
            },
            {
                "entity": "MS The World",
                "confidence_score": "-11.259781"
            },
            {
                "entity": "State of the World",
                "confidence_score": "-12.127407"
            }
        ]
    },
    {
        "mention": "antibiotics",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bacteria",
                "confidence_score": "-5.8078856"
            },
            {
                "entity": "List of antibiotics",
                "confidence_score": "-6.2307806"
            },
            {
                "entity": "Pathogenic bacteria",
                "confidence_score": "-9.393783"
            }
        ]
    },
    {
        "mention": "the sunday times",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Sunday Times",
                "confidence_score": "6.922102"
            },
            {
                "entity": "The Sunday Times (India)",
                "confidence_score": "-6.636953"
            },
            {
                "entity": "The Sunday Times (South Africa)",
                "confidence_score": "-10.273158"
            }
        ]
    },
    {
        "mention": "victim support",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Victim Support",
                "confidence_score": "5.111811"
            },
            {
                "entity": "Victimisation",
                "confidence_score": "-9.337071"
            },
            {
                "entity": "Victim blaming",
                "confidence_score": "-9.988235"
            }
        ]
    },
    {
        "mention": "lucy hastings",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Lucy Hastings",
                "confidence_score": "-0.8045176"
            },
            {
                "entity": "Lucy Sussex",
                "confidence_score": "-10.612635"
            },
            {
                "entity": "Lucy Turnbull",
                "confidence_score": "-11.328981"
            }
        ]
    },
    {
        "mention": "victim support",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Victim Support",
                "confidence_score": "5.111811"
            },
            {
                "entity": "Victimisation",
                "confidence_score": "-9.337071"
            },
            {
                "entity": "Victim blaming",
                "confidence_score": "-9.988235"
            }
        ]
    },
    {
        "mention": "bbc",
        "type": "ORG",
        "predictions": [
            {
                "entity": "BBC",
                "confidence_score": "10.874385"
            },
            {
                "entity": "BBC News",
                "confidence_score": "-3.6971645"
            },
            {
                "entity": "BBC World Service",
                "confidence_score": "-3.7842958"
            }
        ]
    },
    {
        "mention": "thornhill trojans",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Thornhill Trojans",
                "confidence_score": "8.277642"
            },
            {
                "entity": "Dewsbury Celtic",
                "confidence_score": "-11.299085"
            },
            {
                "entity": "Moorends-Thorne Marauders A.R.L.F.C.",
                "confidence_score": "-11.779855"
            }
        ]
    },
    {
        "mention": "national conference league division",
        "type": "ORG",
        "predictions": [
            {
                "entity": "National Conference League",
                "confidence_score": "-3.074806"
            },
            {
                "entity": "Rugby League Conference National Division",
                "confidence_score": "-5.2173705"
            },
            {
                "entity": "National Conference League Premier Division",
                "confidence_score": "-5.564135"
            }
        ]
    },
    {
        "mention": "crosfields",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "John Crosland",
                "confidence_score": "-11.448372"
            },
            {
                "entity": "Crosland Moor",
                "confidence_score": "-11.458675"
            },
            {
                "entity": "Crosfield",
                "confidence_score": "-11.560529"
            }
        ]
    },
    {
        "mention": "mindaugas bendikas",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mindaugas Lukauskis",
                "confidence_score": "-11.73491"
            },
            {
                "entity": "Mindaugas Timinskas",
                "confidence_score": "-11.840137"
            },
            {
                "entity": "Mindaugas \u017dukauskas",
                "confidence_score": "-11.908769"
            }
        ]
    },
    {
        "mention": "liam morley",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Chris Morley",
                "confidence_score": "-10.95963"
            },
            {
                "entity": "Liam Kirk",
                "confidence_score": "-11.579979"
            },
            {
                "entity": "Liam Kirk (ice hockey)",
                "confidence_score": "-11.944601"
            }
        ]
    },
    {
        "mention": "thornhill",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Thornhill, Southampton",
                "confidence_score": "-4.7125278"
            },
            {
                "entity": "Gary Thornhill",
                "confidence_score": "-8.809424"
            },
            {
                "entity": "Thornhill Thunderbirds",
                "confidence_score": "-10.715937"
            }
        ]
    },
    {
        "mention": "danny ratcliffe",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Samuel D. Ratcliffe",
                "confidence_score": "-10.872925"
            },
            {
                "entity": "Kevin Ratcliffe",
                "confidence_score": "-11.1955385"
            },
            {
                "entity": "Peter J. Ratcliffe",
                "confidence_score": "-11.335762"
            }
        ]
    },
    {
        "mention": "joel gibson",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Joel Reynolds",
                "confidence_score": "-11.764317"
            },
            {
                "entity": "Vanessa Gibson",
                "confidence_score": "-11.782187"
            },
            {
                "entity": "Joel Watson",
                "confidence_score": "-11.994231"
            }
        ]
    },
    {
        "mention": "crosfield",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "John Crosfield",
                "confidence_score": "-7.11296"
            },
            {
                "entity": "John Crosland",
                "confidence_score": "-11.566937"
            },
            {
                "entity": "Troy Crosland",
                "confidence_score": "-12.076608"
            }
        ]
    },
    {
        "mention": "nathan taylor",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Nathan Taylor (footballer)",
                "confidence_score": "-4.915008"
            },
            {
                "entity": "Benedict Taylor",
                "confidence_score": "-10.887313"
            },
            {
                "entity": "Jax Taylor",
                "confidence_score": "-10.892753"
            }
        ]
    },
    {
        "mention": "taylor",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Taylor Caniff",
                "confidence_score": "-10.620693"
            },
            {
                "entity": "Taylor Mackenzie",
                "confidence_score": "-11.315099"
            },
            {
                "entity": "2012 PSA World Series Finals",
                "confidence_score": "-11.492108"
            }
        ]
    },
    {
        "mention": "jack reid",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jack Reid",
                "confidence_score": "6.0778146"
            },
            {
                "entity": "Jack Reid (footballer)",
                "confidence_score": "-6.035752"
            },
            {
                "entity": "Jack Steele (footballer)",
                "confidence_score": "-11.658294"
            }
        ]
    },
    {
        "mention": "thornhill",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Thornhill, Southampton",
                "confidence_score": "-4.7125278"
            },
            {
                "entity": "Gary Thornhill",
                "confidence_score": "-8.809424"
            },
            {
                "entity": "Thornhill Thunderbirds",
                "confidence_score": "-10.715937"
            }
        ]
    },
    {
        "mention": "ben kendall",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "FC Kendall",
                "confidence_score": "-11.33881"
            },
            {
                "entity": "Ryan Kendall",
                "confidence_score": "-11.667196"
            },
            {
                "entity": "Ben Eager",
                "confidence_score": "-11.945236"
            }
        ]
    },
    {
        "mention": "luke haigh",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Luke Haines",
                "confidence_score": "-10.785933"
            },
            {
                "entity": "Natalie Haigh",
                "confidence_score": "-11.871213"
            },
            {
                "entity": "Luke Hall (swimmer)",
                "confidence_score": "-12.305131"
            }
        ]
    },
    {
        "mention": "gibson",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "William Gibson",
                "confidence_score": "-5.961008"
            },
            {
                "entity": "Gibson (dog)",
                "confidence_score": "-10.260248"
            },
            {
                "entity": "Eilidh Gibson",
                "confidence_score": "-10.976382"
            }
        ]
    },
    {
        "mention": "reid",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Reid",
                "confidence_score": "-2.7441976"
            },
            {
                "entity": "D. C. Reid",
                "confidence_score": "-6.263761"
            },
            {
                "entity": "John Reid, Baron Reid of Cardowan",
                "confidence_score": "-6.3796444"
            }
        ]
    },
    {
        "mention": "morley",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Morley, West Yorkshire",
                "confidence_score": "-1.9889169"
            },
            {
                "entity": "Morley, Derbyshire",
                "confidence_score": "-5.6363997"
            },
            {
                "entity": "Morley (name)",
                "confidence_score": "-8.003414"
            }
        ]
    },
    {
        "mention": "gibson",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "William Gibson",
                "confidence_score": "-5.961008"
            },
            {
                "entity": "Gibson (dog)",
                "confidence_score": "-10.260248"
            },
            {
                "entity": "Eilidh Gibson",
                "confidence_score": "-10.976382"
            }
        ]
    },
    {
        "mention": "crosfields",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "John Crosland",
                "confidence_score": "-11.448372"
            },
            {
                "entity": "Crosland Moor",
                "confidence_score": "-11.458675"
            },
            {
                "entity": "Crosfield",
                "confidence_score": "-11.560529"
            }
        ]
    },
    {
        "mention": "reid",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Reid",
                "confidence_score": "-2.7441976"
            },
            {
                "entity": "D. C. Reid",
                "confidence_score": "-6.263761"
            },
            {
                "entity": "John Reid, Baron Reid of Cardowan",
                "confidence_score": "-6.3796444"
            }
        ]
    },
    {
        "mention": "gibson",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "William Gibson",
                "confidence_score": "-5.961008"
            },
            {
                "entity": "Gibson (dog)",
                "confidence_score": "-10.260248"
            },
            {
                "entity": "Eilidh Gibson",
                "confidence_score": "-10.976382"
            }
        ]
    },
    {
        "mention": "ian bowden",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Andrew Bowden",
                "confidence_score": "-10.010528"
            },
            {
                "entity": "Ian Bowyer",
                "confidence_score": "-10.988681"
            },
            {
                "entity": "Ian Bowditch",
                "confidence_score": "-11.146291"
            }
        ]
    },
    {
        "mention": "thornhill",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Thornhill, Southampton",
                "confidence_score": "-4.7125278"
            },
            {
                "entity": "Gary Thornhill",
                "confidence_score": "-8.809424"
            },
            {
                "entity": "Thornhill Thunderbirds",
                "confidence_score": "-10.715937"
            }
        ]
    },
    {
        "mention": "morley",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Morley, West Yorkshire",
                "confidence_score": "-1.9889169"
            },
            {
                "entity": "Morley, Derbyshire",
                "confidence_score": "-5.6363997"
            },
            {
                "entity": "Morley (name)",
                "confidence_score": "-8.003414"
            }
        ]
    },
    {
        "mention": "gibson",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "William Gibson",
                "confidence_score": "-5.961008"
            },
            {
                "entity": "Gibson (dog)",
                "confidence_score": "-10.260248"
            },
            {
                "entity": "Eilidh Gibson",
                "confidence_score": "-10.976382"
            }
        ]
    },
    {
        "mention": "thornhill",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Thornhill, Southampton",
                "confidence_score": "-4.7125278"
            },
            {
                "entity": "Gary Thornhill",
                "confidence_score": "-8.809424"
            },
            {
                "entity": "Thornhill Thunderbirds",
                "confidence_score": "-10.715937"
            }
        ]
    },
    {
        "mention": "thornhill",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Thornhill, Southampton",
                "confidence_score": "-4.7125278"
            },
            {
                "entity": "Gary Thornhill",
                "confidence_score": "-8.809424"
            },
            {
                "entity": "Thornhill Thunderbirds",
                "confidence_score": "-10.715937"
            }
        ]
    },
    {
        "mention": "ben kendall",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "FC Kendall",
                "confidence_score": "-11.33881"
            },
            {
                "entity": "Ryan Kendall",
                "confidence_score": "-11.667196"
            },
            {
                "entity": "Ben Eager",
                "confidence_score": "-11.945236"
            }
        ]
    },
    {
        "mention": "crosfields",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "John Crosland",
                "confidence_score": "-11.448372"
            },
            {
                "entity": "Crosland Moor",
                "confidence_score": "-11.458675"
            },
            {
                "entity": "Crosfield",
                "confidence_score": "-11.560529"
            }
        ]
    },
    {
        "mention": "joe moore",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Joe Moore (musician)",
                "confidence_score": "-6.87881"
            },
            {
                "entity": "Joe Moore (speed skater)",
                "confidence_score": "-8.166358"
            },
            {
                "entity": "Joe Moore (politician)",
                "confidence_score": "-9.625218"
            }
        ]
    },
    {
        "mention": "pleasanton - hall of fame",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Pleasanton, California",
                "confidence_score": "-2.8786216"
            },
            {
                "entity": "Pleasanton, Texas",
                "confidence_score": "-6.751242"
            },
            {
                "entity": "Pleasanton Fairgrounds Racetrack",
                "confidence_score": "-7.9324913"
            }
        ]
    },
    {
        "mention": "john madden",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "John Madden",
                "confidence_score": "11.682927"
            },
            {
                "entity": "John Madden (director)",
                "confidence_score": "-3.0436275"
            },
            {
                "entity": "John Madden (ice hockey)",
                "confidence_score": "-8.670011"
            }
        ]
    },
    {
        "mention": "rose hotel",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Rose Hotel",
                "confidence_score": "1.7670436"
            },
            {
                "entity": "Rose Rayhaan by Rotana",
                "confidence_score": "-8.434524"
            },
            {
                "entity": "Mount Rose Station",
                "confidence_score": "-10.874116"
            }
        ]
    },
    {
        "mention": "nfl",
        "type": "ORG",
        "predictions": [
            {
                "entity": "National Football League",
                "confidence_score": "8.221288"
            },
            {
                "entity": "Madden NFL",
                "confidence_score": "6.3371577"
            },
            {
                "entity": "NFL regular season",
                "confidence_score": "-5.845305"
            }
        ]
    },
    {
        "mention": "emmy award",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Emmy Award",
                "confidence_score": "10.869496"
            },
            {
                "entity": "Primetime Emmy Award",
                "confidence_score": "1.5249999"
            },
            {
                "entity": "Sports Emmy Award",
                "confidence_score": "-6.066679"
            }
        ]
    },
    {
        "mention": "nfl",
        "type": "ORG",
        "predictions": [
            {
                "entity": "National Football League",
                "confidence_score": "8.221288"
            },
            {
                "entity": "Madden NFL",
                "confidence_score": "6.3371577"
            },
            {
                "entity": "NFL regular season",
                "confidence_score": "-5.845305"
            }
        ]
    },
    {
        "mention": "jim tomsula",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jim Tomsula",
                "confidence_score": "11.082054"
            },
            {
                "entity": "Michael Ola",
                "confidence_score": "-11.499793"
            },
            {
                "entity": "Jim Haluska",
                "confidence_score": "-11.688012"
            }
        ]
    },
    {
        "mention": "jim harbaugh",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jim Harbaugh",
                "confidence_score": "10.54531"
            },
            {
                "entity": "John Harbaugh",
                "confidence_score": "-6.3046165"
            },
            {
                "entity": "Jay Harbaugh",
                "confidence_score": "-10.029214"
            }
        ]
    },
    {
        "mention": "jim harbaugh",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jim Harbaugh",
                "confidence_score": "10.54531"
            },
            {
                "entity": "John Harbaugh",
                "confidence_score": "-6.3046165"
            },
            {
                "entity": "Jay Harbaugh",
                "confidence_score": "-10.029214"
            }
        ]
    },
    {
        "mention": "kaepernick",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Colin Kaepernick",
                "confidence_score": "5.9220734"
            },
            {
                "entity": "1983 Kentucky Wildcats football team",
                "confidence_score": "-10.899898"
            },
            {
                "entity": "2011 Kentucky Wildcats football team",
                "confidence_score": "-10.9715185"
            }
        ]
    },
    {
        "mention": "the university of nevada",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Nevada Wolf Pack football",
                "confidence_score": "-2.3625658"
            },
            {
                "entity": "2003 Nevada Wolf Pack football team",
                "confidence_score": "-9.184731"
            },
            {
                "entity": "2002 Nevada Wolf Pack football team",
                "confidence_score": "-9.443371"
            }
        ]
    },
    {
        "mention": "patriots",
        "type": "ORG",
        "predictions": [
            {
                "entity": "New England Patriots",
                "confidence_score": "-2.5284922"
            },
            {
                "entity": "1985 New England Patriots season",
                "confidence_score": "-7.544194"
            },
            {
                "entity": "1984 New England Patriots season",
                "confidence_score": "-8.4325695"
            }
        ]
    },
    {
        "mention": "navorro bowman",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "NaVorro Bowman",
                "confidence_score": "8.659656"
            },
            {
                "entity": "Braedon Bowman",
                "confidence_score": "-11.646859"
            },
            {
                "entity": "Christopher N. Bowman",
                "confidence_score": "-11.912799"
            }
        ]
    },
    {
        "mention": "bowman",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bowman Yachts",
                "confidence_score": "-8.199382"
            },
            {
                "entity": "Bowman Automotive",
                "confidence_score": "-8.954215"
            },
            {
                "entity": "Ryan Bowman",
                "confidence_score": "-10.991105"
            }
        ]
    },
    {
        "mention": "smiths",
        "type": "LOC",
        "predictions": [
            {
                "entity": "The Smiths",
                "confidence_score": "-3.8831403"
            },
            {
                "entity": "Smith (surname)",
                "confidence_score": "-5.7986784"
            },
            {
                "entity": "The Smith Brothers",
                "confidence_score": "-9.835806"
            }
        ]
    },
    {
        "mention": "aldon",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Aldon Inc.",
                "confidence_score": "0.4691499"
            },
            {
                "entity": "Aldon Smith",
                "confidence_score": "-8.476578"
            },
            {
                "entity": "Aldor",
                "confidence_score": "-9.614188"
            }
        ]
    },
    {
        "mention": "justin",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Justin Smith (American actor)",
                "confidence_score": "-2.9675663"
            },
            {
                "entity": "Justin Cornwell",
                "confidence_score": "-9.807599"
            },
            {
                "entity": "Justin Smith (linebacker)",
                "confidence_score": "-10.178472"
            }
        ]
    },
    {
        "mention": "justin smith",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Justin Smith (American actor)",
                "confidence_score": "1.785391"
            },
            {
                "entity": "Justin Smith (presenter)",
                "confidence_score": "-0.14893068"
            },
            {
                "entity": "Justin Smith (poker player)",
                "confidence_score": "-5.62257"
            }
        ]
    },
    {
        "mention": "bowman",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bowman Yachts",
                "confidence_score": "-8.199382"
            },
            {
                "entity": "Bowman Automotive",
                "confidence_score": "-8.954215"
            },
            {
                "entity": "Ryan Bowman",
                "confidence_score": "-10.991105"
            }
        ]
    },
    {
        "mention": "patrick",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Patrick Kane",
                "confidence_score": "-9.736839"
            },
            {
                "entity": "Patrick Christopher",
                "confidence_score": "-10.558743"
            },
            {
                "entity": "Patrick Cannon",
                "confidence_score": "-10.991185"
            }
        ]
    },
    {
        "mention": "bowman",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bowman Yachts",
                "confidence_score": "-8.199382"
            },
            {
                "entity": "Bowman Automotive",
                "confidence_score": "-8.954215"
            },
            {
                "entity": "Ryan Bowman",
                "confidence_score": "-10.991105"
            }
        ]
    },
    {
        "mention": "bowman",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bowman Yachts",
                "confidence_score": "-8.199382"
            },
            {
                "entity": "Bowman Automotive",
                "confidence_score": "-8.954215"
            },
            {
                "entity": "Ryan Bowman",
                "confidence_score": "-10.991105"
            }
        ]
    },
    {
        "mention": "matt millen",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Matt Millen",
                "confidence_score": "12.192708"
            },
            {
                "entity": "Greg Millen",
                "confidence_score": "-11.202518"
            },
            {
                "entity": "Corey Millen",
                "confidence_score": "-11.38625"
            }
        ]
    },
    {
        "mention": "justin smith",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Justin Smith (American actor)",
                "confidence_score": "1.785391"
            },
            {
                "entity": "Justin Smith (presenter)",
                "confidence_score": "-0.14893068"
            },
            {
                "entity": "Justin Smith (poker player)",
                "confidence_score": "-5.62257"
            }
        ]
    },
    {
        "mention": "justin smith",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Justin Smith (American actor)",
                "confidence_score": "1.785391"
            },
            {
                "entity": "Justin Smith (presenter)",
                "confidence_score": "-0.14893068"
            },
            {
                "entity": "Justin Smith (poker player)",
                "confidence_score": "-5.62257"
            }
        ]
    },
    {
        "mention": "nfl",
        "type": "ORG",
        "predictions": [
            {
                "entity": "National Football League",
                "confidence_score": "8.221288"
            },
            {
                "entity": "Madden NFL",
                "confidence_score": "6.3371577"
            },
            {
                "entity": "NFL regular season",
                "confidence_score": "-5.845305"
            }
        ]
    },
    {
        "mention": "frank gore",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Frank Gore",
                "confidence_score": "9.271398"
            },
            {
                "entity": "Frank Bush",
                "confidence_score": "-11.868268"
            },
            {
                "entity": "Frank Hare",
                "confidence_score": "-12.068028"
            }
        ]
    },
    {
        "mention": "jarryd hayne",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jarryd Hayne",
                "confidence_score": "-0.44328946"
            },
            {
                "entity": "Shyne",
                "confidence_score": "-11.593733"
            },
            {
                "entity": "Rupert Hine",
                "confidence_score": "-11.79383"
            }
        ]
    },
    {
        "mention": "reggie bush",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Reggie Bush",
                "confidence_score": "5.226647"
            },
            {
                "entity": "Reggie Jackson",
                "confidence_score": "-11.233729"
            },
            {
                "entity": "Reggie Branch",
                "confidence_score": "-12.04716"
            }
        ]
    },
    {
        "mention": "bruce ellington",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bruce Ellington",
                "confidence_score": "4.821603"
            },
            {
                "entity": "Wayne Ellington",
                "confidence_score": "-11.261051"
            },
            {
                "entity": "Bruce Branch",
                "confidence_score": "-11.822601"
            }
        ]
    },
    {
        "mention": "jimmie ward",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jimmie Ward",
                "confidence_score": "3.2252228"
            },
            {
                "entity": "Pete Ward",
                "confidence_score": "-10.838374"
            },
            {
                "entity": "Walter Ward (singer)",
                "confidence_score": "-11.189539"
            }
        ]
    },
    {
        "mention": "mike davis",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mike Davis (wrestler)",
                "confidence_score": "-3.1626894"
            },
            {
                "entity": "Mike Davis (baseball)",
                "confidence_score": "-4.2617025"
            },
            {
                "entity": "Mike Davis (basketball, born 1946)",
                "confidence_score": "-6.8527026"
            }
        ]
    },
    {
        "mention": "reggie bush",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Reggie Bush",
                "confidence_score": "5.226647"
            },
            {
                "entity": "Reggie Jackson",
                "confidence_score": "-11.233729"
            },
            {
                "entity": "Reggie Branch",
                "confidence_score": "-12.04716"
            }
        ]
    },
    {
        "mention": "reggie",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Reggie King",
                "confidence_score": "-8.994649"
            },
            {
                "entity": "Reggie Freeman",
                "confidence_score": "-9.790477"
            },
            {
                "entity": "Reggie Patterson",
                "confidence_score": "-11.228472"
            }
        ]
    },
    {
        "mention": "reggie",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Reggie King",
                "confidence_score": "-8.994649"
            },
            {
                "entity": "Reggie Freeman",
                "confidence_score": "-9.790477"
            },
            {
                "entity": "Reggie Patterson",
                "confidence_score": "-11.228472"
            }
        ]
    },
    {
        "mention": "bush",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bush family",
                "confidence_score": "-4.3312545"
            },
            {
                "entity": "George H. W. Bush",
                "confidence_score": "-4.7943854"
            },
            {
                "entity": "George W. Bush",
                "confidence_score": "-4.846785"
            }
        ]
    },
    {
        "mention": "eddie debartolo",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Edward J. DeBartolo Jr.",
                "confidence_score": "3.492275"
            },
            {
                "entity": "Edward J. DeBartolo Sr.",
                "confidence_score": "-10.626123"
            },
            {
                "entity": "Gavin DeBartolo",
                "confidence_score": "-11.84925"
            }
        ]
    },
    {
        "mention": "the pro football hall of fame",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Pro Football Hall of Fame",
                "confidence_score": "8.166419"
            },
            {
                "entity": "List of Pro Football Hall of Fame inductees",
                "confidence_score": "-2.0063806"
            },
            {
                "entity": "Pittsburgh Pro Football Hall of Fame",
                "confidence_score": "-7.7907467"
            }
        ]
    },
    {
        "mention": "eddie",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Eddie George",
                "confidence_score": "-7.871673"
            },
            {
                "entity": "Eddie Lewis (American football)",
                "confidence_score": "-8.300712"
            },
            {
                "entity": "Eddie Miller (American football)",
                "confidence_score": "-10.711268"
            }
        ]
    },
    {
        "mention": "eddie",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Eddie George",
                "confidence_score": "-7.871673"
            },
            {
                "entity": "Eddie Lewis (American football)",
                "confidence_score": "-8.300712"
            },
            {
                "entity": "Eddie Miller (American football)",
                "confidence_score": "-10.711268"
            }
        ]
    },
    {
        "mention": "paul tagliabue",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Paul Tagliabue",
                "confidence_score": "11.653808"
            },
            {
                "entity": "Paul Guilfoyle",
                "confidence_score": "-10.992255"
            },
            {
                "entity": "Paul Clemente",
                "confidence_score": "-11.905178"
            }
        ]
    },
    {
        "mention": "follow cam",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Cam (singer)",
                "confidence_score": "-9.775342"
            },
            {
                "entity": "CaliCam",
                "confidence_score": "-9.828499"
            },
            {
                "entity": "Cam Kirk",
                "confidence_score": "-10.838531"
            }
        ]
    },
    {
        "mention": "twitter.com/caminman",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Twitter",
                "confidence_score": "0.78926146"
            },
            {
                "entity": "Jukin Media",
                "confidence_score": "-11.758886"
            },
            {
                "entity": "Techinline",
                "confidence_score": "-12.201656"
            }
        ]
    },
    {
        "mention": "rob quinn",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Rob Quinn",
                "confidence_score": "-1.8450193"
            },
            {
                "entity": "Zo\u00eb Quinn",
                "confidence_score": "-11.808153"
            },
            {
                "entity": "Rob Hopkins",
                "confidence_score": "-11.951189"
            }
        ]
    },
    {
        "mention": "newser staff",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Newser",
                "confidence_score": "-2.3410285"
            },
            {
                "entity": "News America Marketing",
                "confidence_score": "-9.516123"
            },
            {
                "entity": "NewsLeecher",
                "confidence_score": "-11.288979"
            }
        ]
    },
    {
        "mention": "mcdonald",
        "type": "ORG",
        "predictions": [
            {
                "entity": "McDonald's",
                "confidence_score": "0.22948773"
            },
            {
                "entity": "McDonald's Canada",
                "confidence_score": "-7.0245466"
            },
            {
                "entity": "McDonald & Co.",
                "confidence_score": "-8.415683"
            }
        ]
    },
    {
        "mention": "mcdonald's",
        "type": "ORG",
        "predictions": [
            {
                "entity": "McDonald's",
                "confidence_score": "8.3873825"
            },
            {
                "entity": "McDonald's Canada",
                "confidence_score": "-6.752019"
            },
            {
                "entity": "McDonald & Co.",
                "confidence_score": "-10.09703"
            }
        ]
    },
    {
        "mention": "bloomberg",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bloomberg Television",
                "confidence_score": "1.7210228"
            },
            {
                "entity": "Bloomberg L.P.",
                "confidence_score": "-3.3349562"
            },
            {
                "entity": "Bloomberg",
                "confidence_score": "-11.71278"
            }
        ]
    },
    {
        "mention": "panda express'",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Panda Express",
                "confidence_score": "6.147899"
            },
            {
                "entity": "Panda Inn",
                "confidence_score": "-11.014731"
            },
            {
                "entity": "Cafe Express",
                "confidence_score": "-11.57026"
            }
        ]
    },
    {
        "mention": "mcdonald",
        "type": "ORG",
        "predictions": [
            {
                "entity": "McDonald's",
                "confidence_score": "0.22948773"
            },
            {
                "entity": "McDonald's Canada",
                "confidence_score": "-7.0245466"
            },
            {
                "entity": "McDonald & Co.",
                "confidence_score": "-8.415683"
            }
        ]
    },
    {
        "mention": "bloomberg",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bloomberg Television",
                "confidence_score": "1.7210228"
            },
            {
                "entity": "Bloomberg L.P.",
                "confidence_score": "-3.3349562"
            },
            {
                "entity": "Bloomberg",
                "confidence_score": "-11.71278"
            }
        ]
    },
    {
        "mention": "europe",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Europe",
                "confidence_score": "3.7369654"
            },
            {
                "entity": "Time in Europe",
                "confidence_score": "-8.074035"
            },
            {
                "entity": "Europe (magazine)",
                "confidence_score": "-10.902366"
            }
        ]
    },
    {
        "mention": "newser.com",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Newser",
                "confidence_score": "1.2090702"
            },
            {
                "entity": "NewsRx",
                "confidence_score": "-10.26019"
            },
            {
                "entity": "NewsON",
                "confidence_score": "-11.118991"
            }
        ]
    },
    {
        "mention": "wildcats",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Northwestern Wildcats football",
                "confidence_score": "-7.7032313"
            },
            {
                "entity": "Kentucky Wildcats football",
                "confidence_score": "-8.67985"
            },
            {
                "entity": "1998 Kentucky Wildcats football team",
                "confidence_score": "-10.795735"
            }
        ]
    },
    {
        "mention": "axtell",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Axtell, Kansas",
                "confidence_score": "-6.10211"
            },
            {
                "entity": "Axtell, Nebraska",
                "confidence_score": "-9.215293"
            },
            {
                "entity": "Axtell, Utah",
                "confidence_score": "-10.152725"
            }
        ]
    },
    {
        "mention": "marysville",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Marysville, Kansas",
                "confidence_score": "-3.769655"
            },
            {
                "entity": "Marysville, Iowa",
                "confidence_score": "-9.3106785"
            },
            {
                "entity": "St. Marys, Kansas",
                "confidence_score": "-10.493655"
            }
        ]
    },
    {
        "mention": "north central kansas league",
        "type": "ORG",
        "predictions": [
            {
                "entity": "North Central Kansas",
                "confidence_score": "-7.4253964"
            },
            {
                "entity": "Central Kansas League",
                "confidence_score": "-9.090131"
            },
            {
                "entity": "North Central Kansas Technical College",
                "confidence_score": "-11.128783"
            }
        ]
    },
    {
        "mention": "marysville",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Marysville, Kansas",
                "confidence_score": "-3.769655"
            },
            {
                "entity": "Marysville, Iowa",
                "confidence_score": "-9.3106785"
            },
            {
                "entity": "St. Marys, Kansas",
                "confidence_score": "-10.493655"
            }
        ]
    },
    {
        "mention": "wamego",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Wamego, Kansas",
                "confidence_score": "-2.3050907"
            },
            {
                "entity": "Wamego USD 320",
                "confidence_score": "-7.2855964"
            },
            {
                "entity": "WAME",
                "confidence_score": "-11.534213"
            }
        ]
    },
    {
        "mention": "tvl",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Pacific View League",
                "confidence_score": "-11.492096"
            },
            {
                "entity": "Fox Lake (Timberlea)",
                "confidence_score": "-11.735805"
            },
            {
                "entity": "Dual County League",
                "confidence_score": "-11.781105"
            }
        ]
    },
    {
        "mention": "valley heights",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Valley Heights, New South Wales",
                "confidence_score": "-9.663199"
            },
            {
                "entity": "Valley Heights, West Virginia",
                "confidence_score": "-11.465603"
            },
            {
                "entity": "Vadnais Heights, Minnesota",
                "confidence_score": "-11.560941"
            }
        ]
    },
    {
        "mention": "onaga",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Onaga, Kansas",
                "confidence_score": "-8.279366"
            },
            {
                "entity": "Onaga Shale",
                "confidence_score": "-11.61366"
            },
            {
                "entity": "Onaga",
                "confidence_score": "-11.9579315"
            }
        ]
    },
    {
        "mention": "washington co.",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Washington County, Oregon",
                "confidence_score": "-5.774222"
            },
            {
                "entity": "Washington County, New York",
                "confidence_score": "-8.597835"
            },
            {
                "entity": "Washington County, Wisconsin",
                "confidence_score": "-8.993506"
            }
        ]
    },
    {
        "mention": "doniphan west 7",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Doniphan County, Kansas",
                "confidence_score": "-9.576579"
            },
            {
                "entity": "Doniphan, Nebraska",
                "confidence_score": "-10.963711"
            },
            {
                "entity": "Doniphan Township, Hall County, Nebraska",
                "confidence_score": "-11.3439"
            }
        ]
    },
    {
        "mention": "tvl",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Pacific View League",
                "confidence_score": "-11.492096"
            },
            {
                "entity": "Fox Lake (Timberlea)",
                "confidence_score": "-11.735805"
            },
            {
                "entity": "Dual County League",
                "confidence_score": "-11.781105"
            }
        ]
    },
    {
        "mention": "axtell",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Axtell, Kansas",
                "confidence_score": "-6.10211"
            },
            {
                "entity": "Axtell, Nebraska",
                "confidence_score": "-9.215293"
            },
            {
                "entity": "Axtell, Utah",
                "confidence_score": "-10.152725"
            }
        ]
    },
    {
        "mention": "tescott",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Tescott, Kansas",
                "confidence_score": "-3.2031944"
            },
            {
                "entity": "Teutopolis Township, Effingham County, Illinois",
                "confidence_score": "-11.717622"
            },
            {
                "entity": "Tebbetts, Missouri",
                "confidence_score": "-12.233113"
            }
        ]
    },
    {
        "mention": "blue valley randolph",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Bluefield, Virginia",
                "confidence_score": "-11.162382"
            },
            {
                "entity": "Bluefield Blue Jays",
                "confidence_score": "-11.388344"
            },
            {
                "entity": "Blueville, West Virginia",
                "confidence_score": "-12.358437"
            }
        ]
    },
    {
        "mention": "linn",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Linn County, Missouri",
                "confidence_score": "-4.0896344"
            },
            {
                "entity": "Linn County, Kansas",
                "confidence_score": "-6.7558284"
            },
            {
                "entity": "Linn County, Iowa",
                "confidence_score": "-6.793494"
            }
        ]
    },
    {
        "mention": "hodgeman co.",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Hodgeman County, Kansas",
                "confidence_score": "-6.6432586"
            },
            {
                "entity": "Hodgenville, Kentucky",
                "confidence_score": "-11.476821"
            },
            {
                "entity": "Hodgeman County Courthouse",
                "confidence_score": "-11.909463"
            }
        ]
    },
    {
        "mention": "lincoln",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Lincoln, Illinois",
                "confidence_score": "-2.6626446"
            },
            {
                "entity": "Lincoln County, Kansas",
                "confidence_score": "-3.1052325"
            },
            {
                "entity": "Lincoln County, Wisconsin",
                "confidence_score": "-5.8368955"
            }
        ]
    },
    {
        "mention": "clifton-clyde",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Clifton, Michigan",
                "confidence_score": "-9.982541"
            },
            {
                "entity": "Clifton, Illinois",
                "confidence_score": "-10.519332"
            },
            {
                "entity": "Clifton, Kansas",
                "confidence_score": "-10.658595"
            }
        ]
    },
    {
        "mention": "beloit",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Beloit, Wisconsin",
                "confidence_score": "-0.0074373754"
            },
            {
                "entity": "Beloit, Kansas",
                "confidence_score": "-2.5297523"
            },
            {
                "entity": "Beloit, Ohio",
                "confidence_score": "-6.949477"
            }
        ]
    },
    {
        "mention": "sacred heart",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Sacred Heart Church (Eau Claire, Wisconsin)",
                "confidence_score": "-10.85104"
            },
            {
                "entity": "Sacred Heart, Minnesota",
                "confidence_score": "-11.17081"
            },
            {
                "entity": "Sacred Heart Church (Chicago)",
                "confidence_score": "-11.214116"
            }
        ]
    },
    {
        "mention": "herington",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Herington, Kansas",
                "confidence_score": "0.05050886"
            },
            {
                "entity": "Herington USD 487",
                "confidence_score": "-5.9283853"
            },
            {
                "entity": "Armington, Illinois",
                "confidence_score": "-11.652167"
            }
        ]
    },
    {
        "mention": "nemaha central 72",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Nemaha County, Kansas",
                "confidence_score": "-9.036435"
            },
            {
                "entity": "Nemaha Central USD 115",
                "confidence_score": "-10.509478"
            },
            {
                "entity": "Nemaha Township, Gage County, Nebraska",
                "confidence_score": "-11.417753"
            }
        ]
    },
    {
        "mention": "holton",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Holton, Kansas",
                "confidence_score": "-7.705096"
            },
            {
                "entity": "Holton, Wisconsin",
                "confidence_score": "-8.294572"
            },
            {
                "entity": "Holton, Indiana",
                "confidence_score": "-10.853353"
            }
        ]
    },
    {
        "mention": "lacrosse",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Lacrosse",
                "confidence_score": "6.935642"
            },
            {
                "entity": "Buick LaCrosse",
                "confidence_score": "-7.970001"
            },
            {
                "entity": "LaCrosse, Arkansas",
                "confidence_score": "-11.743636"
            }
        ]
    },
    {
        "mention": "royal valley 40",
        "type": "LOC",
        "predictions": [
            {
                "entity": "U.S. Route 40 in Colorado",
                "confidence_score": "-9.92212"
            },
            {
                "entity": "Kings County, California",
                "confidence_score": "-10.464883"
            },
            {
                "entity": "North Dakota Highway 40",
                "confidence_score": "-11.728337"
            }
        ]
    },
    {
        "mention": "santa fe trail",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Santa Fe Trail",
                "confidence_score": "4.840313"
            },
            {
                "entity": "Santa Fe Trails",
                "confidence_score": "1.6811886"
            },
            {
                "entity": "Jane Addams Trail",
                "confidence_score": "-11.789247"
            }
        ]
    },
    {
        "mention": "solomon",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Solomon, Kansas",
                "confidence_score": "-8.1391115"
            },
            {
                "entity": "Solomon, Illinois",
                "confidence_score": "-10.944631"
            },
            {
                "entity": "Solomon Lake",
                "confidence_score": "-11.3166895"
            }
        ]
    },
    {
        "mention": "rural vista",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Alta Vista Township, Lincoln County, Minnesota",
                "confidence_score": "-11.813685"
            },
            {
                "entity": "Brook Park Township, Pine County, Minnesota",
                "confidence_score": "-11.891596"
            },
            {
                "entity": "Forest Area Township, Lake of the Woods County, Minnesota",
                "confidence_score": "-12.075404"
            }
        ]
    },
    {
        "mention": "conned",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Connexions Card",
                "confidence_score": "-10.615592"
            },
            {
                "entity": "Cynwyl Elfed",
                "confidence_score": "-10.861091"
            },
            {
                "entity": "ConnNet",
                "confidence_score": "-11.118333"
            }
        ]
    },
    {
        "mention": "kidderminster",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Kidderminster",
                "confidence_score": "7.823537"
            },
            {
                "entity": "Kidderminster railway station",
                "confidence_score": "-7.109732"
            },
            {
                "entity": "Kidderminster Cricket Club",
                "confidence_score": "-7.822868"
            }
        ]
    },
    {
        "mention": "moneygram",
        "type": "ORG",
        "predictions": [
            {
                "entity": "MoneyGram",
                "confidence_score": "4.0431647"
            },
            {
                "entity": "Money Dashboard",
                "confidence_score": "-10.875394"
            },
            {
                "entity": "MoneyPass",
                "confidence_score": "-11.06883"
            }
        ]
    },
    {
        "mention": "moneygram",
        "type": "ORG",
        "predictions": [
            {
                "entity": "MoneyGram",
                "confidence_score": "4.0431647"
            },
            {
                "entity": "Money Dashboard",
                "confidence_score": "-10.875394"
            },
            {
                "entity": "MoneyPass",
                "confidence_score": "-11.06883"
            }
        ]
    },
    {
        "mention": "west mercia",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mercia",
                "confidence_score": "-3.9671574"
            },
            {
                "entity": "West Midlands (region)",
                "confidence_score": "-6.330643"
            },
            {
                "entity": "West Bromwich",
                "confidence_score": "-8.061247"
            }
        ]
    },
    {
        "mention": "moneygram",
        "type": "ORG",
        "predictions": [
            {
                "entity": "MoneyGram",
                "confidence_score": "4.0431647"
            },
            {
                "entity": "Money Dashboard",
                "confidence_score": "-10.875394"
            },
            {
                "entity": "MoneyPass",
                "confidence_score": "-11.06883"
            }
        ]
    },
    {
        "mention": "pok\u00e9mon rumble u 3d publisher",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Pok\u00e9mon Rumble U",
                "confidence_score": "3.670584"
            },
            {
                "entity": "Pok\u00e9mon Rumble World",
                "confidence_score": "-7.3350186"
            },
            {
                "entity": "The 3DO Company",
                "confidence_score": "-9.004238"
            }
        ]
    },
    {
        "mention": "arena battle",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Multiplayer online battle arena",
                "confidence_score": "-6.7238603"
            },
            {
                "entity": "Monster Retsuden Oreca Battle",
                "confidence_score": "-11.108017"
            },
            {
                "entity": "Arena Wars",
                "confidence_score": "-11.513967"
            }
        ]
    },
    {
        "mention": "launch trailer",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Teaser campaign",
                "confidence_score": "-2.0563421"
            },
            {
                "entity": "Trailer (promotion)",
                "confidence_score": "-6.359197"
            },
            {
                "entity": "Pitch trailer",
                "confidence_score": "-10.005776"
            }
        ]
    },
    {
        "mention": "pok\u00e9mon rumble u",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Pok\u00e9mon Rumble U",
                "confidence_score": "9.702672"
            },
            {
                "entity": "Pok\u00e9mon Rumble",
                "confidence_score": "-5.6219697"
            },
            {
                "entity": "Pok\u00e9mon Rumble World",
                "confidence_score": "-6.5124764"
            }
        ]
    },
    {
        "mention": "nfc",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Near-field communication",
                "confidence_score": "6.076406"
            },
            {
                "entity": "NFC-WI",
                "confidence_score": "-10.898404"
            },
            {
                "entity": "NFC",
                "confidence_score": "-11.249118"
            }
        ]
    },
    {
        "mention": "toy pok\u00e9mon",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Toy",
                "confidence_score": "-6.1236506"
            },
            {
                "entity": "Toys-to-life",
                "confidence_score": "-6.7552185"
            },
            {
                "entity": "List of Pok\u00e9mon",
                "confidence_score": "-8.104606"
            }
        ]
    },
    {
        "mention": "defeated pok\u00e9mon",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "List of generation II Pok\u00e9mon",
                "confidence_score": "-10.143596"
            },
            {
                "entity": "List of generation V Pok\u00e9mon",
                "confidence_score": "-10.354275"
            },
            {
                "entity": "List of generation III Pok\u00e9mon",
                "confidence_score": "-10.484416"
            }
        ]
    },
    {
        "mention": "pok\u00e9mon",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Pok\u00e9mon",
                "confidence_score": "9.174415"
            },
            {
                "entity": "Pok\u00e9mon (video game series)",
                "confidence_score": "-2.674263"
            },
            {
                "entity": "List of Pok\u00e9mon",
                "confidence_score": "-3.8813608"
            }
        ]
    },
    {
        "mention": "ny",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Albany, New York",
                "confidence_score": "0.22707973"
            },
            {
                "entity": "New York (state)",
                "confidence_score": "-4.5831227"
            },
            {
                "entity": "New York",
                "confidence_score": "-12.470939"
            }
        ]
    },
    {
        "mention": "swot",
        "type": "ORG",
        "predictions": [
            {
                "entity": "SWOT analysis",
                "confidence_score": "-0.3342776"
            },
            {
                "entity": "SWOT",
                "confidence_score": "-11.008878"
            },
            {
                "entity": "WUOT",
                "confidence_score": "-11.066566"
            }
        ]
    },
    {
        "mention": "seattle public schools",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Seattle Public Schools",
                "confidence_score": "11.890267"
            },
            {
                "entity": "List of schools of the Seattle School District",
                "confidence_score": "-3.2656815"
            },
            {
                "entity": "Seattle Public Utilities",
                "confidence_score": "-8.095007"
            }
        ]
    },
    {
        "mention": "the teachers union",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Trade union",
                "confidence_score": "-5.1864924"
            },
            {
                "entity": "Teachers Union",
                "confidence_score": "-7.407612"
            },
            {
                "entity": "Japan Teachers Union",
                "confidence_score": "-10.578734"
            }
        ]
    },
    {
        "mention": "phyllis campano",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Phyllis Young",
                "confidence_score": "-11.43614"
            },
            {
                "entity": "The Campana Company",
                "confidence_score": "-11.820332"
            },
            {
                "entity": "Phyllis Krasilovsky",
                "confidence_score": "-11.885126"
            }
        ]
    },
    {
        "mention": "bureau of labor statistics",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Bureau of Labor Statistics",
                "confidence_score": "11.860587"
            },
            {
                "entity": "United States Department of Labor",
                "confidence_score": "-5.607055"
            },
            {
                "entity": "Occupational Employment Statistics",
                "confidence_score": "-8.494426"
            }
        ]
    },
    {
        "mention": "the national compensation survey",
        "type": "ORG",
        "predictions": [
            {
                "entity": "National Compensation Survey",
                "confidence_score": "11.324277"
            },
            {
                "entity": "National Longitudinal Surveys",
                "confidence_score": "-8.872485"
            },
            {
                "entity": "National Council on Compensation Insurance",
                "confidence_score": "-9.801768"
            }
        ]
    },
    {
        "mention": "richard vedder",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Richard Vedder",
                "confidence_score": "10.057748"
            },
            {
                "entity": "Richard Verrall",
                "confidence_score": "-11.089509"
            },
            {
                "entity": "Richard Venable",
                "confidence_score": "-11.5123825"
            }
        ]
    },
    {
        "mention": "bls",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Bureau of Labor Statistics",
                "confidence_score": "7.153271"
            },
            {
                "entity": "BLS AG",
                "confidence_score": "-4.6875253"
            },
            {
                "entity": "Nielsen Broadcast Data Systems",
                "confidence_score": "-5.930637"
            }
        ]
    },
    {
        "mention": "ncs",
        "type": "ORG",
        "predictions": [
            {
                "entity": "National Council for Science and the Environment",
                "confidence_score": "-10.53114"
            },
            {
                "entity": "National Coalition of Independent Scholars",
                "confidence_score": "-10.560463"
            },
            {
                "entity": "National Council for the Social Studies",
                "confidence_score": "-11.066806"
            }
        ]
    },
    {
        "mention": "the manhattan institute\u2019s",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Manhattan Institute for Policy Research",
                "confidence_score": "-4.041488"
            },
            {
                "entity": "Manhattan Research",
                "confidence_score": "-10.145362"
            },
            {
                "entity": "Hudson Institute",
                "confidence_score": "-10.782458"
            }
        ]
    },
    {
        "mention": "jay greene",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jay Greene",
                "confidence_score": "7.676336"
            },
            {
                "entity": "Jay P. Greene",
                "confidence_score": "-3.9344635"
            },
            {
                "entity": "John C. Greene",
                "confidence_score": "-11.0415945"
            }
        ]
    },
    {
        "mention": "marcus winters",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Dean Winters",
                "confidence_score": "-10.798395"
            },
            {
                "entity": "Bill Winters",
                "confidence_score": "-10.941709"
            },
            {
                "entity": "John W. Winters",
                "confidence_score": "-11.923697"
            }
        ]
    },
    {
        "mention": "bureau of labor statistics",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Bureau of Labor Statistics",
                "confidence_score": "11.860587"
            },
            {
                "entity": "United States Department of Labor",
                "confidence_score": "-5.607055"
            },
            {
                "entity": "Occupational Employment Statistics",
                "confidence_score": "-8.494426"
            }
        ]
    },
    {
        "mention": "seattle teachers on strike",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Seattle General Strike",
                "confidence_score": "-8.813672"
            },
            {
                "entity": "Seattle head tax",
                "confidence_score": "-11.078513"
            },
            {
                "entity": "Puget Sound fishermen's strike of 1949",
                "confidence_score": "-11.917005"
            }
        ]
    },
    {
        "mention": "redstate",
        "type": "ORG",
        "predictions": [
            {
                "entity": "RedState",
                "confidence_score": "11.244413"
            },
            {
                "entity": "Presbytery of Redstone",
                "confidence_score": "-11.123872"
            },
            {
                "entity": "Louisville, Cincinnati and Lexington Railroad",
                "confidence_score": "-11.49285"
            }
        ]
    },
    {
        "mention": "bermuda",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bermuda",
                "confidence_score": "8.81509"
            },
            {
                "entity": "Bermuda Day",
                "confidence_score": "-6.9112945"
            },
            {
                "entity": "Same-sex marriage in Bermuda",
                "confidence_score": "-10.101247"
            }
        ]
    },
    {
        "mention": "john fredriksen",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "John Fredriksen",
                "confidence_score": "10.60584"
            },
            {
                "entity": "Jan-Henrik Fredriksen",
                "confidence_score": "-11.810299"
            },
            {
                "entity": "John Christen Johansen",
                "confidence_score": "-11.845063"
            }
        ]
    },
    {
        "mention": "kate blankenship",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jane Blankenship",
                "confidence_score": "-10.726731"
            },
            {
                "entity": "Erin Blankenship",
                "confidence_score": "-11.195629"
            },
            {
                "entity": "J. Richard Blankenship",
                "confidence_score": "-11.254931"
            }
        ]
    },
    {
        "mention": "georgina e. sousa",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Georgina Mellor",
                "confidence_score": "-11.727206"
            },
            {
                "entity": "Georgina Wilson",
                "confidence_score": "-11.809529"
            },
            {
                "entity": "Georgina Oliva",
                "confidence_score": "-12.251878"
            }
        ]
    },
    {
        "mention": "jens martin jensen",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jens Fink-Jensen",
                "confidence_score": "-11.143422"
            },
            {
                "entity": "Martin Jensen (cricketer)",
                "confidence_score": "-11.277481"
            },
            {
                "entity": "Jens Kristian Jensen",
                "confidence_score": "-11.2846575"
            }
        ]
    },
    {
        "mention": "paul m. leand",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Sonja Ludvigsen",
                "confidence_score": "-11.659074"
            },
            {
                "entity": "Paul Lindemark J\u00f8rgensen",
                "confidence_score": "-11.939433"
            },
            {
                "entity": "Paul Jungbluth",
                "confidence_score": "-12.085913"
            }
        ]
    },
    {
        "mention": "hans petter aas",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Hans Aas",
                "confidence_score": "-9.328599"
            },
            {
                "entity": "Jan Petter Aas",
                "confidence_score": "-11.838099"
            },
            {
                "entity": "Johan Aas",
                "confidence_score": "-11.856629"
            }
        ]
    },
    {
        "mention": "the company's board of directors",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Board of directors",
                "confidence_score": "0.49285454"
            },
            {
                "entity": "Supervisory board",
                "confidence_score": "-1.6010246"
            },
            {
                "entity": "Directors register",
                "confidence_score": "-8.835935"
            }
        ]
    },
    {
        "mention": "hamilton",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Hamilton, Bermuda",
                "confidence_score": "3.0234861"
            },
            {
                "entity": "Bermuda",
                "confidence_score": "-4.3130937"
            },
            {
                "entity": "Hamilton Island, Bermuda",
                "confidence_score": "-7.630141"
            }
        ]
    },
    {
        "mention": "bermuda",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bermuda",
                "confidence_score": "8.81509"
            },
            {
                "entity": "Bermuda Day",
                "confidence_score": "-6.9112945"
            },
            {
                "entity": "Same-sex marriage in Bermuda",
                "confidence_score": "-10.101247"
            }
        ]
    },
    {
        "mention": "frontline",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Frontline (American TV program)",
                "confidence_score": "-4.1603403"
            },
            {
                "entity": "Front running",
                "confidence_score": "-9.627088"
            },
            {
                "entity": "Frontline Test Equipment",
                "confidence_score": "-10.660607"
            }
        ]
    },
    {
        "mention": "the securities and exchange commission",
        "type": "ORG",
        "predictions": [
            {
                "entity": "U.S. Securities and Exchange Commission",
                "confidence_score": "5.00856"
            },
            {
                "entity": "Securities and Exchange Commission (Philippines)",
                "confidence_score": "-2.4391124"
            },
            {
                "entity": "Securities commission",
                "confidence_score": "-6.4102306"
            }
        ]
    },
    {
        "mention": "sec",
        "type": "ORG",
        "predictions": [
            {
                "entity": "U.S. Securities and Exchange Commission",
                "confidence_score": "11.701374"
            },
            {
                "entity": "Securities and Exchange Commission (Philippines)",
                "confidence_score": "6.4866815"
            },
            {
                "entity": "Federal Communications Commission",
                "confidence_score": "-5.7850766"
            }
        ]
    },
    {
        "mention": "sec",
        "type": "ORG",
        "predictions": [
            {
                "entity": "U.S. Securities and Exchange Commission",
                "confidence_score": "11.701374"
            },
            {
                "entity": "Securities and Exchange Commission (Philippines)",
                "confidence_score": "6.4866815"
            },
            {
                "entity": "Federal Communications Commission",
                "confidence_score": "-5.7850766"
            }
        ]
    },
    {
        "mention": "investors",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Investor",
                "confidence_score": "4.1453314"
            },
            {
                "entity": "Investment",
                "confidence_score": "-2.0992265"
            },
            {
                "entity": "Institutional investor",
                "confidence_score": "-6.5157657"
            }
        ]
    },
    {
        "mention": "sec",
        "type": "ORG",
        "predictions": [
            {
                "entity": "U.S. Securities and Exchange Commission",
                "confidence_score": "11.701374"
            },
            {
                "entity": "Securities and Exchange Commission (Philippines)",
                "confidence_score": "6.4866815"
            },
            {
                "entity": "Federal Communications Commission",
                "confidence_score": "-5.7850766"
            }
        ]
    },
    {
        "mention": "sec",
        "type": "ORG",
        "predictions": [
            {
                "entity": "U.S. Securities and Exchange Commission",
                "confidence_score": "11.701374"
            },
            {
                "entity": "Securities and Exchange Commission (Philippines)",
                "confidence_score": "6.4866815"
            },
            {
                "entity": "Federal Communications Commission",
                "confidence_score": "-5.7850766"
            }
        ]
    },
    {
        "mention": "frontline",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Frontline (American TV program)",
                "confidence_score": "-4.1603403"
            },
            {
                "entity": "Front running",
                "confidence_score": "-9.627088"
            },
            {
                "entity": "Frontline Test Equipment",
                "confidence_score": "-10.660607"
            }
        ]
    },
    {
        "mention": "sec",
        "type": "ORG",
        "predictions": [
            {
                "entity": "U.S. Securities and Exchange Commission",
                "confidence_score": "11.701374"
            },
            {
                "entity": "Securities and Exchange Commission (Philippines)",
                "confidence_score": "6.4866815"
            },
            {
                "entity": "Federal Communications Commission",
                "confidence_score": "-5.7850766"
            }
        ]
    },
    {
        "mention": "sec",
        "type": "ORG",
        "predictions": [
            {
                "entity": "U.S. Securities and Exchange Commission",
                "confidence_score": "11.701374"
            },
            {
                "entity": "Securities and Exchange Commission (Philippines)",
                "confidence_score": "6.4866815"
            },
            {
                "entity": "Federal Communications Commission",
                "confidence_score": "-5.7850766"
            }
        ]
    },
    {
        "mention": "frontline",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Frontline (American TV program)",
                "confidence_score": "-4.1603403"
            },
            {
                "entity": "Front running",
                "confidence_score": "-9.627088"
            },
            {
                "entity": "Frontline Test Equipment",
                "confidence_score": "-10.660607"
            }
        ]
    },
    {
        "mention": "frontline",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Frontline (American TV program)",
                "confidence_score": "-4.1603403"
            },
            {
                "entity": "Front running",
                "confidence_score": "-9.627088"
            },
            {
                "entity": "Frontline Test Equipment",
                "confidence_score": "-10.660607"
            }
        ]
    },
    {
        "mention": "sec",
        "type": "ORG",
        "predictions": [
            {
                "entity": "U.S. Securities and Exchange Commission",
                "confidence_score": "11.701374"
            },
            {
                "entity": "Securities and Exchange Commission (Philippines)",
                "confidence_score": "6.4866815"
            },
            {
                "entity": "Federal Communications Commission",
                "confidence_score": "-5.7850766"
            }
        ]
    },
    {
        "mention": "nasdaq omx corporate solutions",
        "type": "ORG",
        "predictions": [
            {
                "entity": "NASDAQ",
                "confidence_score": "-4.8444333"
            },
            {
                "entity": "Nasdaq, Inc.",
                "confidence_score": "-6.405333"
            },
            {
                "entity": "Nasdaq Nordic",
                "confidence_score": "-8.659341"
            }
        ]
    },
    {
        "mention": "nasdaq omx corporate solutions",
        "type": "ORG",
        "predictions": [
            {
                "entity": "NASDAQ",
                "confidence_score": "-4.8444333"
            },
            {
                "entity": "Nasdaq, Inc.",
                "confidence_score": "-6.405333"
            },
            {
                "entity": "Nasdaq Nordic",
                "confidence_score": "-8.659341"
            }
        ]
    }
]


check_entity_type(predictions)
```

```python
import requests
from datetime import datetime, timedelta

def get_wikipedia_pageviews(page_titles):
    # Format today's date and the date 30 days ago for the API request
    end_date = datetime.utcnow().date()
    start_date = end_date - timedelta(days=30)
    formatted_start_date = start_date.strftime('%Y%m%d')
    formatted_end_date = end_date.strftime('%Y%m%d')

    # Base URL for the Wikipedia Pageviews API
    base_url = "https://wikimedia.org/api/rest_v1/metrics/pageviews/per-article/en.wikipedia/all-access/user/"

    # Dictionary to store page views
    page_views = {}

    # Loop through each page title to fetch the page views
    for title in page_titles:
        # Replace spaces with underscores for URL formatting
        formatted_title = title.replace(" ", "_")
        # Construct the full API URL
        full_url = f"{base_url}{formatted_title}/daily/{formatted_start_date}/{formatted_end_date}"

        # Make the API request
        response = requests.get(full_url)

        # Check if the request was successful
        if response.status_code == 200:
            data = response.json()
            total_views = sum(item['views'] for item in data['items'])
            page_views[title] = total_views
        else:
            page_views[title] = 'Error fetching data'

    return page_views

# List of Wikipedia page titles to check
pages_to_check = [
    "Chillicothe, Missouri",
    "Rob Weingartner",
    "Pumping station",
    "Luther Burbank Center for the Arts",
    "Kevin Coleman (aviator)",
    "Circleville, Ohio"
]

# Get page views for the list of pages
views = get_wikipedia_pageviews(pages_to_check)

# Print the results
for page, view_count in views.items():
    print(f"{page}: {view_count} views in the last 30 days")
```

```python
import difflib

def calculate_overlap(mention, entity):
    """
    Calculate the percentage overlap between the mention and the entity name based on the average ratio.
    """
    # Normalize strings by lowering case
    mention_normalized = mention.lower()
    entity_normalized = entity.lower()

    # Create a sequence matcher
    s = difflib.SequenceMatcher(None, mention_normalized, entity_normalized)

    # Calculate the number of matching characters
    match = sum(n for i, j, n in s.get_matching_blocks())

    # Calculate the average length of the mention and the entity
    average_length = (len(mention_normalized) + len(entity_normalized)) / 2

    # Calculate the percentage overlap based on the average length
    percentage_overlap = (match / average_length) * 100
    return percentage_overlap

# List of entity samples, each with only one prediction
entity_samples = [
    {
        "mention": "chillicothe's",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Chillicothe, Missouri",
                "confidence_score": "-5.2446394"
            }
        ]
    },
    {
        "mention": "rob weingartner",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Rob Weingartner",
                "confidence_score": "-0.047881465"
            }
        ]
    },
    {
        "mention": "pump house",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Pumping station",
                "confidence_score": "-3.9107552"
            }
        ]
    },
    {
        "mention": "the pump house center for the arts",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Luther Burbank Center for the Arts",
                "confidence_score": "-11.539211"
            }
        ]
    },
    {
        "mention": "kevin coleman",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kevin Coleman (aviator)",
                "confidence_score": "-9.057244"
            }
        ]
    },
    {
        "mention": "chillicothe & circleville",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Circleville, Ohio",
                "confidence_score": "-9.384942"
            }
        ]
    },
    {
        "mention": "oscar",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Academy Awards",
                "confidence_score": "2.9287302"
            }
        ]
    },
    {
        "mention": "everest",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Mount Everest",
                "confidence_score": "-5.034682"
            }
        ]
    },
    {
        "mention": "jason clarke",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jason Clarke",
                "confidence_score": "10.829769"
            }
        ]
    },
    {
        "mention": "rob hall",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Rob Hall",
                "confidence_score": "4.9627304"
            }
        ]
    },
    {
        "mention": "einar wegener",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Gerda Wegener",
                "confidence_score": "-1.7132143"
            }
        ]
    },
    {
        "mention": "lili elbe",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Lili Elbe",
                "confidence_score": "12.226471"
            }
        ]
    },
    {
        "mention": "alicia vikander",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Alicia Vikander",
                "confidence_score": "7.7965236"
            }
        ]
    },
    {
        "mention": "gerda wegener",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Gerda Wegener",
                "confidence_score": "5.5195546"
            }
        ]
    },
    {
        "mention": "jennifer lawrence",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jennifer Lawrence",
                "confidence_score": "12.278758"
            }
        ]
    },
    {
        "mention": "bradley cooper",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Bradley Cooper",
                "confidence_score": "12.290053"
            }
        ]
    },
    {
        "mention": "david o. russell",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "David O. Russell",
                "confidence_score": "11.249957"
            }
        ]
    },
    {
        "mention": "lawrence",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Lawrence Elion",
                "confidence_score": "-6.7262163"
            }
        ]
    },
    {
        "mention": "joy",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Joy Brook",
                "confidence_score": "-7.3752456"
            }
        ]
    },
    {
        "mention": "tom hardy",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Tom Hardy",
                "confidence_score": "8.108426"
            }
        ]
    },
    {
        "mention": "the kray twins",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Kray twins",
                "confidence_score": "1.1147673"
            }
        ]
    },
    {
        "mention": "kray",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kray twins",
                "confidence_score": "-1.2573466"
            }
        ]
    },
    {
        "mention": "kemp",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Jenny Kemp",
                "confidence_score": "-10.142403"
            }
        ]
    },
    {
        "mention": "patricia highsmith",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Patricia Highsmith",
                "confidence_score": "12.286905"
            }
        ]
    },
    {
        "mention": "rooney mara",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Rooney Mara",
                "confidence_score": "11.925614"
            }
        ]
    },
    {
        "mention": "rooney",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mickey Rooney",
                "confidence_score": "0.5553501"
            }
        ]
    },
    {
        "mention": "hugh glass",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Hugh Glass",
                "confidence_score": "3.4003975"
            }
        ]
    },
    {
        "mention": "whitey bulger",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Whitey Bulger",
                "confidence_score": "12.145009"
            }
        ]
    },
    {
        "mention": "benedict cumberbatch",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Benedict Cumberbatch",
                "confidence_score": "2.1088939"
            }
        ]
    },
    {
        "mention": "fbi",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Federal Bureau of Investigation",
                "confidence_score": "9.418494"
            }
        ]
    },
    {
        "mention": "steve jobs",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Steve Jobs",
                "confidence_score": "11.6573305"
            }
        ]
    },
    {
        "mention": "sony",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sony",
                "confidence_score": "10.630389"
            }
        ]
    },
    {
        "mention": "michael fassbender",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Michael Fassbender",
                "confidence_score": "2.4499412"
            }
        ]
    },
    {
        "mention": "jobs",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Job",
                "confidence_score": "7.691918"
            }
        ]
    },
    {
        "mention": "maud watts",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Kate Watts",
                "confidence_score": "-10.434687"
            }
        ]
    },
    {
        "mention": "carey mulligan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Carey Mulligan",
                "confidence_score": "12.278102"
            }
        ]
    },
    {
        "mention": "meryl streep",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Meryl Streep",
                "confidence_score": "12.702667"
            }
        ]
    },
    {
        "mention": "emmeline pankhurst",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Emmeline Pankhurst",
                "confidence_score": "8.254952"
            }
        ]
    },
    {
        "mention": "shakespeare",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "William Shakespeare",
                "confidence_score": "7.219626"
            }
        ]
    },
    {
        "mention": "marion cotillard",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Marion Cotillard",
                "confidence_score": "11.856938"
            }
        ]
    },
    {
        "mention": "fassbender",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Michael Fassbender",
                "confidence_score": "-4.6443315"
            }
        ]
    },
    {
        "mention": "julianne moore",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Julianne Moore",
                "confidence_score": "11.603573"
            }
        ]
    },
    {
        "mention": "laurel hester",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Laurel Hester",
                "confidence_score": "12.090917"
            }
        ]
    },
    {
        "mention": "stacie andree",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Stacie Lynn Renna",
                "confidence_score": "-10.795066"
            }
        ]
    },
    {
        "mention": "ellen page",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ellen Page",
                "confidence_score": "11.127679"
            }
        ]
    },
    {
        "mention": "cannes",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Cannes",
                "confidence_score": "8.212696"
            }
        ]
    },
    {
        "mention": "michael caine",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Michael Caine",
                "confidence_score": "12.066547"
            }
        ]
    },
    {
        "mention": "fred ballinger",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Adrian Ballinger",
                "confidence_score": "-10.5318365"
            }
        ]
    },
    {
        "mention": "alps",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Alps",
                "confidence_score": "1.4700428"
            }
        ]
    },
    {
        "mention": "mick boyle",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Frankie Boyle",
                "confidence_score": "-9.869424"
            }
        ]
    },
    {
        "mention": "harvey keitel",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Harvey Keitel",
                "confidence_score": "12.7008915"
            }
        ]
    },
    {
        "mention": "zomato",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Zomato",
                "confidence_score": "10.992883"
            }
        ]
    },
    {
        "mention": "sequoia india",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sequoia Capital",
                "confidence_score": "-6.7231884"
            }
        ]
    },
    {
        "mention": "vy capital",
        "type": "ORG",
        "predictions": [
            {
                "entity": "E.ventures",
                "confidence_score": "-10.503558"
            }
        ]
    },
    {
        "mention": "zomato founder",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "List of Internet entrepreneurs",
                "confidence_score": "-7.58396"
            }
        ]
    },
    {
        "mention": "deepinder goyal",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Des Raj Goyal",
                "confidence_score": "-8.819308"
            }
        ]
    },
    {
        "mention": "goyal",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Naresh Goyal",
                "confidence_score": "-5.3655577"
            }
        ]
    },
    {
        "mention": "chelsea",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Chelsea F.C.",
                "confidence_score": "0.9543292"
            }
        ]
    },
    {
        "mention": "aitor karanka",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Aitor Karanka",
                "confidence_score": "3.9784498"
            }
        ]
    },
    {
        "mention": "jose mourinho",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jos\u00e9 Mourinho",
                "confidence_score": "4.1757812"
            }
        ]
    },
    {
        "mention": "karanka",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Karanka",
                "confidence_score": "-6.5544124"
            }
        ]
    },
    {
        "mention": "mourinho",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Jos\u00e9 Mourinho",
                "confidence_score": "-3.9383771"
            }
        ]
    },
    {
        "mention": "ame info staff",
        "type": "ORG",
        "predictions": [
            {
                "entity": "AME Accounting Software",
                "confidence_score": "-10.947708"
            }
        ]
    },
    {
        "mention": "ppia",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Professional Photographers of America",
                "confidence_score": "-10.027242"
            }
        ]
    },
    {
        "mention": "matt",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Matt Damon",
                "confidence_score": "-3.689821"
            }
        ]
    },
    {
        "mention": "red rock canyon'",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Red Rock Canyon State Park (California)",
                "confidence_score": "-7.0642753"
            }
        ]
    },
    {
        "mention": "kraft boulders'",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Kraft Group",
                "confidence_score": "-9.100753"
            }
        ]
    },
    {
        "mention": "two sizes",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Canoeing at the 2016 Summer Olympics \u2013 Men's K-2 200 metres",
                "confidence_score": "-12.196137"
            }
        ]
    },
    {
        "mention": "new colgate total daily repair",
        "type": "ORG",
        "predictions": [
            {
                "entity": "All Pro Sound",
                "confidence_score": "-11.041748"
            }
        ]
    },
    {
        "mention": "monoprice hi-fi dj style acoustic pro studio headphones",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Monoprice",
                "confidence_score": "-7.9346685"
            }
        ]
    },
    {
        "mention": "credit line",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Credit card",
                "confidence_score": "-3.838872"
            }
        ]
    },
    {
        "mention": "queensland's",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Queensland",
                "confidence_score": "8.801793"
            }
        ]
    },
    {
        "mention": "the gold coast",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Gold Coast, Queensland",
                "confidence_score": "4.5411906"
            }
        ]
    },
    {
        "mention": "intuit inc.",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Intuit",
                "confidence_score": "9.700464"
            }
        ]
    },
    {
        "mention": "intuit workforce",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Intuit",
                "confidence_score": "-4.8245764"
            }
        ]
    },
    {
        "mention": "deliv",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Deliv",
                "confidence_score": "9.213618"
            }
        ]
    },
    {
        "mention": "orderup",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "OrderUp",
                "confidence_score": "8.679609"
            }
        ]
    },
    {
        "mention": "intuit",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Intuit",
                "confidence_score": "12.278062"
            }
        ]
    },
    {
        "mention": "intuit inc",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Intuit",
                "confidence_score": "11.465545"
            }
        ]
    },
    {
        "mention": "related stocks",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Related party transaction",
                "confidence_score": "-9.29873"
            }
        ]
    },
    {
        "mention": "pope francis",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Pope Francis",
                "confidence_score": "11.410729"
            }
        ]
    },
    {
        "mention": "francis",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Pope Francis",
                "confidence_score": "9.714959"
            }
        ]
    },
    {
        "mention": "eu",
        "type": "ORG",
        "predictions": [
            {
                "entity": "European Union",
                "confidence_score": "8.82327"
            }
        ]
    },
    {
        "mention": "donald tusk",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Donald Tusk",
                "confidence_score": "7.318099"
            }
        ]
    },
    {
        "mention": "europe",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Europe",
                "confidence_score": "4.80574"
            }
        ]
    },
    {
        "mention": "tusk",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Donald Tusk",
                "confidence_score": "-0.50873744"
            }
        ]
    },
    {
        "mention": "kolinda grabar-kitarovic",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kolinda Grabar-Kitarovi\u0107",
                "confidence_score": "11.856862"
            }
        ]
    },
    {
        "mention": "xenophobic",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Xenophobia",
                "confidence_score": "7.575575"
            }
        ]
    },
    {
        "mention": "angela merkel",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Angela Merkel",
                "confidence_score": "11.385147"
            }
        ]
    },
    {
        "mention": "croatia",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Croatia",
                "confidence_score": "6.6205053"
            }
        ]
    },
    {
        "mention": "european union",
        "type": "ORG",
        "predictions": [
            {
                "entity": "European Union",
                "confidence_score": "9.689422"
            }
        ]
    },
    {
        "mention": "grabar-kitarovic",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kolinda Grabar-Kitarovi\u0107",
                "confidence_score": "5.1088285"
            }
        ]
    },
    {
        "mention": "mediterranean",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Mediterranean Sea",
                "confidence_score": "2.0298774"
            }
        ]
    },
    {
        "mention": "the international organization for migration",
        "type": "ORG",
        "predictions": [
            {
                "entity": "International Organization for Migration",
                "confidence_score": "-1.8067393"
            }
        ]
    },
    {
        "mention": "beyond security",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Beyond Fear: Thinking Sensibly About Security in an Uncertain World",
                "confidence_score": "-4.483824"
            }
        ]
    },
    {
        "mention": "alp",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "9.75735"
            }
        ]
    },
    {
        "mention": "bill shorten",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bill Shorten",
                "confidence_score": "11.175131"
            }
        ]
    },
    {
        "mention": "malcolm turnbull",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Malcolm Turnbull",
                "confidence_score": "10.568268"
            }
        ]
    },
    {
        "mention": "the morgan poll",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Roy Morgan Research",
                "confidence_score": "-7.930781"
            }
        ]
    },
    {
        "mention": "labor",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Labor Party",
                "confidence_score": "4.0723853"
            }
        ]
    },
    {
        "mention": "morgan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Charles Morgan (Australian politician)",
                "confidence_score": "-5.300947"
            }
        ]
    },
    {
        "mention": "shorten",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Bill Shorten",
                "confidence_score": "2.9178472"
            }
        ]
    },
    {
        "mention": "yarralumla",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Yarralumla, Australian Capital Territory",
                "confidence_score": "0.2666576"
            }
        ]
    },
    {
        "mention": "nationals",
        "type": "LOC",
        "predictions": [
            {
                "entity": "National Party of Australia",
                "confidence_score": "8.716648"
            }
        ]
    },
    {
        "mention": "reachtel",
        "type": "ORG",
        "predictions": [
            {
                "entity": "ReachTEL",
                "confidence_score": "9.550408"
            }
        ]
    },
    {
        "mention": "western australia\u2019s",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Western Australia",
                "confidence_score": "7.843019"
            }
        ]
    },
    {
        "mention": "hawke",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bob Hawke",
                "confidence_score": "4.7292747"
            }
        ]
    },
    {
        "mention": "canning",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Division of Canning",
                "confidence_score": "1.3687484"
            }
        ]
    },
    {
        "mention": "liberal mp",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Liberal Party of Australia",
                "confidence_score": "0.29603356"
            }
        ]
    },
    {
        "mention": "don randall",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Don Randall",
                "confidence_score": "12.53306"
            }
        ]
    },
    {
        "mention": "liberals",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Liberal Party of Australia (New South Wales Division)",
                "confidence_score": "7.5084767"
            }
        ]
    },
    {
        "mention": "howard",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "John Howard",
                "confidence_score": "3.7991908"
            }
        ]
    },
    {
        "mention": "lindsay",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Division of Lindsay",
                "confidence_score": "3.1409073"
            }
        ]
    },
    {
        "mention": "nsw",
        "type": "ORG",
        "predictions": [
            {
                "entity": "New South Wales",
                "confidence_score": "7.0234013"
            }
        ]
    },
    {
        "mention": "galaxy",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Galaxy (Australian television)",
                "confidence_score": "1.002422"
            }
        ]
    },
    {
        "mention": "tony abbott",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Tony Abbott",
                "confidence_score": "11.302775"
            }
        ]
    },
    {
        "mention": "turnbull",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Malcolm Turnbull",
                "confidence_score": "-0.58703876"
            }
        ]
    },
    {
        "mention": "abbott",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Abbott Government",
                "confidence_score": "-3.1683836"
            }
        ]
    },
    {
        "mention": "kevin rudd",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kevin Rudd",
                "confidence_score": "11.032903"
            }
        ]
    },
    {
        "mention": "the house of representatives",
        "type": "ORG",
        "predictions": [
            {
                "entity": "United States House of Representatives",
                "confidence_score": "-2.7429712"
            }
        ]
    },
    {
        "mention": "john howard",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "John Howard",
                "confidence_score": "10.323403"
            }
        ]
    },
    {
        "mention": "rudd",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kevin Rudd",
                "confidence_score": "0.9712579"
            }
        ]
    },
    {
        "mention": "wayne swan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Wayne Swan",
                "confidence_score": "12.203911"
            }
        ]
    },
    {
        "mention": "julia gillard",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Julia Gillard",
                "confidence_score": "11.079709"
            }
        ]
    },
    {
        "mention": "actu",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian Council of Trade Unions",
                "confidence_score": "4.713805"
            }
        ]
    },
    {
        "mention": "kindy",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Alex Kindy",
                "confidence_score": "-9.07692"
            }
        ]
    },
    {
        "mention": "john black",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "John Black (Australian politician)",
                "confidence_score": "6.4515595"
            }
        ]
    },
    {
        "mention": "australian development strategies",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Committee for Economic Development of Australia",
                "confidence_score": "-8.79475"
            }
        ]
    },
    {
        "mention": "davis county school district",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Davis School District",
                "confidence_score": "0.26024666"
            }
        ]
    },
    {
        "mention": "dui",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Driving under the influence",
                "confidence_score": "3.0896108"
            }
        ]
    },
    {
        "mention": "brightgreen",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Shades of yellow",
                "confidence_score": "-11.037421"
            }
        ]
    },
    {
        "mention": "the d700+",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Flyer New Look bus",
                "confidence_score": "-10.521388"
            }
        ]
    },
    {
        "mention": "tru-colour",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "8-bit color",
                "confidence_score": "-7.0598793"
            }
        ]
    },
    {
        "mention": "jeremy corbyn",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jeremy Corbyn",
                "confidence_score": "12.129742"
            }
        ]
    },
    {
        "mention": "labour",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Labour Party (UK)",
                "confidence_score": "2.7017815"
            }
        ]
    },
    {
        "mention": "corbyn",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jeremy Corbyn",
                "confidence_score": "0.5464574"
            }
        ]
    },
    {
        "mention": "jez",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jez",
                "confidence_score": "-9.403179"
            }
        ]
    },
    {
        "mention": "green party",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Green party",
                "confidence_score": "5.3362203"
            }
        ]
    },
    {
        "mention": "natalie bennett",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Natalie Bennett",
                "confidence_score": "12.218274"
            }
        ]
    },
    {
        "mention": "david cameron",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "David Cameron",
                "confidence_score": "10.9060545"
            }
        ]
    },
    {
        "mention": "carol turner",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Carol I. Turner",
                "confidence_score": "-10.652766"
            }
        ]
    },
    {
        "mention": "stop the war coalition",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Stop the War Coalition",
                "confidence_score": "10.414976"
            }
        ]
    },
    {
        "mention": "parliament",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Parliament",
                "confidence_score": "7.3973947"
            }
        ]
    },
    {
        "mention": "walter wolfgang",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Walter Wolfgang",
                "confidence_score": "12.015021"
            }
        ]
    },
    {
        "mention": "cnd",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Campaign for Nuclear Disarmament",
                "confidence_score": "4.9109015"
            }
        ]
    },
    {
        "mention": "wolfgang",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Walter Wolfgang",
                "confidence_score": "7.731377"
            }
        ]
    },
    {
        "mention": "download) civil services mentor magazine",
        "type": "ORG",
        "predictions": [
            {
                "entity": "25 Magazine",
                "confidence_score": "-11.536889"
            }
        ]
    },
    {
        "mention": "pdf yojana magazine",
        "type": "ORG",
        "predictions": [
            {
                "entity": "PDF",
                "confidence_score": "-4.29032"
            }
        ]
    },
    {
        "mention": "current public administration magazine",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Public Administration (journal)",
                "confidence_score": "-10.280338"
            }
        ]
    },
    {
        "mention": "modi government",
        "type": "ORG",
        "predictions": [
            {
                "entity": "First Modi ministry",
                "confidence_score": "-3.1535122"
            }
        ]
    },
    {
        "mention": "aparajita priyadarshini behera",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Aparna Dixit",
                "confidence_score": "-11.457622"
            }
        ]
    },
    {
        "mention": "indian economic service",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Indian Economic Service",
                "confidence_score": "11.819936"
            }
        ]
    },
    {
        "mention": "upsc exams",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Union Public Service Commission",
                "confidence_score": "-3.0394003"
            }
        ]
    },
    {
        "mention": "mahesh sharma",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mahesh Sharma",
                "confidence_score": "10.007316"
            }
        ]
    },
    {
        "mention": "\u0930\u093e\u0937\u094d\u091f\u094d\u0930\u092e\u0902\u0921\u0932 \u0916\u0947\u0932",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Jhala Nath Khanal",
                "confidence_score": "-9.743552"
            }
        ]
    },
    {
        "mention": "amazon click",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Amazon (company)",
                "confidence_score": "-5.560904"
            }
        ]
    },
    {
        "mention": "digital india socio economic",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Digital India",
                "confidence_score": "-4.382664"
            }
        ]
    },
    {
        "mention": "national issues international issues economy india & the world science & technology sports awards & prizes persons",
        "type": "ORG",
        "predictions": [
            {
                "entity": "2015 National Games of India",
                "confidence_score": "-8.429058"
            }
        ]
    },
    {
        "mention": "yojana magazine",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Yojana Malayalam",
                "confidence_score": "-7.3992524"
            }
        ]
    },
    {
        "mention": "modi government posted",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Narendra Modi",
                "confidence_score": "-3.3587415"
            }
        ]
    },
    {
        "mention": "current public administration magazine welfare administration major",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Currentweek",
                "confidence_score": "-10.805299"
            }
        ]
    },
    {
        "mention": "modi government ever",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Narendra Modi",
                "confidence_score": "-4.494273"
            }
        ]
    },
    {
        "mention": "lok sabha",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Lok Sabha",
                "confidence_score": "10.685716"
            }
        ]
    },
    {
        "mention": "pradhan mantri",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Pradhanmantri",
                "confidence_score": "8.048137"
            }
        ]
    },
    {
        "mention": "sansad",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sansad",
                "confidence_score": "1.8994746"
            }
        ]
    },
    {
        "mention": "gram yojana",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Sansad Adarsh Gram Yojana",
                "confidence_score": "3.4763048"
            }
        ]
    },
    {
        "mention": "greenery",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Landscaping",
                "confidence_score": "-2.6385872"
            }
        ]
    },
    {
        "mention": "namami ganga yojana",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Pradhan Mantri Gramin Awaas Yojana",
                "confidence_score": "-11.141009"
            }
        ]
    },
    {
        "mention": "the integrated ganga conservation mission",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Integrated Conservation and Development Project",
                "confidence_score": "-11.5035515"
            }
        ]
    },
    {
        "mention": "namami ganga yojana'",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jogi",
                "confidence_score": "-9.663323"
            }
        ]
    },
    {
        "mention": "the modi government",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Narendra Modi",
                "confidence_score": "-5.288714"
            }
        ]
    },
    {
        "mention": "namami ganga'",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Namli",
                "confidence_score": "-10.992118"
            }
        ]
    },
    {
        "mention": "krishi sinchayee",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Krishnendu Adhikari",
                "confidence_score": "-11.106698"
            }
        ]
    },
    {
        "mention": "yojana",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Yojana",
                "confidence_score": "7.5459175"
            }
        ]
    },
    {
        "mention": "random inspection scheme",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Random assignment",
                "confidence_score": "-10.538868"
            }
        ]
    },
    {
        "mention": "provident fund",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Provident fund",
                "confidence_score": "9.231428"
            }
        ]
    },
    {
        "mention": "apprentice protsahan yojana:",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Swarna Jayanti Shahari Rozgar Yojana",
                "confidence_score": "-10.846993"
            }
        ]
    },
    {
        "mention": "rashtriya swasthya bima yojana",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Rashtriya Swasthya Bima Yojana",
                "confidence_score": "8.013553"
            }
        ]
    },
    {
        "mention": "a smart card",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Smart card",
                "confidence_score": "5.4725575"
            }
        ]
    },
    {
        "mention": "ies",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Indian Economic Service",
                "confidence_score": "11.033611"
            }
        ]
    },
    {
        "mention": "the indian economic service",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Indian Economic Service",
                "confidence_score": "9.033747"
            }
        ]
    },
    {
        "mention": "upsc",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Union Public Service Commission",
                "confidence_score": "7.877064"
            }
        ]
    },
    {
        "mention": "mahakalapada",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Mahakalapada",
                "confidence_score": "4.680041"
            }
        ]
    },
    {
        "mention": "phd",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Doctor of Philosophy",
                "confidence_score": "10.021392"
            }
        ]
    },
    {
        "mention": "jawaharlal nehru university",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Jawaharlal Nehru University",
                "confidence_score": "10.6941595"
            }
        ]
    },
    {
        "mention": "aparajita",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Aparajita",
                "confidence_score": "0.86721766"
            }
        ]
    },
    {
        "mention": "kapileswar high school",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Government High School, Karwar",
                "confidence_score": "-11.090369"
            }
        ]
    },
    {
        "mention": "post-graduation in economics",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Master of Economics",
                "confidence_score": "-9.636912"
            }
        ]
    },
    {
        "mention": "ira singhal",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ira Singhal",
                "confidence_score": "12.27744"
            }
        ]
    },
    {
        "mention": "indian administrative service",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Indian Administrative Service",
                "confidence_score": "11.102447"
            }
        ]
    },
    {
        "mention": "vandana rao",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Vandana Rao",
                "confidence_score": "5.897549"
            }
        ]
    },
    {
        "mention": "renu raj",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Renu Kumari Singh",
                "confidence_score": "-10.885899"
            }
        ]
    },
    {
        "mention": "suharsha bhagat",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Sudarshan Bhagat",
                "confidence_score": "-10.554279"
            }
        ]
    },
    {
        "mention": "department of personnel and training",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Ministry of Personnel, Public Grievances and Pensions",
                "confidence_score": "-6.2770805"
            }
        ]
    },
    {
        "mention": "nidhi gupta",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Nidhi Singh",
                "confidence_score": "-11.313328"
            }
        ]
    },
    {
        "mention": "haryana",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Haryana",
                "confidence_score": "8.280659"
            }
        ]
    },
    {
        "mention": "discuss your answer",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Question",
                "confidence_score": "-7.3469243"
            }
        ]
    },
    {
        "mention": "comment box",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Comment (computer programming)",
                "confidence_score": "-8.036764"
            }
        ]
    },
    {
        "mention": "hamid ansari's",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mohammad Hamid Ansari",
                "confidence_score": "-4.345567"
            }
        ]
    },
    {
        "mention": "the special investigation team",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Inter-Services Intelligence",
                "confidence_score": "-8.863917"
            }
        ]
    },
    {
        "mention": "participatory notes",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Participatory note",
                "confidence_score": "7.417099"
            }
        ]
    },
    {
        "mention": "p-notes",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Participatory note",
                "confidence_score": "3.3666077"
            }
        ]
    },
    {
        "mention": "g subba rao",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "G. S. R. Subba Rao",
                "confidence_score": "-0.2877108"
            }
        ]
    },
    {
        "mention": "p n roy",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Samarendra Nath Roy",
                "confidence_score": "-7.9348793"
            }
        ]
    },
    {
        "mention": "iii & iv",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Indian Institutes of Science Education and Research",
                "confidence_score": "-10.38115"
            }
        ]
    },
    {
        "mention": "daily questions challenge",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Helsinki Challenge",
                "confidence_score": "-10.7572975"
            }
        ]
    },
    {
        "mention": "tamil nadu",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Tamil Nadu",
                "confidence_score": "6.1177793"
            }
        ]
    },
    {
        "mention": "jaduguda",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jadugora",
                "confidence_score": "3.0001957"
            }
        ]
    },
    {
        "mention": "pressurized water",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Pressurized water reactor",
                "confidence_score": "6.6550784"
            }
        ]
    },
    {
        "mention": "pressurized heavy",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Pressurized heavy-water reactor",
                "confidence_score": "4.3804383"
            }
        ]
    },
    {
        "mention": "fast breeder",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Fast Breeder Test Reactor",
                "confidence_score": "-4.5922112"
            }
        ]
    },
    {
        "mention": "advanced heavy",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Advanced heavy-water reactor",
                "confidence_score": "-7.387637"
            }
        ]
    },
    {
        "mention": "kamini",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "KAMINI",
                "confidence_score": "7.050175"
            }
        ]
    },
    {
        "mention": "rohini click",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Rohini Kesavan Srihari",
                "confidence_score": "-10.762311"
            }
        ]
    },
    {
        "mention": "current affairs mcq",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Global Affairs Canada",
                "confidence_score": "-10.564147"
            }
        ]
    },
    {
        "mention": "diclofenac",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Diclofenac",
                "confidence_score": "12.24802"
            }
        ]
    },
    {
        "mention": "asia",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Asia",
                "confidence_score": "3.6320422"
            }
        ]
    },
    {
        "mention": "current affairs for",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Current affairs (news format)",
                "confidence_score": "-4.1939354"
            }
        ]
    },
    {
        "mention": "greenpeace india stating",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Greenpeace India",
                "confidence_score": "2.6043289"
            }
        ]
    },
    {
        "mention": "fcra",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Fair Credit Reporting Act",
                "confidence_score": "3.878489"
            }
        ]
    },
    {
        "mention": "the ministry of home affairs",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Ministry of Home Affairs (India)",
                "confidence_score": "0.2794178"
            }
        ]
    },
    {
        "mention": "ngo",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Non-governmental organization",
                "confidence_score": "7.5985384"
            }
        ]
    },
    {
        "mention": "the foreign contribution regulation act",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Foreign Contribution (Regulation) Act, 2010",
                "confidence_score": "5.941937"
            }
        ]
    },
    {
        "mention": "the delhi high court",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Delhi High Court",
                "confidence_score": "4.571439"
            }
        ]
    },
    {
        "mention": "mha",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Master of Health Administration",
                "confidence_score": "-2.552178"
            }
        ]
    },
    {
        "mention": "greenpeace india",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Greenpeace India",
                "confidence_score": "10.40537"
            }
        ]
    },
    {
        "mention": "idbi",
        "type": "ORG",
        "predictions": [
            {
                "entity": "IDBI Bank",
                "confidence_score": "1.6861428"
            }
        ]
    },
    {
        "mention": "icici",
        "type": "ORG",
        "predictions": [
            {
                "entity": "ICICI Bank",
                "confidence_score": "5.071098"
            }
        ]
    },
    {
        "mention": "bank",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Bank",
                "confidence_score": "4.9896016"
            }
        ]
    },
    {
        "mention": "euros",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Euro",
                "confidence_score": "9.414692"
            }
        ]
    },
    {
        "mention": "stichting greenpeace council",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Greenpeace",
                "confidence_score": "-3.4288359"
            }
        ]
    },
    {
        "mention": "cameron",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "David Cameron",
                "confidence_score": "0.045297895"
            }
        ]
    },
    {
        "mention": "un",
        "type": "ORG",
        "predictions": [
            {
                "entity": "United Nations",
                "confidence_score": "7.8057218"
            }
        ]
    },
    {
        "mention": "pdt current affairs daily voice",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Daily Voice (South African newspaper)",
                "confidence_score": "-10.9162855"
            }
        ]
    },
    {
        "mention": "afs interviewer",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Inter Press Service",
                "confidence_score": "-11.325374"
            }
        ]
    },
    {
        "mention": "sindhu jain bhattacharya",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Basu Bhattacharya",
                "confidence_score": "-11.010752"
            }
        ]
    },
    {
        "mention": "\u0928\u093f\u0930\u094d\u0923\u092f \u092a\u0930",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Pandit Narendra Sharma",
                "confidence_score": "-11.24919"
            }
        ]
    },
    {
        "mention": "hdp",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Peoples' Democratic Party (Turkey)",
                "confidence_score": "4.72046"
            }
        ]
    },
    {
        "mention": "raymond mould",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Raymond Mould",
                "confidence_score": "12.062735"
            }
        ]
    },
    {
        "mention": "the grand national",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Grand National",
                "confidence_score": "-1.3877977"
            }
        ]
    },
    {
        "mention": "champion chase",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Champion Hurdle",
                "confidence_score": "-4.991761"
            }
        ]
    },
    {
        "mention": "george vi chase",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "King George VI Chase",
                "confidence_score": "-1.6328759"
            }
        ]
    },
    {
        "mention": "mould",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Philip Mould",
                "confidence_score": "-2.245812"
            }
        ]
    },
    {
        "mention": "sprinter sacre",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sprinter Sacre",
                "confidence_score": "3.6181815"
            }
        ]
    },
    {
        "mention": "charter party",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Charter Party",
                "confidence_score": "5.193508"
            }
        ]
    },
    {
        "mention": "barton bank",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Barton Bank",
                "confidence_score": "1.6099434"
            }
        ]
    },
    {
        "mention": "pillar properties",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Capita Property and Infrastructure",
                "confidence_score": "-10.482668"
            }
        ]
    },
    {
        "mention": "guiting power",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Guiting Power",
                "confidence_score": "8.496997"
            }
        ]
    },
    {
        "mention": "gloucestershire",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Gloucestershire",
                "confidence_score": "6.0391774"
            }
        ]
    },
    {
        "mention": "david nicholson",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "David Nicholson (journalist)",
                "confidence_score": "1.4030061"
            }
        ]
    },
    {
        "mention": "colin",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Colin Phipps",
                "confidence_score": "-7.427876"
            }
        ]
    },
    {
        "mention": "clare smith",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Clare Calbraith",
                "confidence_score": "-11.647167"
            }
        ]
    },
    {
        "mention": "nigel twiston-davies",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Nigel Twiston-Davies",
                "confidence_score": "11.401046"
            }
        ]
    },
    {
        "mention": "twiston-davies",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Nigel Twiston-Davies",
                "confidence_score": "1.4381024"
            }
        ]
    },
    {
        "mention": "raymond",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Raymond",
                "confidence_score": "-5.1941285"
            }
        ]
    },
    {
        "mention": "tipping tim",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Tip Tipping",
                "confidence_score": "-9.992383"
            }
        ]
    },
    {
        "mention": "caroline",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Caroline Pierce",
                "confidence_score": "-11.365865"
            }
        ]
    },
    {
        "mention": "the queen mother champion chase",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Queen Mother Champion Chase",
                "confidence_score": "9.97849"
            }
        ]
    },
    {
        "mention": "the timeform organisation",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Timeform",
                "confidence_score": "-1.4261535"
            }
        ]
    },
    {
        "mention": "national hunt",
        "type": "ORG",
        "predictions": [
            {
                "entity": "National Hunt racing",
                "confidence_score": "6.0748844"
            }
        ]
    },
    {
        "mention": "flyingbolt",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Flyingbolt",
                "confidence_score": "9.404755"
            }
        ]
    },
    {
        "mention": "mn",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Minnesota",
                "confidence_score": "0.008507854"
            }
        ]
    },
    {
        "mention": "the goodman group",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Goodman Group",
                "confidence_score": "-7.1211195"
            }
        ]
    },
    {
        "mention": "jennifer ballantine",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jennifer Crystal Foley",
                "confidence_score": "-11.160103"
            }
        ]
    },
    {
        "mention": "able palms home health of minneapolis",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Jonathan, Minnesota",
                "confidence_score": "-10.24767"
            }
        ]
    },
    {
        "mention": "ballantine",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Ballantine, Montana",
                "confidence_score": "-9.58007"
            }
        ]
    },
    {
        "mention": "craig edinger",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Evan Edinger",
                "confidence_score": "-11.7964945"
            }
        ]
    },
    {
        "mention": "senior living & health care",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Elderly care",
                "confidence_score": "-8.198955"
            }
        ]
    },
    {
        "mention": "jennifer",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jennifer Aniston",
                "confidence_score": "-9.237152"
            }
        ]
    },
    {
        "mention": "edinger",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Edward F. Edinger",
                "confidence_score": "-3.9674938"
            }
        ]
    },
    {
        "mention": "recover health",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Recovery International",
                "confidence_score": "-8.6431875"
            }
        ]
    },
    {
        "mention": "medicare",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Medicare (United States)",
                "confidence_score": "8.6975155"
            }
        ]
    },
    {
        "mention": "metropolitan state university",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Metropolitan State University",
                "confidence_score": "11.63488"
            }
        ]
    },
    {
        "mention": "twin cities",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Twin cities",
                "confidence_score": "5.7361817"
            }
        ]
    },
    {
        "mention": "community health accreditation partner",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Community Health Accreditation Program",
                "confidence_score": "-1.7553334"
            }
        ]
    },
    {
        "mention": "able palms home & health care services",
        "type": "ORG",
        "predictions": [
            {
                "entity": "PSA Healthcare",
                "confidence_score": "-10.8723345"
            }
        ]
    },
    {
        "mention": "arch coal",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Arch Coal",
                "confidence_score": "9.177982"
            }
        ]
    },
    {
        "mention": "aci",
        "type": "ORG",
        "predictions": [
            {
                "entity": "American Competitiveness Initiative",
                "confidence_score": "-0.014987714"
            }
        ]
    },
    {
        "mention": "free",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Free software",
                "confidence_score": "-3.2891412"
            }
        ]
    },
    {
        "mention": "gazette",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Gazette",
                "confidence_score": "11.215565"
            }
        ]
    },
    {
        "mention": "carmen de miguel",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Carmen Ejogo",
                "confidence_score": "-11.072501"
            }
        ]
    },
    {
        "mention": "the university of alabama at birmingham",
        "type": "ORG",
        "predictions": [
            {
                "entity": "University of Alabama at Birmingham",
                "confidence_score": "4.5339465"
            }
        ]
    },
    {
        "mention": "de miguel",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Miguel de Miguel",
                "confidence_score": "-6.6036286"
            }
        ]
    },
    {
        "mention": "bmi",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Body mass index",
                "confidence_score": "3.268693"
            }
        ]
    },
    {
        "mention": "de miguel c",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Miguel de Miguel",
                "confidence_score": "-7.783598"
            }
        ]
    },
    {
        "mention": "xu x",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Xu Yuhua",
                "confidence_score": "-11.412566"
            }
        ]
    },
    {
        "mention": "pan y",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Pan Entertainment",
                "confidence_score": "-10.940717"
            }
        ]
    },
    {
        "mention": "teen",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Adolescence",
                "confidence_score": "-1.9932653"
            }
        ]
    },
    {
        "mention": "american heart association abstract",
        "type": "ORG",
        "predictions": [
            {
                "entity": "American Heart Association",
                "confidence_score": "-2.8352323"
            }
        ]
    },
    {
        "mention": "cine kolly deivamagal - 26-09-2015",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Cin\u00e9+",
                "confidence_score": "-10.647243"
            }
        ]
    },
    {
        "mention": "sun tv deivamagal",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Deivamagal",
                "confidence_score": "1.929848"
            }
        ]
    },
    {
        "mention": "sun tv serials",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Tamil television soap opera",
                "confidence_score": "-3.1146088"
            }
        ]
    },
    {
        "mention": "youtube video deivamagal",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "YouTube",
                "confidence_score": "-4.0530386"
            }
        ]
    },
    {
        "mention": "vikatan tele vistas",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Vu Televisions",
                "confidence_score": "-10.618415"
            }
        ]
    },
    {
        "mention": "thirumathi selvam",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Thirumathi Selvam",
                "confidence_score": "12.114499"
            }
        ]
    },
    {
        "mention": "sun",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sun Microsystems",
                "confidence_score": "-3.78396"
            }
        ]
    },
    {
        "mention": "post comments",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Comments section",
                "confidence_score": "-3.9577708"
            }
        ]
    },
    {
        "mention": "los angeles professional employment forecast",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Los Angeles Business Journal",
                "confidence_score": "-10.098554"
            }
        ]
    },
    {
        "mention": "robert half indicates",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Robert Half International",
                "confidence_score": "-3.875004"
            }
        ]
    },
    {
        "mention": "freezing",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Freezing",
                "confidence_score": "6.6845775"
            }
        ]
    },
    {
        "mention": "brandi britton",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Brandi Britton",
                "confidence_score": "10.034802"
            }
        ]
    },
    {
        "mention": "robert half",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Robert Half International",
                "confidence_score": "4.106703"
            }
        ]
    },
    {
        "mention": "robert half's",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Robert Half International",
                "confidence_score": "5.643333"
            }
        ]
    },
    {
        "mention": "the professional employment forecast",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Ten-year occupational employment projections",
                "confidence_score": "-10.643827"
            }
        ]
    },
    {
        "mention": "compensation upgrades",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Golden boot compensation",
                "confidence_score": "-10.370938"
            }
        ]
    },
    {
        "mention": "roberthalf.com/los-angeles",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Los Angeles",
                "confidence_score": "-1.6272929"
            }
        ]
    },
    {
        "mention": "pr newswire",
        "type": "ORG",
        "predictions": [
            {
                "entity": "PR Newswire",
                "confidence_score": "12.28939"
            }
        ]
    },
    {
        "mention": "paul a. eisenstein",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Melvin A. Eisenberg",
                "confidence_score": "-11.520937"
            }
        ]
    },
    {
        "mention": "cameroon",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Cameroon",
                "confidence_score": "0.70326823"
            }
        ]
    },
    {
        "mention": "the international organization of motor vehicle manufacturers",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Organisation Internationale des Constructeurs d'Automobiles",
                "confidence_score": "-6.320154"
            }
        ]
    },
    {
        "mention": "oica",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Organisation Internationale des Constructeurs d'Automobiles",
                "confidence_score": "3.1770308"
            }
        ]
    },
    {
        "mention": "africa",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Africa",
                "confidence_score": "4.098563"
            }
        ]
    },
    {
        "mention": "americas",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Americas",
                "confidence_score": "5.7528462"
            }
        ]
    },
    {
        "mention": "matthias wissmann",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Matthias Wissmann",
                "confidence_score": "11.339347"
            }
        ]
    },
    {
        "mention": "the german association of the automotive industry",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Verband der Automobilindustrie",
                "confidence_score": "5.5893908"
            }
        ]
    },
    {
        "mention": "frankfurt messe",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Messe Frankfurt",
                "confidence_score": "1.348766"
            }
        ]
    },
    {
        "mention": "renault/nissan",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Nissan",
                "confidence_score": "-2.7255714"
            }
        ]
    },
    {
        "mention": "carlos ghosn",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Carlos Ghosn",
                "confidence_score": "10.845214"
            }
        ]
    },
    {
        "mention": "gm",
        "type": "ORG",
        "predictions": [
            {
                "entity": "General Motors",
                "confidence_score": "8.120402"
            }
        ]
    },
    {
        "mention": "justice department",
        "type": "ORG",
        "predictions": [
            {
                "entity": "United States Department of Justice",
                "confidence_score": "5.0132995"
            }
        ]
    },
    {
        "mention": "daimler",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Daimler AG",
                "confidence_score": "4.9833813"
            }
        ]
    },
    {
        "mention": "honda",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Honda",
                "confidence_score": "9.132179"
            }
        ]
    },
    {
        "mention": "click here",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Nissan (disambiguation)",
                "confidence_score": "-11.386483"
            }
        ]
    },
    {
        "mention": "renault/nissan\u2019s",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Nissan",
                "confidence_score": "-1.1531711"
            }
        ]
    },
    {
        "mention": "automobiles",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Automobile (magazine)",
                "confidence_score": "-4.559293"
            }
        ]
    },
    {
        "mention": "nissan",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Nissan",
                "confidence_score": "4.688967"
            }
        ]
    },
    {
        "mention": "joey wang",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Joey Wong",
                "confidence_score": "1.3677337"
            }
        ]
    },
    {
        "mention": "lanniao",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Nissan Lannia",
                "confidence_score": "-4.159739"
            }
        ]
    },
    {
        "mention": "the nissan china design center (ndc",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Nintendo System Development",
                "confidence_score": "-11.417943"
            }
        ]
    },
    {
        "mention": "the nissan global design center",
        "type": "ORG",
        "predictions": [
            {
                "entity": "International Technical Development Center",
                "confidence_score": "-10.580587"
            }
        ]
    },
    {
        "mention": "ngdc",
        "type": "ORG",
        "predictions": [
            {
                "entity": "National Commission for Denotified, Nomadic and Semi-Nomadic Tribes",
                "confidence_score": "-11.838946"
            }
        ]
    },
    {
        "mention": "dongfeng-nissan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Dongfeng Motor Co., Ltd.",
                "confidence_score": "-4.691514"
            }
        ]
    },
    {
        "mention": "fuzzy",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Fuzzy classification",
                "confidence_score": "-7.445805"
            }
        ]
    },
    {
        "mention": "loading",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Loader (computing)",
                "confidence_score": "-7.0049486"
            }
        ]
    },
    {
        "mention": "dongfeng-nissan nissan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Nissan",
                "confidence_score": "-3.6953557"
            }
        ]
    },
    {
        "mention": "ritz",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Ritz Tower",
                "confidence_score": "-9.453444"
            }
        ]
    },
    {
        "mention": "the ritz-carlton \u2018s",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Ritz-Carlton Hotel Company",
                "confidence_score": "-3.5609815"
            }
        ]
    },
    {
        "mention": "the mexican embassy",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Embassy of Mexico, Paris",
                "confidence_score": "-9.635977"
            }
        ]
    },
    {
        "mention": "the chile en nogada",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Chile",
                "confidence_score": "-4.134796"
            }
        ]
    },
    {
        "mention": "nogada",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Negada",
                "confidence_score": "-11.620681"
            }
        ]
    },
    {
        "mention": "jorge carlos gonz\u00e1les",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Rodolfo Gonzales",
                "confidence_score": "-11.635093"
            }
        ]
    },
    {
        "mention": "nett",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Net (economics)",
                "confidence_score": "4.3096323"
            }
        ]
    },
    {
        "mention": "jibby & co.",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Robinsons & Co.",
                "confidence_score": "-10.038628"
            }
        ]
    },
    {
        "mention": "juicy",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Sweetness",
                "confidence_score": "-7.4782825"
            }
        ]
    },
    {
        "mention": "meaty",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Meaty",
                "confidence_score": "-10.120429"
            }
        ]
    },
    {
        "mention": "the jibby & co.",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Robinsons & Co.",
                "confidence_score": "-9.274943"
            }
        ]
    },
    {
        "mention": "empire shopping gallery",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Empire City Mall",
                "confidence_score": "-9.864724"
            }
        ]
    },
    {
        "mention": "jason chan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jason Chan Chi-san",
                "confidence_score": "4.2236667"
            }
        ]
    },
    {
        "mention": "cheesy",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Cheetos",
                "confidence_score": "-8.672715"
            }
        ]
    },
    {
        "mention": "al dente",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Al dente",
                "confidence_score": "5.1736054"
            }
        ]
    },
    {
        "mention": "publika\u2019s white box",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Bali Post Media Group",
                "confidence_score": "-11.758748"
            }
        ]
    },
    {
        "mention": "kl",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kuala Lumpur",
                "confidence_score": "6.567741"
            }
        ]
    },
    {
        "mention": "kl editor",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Contributing editor",
                "confidence_score": "-8.451207"
            }
        ]
    },
    {
        "mention": "dee may",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Deems May",
                "confidence_score": "-9.396798"
            }
        ]
    },
    {
        "mention": "mia wasikowska",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mia Wasikowska",
                "confidence_score": "12.496398"
            }
        ]
    },
    {
        "mention": "alice",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Alice (Alice's Adventures in Wonderland)",
                "confidence_score": "-5.663775"
            }
        ]
    },
    {
        "mention": "crimson peak",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Crimson Peak",
                "confidence_score": "12.409198"
            }
        ]
    },
    {
        "mention": "edith cushing",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Catherine Chisholm Cushing",
                "confidence_score": "-10.852707"
            }
        ]
    },
    {
        "mention": "tim burton",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Tim Burton",
                "confidence_score": "12.258313"
            }
        ]
    },
    {
        "mention": "edith",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Edith",
                "confidence_score": "0.99857825"
            }
        ]
    },
    {
        "mention": "guillermo del toro",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Guillermo del Toro",
                "confidence_score": "11.722655"
            }
        ]
    },
    {
        "mention": "frankenstein",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Frankenstein",
                "confidence_score": "6.1125445"
            }
        ]
    },
    {
        "mention": "callum greene",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Sonia Greene",
                "confidence_score": "-10.847867"
            }
        ]
    },
    {
        "mention": "mia",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mia Wasikowska",
                "confidence_score": "0.42582998"
            }
        ]
    },
    {
        "mention": "universal pictures",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Universal Pictures",
                "confidence_score": "10.809438"
            }
        ]
    },
    {
        "mention": "academy award\u00ae",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Academy Awards",
                "confidence_score": "6.936744"
            }
        ]
    },
    {
        "mention": "jessica chastain",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jessica Chastain",
                "confidence_score": "11.668368"
            }
        ]
    },
    {
        "mention": "interstellar",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Interstellar (film)",
                "confidence_score": "5.6974053"
            }
        ]
    },
    {
        "mention": "tom hiddleston",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Tom Hiddleston",
                "confidence_score": "11.254449"
            }
        ]
    },
    {
        "mention": "charlie hunnam",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Charlie Hunnam",
                "confidence_score": "8.1086855"
            }
        ]
    },
    {
        "mention": "united international pictures",
        "type": "ORG",
        "predictions": [
            {
                "entity": "United International Pictures",
                "confidence_score": "9.586531"
            }
        ]
    },
    {
        "mention": "columbia pictures",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Columbia Pictures",
                "confidence_score": "11.634309"
            }
        ]
    },
    {
        "mention": "steel dynamics, inc.",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Steel Dynamics",
                "confidence_score": "10.771548"
            }
        ]
    },
    {
        "mention": "thomson reuters",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Thomson Reuters",
                "confidence_score": "9.638311"
            }
        ]
    },
    {
        "mention": "ratings network.com",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Nielsen Media Research",
                "confidence_score": "-5.9906774"
            }
        ]
    },
    {
        "mention": "steel dynamics",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Steel Dynamics",
                "confidence_score": "10.97265"
            }
        ]
    },
    {
        "mention": "bank of america",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Bank of America",
                "confidence_score": "10.363298"
            }
        ]
    },
    {
        "mention": "thestreet",
        "type": "ORG",
        "predictions": [
            {
                "entity": "TheStreet.com",
                "confidence_score": "6.1778307"
            }
        ]
    },
    {
        "mention": "jefferies group",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Jefferies Group",
                "confidence_score": "10.725219"
            }
        ]
    },
    {
        "mention": "zacks cut steel dynamics",
        "type": "ORG",
        "predictions": [
            {
                "entity": "S&Z Metalworks",
                "confidence_score": "-10.891122"
            }
        ]
    },
    {
        "mention": "macquarie",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Macquarie Group",
                "confidence_score": "-3.2178836"
            }
        ]
    },
    {
        "mention": "buy",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Buy\u2013sell agreement",
                "confidence_score": "-6.094321"
            }
        ]
    },
    {
        "mention": "john c. bates",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "John C. Bates",
                "confidence_score": "9.03237"
            }
        ]
    },
    {
        "mention": "the securities & exchange commission",
        "type": "ORG",
        "predictions": [
            {
                "entity": "U.S. Securities and Exchange Commission",
                "confidence_score": "9.095315"
            }
        ]
    },
    {
        "mention": "sec",
        "type": "ORG",
        "predictions": [
            {
                "entity": "U.S. Securities and Exchange Commission",
                "confidence_score": "11.189518"
            }
        ]
    },
    {
        "mention": "company",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Company",
                "confidence_score": "9.122138"
            }
        ]
    },
    {
        "mention": "galvalume",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Leonard Green & Partners",
                "confidence_score": "-11.019694"
            }
        ]
    },
    {
        "mention": "galfan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Galcon",
                "confidence_score": "-11.121753"
            }
        ]
    },
    {
        "mention": "warners bay",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Warners Bay, New South Wales",
                "confidence_score": "0.68160397"
            }
        ]
    },
    {
        "mention": "nsw police",
        "type": "ORG",
        "predictions": [
            {
                "entity": "New South Wales Police Force",
                "confidence_score": "6.0501738"
            }
        ]
    },
    {
        "mention": "postmans lane",
        "type": "ORG",
        "predictions": [
            {
                "entity": "General Holmes Drive",
                "confidence_score": "-10.702782"
            }
        ]
    },
    {
        "mention": "the professional standards command",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Standards Australia",
                "confidence_score": "-8.421635"
            }
        ]
    },
    {
        "mention": "australia post",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australia Post",
                "confidence_score": "11.573691"
            }
        ]
    },
    {
        "mention": "cheryl taylor",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Cheryl Taylor",
                "confidence_score": "9.581741"
            }
        ]
    },
    {
        "mention": "the newcastle herald",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Newcastle Herald",
                "confidence_score": "10.175796"
            }
        ]
    },
    {
        "mention": "dawson\u2019s oysters",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Bentley's Oyster Bar and Grill",
                "confidence_score": "-11.092195"
            }
        ]
    },
    {
        "mention": "king street",
        "type": "ORG",
        "predictions": [
            {
                "entity": "King Street, Sydney",
                "confidence_score": "-3.0228975"
            }
        ]
    },
    {
        "mention": "newcastle herald",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Newcastle Herald",
                "confidence_score": "9.92756"
            }
        ]
    },
    {
        "mention": "the islamic state",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Islamic State of Iraq and the Levant",
                "confidence_score": "-4.269014"
            }
        ]
    },
    {
        "mention": "kashmir",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Jammu and Kashmir",
                "confidence_score": "-2.3491042"
            }
        ]
    },
    {
        "mention": "union home",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Union territory",
                "confidence_score": "-2.4101503"
            }
        ]
    },
    {
        "mention": "rajnath singh",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Rajnath Singh",
                "confidence_score": "11.382765"
            }
        ]
    },
    {
        "mention": "home",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Home",
                "confidence_score": "6.4426885"
            }
        ]
    },
    {
        "mention": "ani",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Asian News International",
                "confidence_score": "5.5771437"
            }
        ]
    },
    {
        "mention": "malir's rifa-i-aam society",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Pakistan",
                "confidence_score": "-2.2003558"
            }
        ]
    },
    {
        "mention": "farhan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Farhan",
                "confidence_score": "0.2850706"
            }
        ]
    },
    {
        "mention": "pehlwan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Pehlwani",
                "confidence_score": "-1.5431557"
            }
        ]
    },
    {
        "mention": "rangers",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Pakistan Rangers",
                "confidence_score": "-0.046780016"
            }
        ]
    },
    {
        "mention": "owen eagles",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Owen Stephens",
                "confidence_score": "-9.763269"
            }
        ]
    },
    {
        "mention": "ski express nz/japan",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Kenichi Yamamoto (skier)",
                "confidence_score": "-8.078839"
            }
        ]
    },
    {
        "mention": "kirra holidays",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "KILROY International",
                "confidence_score": "-11.138485"
            }
        ]
    },
    {
        "mention": "toni portmann",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Toni Ko",
                "confidence_score": "-9.871561"
            }
        ]
    },
    {
        "mention": "dhisco inc.",
        "type": "ORG",
        "predictions": [
            {
                "entity": "DSG International Ltd.",
                "confidence_score": "-11.146859"
            }
        ]
    },
    {
        "mention": "karishma tulsidas",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Karishma Naina Sharma",
                "confidence_score": "-10.62347"
            }
        ]
    },
    {
        "mention": "jewelfest singapore",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Jewel Changi Airport",
                "confidence_score": "-9.446652"
            }
        ]
    },
    {
        "mention": "the singapore jewellery & gem fair",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Singapore Jewellery Design Award",
                "confidence_score": "-9.138576"
            }
        ]
    },
    {
        "mention": "the singapore international jewelry expo",
        "type": "ORG",
        "predictions": [
            {
                "entity": "World Jewellery Confederation",
                "confidence_score": "-9.491719"
            }
        ]
    },
    {
        "mention": "singapore tatler jewels & time",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Straits Times",
                "confidence_score": "-8.812173"
            }
        ]
    },
    {
        "mention": "michael koh",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Michael Kum",
                "confidence_score": "-8.994506"
            }
        ]
    },
    {
        "mention": "caratell",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kimi ni Aitaku Naru Kara",
                "confidence_score": "-11.640695"
            }
        ]
    },
    {
        "mention": "belinda chua",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Paige Chua",
                "confidence_score": "-9.954881"
            }
        ]
    },
    {
        "mention": "flower diamond",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Erica Diamond",
                "confidence_score": "-11.400855"
            }
        ]
    },
    {
        "mention": "belinda",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Belinda",
                "confidence_score": "-5.327825"
            }
        ]
    },
    {
        "mention": "ritu aurora",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Aurora",
                "confidence_score": "-7.579847"
            }
        ]
    },
    {
        "mention": "jaipore jewels",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jewel BEM Hunter Lime",
                "confidence_score": "-9.781619"
            }
        ]
    },
    {
        "mention": "jaipore",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Chandrapore",
                "confidence_score": "-8.100399"
            }
        ]
    },
    {
        "mention": "the patron jeweller",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Pearl Source",
                "confidence_score": "-11.3287"
            }
        ]
    },
    {
        "mention": "the islamic fashion festival",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Islamic fashion",
                "confidence_score": "-8.134654"
            }
        ]
    },
    {
        "mention": "goh shuet-li",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Shueh-li Ong",
                "confidence_score": "-10.876456"
            }
        ]
    },
    {
        "mention": "la putri",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "La (Javanese)",
                "confidence_score": "-11.232955"
            }
        ]
    },
    {
        "mention": "marilyn tan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Vincent Tan",
                "confidence_score": "-9.297695"
            }
        ]
    },
    {
        "mention": "marilyn tan jewellery",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jewellery Design and Management International School",
                "confidence_score": "-9.426469"
            }
        ]
    },
    {
        "mention": "marilyn",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Marilyn Monroe",
                "confidence_score": "-5.549219"
            }
        ]
    },
    {
        "mention": "sara taseer",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Taman Tasik Prima",
                "confidence_score": "-10.51924"
            }
        ]
    },
    {
        "mention": "simone ng",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Simone Singh",
                "confidence_score": "-11.119492"
            }
        ]
    },
    {
        "mention": "nayantara dhillon",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Navneet Kaur Dhillon",
                "confidence_score": "-7.8974285"
            }
        ]
    },
    {
        "mention": "michelle buxani",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Michelle Brooke-Marciniak",
                "confidence_score": "-11.462352"
            }
        ]
    },
    {
        "mention": "temptations",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "The Temptations",
                "confidence_score": "-1.466572"
            }
        ]
    },
    {
        "mention": "vihari poddar",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Abhishek Poddar",
                "confidence_score": "-8.586196"
            }
        ]
    },
    {
        "mention": "vihari jewels",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jewels (mixed martial arts)",
                "confidence_score": "-10.95498"
            }
        ]
    },
    {
        "mention": "gia",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Gia Johnson",
                "confidence_score": "-5.600068"
            }
        ]
    },
    {
        "mention": "jessie foo",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jessie Cave",
                "confidence_score": "-10.932743"
            }
        ]
    },
    {
        "mention": "yuli inc \u2013 fine jewellery",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Xiaomi",
                "confidence_score": "-8.274638"
            }
        ]
    },
    {
        "mention": "jessie",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jessie Chung",
                "confidence_score": "-8.458671"
            }
        ]
    },
    {
        "mention": "jewellery",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jewellery",
                "confidence_score": "8.436512"
            }
        ]
    },
    {
        "mention": "friends of singapore tatler",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Malaysian Insider",
                "confidence_score": "-10.598672"
            }
        ]
    },
    {
        "mention": "flawless",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Flawless (dance troupe)",
                "confidence_score": "-7.6733947"
            }
        ]
    },
    {
        "mention": "bulgari",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bulgari",
                "confidence_score": "8.936749"
            }
        ]
    },
    {
        "mention": "detroit",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Detroit",
                "confidence_score": "9.890772"
            }
        ]
    },
    {
        "mention": "atlantic",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Atlantic Ocean",
                "confidence_score": "4.698329"
            }
        ]
    },
    {
        "mention": "the east coast",
        "type": "LOC",
        "predictions": [
            {
                "entity": "East Coast of the United States",
                "confidence_score": "-2.0340664"
            }
        ]
    },
    {
        "mention": "the great lakes",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Great Lakes",
                "confidence_score": "1.6418152"
            }
        ]
    },
    {
        "mention": "lake huron",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Lake Huron",
                "confidence_score": "9.36548"
            }
        ]
    },
    {
        "mention": "showers",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Shower",
                "confidence_score": "-1.2876487"
            }
        ]
    },
    {
        "mention": "desperately",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Desperately (Don Williams song)",
                "confidence_score": "-11.312842"
            }
        ]
    },
    {
        "mention": "lisa herbold",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Lisa Herbold",
                "confidence_score": "12.056472"
            }
        ]
    },
    {
        "mention": "councilmember licata's office",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Nick Licata (politician)",
                "confidence_score": "-8.631442"
            }
        ]
    },
    {
        "mention": "councilmember sawant's office",
        "type": "ORG",
        "predictions": [
            {
                "entity": "List of Nunavut Legislative Assemblies",
                "confidence_score": "-11.901475"
            }
        ]
    },
    {
        "mention": "dana robinson slote",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "List of census-designated places in North Dakota",
                "confidence_score": "-11.51553"
            }
        ]
    },
    {
        "mention": "council  committee",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Committee",
                "confidence_score": "-1.3599354"
            }
        ]
    },
    {
        "mention": "limiting rent increases",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Brooke Amendment",
                "confidence_score": "-10.612016"
            }
        ]
    },
    {
        "mention": "seattle - council's housing affordability, human services and economic  resiliency committee",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Seattle",
                "confidence_score": "-2.8440382"
            }
        ]
    },
    {
        "mention": "councilmembers",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Councillor",
                "confidence_score": "-2.4868972"
            }
        ]
    },
    {
        "mention": "kshama sawant",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kshama Sawant",
                "confidence_score": "12.244417"
            }
        ]
    },
    {
        "mention": "nick licata",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Nick Licata (politician)",
                "confidence_score": "11.648968"
            }
        ]
    },
    {
        "mention": "washington state",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Washington (state)",
                "confidence_score": "5.5801234"
            }
        ]
    },
    {
        "mention": "councilmembers  sawant",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kshama Sawant",
                "confidence_score": "-2.0385404"
            }
        ]
    },
    {
        "mention": "licata",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Nick Licata (politician)",
                "confidence_score": "5.988609"
            }
        ]
    },
    {
        "mention": "council",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Council",
                "confidence_score": "4.339744"
            }
        ]
    },
    {
        "mention": "councilmembers licata",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Nick Licata (politician)",
                "confidence_score": "0.46282932"
            }
        ]
    },
    {
        "mention": "full council",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Seattle City Council",
                "confidence_score": "-1.451632"
            }
        ]
    },
    {
        "mention": "councilmembers  licata",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Nick Licata (politician)",
                "confidence_score": "0.43720028"
            }
        ]
    },
    {
        "mention": "the council's housing affordability, human services",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Seattle City Council",
                "confidence_score": "-4.708395"
            }
        ]
    },
    {
        "mention": "economic resiliency committee",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Community resilience",
                "confidence_score": "-9.28616"
            }
        ]
    },
    {
        "mention": "second floor",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Stairs",
                "confidence_score": "-7.641441"
            }
        ]
    },
    {
        "mention": "the council\u2019s housing affordability, human services",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Housing and Community Development Act of 1992",
                "confidence_score": "-10.258757"
            }
        ]
    },
    {
        "mention": "arya stark",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Arya Stark",
                "confidence_score": "10.690959"
            }
        ]
    },
    {
        "mention": "maisie williams",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Maisie Williams",
                "confidence_score": "12.085702"
            }
        ]
    },
    {
        "mention": "youtube",
        "type": "ORG",
        "predictions": [
            {
                "entity": "YouTube",
                "confidence_score": "10.731659"
            }
        ]
    },
    {
        "mention": "maisie",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Maisie Williams",
                "confidence_score": "8.9909"
            }
        ]
    },
    {
        "mention": "ms williams",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Venus Williams",
                "confidence_score": "-5.8795123"
            }
        ]
    },
    {
        "mention": "maisie williams/youtube",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Maisie Williams",
                "confidence_score": "4.3699503"
            }
        ]
    },
    {
        "mention": "sophie",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sophie, Countess of Wessex",
                "confidence_score": "-6.304443"
            }
        ]
    },
    {
        "mention": "maisie\u2019s",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Maisie Williams",
                "confidence_score": "-0.7427547"
            }
        ]
    },
    {
        "mention": "@maisie_williams",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Maisie Williams",
                "confidence_score": "-0.8978305"
            }
        ]
    },
    {
        "mention": "mughal",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mughal Empire",
                "confidence_score": "6.9409075"
            }
        ]
    },
    {
        "mention": "the delhi cantonment",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Delhi Cantonment",
                "confidence_score": "0.044629987"
            }
        ]
    },
    {
        "mention": "agra",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Agra",
                "confidence_score": "9.960967"
            }
        ]
    },
    {
        "mention": "jantar mantar",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Jantar Mantar",
                "confidence_score": "9.296149"
            }
        ]
    },
    {
        "mention": "mahabharata",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Mahabharata",
                "confidence_score": "10.583041"
            }
        ]
    },
    {
        "mention": "ka mah",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mah",
                "confidence_score": "-8.022498"
            }
        ]
    },
    {
        "mention": "jhandewalan metro station",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jhandewalan metro station",
                "confidence_score": "8.67785"
            }
        ]
    },
    {
        "mention": "shah tughlaq",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Firuz Shah Tughlaq",
                "confidence_score": "-0.82747626"
            }
        ]
    },
    {
        "mention": "mahal",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Mahal (palace)",
                "confidence_score": "6.642782"
            }
        ]
    },
    {
        "mention": "ka mahal",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Taj Mahal",
                "confidence_score": "-8.960652"
            }
        ]
    },
    {
        "mention": "daily mail\u2019s",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Daily Mail",
                "confidence_score": "11.0001955"
            }
        ]
    },
    {
        "mention": "macmillan",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Harold Macmillan",
                "confidence_score": "-4.0180745"
            }
        ]
    },
    {
        "mention": "stanley crook",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Stanley Crook",
                "confidence_score": "7.563377"
            }
        ]
    },
    {
        "mention": "charlene harker",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Baker Street Babes",
                "confidence_score": "-11.275212"
            }
        ]
    },
    {
        "mention": "beryl",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Beryl",
                "confidence_score": "4.0090036"
            }
        ]
    },
    {
        "mention": "ellie harker",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ellie Darcey-Alden",
                "confidence_score": "-8.318237"
            }
        ]
    },
    {
        "mention": "harry weatherald",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Tim Weatherald",
                "confidence_score": "-10.089182"
            }
        ]
    },
    {
        "mention": "stanley crook primary school",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Stanley Crook",
                "confidence_score": "-7.3152614"
            }
        ]
    },
    {
        "mention": "leon",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Leon Jackson",
                "confidence_score": "-7.4546633"
            }
        ]
    },
    {
        "mention": "luke mason",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Luke Lennox",
                "confidence_score": "-11.141234"
            }
        ]
    },
    {
        "mention": "connie herron",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Connie Martinson",
                "confidence_score": "-11.014575"
            }
        ]
    },
    {
        "mention": "joanne laundy",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Joanne Salley",
                "confidence_score": "-11.110135"
            }
        ]
    },
    {
        "mention": "billy row",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Billy Row",
                "confidence_score": "1.4891739"
            }
        ]
    },
    {
        "mention": "julie proud",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Julie Berry",
                "confidence_score": "-11.103684"
            }
        ]
    },
    {
        "mention": "lisa brewster",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Lisa Joann Thompson",
                "confidence_score": "-11.009149"
            }
        ]
    },
    {
        "mention": "donna thompson",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Donna Simpson (internet celebrity)",
                "confidence_score": "-10.867162"
            }
        ]
    },
    {
        "mention": "leanne harker",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Leanne Battersby",
                "confidence_score": "-9.964695"
            }
        ]
    },
    {
        "mention": "amy howcroft",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Amy Willerton",
                "confidence_score": "-10.411445"
            }
        ]
    },
    {
        "mention": "john thompson",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "John Thompson (sociologist)",
                "confidence_score": "-2.5995758"
            }
        ]
    },
    {
        "mention": "jane herron",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jane Danson",
                "confidence_score": "-7.5946956"
            }
        ]
    },
    {
        "mention": "baltimore orioles",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Baltimore Orioles",
                "confidence_score": "8.931406"
            }
        ]
    },
    {
        "mention": "manny machado",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Manny Machado",
                "confidence_score": "9.632159"
            }
        ]
    },
    {
        "mention": "washington nationals",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Washington Nationals",
                "confidence_score": "8.68574"
            }
        ]
    },
    {
        "mention": "jonathan papelbon",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jonathan Papelbon",
                "confidence_score": "9.717675"
            }
        ]
    },
    {
        "mention": "max scherzer",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Max Scherzer",
                "confidence_score": "9.35327"
            }
        ]
    },
    {
        "mention": "papelbon",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jonathan Papelbon",
                "confidence_score": "-4.295975"
            }
        ]
    },
    {
        "mention": "reuters",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Reuters",
                "confidence_score": "9.007746"
            }
        ]
    },
    {
        "mention": "the u.s. federal reserve",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Federal Reserve",
                "confidence_score": "0.58824587"
            }
        ]
    },
    {
        "mention": "fed",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Federal Reserve",
                "confidence_score": "8.06001"
            }
        ]
    },
    {
        "mention": "csi300",
        "type": "ORG",
        "predictions": [
            {
                "entity": "ISO 9000",
                "confidence_score": "-9.439532"
            }
        ]
    },
    {
        "mention": "gmt",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Greenwich Mean Time",
                "confidence_score": "5.4173503"
            }
        ]
    },
    {
        "mention": "the shanghai composite index",
        "type": "ORG",
        "predictions": [
            {
                "entity": "SSE Composite Index",
                "confidence_score": "-4.061337"
            }
        ]
    },
    {
        "mention": "hang seng",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Hang Seng Bank",
                "confidence_score": "3.9861526"
            }
        ]
    },
    {
        "mention": "the china securities regulatory commission",
        "type": "ORG",
        "predictions": [
            {
                "entity": "China Securities Regulatory Commission",
                "confidence_score": "8.079721"
            }
        ]
    },
    {
        "mention": "samuel shen",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "H.J. Shen",
                "confidence_score": "-10.385303"
            }
        ]
    },
    {
        "mention": "pete sweeney",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Pete Sweeney",
                "confidence_score": "0.26639315"
            }
        ]
    },
    {
        "mention": "fenway",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Fenway Park",
                "confidence_score": "0.040342793"
            }
        ]
    },
    {
        "mention": "red sox",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Boston Red Sox",
                "confidence_score": "0.58450556"
            }
        ]
    },
    {
        "mention": "harvard",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Harvard University",
                "confidence_score": "4.555863"
            }
        ]
    },
    {
        "mention": "harvard yard",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Harvard Yard",
                "confidence_score": "11.096471"
            }
        ]
    },
    {
        "mention": "ucf",
        "type": "ORG",
        "predictions": [
            {
                "entity": "UNCF",
                "confidence_score": "-9.81387"
            }
        ]
    },
    {
        "mention": "curious george",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Curious George",
                "confidence_score": "0.87910205"
            }
        ]
    },
    {
        "mention": "jillian",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jillian",
                "confidence_score": "-8.265801"
            }
        ]
    },
    {
        "mention": "cracker jack",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Cracker Jack",
                "confidence_score": "8.608442"
            }
        ]
    },
    {
        "mention": "take me",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Take Me (film)",
                "confidence_score": "-7.437864"
            }
        ]
    },
    {
        "mention": "ballgame",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ball game",
                "confidence_score": "3.4378371"
            }
        ]
    },
    {
        "mention": "sweet caroline",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Sweet Caroline",
                "confidence_score": "-2.8346088"
            }
        ]
    },
    {
        "mention": "faneuil hall",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Faneuil Hall",
                "confidence_score": "2.4458926"
            }
        ]
    },
    {
        "mention": "the north end",
        "type": "LOC",
        "predictions": [
            {
                "entity": "North End, Boston",
                "confidence_score": "-3.4153123"
            }
        ]
    },
    {
        "mention": "north end",
        "type": "LOC",
        "predictions": [
            {
                "entity": "North End, Boston",
                "confidence_score": "6.2282286"
            }
        ]
    },
    {
        "mention": "hubs",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Airline hub",
                "confidence_score": "-4.1495557"
            }
        ]
    },
    {
        "mention": "mike",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "MIKE",
                "confidence_score": "-8.402079"
            }
        ]
    },
    {
        "mention": "modern",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Modern history",
                "confidence_score": "-5.992842"
            }
        ]
    },
    {
        "mention": "oktoberfest",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Oktoberfest",
                "confidence_score": "11.591672"
            }
        ]
    },
    {
        "mention": "windhoek",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Windhoek",
                "confidence_score": "11.715169"
            }
        ]
    },
    {
        "mention": "namibia breweries",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Namibia Breweries Limited",
                "confidence_score": "2.8188846"
            }
        ]
    },
    {
        "mention": "skw",
        "type": "ORG",
        "predictions": [
            {
                "entity": "SKW",
                "confidence_score": "-10.299478"
            }
        ]
    },
    {
        "mention": "cathy mcgowan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Cathy McGowan (politician)",
                "confidence_score": "9.276536"
            }
        ]
    },
    {
        "mention": "anzac",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Anzac Day",
                "confidence_score": "0.006327114"
            }
        ]
    },
    {
        "mention": "indi cathy mcgowan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Cathy McGowan (politician)",
                "confidence_score": "0.6377952"
            }
        ]
    },
    {
        "mention": "spirit",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Spirit",
                "confidence_score": "4.9421077"
            }
        ]
    },
    {
        "mention": "ms mcgowan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Cathy McGowan (politician)",
                "confidence_score": "-4.9800916"
            }
        ]
    },
    {
        "mention": "indi",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Division of Indi",
                "confidence_score": "4.2555566"
            }
        ]
    },
    {
        "mention": "sophie mirabella",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Sophie Mirabella",
                "confidence_score": "11.693441"
            }
        ]
    },
    {
        "mention": "liberal party",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Liberal Party of Australia",
                "confidence_score": "5.5795703"
            }
        ]
    },
    {
        "mention": "liberal",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Liberal Party of Canada",
                "confidence_score": "-6.08277"
            }
        ]
    },
    {
        "mention": "mrs mirabella",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Sophie Mirabella",
                "confidence_score": "2.030365"
            }
        ]
    },
    {
        "mention": "lou lieberman",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Lou Lieberman",
                "confidence_score": "11.255984"
            }
        ]
    },
    {
        "mention": "brendan nelson",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Brendan Nelson",
                "confidence_score": "10.046449"
            }
        ]
    },
    {
        "mention": "the australian war memorial",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australian War Memorial",
                "confidence_score": "5.2633767"
            }
        ]
    },
    {
        "mention": "wodonga sports",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Wodonga",
                "confidence_score": "-4.5876737"
            }
        ]
    },
    {
        "mention": "leisure centre",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Leisure centre",
                "confidence_score": "5.74202"
            }
        ]
    },
    {
        "mention": "jenny o'connor",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jennifer O'Connor (netball)",
                "confidence_score": "-11.215902"
            }
        ]
    },
    {
        "mention": "ms o'connor",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Frances O'Connor",
                "confidence_score": "-7.157365"
            }
        ]
    },
    {
        "mention": "farrer sussan ley",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sussan Ley",
                "confidence_score": "1.9322057"
            }
        ]
    },
    {
        "mention": "wodonga",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Wodonga",
                "confidence_score": "8.673154"
            }
        ]
    },
    {
        "mention": "rod wangman",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Dennis Rodman",
                "confidence_score": "-10.591179"
            }
        ]
    },
    {
        "mention": "commonwealth bank",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Commonwealth Bank",
                "confidence_score": "12.094094"
            }
        ]
    },
    {
        "mention": "ian narev",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ian Narev",
                "confidence_score": "12.340396"
            }
        ]
    },
    {
        "mention": "andrew penn",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Andrew Penn (cricketer)",
                "confidence_score": "-7.430454"
            }
        ]
    },
    {
        "mention": "tim fischer",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Tim Fischer",
                "confidence_score": "11.266134"
            }
        ]
    },
    {
        "mention": "andy fairweather",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Andy Fairweather Low",
                "confidence_score": "6.1999736"
            }
        ]
    },
    {
        "mention": "welsh",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Welsh language",
                "confidence_score": "6.5837812"
            }
        ]
    },
    {
        "mention": "hello suzy",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Hello (band)",
                "confidence_score": "-10.538263"
            }
        ]
    },
    {
        "mention": "paradise",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Paradise",
                "confidence_score": "1.9678035"
            }
        ]
    },
    {
        "mention": "andy",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Andy Revell",
                "confidence_score": "-7.235394"
            }
        ]
    },
    {
        "mention": "eric clapton",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Eric Clapton",
                "confidence_score": "12.659146"
            }
        ]
    },
    {
        "mention": "george harrison",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "George Harrison",
                "confidence_score": "11.29125"
            }
        ]
    },
    {
        "mention": "bob dylan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bob Dylan",
                "confidence_score": "11.524373"
            }
        ]
    },
    {
        "mention": "jimi hendrix",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jimi Hendrix",
                "confidence_score": "11.506654"
            }
        ]
    },
    {
        "mention": "roger waters",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Roger Waters",
                "confidence_score": "11.70103"
            }
        ]
    },
    {
        "mention": "joe cocker",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Joe Cocker",
                "confidence_score": "11.606954"
            }
        ]
    },
    {
        "mention": "steve winwood",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Steve Winwood",
                "confidence_score": "12.149531"
            }
        ]
    },
    {
        "mention": "rockwell medical",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Rockwell Medical",
                "confidence_score": "11.957061"
            }
        ]
    },
    {
        "mention": "trade-ideas",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Trade idea",
                "confidence_score": "7.7831807"
            }
        ]
    },
    {
        "mention": "rmti",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Indian rupee",
                "confidence_score": "-5.285209"
            }
        ]
    },
    {
        "mention": "rockwell medical, inc.",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Rockwell Medical",
                "confidence_score": "7.2292705"
            }
        ]
    },
    {
        "mention": "jim cramer's",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jim Cramer",
                "confidence_score": "-2.5926697"
            }
        ]
    },
    {
        "mention": "rockwell medical inc",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Rockwell Medical",
                "confidence_score": "9.856086"
            }
        ]
    },
    {
        "mention": "rockwell medical ratings report",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Rockwell Medical",
                "confidence_score": "-8.667266"
            }
        ]
    },
    {
        "mention": "jim cramer",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jim Cramer",
                "confidence_score": "12.371788"
            }
        ]
    },
    {
        "mention": "jack mohr",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Gordon \"Jack\" Mohr",
                "confidence_score": "-0.2828504"
            }
        ]
    },
    {
        "mention": "large-cap",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Market capitalization",
                "confidence_score": "-5.15095"
            }
        ]
    },
    {
        "mention": "cramer    quant ratings",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Nielsen Media Research",
                "confidence_score": "-8.035184"
            }
        ]
    },
    {
        "mention": "david peltier",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "David Peltier",
                "confidence_score": "-1.7603106"
            }
        ]
    },
    {
        "mention": "hold",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Buy and hold",
                "confidence_score": "-6.0823092"
            }
        ]
    },
    {
        "mention": "model portfolio  intra day trade",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mark to model",
                "confidence_score": "-9.335326"
            }
        ]
    },
    {
        "mention": "manufacturer",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Manufacturing",
                "confidence_score": "4.504576"
            }
        ]
    },
    {
        "mention": "evektor-aerotechnik",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Evektor-Aerotechnik",
                "confidence_score": "8.827137"
            }
        ]
    },
    {
        "mention": "lsa",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Light-sport aircraft",
                "confidence_score": "-0.034614634"
            }
        ]
    },
    {
        "mention": "harmony lsa",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Civil Aviation Agency Slovenia",
                "confidence_score": "-11.409558"
            }
        ]
    },
    {
        "mention": "usa faa",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Federal Aviation Administration",
                "confidence_score": "-0.32924747"
            }
        ]
    },
    {
        "mention": "the directorate general of civil aviation of chile",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Directorate General of Civil Aviation (Chile)",
                "confidence_score": "4.4458036"
            }
        ]
    },
    {
        "mention": "harmony\revektor sportstar",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Evektor SportStar",
                "confidence_score": "-4.6328044"
            }
        ]
    },
    {
        "mention": "rtc aircraft",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Radio Technical Commission for Aeronautics",
                "confidence_score": "-9.259475"
            }
        ]
    },
    {
        "mention": "sportstar",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Evektor SportStar",
                "confidence_score": "9.221049"
            }
        ]
    },
    {
        "mention": "rtc",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Russian State Scientific Center for Robotics and Technical Cybernetics",
                "confidence_score": "0.9966831"
            }
        ]
    },
    {
        "mention": "harmony",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Harmony (ISS module)",
                "confidence_score": "-6.422423"
            }
        ]
    },
    {
        "mention": "vortex",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Vortex (satellite)",
                "confidence_score": "-6.224005"
            }
        ]
    },
    {
        "mention": "bwf - industrial felts for the aerospace industry",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Albany International",
                "confidence_score": "-7.4418178"
            }
        ]
    },
    {
        "mention": "the bwf group",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Badminton World Federation",
                "confidence_score": "-8.058932"
            }
        ]
    },
    {
        "mention": "kuerzi avionics - avionics components and services",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Kuki Inc.",
                "confidence_score": "-10.405135"
            }
        ]
    },
    {
        "mention": "kuerzi avionics",
        "type": "ORG",
        "predictions": [
            {
                "entity": "JSC Kuznetsov",
                "confidence_score": "-11.151296"
            }
        ]
    },
    {
        "mention": "ems",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Emergency medical services",
                "confidence_score": "6.5095267"
            }
        ]
    },
    {
        "mention": "vfr",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Visual flight rules",
                "confidence_score": "10.907287"
            }
        ]
    },
    {
        "mention": "propuls aes",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Rolls-Royce Exe",
                "confidence_score": "-11.276854"
            }
        ]
    },
    {
        "mention": "the rotax 912",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Rotax 912",
                "confidence_score": "9.701241"
            }
        ]
    },
    {
        "mention": "elizabeth armstrong moore",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Elizabeth Moore (historian)",
                "confidence_score": "-11.345449"
            }
        ]
    },
    {
        "mention": "newser staff",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Newser",
                "confidence_score": "-3.7418249"
            }
        ]
    },
    {
        "mention": "the\u00a0 ",
        "type": "ORG",
        "predictions": [
            {
                "entity": "University of Colorado Boulder",
                "confidence_score": "-0.4627882"
            }
        ]
    },
    {
        "mention": "university of colorado",
        "type": "ORG",
        "predictions": [
            {
                "entity": "University of Colorado Boulder",
                "confidence_score": "9.624495"
            }
        ]
    },
    {
        "mention": "science translational medicine",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Science Translational Medicine",
                "confidence_score": "6.06595"
            }
        ]
    },
    {
        "mention": "npr",
        "type": "ORG",
        "predictions": [
            {
                "entity": "NPR",
                "confidence_score": "11.77745"
            }
        ]
    },
    {
        "mention": "newser.com",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Newser",
                "confidence_score": "5.7257447"
            }
        ]
    },
    {
        "mention": "frank",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Frank Endacott",
                "confidence_score": "-5.5879807"
            }
        ]
    },
    {
        "mention": "nrl",
        "type": "ORG",
        "predictions": [
            {
                "entity": "National Rugby League",
                "confidence_score": "5.3634133"
            }
        ]
    },
    {
        "mention": "the sydney roosters",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sydney Roosters",
                "confidence_score": "8.519383"
            }
        ]
    },
    {
        "mention": "des hasler's",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Des Hasler",
                "confidence_score": "6.8061156"
            }
        ]
    },
    {
        "mention": "st george illawarra",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "St. George Illawarra Dragons",
                "confidence_score": "5.7417"
            }
        ]
    },
    {
        "mention": "gareth widdop",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Gareth Widdop",
                "confidence_score": "12.057835"
            }
        ]
    },
    {
        "mention": "bronson harrison",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bronson Harrison",
                "confidence_score": "10.848612"
            }
        ]
    },
    {
        "mention": "mal cochrane",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mal Cochrane",
                "confidence_score": "3.6471572"
            }
        ]
    },
    {
        "mention": "sean garlick",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Sean Garlick",
                "confidence_score": "11.185458"
            }
        ]
    },
    {
        "mention": "the roosters at allianz stadium",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sydney Football Stadium",
                "confidence_score": "-3.7779624"
            }
        ]
    },
    {
        "mention": "widdop",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Gareth Widdop",
                "confidence_score": "0.22250405"
            }
        ]
    },
    {
        "mention": "bulldogs",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Canterbury-Bankstown Bulldogs",
                "confidence_score": "-3.9203165"
            }
        ]
    },
    {
        "mention": "nick ghabar",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Khabar, Shahr-e Babak",
                "confidence_score": "-11.81408"
            }
        ]
    },
    {
        "mention": "hasler with prtichard",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Des Hasler",
                "confidence_score": "-10.453496"
            }
        ]
    },
    {
        "mention": "frankie",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Frankie Boyle",
                "confidence_score": "-9.016181"
            }
        ]
    },
    {
        "mention": "james graham",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "James Graham (playwright)",
                "confidence_score": "0.37251776"
            }
        ]
    },
    {
        "mention": "xperia blog",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Xperia Play",
                "confidence_score": "-10.125974"
            }
        ]
    },
    {
        "mention": "xperia z5 compact",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Sony Xperia Z5 Compact",
                "confidence_score": "7.580156"
            }
        ]
    },
    {
        "mention": "the xperia z1",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Sony Xperia Z1",
                "confidence_score": "5.4181767"
            }
        ]
    },
    {
        "mention": "southern ca",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Southern California",
                "confidence_score": "-3.0381234"
            }
        ]
    },
    {
        "mention": "summer",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Summer",
                "confidence_score": "7.8512044"
            }
        ]
    },
    {
        "mention": "the stylish housewife",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Today's Housewife",
                "confidence_score": "-9.949436"
            }
        ]
    },
    {
        "mention": "loading inlinkz",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "LoadingReadyRun",
                "confidence_score": "-10.150963"
            }
        ]
    },
    {
        "mention": "the pleated poppy",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Poppy (entertainer)",
                "confidence_score": "-9.360101"
            }
        ]
    },
    {
        "mention": "mix & match fashion",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Matches Fashion",
                "confidence_score": "-7.451032"
            }
        ]
    },
    {
        "mention": "ali larijani",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ali Larijani",
                "confidence_score": "10.7804575"
            }
        ]
    },
    {
        "mention": "larijani",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Mohammad-Javad Larijani",
                "confidence_score": "5.8656707"
            }
        ]
    },
    {
        "mention": "the u.s. congress",
        "type": "ORG",
        "predictions": [
            {
                "entity": "United States Congress",
                "confidence_score": "3.6461933"
            }
        ]
    },
    {
        "mention": "obama",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Barack Obama",
                "confidence_score": "6.637609"
            }
        ]
    },
    {
        "mention": "state",
        "type": "ORG",
        "predictions": [
            {
                "entity": "U.S. state",
                "confidence_score": "3.0444102"
            }
        ]
    },
    {
        "mention": "john kerry",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "John Kerry",
                "confidence_score": "10.6447115"
            }
        ]
    },
    {
        "mention": "hezbollah",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Hezbollah",
                "confidence_score": "8.130948"
            }
        ]
    },
    {
        "mention": "hassan rouhani",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Hassan Rouhani",
                "confidence_score": "12.025309"
            }
        ]
    },
    {
        "mention": "jason rezaian",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jason Rezaian",
                "confidence_score": "12.271817"
            }
        ]
    },
    {
        "mention": "saddam hussein",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Saddam Hussein",
                "confidence_score": "11.555086"
            }
        ]
    },
    {
        "mention": "ash carter",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ash Carter",
                "confidence_score": "10.3207"
            }
        ]
    },
    {
        "mention": "farsi",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Persian language",
                "confidence_score": "8.880752"
            }
        ]
    },
    {
        "mention": "login",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Login",
                "confidence_score": "11.363369"
            }
        ]
    },
    {
        "mention": "username & password required",
        "type": "ORG",
        "predictions": [
            {
                "entity": "User (computing)",
                "confidence_score": "-5.4775205"
            }
        ]
    },
    {
        "mention": "maruti suzuki india",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Maruti Suzuki",
                "confidence_score": "3.8024914"
            }
        ]
    },
    {
        "mention": "gurgaon",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Gurgaon",
                "confidence_score": "10.069616"
            }
        ]
    },
    {
        "mention": "blue jeans network",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Blue Book Network",
                "confidence_score": "-11.08869"
            }
        ]
    },
    {
        "mention": "microsoft skype/lync",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Skype for Business Server",
                "confidence_score": "-0.31758332"
            }
        ]
    },
    {
        "mention": "zoom",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Zoom (company)",
                "confidence_score": "-2.1759977"
            }
        ]
    },
    {
        "mention": "blue jeans",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Jeans",
                "confidence_score": "2.282659"
            }
        ]
    },
    {
        "mention": "lync, hangouts",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Google Hangouts",
                "confidence_score": "-9.4589815"
            }
        ]
    },
    {
        "mention": "polycom",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Polycom",
                "confidence_score": "11.761879"
            }
        ]
    },
    {
        "mention": "blue jimp",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Blue Byte",
                "confidence_score": "-10.849291"
            }
        ]
    },
    {
        "mention": "facebook, gopro, pandora",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Facebook",
                "confidence_score": "-0.023043904"
            }
        ]
    },
    {
        "mention": "stanford university",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Stanford University",
                "confidence_score": "10.495671"
            }
        ]
    },
    {
        "mention": "primetime",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Primetime (American TV program)",
                "confidence_score": "-4.843532"
            }
        ]
    },
    {
        "mention": "new enterprise associates",
        "type": "ORG",
        "predictions": [
            {
                "entity": "New Enterprise Associates",
                "confidence_score": "9.0593815"
            }
        ]
    },
    {
        "mention": "accel partners",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Accel (venture capital firm)",
                "confidence_score": "9.500475"
            }
        ]
    },
    {
        "mention": "battery ventures",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Battery Ventures",
                "confidence_score": "12.420905"
            }
        ]
    },
    {
        "mention": "glynn capital",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Alden Global Capital",
                "confidence_score": "-11.189538"
            }
        ]
    },
    {
        "mention": "norwest venture partners",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Norwest Venture Partners",
                "confidence_score": "12.630581"
            }
        ]
    },
    {
        "mention": "quadrille capital",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Quadrangle Group",
                "confidence_score": "-9.868115"
            }
        ]
    },
    {
        "mention": "new york yankees",
        "type": "ORG",
        "predictions": [
            {
                "entity": "New York Yankees",
                "confidence_score": "3.79796"
            }
        ]
    },
    {
        "mention": "derek jeter",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Derek Jeter",
                "confidence_score": "12.209455"
            }
        ]
    },
    {
        "mention": "vbprofiles",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Bank of America",
                "confidence_score": "-7.613111"
            }
        ]
    },
    {
        "mention": "antoinette jadaone",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Antoinette Jadaone",
                "confidence_score": "11.791148"
            }
        ]
    },
    {
        "mention": "jojo saguin",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "JoJo (singer)",
                "confidence_score": "-10.462535"
            }
        ]
    },
    {
        "mention": "james reid",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "James Reid (actor)",
                "confidence_score": "7.0040193"
            }
        ]
    },
    {
        "mention": "nadine lustre",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Nadine Lustre",
                "confidence_score": "11.469524"
            }
        ]
    },
    {
        "mention": "reid",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Reid",
                "confidence_score": "-2.6229036"
            }
        ]
    },
    {
        "mention": "lustre",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Nadine Lustre",
                "confidence_score": "9.068319"
            }
        ]
    },
    {
        "mention": "ysabel ortega",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ysabel Ortega",
                "confidence_score": "10.637683"
            }
        ]
    },
    {
        "mention": "abs-cbn",
        "type": "ORG",
        "predictions": [
            {
                "entity": "ABS-CBN",
                "confidence_score": "11.421739"
            }
        ]
    },
    {
        "mention": "the filipino channel",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Filipino Channel",
                "confidence_score": "10.545868"
            }
        ]
    },
    {
        "mention": "clark",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Clark",
                "confidence_score": "-1.728861"
            }
        ]
    },
    {
        "mention": "clark medina",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Clark Middleton",
                "confidence_score": "-11.3064375"
            }
        ]
    },
    {
        "mention": "leah olivar",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Roger Olivares",
                "confidence_score": "-10.071152"
            }
        ]
    },
    {
        "mention": "filipina",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Philippines",
                "confidence_score": "-3.3341453"
            }
        ]
    },
    {
        "mention": "leah",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Leah",
                "confidence_score": "-3.7191734"
            }
        ]
    },
    {
        "mention": "rona",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Rona, Inc.",
                "confidence_score": "-7.7967706"
            }
        ]
    },
    {
        "mention": "sol",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Sol (musician)",
                "confidence_score": "-9.190414"
            }
        ]
    },
    {
        "mention": "jack",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jack (webcomic)",
                "confidence_score": "-8.818204"
            }
        ]
    },
    {
        "mention": "jiggs",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jiggs (chimpanzee)",
                "confidence_score": "-5.848668"
            }
        ]
    },
    {
        "mention": "ofelia",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Typhoon Ofelia",
                "confidence_score": "-6.3685555"
            }
        ]
    },
    {
        "mention": "the immigration and naturalization service",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Immigration and Naturalization Service",
                "confidence_score": "4.455292"
            }
        ]
    },
    {
        "mention": "pinoy tambayan",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Pinoy Weekly",
                "confidence_score": "-8.183962"
            }
        ]
    },
    {
        "mention": "tambayan replay",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Virtual replay",
                "confidence_score": "-10.510716"
            }
        ]
    },
    {
        "mention": "harry burlton",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Harry Bickerton",
                "confidence_score": "-8.912656"
            }
        ]
    },
    {
        "mention": "zoe",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Zoe Williams",
                "confidence_score": "-6.901143"
            }
        ]
    },
    {
        "mention": "soh mui mui",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Tan Chui Mui",
                "confidence_score": "-11.229165"
            }
        ]
    },
    {
        "mention": "hi-lily",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Lily Ho Ngo Yee",
                "confidence_score": "-9.918379"
            }
        ]
    },
    {
        "mention": "soh kau kau",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kao Chin Su-mei",
                "confidence_score": "-11.257447"
            }
        ]
    },
    {
        "mention": "ching lai fong",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ching Cheong",
                "confidence_score": "-10.963683"
            }
        ]
    },
    {
        "mention": "soh mui mui's",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "MUI Group",
                "confidence_score": "-5.5312004"
            }
        ]
    },
    {
        "mention": "jm",
        "type": "ORG",
        "predictions": [
            {
                "entity": "JMJ Records",
                "confidence_score": "-9.975553"
            }
        ]
    },
    {
        "mention": "jms song bird",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "JHawk",
                "confidence_score": "-11.190145"
            }
        ]
    },
    {
        "mention": "circle",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Circle (band)",
                "confidence_score": "-10.755982"
            }
        ]
    },
    {
        "mention": "silent gorilla",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "The Silent Warrior",
                "confidence_score": "-10.633433"
            }
        ]
    },
    {
        "mention": "poker face",
        "type": "ORG",
        "predictions": [
            {
                "entity": "PokerFace",
                "confidence_score": "7.564985"
            }
        ]
    },
    {
        "mention": "warlord",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Warlord",
                "confidence_score": "4.802262"
            }
        ]
    },
    {
        "mention": "messy",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Steve Huffman",
                "confidence_score": "-10.678642"
            }
        ]
    },
    {
        "mention": "mango",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Mango",
                "confidence_score": "3.5163772"
            }
        ]
    },
    {
        "mention": "loud chick",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Football at the 2017 Southeast Asian Games \u2013 Women's team squads",
                "confidence_score": "-11.868988"
            }
        ]
    },
    {
        "mention": "soh miow miow",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ki Su U Ma I (Kiss Your Mind) / S.O.S. (Smile on Smile)",
                "confidence_score": "-10.563547"
            }
        ]
    },
    {
        "mention": "jms",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "JMSN",
                "confidence_score": "-10.70329"
            }
        ]
    },
    {
        "mention": "michael",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Michael Jackson",
                "confidence_score": "-5.9904094"
            }
        ]
    },
    {
        "mention": "charles tan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Charles Chong",
                "confidence_score": "-8.436384"
            }
        ]
    },
    {
        "mention": "tongkat",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Kitkat (comedian)",
                "confidence_score": "-10.508563"
            }
        ]
    },
    {
        "mention": "kirie kimono",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kiko Mizuhara",
                "confidence_score": "-11.392098"
            }
        ]
    },
    {
        "mention": "chin",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Chin people",
                "confidence_score": "-0.3282439"
            }
        ]
    },
    {
        "mention": "styrofoam",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Styrofoam",
                "confidence_score": "8.773667"
            }
        ]
    },
    {
        "mention": "greg inglis",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Greg Inglis",
                "confidence_score": "12.188951"
            }
        ]
    },
    {
        "mention": "rabbitohs",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "South Sydney Rabbitohs",
                "confidence_score": "5.8960776"
            }
        ]
    },
    {
        "mention": "brisbane",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Brisbane",
                "confidence_score": "7.9944043"
            }
        ]
    },
    {
        "mention": "roosters",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Roosters",
                "confidence_score": "-0.4277949"
            }
        ]
    },
    {
        "mention": "michael maguire's",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Michael Maguire (rugby league)",
                "confidence_score": "4.5935664"
            }
        ]
    },
    {
        "mention": "cronulla",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Cronulla-Sutherland Sharks",
                "confidence_score": "0.72116846"
            }
        ]
    },
    {
        "mention": "dylan walker",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Dylan Walker",
                "confidence_score": "4.8928905"
            }
        ]
    },
    {
        "mention": "alex [johnston]",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Alex Johnston (rugby league)",
                "confidence_score": "-6.1222034"
            }
        ]
    },
    {
        "mention": "walker",
        "type": "ORG",
        "predictions": [
            {
                "entity": "2014 Dubai 24 Hour",
                "confidence_score": "-11.598541"
            }
        ]
    },
    {
        "mention": "adam reynolds",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Adam Reynolds",
                "confidence_score": "8.54242"
            }
        ]
    },
    {
        "mention": "souths",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "South Sydney Rabbitohs",
                "confidence_score": "6.1558805"
            }
        ]
    },
    {
        "mention": "george burgess",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "George Burgess (rugby league)",
                "confidence_score": "-0.4564182"
            }
        ]
    },
    {
        "mention": "kane evans",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Kane Evans",
                "confidence_score": "1.5371611"
            }
        ]
    },
    {
        "mention": "glenn stewart's",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Glenn Stewart",
                "confidence_score": "1.8319211"
            }
        ]
    },
    {
        "mention": "the catalans dragons",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Catalans Dragons",
                "confidence_score": "6.651554"
            }
        ]
    },
    {
        "mention": "super league",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Super League",
                "confidence_score": "8.395976"
            }
        ]
    },
    {
        "mention": "burgess",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Tom Burgess (rugby league)",
                "confidence_score": "-4.5346656"
            }
        ]
    },
    {
        "mention": "reynolds",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Josh Reynolds",
                "confidence_score": "-5.951422"
            }
        ]
    },
    {
        "mention": "stewart",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Wayne Stewart",
                "confidence_score": "-9.795752"
            }
        ]
    },
    {
        "mention": "liz raftery",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Raftery",
                "confidence_score": "-9.703352"
            }
        ]
    },
    {
        "mention": "sam smith",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Sam Smith (singer)",
                "confidence_score": "-3.1793518"
            }
        ]
    },
    {
        "mention": "james bond",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "James Bond",
                "confidence_score": "8.675059"
            }
        ]
    },
    {
        "mention": "spectre",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "SPECTRE",
                "confidence_score": "7.600545"
            }
        ]
    },
    {
        "mention": "itunes",
        "type": "ORG",
        "predictions": [
            {
                "entity": "ITunes",
                "confidence_score": "10.766058"
            }
        ]
    },
    {
        "mention": "bond",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Edward Bond",
                "confidence_score": "-4.6062846"
            }
        ]
    },
    {
        "mention": "michael g. wilson",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Michael G. Wilson",
                "confidence_score": "10.987083"
            }
        ]
    },
    {
        "mention": "barbara broccoli",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Barbara Broccoli",
                "confidence_score": "11.989828"
            }
        ]
    },
    {
        "mention": "the lonely hour",
        "type": "LOC",
        "predictions": [
            {
                "entity": "In the Lonely Hour",
                "confidence_score": "-4.666655"
            }
        ]
    },
    {
        "mention": "adele",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Adele",
                "confidence_score": "9.025246"
            }
        ]
    },
    {
        "mention": "golden globe",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "69th Golden Globe Awards",
                "confidence_score": "-6.758411"
            }
        ]
    },
    {
        "mention": "halo trust",
        "type": "ORG",
        "predictions": [
            {
                "entity": "HALO Trust",
                "confidence_score": "6.8039308"
            }
        ]
    },
    {
        "mention": "mozambique",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Mozambique",
                "confidence_score": "4.787083"
            }
        ]
    },
    {
        "mention": "the indian ocean",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Indian Ocean",
                "confidence_score": "1.915093"
            }
        ]
    },
    {
        "mention": "ed stoddard",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "George D. Stoddard",
                "confidence_score": "-10.169612"
            }
        ]
    },
    {
        "mention": "joe brock",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Pete Brock",
                "confidence_score": "-10.515062"
            }
        ]
    },
    {
        "mention": "george brandis",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "George Brandis",
                "confidence_score": "10.065862"
            }
        ]
    },
    {
        "mention": "qc",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Queen's Counsel",
                "confidence_score": "5.5903077"
            }
        ]
    },
    {
        "mention": "brandis",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "George Brandis",
                "confidence_score": "-4.5354633"
            }
        ]
    },
    {
        "mention": "dubbo",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Dubbo",
                "confidence_score": "9.10349"
            }
        ]
    },
    {
        "mention": "mallee",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mallee (Victoria)",
                "confidence_score": "7.8310328"
            }
        ]
    },
    {
        "mention": "alice springs",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Alice Springs",
                "confidence_score": "11.385169"
            }
        ]
    },
    {
        "mention": "the northern territory",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Northern Territory",
                "confidence_score": "8.391966"
            }
        ]
    },
    {
        "mention": "kimberley",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kimberley (Western Australia)",
                "confidence_score": "3.6679995"
            }
        ]
    },
    {
        "mention": "inner melbourne community legal's '",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Community Legal Centre",
                "confidence_score": "-8.219142"
            }
        ]
    },
    {
        "mention": "the royal women's hospital",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Royal Women's Hospital",
                "confidence_score": "4.0549836"
            }
        ]
    },
    {
        "mention": "the productivity commission's",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Productivity Commission",
                "confidence_score": "1.3167756"
            }
        ]
    },
    {
        "mention": "the sydney morning herald",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Sydney Morning Herald",
                "confidence_score": "9.551835"
            }
        ]
    },
    {
        "mention": "the connecticut coalition against domestic violence",
        "type": "ORG",
        "predictions": [
            {
                "entity": "National Coalition Against Domestic Violence",
                "confidence_score": "-9.462857"
            }
        ]
    },
    {
        "mention": "arthur hapgood sr.",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Arthur Doe Jr.",
                "confidence_score": "-12.211156"
            }
        ]
    },
    {
        "mention": "zaniyah calloway",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Lil Mama",
                "confidence_score": "-10.77662"
            }
        ]
    },
    {
        "mention": "hapgood",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Hutchins Hapgood",
                "confidence_score": "-6.6143374"
            }
        ]
    },
    {
        "mention": "marijuana",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Cannabis (drug)",
                "confidence_score": "1.556718"
            }
        ]
    },
    {
        "mention": "pcp",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Phencyclidine",
                "confidence_score": "3.990393"
            }
        ]
    },
    {
        "mention": "gncc",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Grand National Curling Club",
                "confidence_score": "-4.1400294"
            }
        ]
    },
    {
        "mention": "skyshot",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Skywriting",
                "confidence_score": "-9.857505"
            }
        ]
    },
    {
        "mention": "michael swift",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Michael Swift (ice hockey)",
                "confidence_score": "-8.653503"
            }
        ]
    },
    {
        "mention": "kevin trantham",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Roy Trantham",
                "confidence_score": "-11.33381"
            }
        ]
    },
    {
        "mention": "maxxis tires",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Max Records",
                "confidence_score": "-11.328092"
            }
        ]
    },
    {
        "mention": "trantham",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Roy Trantham",
                "confidence_score": "-8.62604"
            }
        ]
    },
    {
        "mention": "polaris scrambler 850xp",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Australis-OSCAR 5",
                "confidence_score": "-11.665299"
            }
        ]
    },
    {
        "mention": "don higbee",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Scott Donohue",
                "confidence_score": "-11.231226"
            }
        ]
    },
    {
        "mention": "graham widdicombe",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Angus Widdicombe",
                "confidence_score": "-11.699879"
            }
        ]
    },
    {
        "mention": "polaris scrambler 850xp\u2019s",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Antarctic Ice Marathon & 100k ultra race",
                "confidence_score": "-11.818307"
            }
        ]
    },
    {
        "mention": "scrambler xp1000",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ski jumping at the 2018 Winter Olympics \u2013 Men's large hill team",
                "confidence_score": "-12.131634"
            }
        ]
    },
    {
        "mention": "jeff wright",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jeff Wright (politician)",
                "confidence_score": "-8.894492"
            }
        ]
    },
    {
        "mention": "swift",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Swift Aircraft",
                "confidence_score": "-3.6940126"
            }
        ]
    },
    {
        "mention": "polaris scrambler xp850",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "1:10 radio-controlled off-road buggy",
                "confidence_score": "-11.956845"
            }
        ]
    },
    {
        "mention": "team uxc\u2019s",
        "type": "ORG",
        "predictions": [
            {
                "entity": "UXC",
                "confidence_score": "-6.770965"
            }
        ]
    },
    {
        "mention": "the customaxis shocks",
        "type": "ORG",
        "predictions": [
            {
                "entity": "List of Super Powers minicomics",
                "confidence_score": "-11.216742"
            }
        ]
    },
    {
        "mention": "scrambler",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Scrambler (comics)",
                "confidence_score": "3.3512099"
            }
        ]
    },
    {
        "mention": "maxxis",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Max Collins (musician)",
                "confidence_score": "-11.073919"
            }
        ]
    },
    {
        "mention": "the dp brakes",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Brakes",
                "confidence_score": "-8.312257"
            }
        ]
    },
    {
        "mention": "somerset",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Somerset",
                "confidence_score": "1.4524075"
            }
        ]
    },
    {
        "mention": "polaris",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Polaris Industries",
                "confidence_score": "2.0648048"
            }
        ]
    },
    {
        "mention": "driven powersports",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Power Challenge",
                "confidence_score": "-11.021035"
            }
        ]
    },
    {
        "mention": "fly racing",
        "type": "ORG",
        "predictions": [
            {
                "entity": "FlyQuest",
                "confidence_score": "-10.329416"
            }
        ]
    },
    {
        "mention": "evans coolant",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Evans Tuning",
                "confidence_score": "-10.281847"
            }
        ]
    },
    {
        "mention": "rox speed fx",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Roxor Games",
                "confidence_score": "-11.313931"
            }
        ]
    },
    {
        "mention": "dirtworks, cool-it thermo tech",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Dirt 3",
                "confidence_score": "-11.3040495"
            }
        ]
    },
    {
        "mention": "leatt",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Leatt-Brace",
                "confidence_score": "-7.5999446"
            }
        ]
    },
    {
        "mention": "ricochet off road",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Evo Street Racers",
                "confidence_score": "-10.328282"
            }
        ]
    },
    {
        "mention": "spider",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Daniel Quirk",
                "confidence_score": "-5.498012"
            }
        ]
    },
    {
        "mention": "moto pro training",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "MotoAmerica",
                "confidence_score": "-9.352872"
            }
        ]
    },
    {
        "mention": "scott goggles",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Scott Sports",
                "confidence_score": "-10.074452"
            }
        ]
    },
    {
        "mention": "k&n kilters, america\u2019s",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Johnson & Johnson",
                "confidence_score": "-7.301103"
            }
        ]
    },
    {
        "mention": "rigid industries",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Bentley Systems",
                "confidence_score": "-11.13816"
            }
        ]
    },
    {
        "mention": "holz racing products",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Speedway Motorsports",
                "confidence_score": "-9.86752"
            }
        ]
    },
    {
        "mention": "teixeira tech",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Tektronix",
                "confidence_score": "-10.9082365"
            }
        ]
    },
    {
        "mention": "quad tec",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Quad (unit)",
                "confidence_score": "-8.097997"
            }
        ]
    },
    {
        "mention": "alco cleaner",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Alcoa",
                "confidence_score": "-8.507862"
            }
        ]
    },
    {
        "mention": "polaris power generators",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Polaris Industries",
                "confidence_score": "-8.722072"
            }
        ]
    },
    {
        "mention": "omf wheels",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "OM Group",
                "confidence_score": "-10.415758"
            }
        ]
    },
    {
        "mention": "muzzy performance exhaust",
        "type": "ORG",
        "predictions": [
            {
                "entity": "HMF Engineering",
                "confidence_score": "-9.097611"
            }
        ]
    },
    {
        "mention": "polaris engineered lubricants",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Polaris Industries",
                "confidence_score": "-9.29434"
            }
        ]
    },
    {
        "mention": "rcv",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Reactivity controlled compression ignition",
                "confidence_score": "-11.202815"
            }
        ]
    },
    {
        "mention": "airdam clutching, santor design company",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Jacobs Vehicle Systems",
                "confidence_score": "-11.418008"
            }
        ]
    },
    {
        "mention": "quad tech",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Quad 4 engine",
                "confidence_score": "-10.777415"
            }
        ]
    },
    {
        "mention": "alex trebek",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Alex Trebek",
                "confidence_score": "12.358435"
            }
        ]
    },
    {
        "mention": "turd ferguson",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "List of characters in Call of Duty 6: Modern Warfare 2",
                "confidence_score": "-11.3215"
            }
        ]
    },
    {
        "mention": "mara montalbano",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Rodriguez, Rizal",
                "confidence_score": "-7.7385907"
            }
        ]
    },
    {
        "mention": "@maramontalbano",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Adam Montoya",
                "confidence_score": "-11.358433"
            }
        ]
    },
    {
        "mention": "katherine rees",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Katie Rees",
                "confidence_score": "2.9151437"
            }
        ]
    },
    {
        "mention": "kevin \"dj kulprit\" sims",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kevin Federline",
                "confidence_score": "-10.7784605"
            }
        ]
    },
    {
        "mention": "miss nevada usa",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Miss Nevada USA",
                "confidence_score": "10.774248"
            }
        ]
    },
    {
        "mention": "sims",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Sims",
                "confidence_score": "-0.3109095"
            }
        ]
    },
    {
        "mention": "east pebble road",
        "type": "LOC",
        "predictions": [
            {
                "entity": "List of county routes in Elko County, Nevada",
                "confidence_score": "-11.608787"
            }
        ]
    },
    {
        "mention": "xanax",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Alprazolam",
                "confidence_score": "0.63187575"
            }
        ]
    },
    {
        "mention": "oxycontin",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Oxycodone",
                "confidence_score": "4.1961436"
            }
        ]
    },
    {
        "mention": "rees",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Dave Rees",
                "confidence_score": "-7.468818"
            }
        ]
    },
    {
        "mention": "warren j. geller",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kenneth Geller",
                "confidence_score": "-8.536869"
            }
        ]
    },
    {
        "mention": "the sixth circuit",
        "type": "ORG",
        "predictions": [
            {
                "entity": "United States Court of Appeals for the Sixth Circuit",
                "confidence_score": "-5.0467796"
            }
        ]
    },
    {
        "mention": "icp",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Interpol",
                "confidence_score": "-5.059827"
            }
        ]
    },
    {
        "mention": "juggalos",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Juggalo",
                "confidence_score": "1.7451768"
            }
        ]
    },
    {
        "mention": "2011 national gang threat assessment",
        "type": "ORG",
        "predictions": [
            {
                "entity": "National Gang Intelligence Center",
                "confidence_score": "-10.7383585"
            }
        ]
    },
    {
        "mention": "nbc news",
        "type": "ORG",
        "predictions": [
            {
                "entity": "NBC News",
                "confidence_score": "10.77444"
            }
        ]
    },
    {
        "mention": "violent j",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Violent J",
                "confidence_score": "11.863929"
            }
        ]
    },
    {
        "mention": "forab",
        "type": "ORG",
        "predictions": [
            {
                "entity": "List of living former members of the United States House of Representatives (D)",
                "confidence_score": "-12.532015"
            }
        ]
    },
    {
        "mention": "supreme court",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Supreme court",
                "confidence_score": "7.4486923"
            }
        ]
    },
    {
        "mention": "doj",
        "type": "ORG",
        "predictions": [
            {
                "entity": "United States Department of Justice",
                "confidence_score": "10.679116"
            }
        ]
    },
    {
        "mention": "juggalo",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Juggalo",
                "confidence_score": "6.203214"
            }
        ]
    },
    {
        "mention": "army",
        "type": "ORG",
        "predictions": [
            {
                "entity": "United States Army",
                "confidence_score": "6.5852256"
            }
        ]
    },
    {
        "mention": "shaggy 2 dope",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Shaggy 2 Dope",
                "confidence_score": "12.713063"
            }
        ]
    },
    {
        "mention": "aclu",
        "type": "ORG",
        "predictions": [
            {
                "entity": "American Civil Liberties Union",
                "confidence_score": "6.4272265"
            }
        ]
    },
    {
        "mention": "pioneer energy services",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Pioneer Natural Resources",
                "confidence_score": "-4.371561"
            }
        ]
    },
    {
        "mention": "the securities and exchange commission",
        "type": "ORG",
        "predictions": [
            {
                "entity": "U.S. Securities and Exchange Commission",
                "confidence_score": "3.6327417"
            }
        ]
    },
    {
        "mention": "aquib javed",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Aaqib Javed",
                "confidence_score": "-7.096915"
            }
        ]
    },
    {
        "mention": "mohammad aamir",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Aamir Khan",
                "confidence_score": "0.44650868"
            }
        ]
    },
    {
        "mention": "salman butt",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Salman Butt",
                "confidence_score": "8.982509"
            }
        ]
    },
    {
        "mention": "javed",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mohammad Javed",
                "confidence_score": "-5.6946874"
            }
        ]
    },
    {
        "mention": "mohammad azharuddin",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mohammad Azharuddin",
                "confidence_score": "9.942309"
            }
        ]
    },
    {
        "mention": "saleem malik",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Saleem Malik",
                "confidence_score": "10.528774"
            }
        ]
    },
    {
        "mention": "ipl",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Indian Premier League",
                "confidence_score": "1.0439676"
            }
        ]
    },
    {
        "mention": "wwe",
        "type": "ORG",
        "predictions": [
            {
                "entity": "WWE",
                "confidence_score": "10.244563"
            }
        ]
    },
    {
        "mention": "aamir",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Aamir Khan",
                "confidence_score": "4.9611077"
            }
        ]
    },
    {
        "mention": "sachin tendulkar",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Sachin Tendulkar",
                "confidence_score": "10.014631"
            }
        ]
    },
    {
        "mention": "younis khan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Younis Khan",
                "confidence_score": "11.047756"
            }
        ]
    },
    {
        "mention": "to bharat sunderesan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bharat Desai",
                "confidence_score": "-11.397184"
            }
        ]
    },
    {
        "mention": "leeds",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Leeds",
                "confidence_score": "7.7471566"
            }
        ]
    },
    {
        "mention": "guiseley school",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Guiseley School",
                "confidence_score": "1.6135678"
            }
        ]
    },
    {
        "mention": "paul clayton",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Paul Clayton (actor)",
                "confidence_score": "-5.56202"
            }
        ]
    },
    {
        "mention": "the red cross",
        "type": "ORG",
        "predictions": [
            {
                "entity": "International Red Cross and Red Crescent Movement",
                "confidence_score": "-1.5281973"
            }
        ]
    },
    {
        "mention": "twitter",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Twitter",
                "confidence_score": "9.86045"
            }
        ]
    },
    {
        "mention": "all south group",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Allman Brothers Band",
                "confidence_score": "-11.1800995"
            }
        ]
    },
    {
        "mention": "adam whitman",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Adam Yauch",
                "confidence_score": "-11.558134"
            }
        ]
    },
    {
        "mention": "sparta live",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Sparta (band)",
                "confidence_score": "-8.174508"
            }
        ]
    },
    {
        "mention": "dhanna resort",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Dubai Parks and Resorts",
                "confidence_score": "-8.102038"
            }
        ]
    },
    {
        "mention": "lafayette gourmet food festival",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Gourmet",
                "confidence_score": "-6.428368"
            }
        ]
    },
    {
        "mention": "underwater zoo",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Marine mammal park",
                "confidence_score": "-7.877282"
            }
        ]
    },
    {
        "mention": "dulce ortega",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Delfy de Ortega",
                "confidence_score": "-11.483848"
            }
        ]
    },
    {
        "mention": "officers club",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Military officers' club",
                "confidence_score": "-2.5156572"
            }
        ]
    },
    {
        "mention": "mea",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Minimum en route altitude",
                "confidence_score": "-3.4562898"
            }
        ]
    },
    {
        "mention": "intercontinental\u00ae hotels & resorts redefines luxury",
        "type": "ORG",
        "predictions": [
            {
                "entity": "InterContinental",
                "confidence_score": "2.4810255"
            }
        ]
    },
    {
        "mention": "club intercontinental",
        "type": "ORG",
        "predictions": [
            {
                "entity": "InterContinental",
                "confidence_score": "-6.314486"
            }
        ]
    },
    {
        "mention": "turnout",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Voter turnout",
                "confidence_score": "-1.7916319"
            }
        ]
    },
    {
        "mention": "yas mall",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Yas Island",
                "confidence_score": "-10.15635"
            }
        ]
    },
    {
        "mention": "al jimi mall",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Al Jimi Mall",
                "confidence_score": "10.139295"
            }
        ]
    },
    {
        "mention": "mercedes",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Mercedes-Benz",
                "confidence_score": "-2.388546"
            }
        ]
    },
    {
        "mention": "waldorf astoria",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Waldorf Astoria Hotels & Resorts",
                "confidence_score": "-1.6795684"
            }
        ]
    },
    {
        "mention": "jason atherton",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jason Atherton",
                "confidence_score": "9.923142"
            }
        ]
    },
    {
        "mention": "bushman",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bushman (reggae singer)",
                "confidence_score": "-10.298779"
            }
        ]
    },
    {
        "mention": "momcinchanz show",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Tech Toyz",
                "confidence_score": "-11.891408"
            }
        ]
    },
    {
        "mention": "healthcare",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Health care",
                "confidence_score": "7.8033137"
            }
        ]
    },
    {
        "mention": "raghu das",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Raghubar Das",
                "confidence_score": "8.536983"
            }
        ]
    },
    {
        "mention": "rick diamond/getty",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Getty family",
                "confidence_score": "-7.638806"
            }
        ]
    },
    {
        "mention": "eric",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jessie James Decker",
                "confidence_score": "7.4153786"
            }
        ]
    },
    {
        "mention": "jessie james decker \u2018s",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jessie James Decker",
                "confidence_score": "11.179743"
            }
        ]
    },
    {
        "mention": "vivianne rose",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Arden Rose",
                "confidence_score": "-11.2634"
            }
        ]
    },
    {
        "mention": "vivi",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Vivi (magazine)",
                "confidence_score": "-4.594382"
            }
        ]
    },
    {
        "mention": "james decker",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jessie James Decker",
                "confidence_score": "3.1656404"
            }
        ]
    },
    {
        "mention": "vivianne",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Vivianne Crowley",
                "confidence_score": "-4.522098"
            }
        ]
    },
    {
        "mention": "jessie james decker",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jessie James Decker",
                "confidence_score": "11.4922085"
            }
        ]
    },
    {
        "mention": "@jessiejamesdecker",
        "type": "ORG",
        "predictions": [
            {
                "entity": "American Ninja Warrior (season 11)",
                "confidence_score": "-11.792045"
            }
        ]
    },
    {
        "mention": "eric decker",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Eric Decker",
                "confidence_score": "5.6499677"
            }
        ]
    },
    {
        "mention": "anya leon",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Anya Jenkins",
                "confidence_score": "-9.634597"
            }
        ]
    },
    {
        "mention": "prweb",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Population Reference Bureau",
                "confidence_score": "-7.6139197"
            }
        ]
    },
    {
        "mention": "the american cancer society",
        "type": "ORG",
        "predictions": [
            {
                "entity": "American Cancer Society",
                "confidence_score": "10.193181"
            }
        ]
    },
    {
        "mention": "michael lazar",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Michael Lazarou",
                "confidence_score": "-8.945752"
            }
        ]
    },
    {
        "mention": "hifu",
        "type": "ORG",
        "predictions": [
            {
                "entity": "High-intensity focused ultrasound",
                "confidence_score": "-0.70003664"
            }
        ]
    },
    {
        "mention": "psa",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Photographic Society of America",
                "confidence_score": "0.73007566"
            }
        ]
    },
    {
        "mention": "fda",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Food and Drug Administration",
                "confidence_score": "9.530396"
            }
        ]
    },
    {
        "mention": "lazar",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Seymour Lazar",
                "confidence_score": "-8.251321"
            }
        ]
    },
    {
        "mention": "santa rosa",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Santa Rosa, California",
                "confidence_score": "10.077419"
            }
        ]
    },
    {
        "mention": "hazmat",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Hazmat suit",
                "confidence_score": "-6.1803145"
            }
        ]
    },
    {
        "mention": "sauvignon blanc",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sauvignon blanc",
                "confidence_score": "9.249314"
            }
        ]
    },
    {
        "mention": "craigy",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Craigkelly transmitting station",
                "confidence_score": "-6.8451667"
            }
        ]
    },
    {
        "mention": "mills & boon",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Mills & Boon",
                "confidence_score": "10.739887"
            }
        ]
    },
    {
        "mention": "jack reacher",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jack Reacher",
                "confidence_score": "8.516049"
            }
        ]
    },
    {
        "mention": "wiine",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Wii",
                "confidence_score": "-5.0936704"
            }
        ]
    },
    {
        "mention": "insanity challenge",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Insanity Fight",
                "confidence_score": "-11.027833"
            }
        ]
    },
    {
        "mention": "street fighter",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Street Fighter",
                "confidence_score": "3.974343"
            }
        ]
    },
    {
        "mention": "mummy",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mummy",
                "confidence_score": "4.888062"
            }
        ]
    },
    {
        "mention": "netflix",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Netflix",
                "confidence_score": "11.158345"
            }
        ]
    },
    {
        "mention": "captain america",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Captain America",
                "confidence_score": "-0.21371786"
            }
        ]
    },
    {
        "mention": "dublin",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Dublin",
                "confidence_score": "7.3117948"
            }
        ]
    },
    {
        "mention": "research and markets",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Research and development",
                "confidence_score": "-7.857689"
            }
        ]
    },
    {
        "mention": "the global molecular diagnostics",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Molecular diagnostics",
                "confidence_score": "-5.3375115"
            }
        ]
    },
    {
        "mention": "the securities exchange commission",
        "type": "ORG",
        "predictions": [
            {
                "entity": "U.S. Securities and Exchange Commission",
                "confidence_score": "2.205327"
            }
        ]
    },
    {
        "mention": "media contact:",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Media relations",
                "confidence_score": "-7.3452377"
            }
        ]
    },
    {
        "mention": "laura wood",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Laura Woods",
                "confidence_score": "0.2341895"
            }
        ]
    },
    {
        "mention": "the thomas edison",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Thomas Edison",
                "confidence_score": "-1.5791097"
            }
        ]
    },
    {
        "mention": "henry ford",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Henry Ford",
                "confidence_score": "7.3060913"
            }
        ]
    },
    {
        "mention": "silicon valley",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Silicon Valley",
                "confidence_score": "7.6953807"
            }
        ]
    },
    {
        "mention": "robert noyce",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Robert Noyce",
                "confidence_score": "12.682679"
            }
        ]
    },
    {
        "mention": "leslie berlin",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Leslie Berlin",
                "confidence_score": "12.722016"
            }
        ]
    },
    {
        "mention": "the silicon valley archives",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Hoover Institution Library and Archives",
                "confidence_score": "-6.5420394"
            }
        ]
    },
    {
        "mention": "jeff frick",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Stephen Frick",
                "confidence_score": "-10.256689"
            }
        ]
    },
    {
        "mention": "intel",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Intel",
                "confidence_score": "11.424021"
            }
        ]
    },
    {
        "mention": "the invention of silicon valley",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Silicon Valley",
                "confidence_score": "-5.2629905"
            }
        ]
    },
    {
        "mention": "noyce",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Robert Noyce",
                "confidence_score": "7.0931554"
            }
        ]
    },
    {
        "mention": "northern california",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Northern California",
                "confidence_score": "9.65029"
            }
        ]
    },
    {
        "mention": "fairchild semiconductor international, inc.",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Fairchild Semiconductor",
                "confidence_score": "9.180472"
            }
        ]
    },
    {
        "mention": "intel corp.",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Intel",
                "confidence_score": "10.321364"
            }
        ]
    },
    {
        "mention": "cloud",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Cloud",
                "confidence_score": "3.8923333"
            }
        ]
    },
    {
        "mention": "marlene den bleyker marlene",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Blokwatch",
                "confidence_score": "-11.48195"
            }
        ]
    },
    {
        "mention": "siliconangle media's",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Digital media",
                "confidence_score": "-9.741878"
            }
        ]
    },
    {
        "mention": "marlene den bleyker",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Walter De Brouwer",
                "confidence_score": "-11.514417"
            }
        ]
    },
    {
        "mention": "apple",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Apple Inc.",
                "confidence_score": "4.0746527"
            }
        ]
    },
    {
        "mention": "the siliconangle newsletter",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Newsletter",
                "confidence_score": "0.29155672"
            }
        ]
    },
    {
        "mention": "kristofer martin",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kristofer Martin",
                "confidence_score": "11.6727295"
            }
        ]
    },
    {
        "mention": "joyce ching",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Joyce Ching",
                "confidence_score": "11.941098"
            }
        ]
    },
    {
        "mention": "mickey ferriols",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mickey Ferriols",
                "confidence_score": "12.517222"
            }
        ]
    },
    {
        "mention": "kailan ba tama ang mali",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kailan Ba Tama ang Mali?",
                "confidence_score": "5.7923036"
            }
        ]
    },
    {
        "mention": "gma pinoy tv",
        "type": "ORG",
        "predictions": [
            {
                "entity": "GMA Pinoy TV",
                "confidence_score": "11.675157"
            }
        ]
    },
    {
        "mention": "the executive remuneration working group",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Efficiency and Reform Group",
                "confidence_score": "-11.074065"
            }
        ]
    },
    {
        "mention": "the investment association",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Investment Association",
                "confidence_score": "3.7977734"
            }
        ]
    },
    {
        "mention": "daniel godfrey",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Dan Godfrey",
                "confidence_score": "-1.7300365"
            }
        ]
    },
    {
        "mention": "nigel wilson",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Nigel Wilson (businessman)",
                "confidence_score": "12.405133"
            }
        ]
    },
    {
        "mention": "legal & general",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Legal & General",
                "confidence_score": "12.504683"
            }
        ]
    },
    {
        "mention": "ftse",
        "type": "ORG",
        "predictions": [
            {
                "entity": "FTSE Group",
                "confidence_score": "-0.10674975"
            }
        ]
    },
    {
        "mention": "david tyler",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "David Tyler (businessman)",
                "confidence_score": "9.496988"
            }
        ]
    },
    {
        "mention": "sainsbury",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sainsbury's",
                "confidence_score": "5.6088643"
            }
        ]
    },
    {
        "mention": "burberry",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Burberry",
                "confidence_score": "10.763467"
            }
        ]
    },
    {
        "mention": "helena morrissey",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Helena Morrissey",
                "confidence_score": "12.4708185"
            }
        ]
    },
    {
        "mention": "bny mellon's",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Bank of New York Mellon",
                "confidence_score": "2.5827045"
            }
        ]
    },
    {
        "mention": "newton investment management the investment association's",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Newton Investment Management",
                "confidence_score": "-0.90514743"
            }
        ]
    },
    {
        "mention": "edi truell",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "ICG Enterprise Trust",
                "confidence_score": "-10.817267"
            }
        ]
    },
    {
        "mention": "boris johnson",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Boris Johnson",
                "confidence_score": "11.597151"
            }
        ]
    },
    {
        "mention": "russell king",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Russell King (fraudster)",
                "confidence_score": "1.6295776"
            }
        ]
    },
    {
        "mention": "aggreko and spectris",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Spectris",
                "confidence_score": "-1.3166323"
            }
        ]
    },
    {
        "mention": "self-improvement month",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Self-help",
                "confidence_score": "-4.678049"
            }
        ]
    },
    {
        "mention": "ny",
        "type": "ORG",
        "predictions": [
            {
                "entity": "New York City",
                "confidence_score": "1.142284"
            }
        ]
    },
    {
        "mention": "john lanzetta",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "John de Lancie",
                "confidence_score": "-10.304204"
            }
        ]
    },
    {
        "mention": "chris chondrogiannis",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Giannis Chondrogiannis",
                "confidence_score": "-11.004365"
            }
        ]
    },
    {
        "mention": "alyssa gursky",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Yuri Gurski",
                "confidence_score": "-11.014455"
            }
        ]
    },
    {
        "mention": "matthew neary",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Matthew Causey",
                "confidence_score": "-10.513783"
            }
        ]
    },
    {
        "mention": "lanzetta",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "James J. Lanzetta",
                "confidence_score": "-3.904218"
            }
        ]
    },
    {
        "mention": "chondrogiannis",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Chondroyannos",
                "confidence_score": "-0.65183055"
            }
        ]
    },
    {
        "mention": "gursky",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Samuel Gursky",
                "confidence_score": "-2.0000727"
            }
        ]
    },
    {
        "mention": "neary",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Shay Neary",
                "confidence_score": "-6.900327"
            }
        ]
    },
    {
        "mention": "drs",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Physician",
                "confidence_score": "-1.9132121"
            }
        ]
    },
    {
        "mention": "tampa bay",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Tampa Bay",
                "confidence_score": "7.647836"
            }
        ]
    },
    {
        "mention": "smoke",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Smoke",
                "confidence_score": "7.133982"
            }
        ]
    },
    {
        "mention": "poutinerie",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Smoke's Poutinerie",
                "confidence_score": "-5.135813"
            }
        ]
    },
    {
        "mention": "amalie arena",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Amalie Arena",
                "confidence_score": "6.3789687"
            }
        ]
    },
    {
        "mention": "ryan smolkin",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ryan Flinn (ice hockey)",
                "confidence_score": "-12.153287"
            }
        ]
    },
    {
        "mention": "non-league",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Non-League football",
                "confidence_score": "-3.3578646"
            }
        ]
    },
    {
        "mention": "cornwall",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Cornwall",
                "confidence_score": "0.06247021"
            }
        ]
    },
    {
        "mention": "monroe-woodbury",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Monroe County, New York",
                "confidence_score": "-8.621379"
            }
        ]
    },
    {
        "mention": "nash",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Nash County, North Carolina",
                "confidence_score": "-7.8910613"
            }
        ]
    },
    {
        "mention": "erica bachiller",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Erica Bougard",
                "confidence_score": "-11.63207"
            }
        ]
    },
    {
        "mention": "mackenzie tuttle",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bill Tuttle (swimmer)",
                "confidence_score": "-11.621792"
            }
        ]
    },
    {
        "mention": "allison dubick",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Allison Jones (athlete)",
                "confidence_score": "-11.291356"
            }
        ]
    },
    {
        "mention": "samantha reichman",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Samantha Futerman",
                "confidence_score": "-11.352302"
            }
        ]
    },
    {
        "mention": "lauren beneat",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Candice Bennatt",
                "confidence_score": "-11.597295"
            }
        ]
    },
    {
        "mention": "ashley ayala",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ashley Greene",
                "confidence_score": "-11.2443695"
            }
        ]
    },
    {
        "mention": "bachiller, beneat",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Chris Bachelder",
                "confidence_score": "-11.86659"
            }
        ]
    },
    {
        "mention": "christman, tuttle",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Christ Church USA",
                "confidence_score": "-10.486736"
            }
        ]
    },
    {
        "mention": "jessica ebner",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jessica Jerome",
                "confidence_score": "-11.609344"
            }
        ]
    },
    {
        "mention": "leishman",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Thomas Leishman",
                "confidence_score": "-6.654786"
            }
        ]
    },
    {
        "mention": "bush",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "George W. Bush",
                "confidence_score": "-4.3197527"
            }
        ]
    },
    {
        "mention": "lapinski",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Lisa Lapinski",
                "confidence_score": "-7.2525544"
            }
        ]
    },
    {
        "mention": "pfeifer",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Scott Pfeifer",
                "confidence_score": "-6.657193"
            }
        ]
    },
    {
        "mention": "alexander",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Alexander Hill (rower)",
                "confidence_score": "-11.330859"
            }
        ]
    },
    {
        "mention": "corinne debeauvernet",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Jason De Carteret",
                "confidence_score": "-11.632197"
            }
        ]
    },
    {
        "mention": "pb",
        "type": "ORG",
        "predictions": [
            {
                "entity": "P",
                "confidence_score": "-10.254656"
            }
        ]
    },
    {
        "mention": "rachel shih",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Rachel Komisarz",
                "confidence_score": "-11.299937"
            }
        ]
    },
    {
        "mention": "nicole lapinski",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Lisa Lapinski",
                "confidence_score": "-10.767877"
            }
        ]
    },
    {
        "mention": "laurel parker-chan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "2865 Laurel",
                "confidence_score": "-11.932832"
            }
        ]
    },
    {
        "mention": "samantha hufcut",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Samantha Jo",
                "confidence_score": "-11.668341"
            }
        ]
    },
    {
        "mention": "jamie alexander",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Scott Alexander",
                "confidence_score": "-11.429856"
            }
        ]
    },
    {
        "mention": "korianne moore",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Carissa Moore",
                "confidence_score": "-11.223605"
            }
        ]
    },
    {
        "mention": "caceres, caceres, caceres, caceres",
        "type": "ORG",
        "predictions": [
            {
                "entity": "C\u00e1ceres, Spain",
                "confidence_score": "-5.076136"
            }
        ]
    },
    {
        "mention": "dexter",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Dexter, Maine",
                "confidence_score": "-6.3326216"
            }
        ]
    },
    {
        "mention": "martin",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Martin, Tennessee",
                "confidence_score": "-4.9932456"
            }
        ]
    },
    {
        "mention": "wilson",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Wilson County, Tennessee",
                "confidence_score": "-6.240399"
            }
        ]
    },
    {
        "mention": "thompson",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Thompson (electoral district)",
                "confidence_score": "-7.4668307"
            }
        ]
    },
    {
        "mention": "jillian donlon",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Roger Donlon",
                "confidence_score": "-11.569699"
            }
        ]
    },
    {
        "mention": "julia laquidara",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Laquidara",
                "confidence_score": "-11.994839"
            }
        ]
    },
    {
        "mention": "erin mcgoldrick",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ryan McGoldrick",
                "confidence_score": "-10.974271"
            }
        ]
    },
    {
        "mention": "dana schmidt",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Dana Vollmer",
                "confidence_score": "-11.212751"
            }
        ]
    },
    {
        "mention": "kaitlyn smolar",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Eugeniusz Smolar",
                "confidence_score": "-11.401817"
            }
        ]
    },
    {
        "mention": "molly miller",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Molly Miller",
                "confidence_score": "0.7731775"
            }
        ]
    },
    {
        "mention": "jordan thompson",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jordan Thompson (American football)",
                "confidence_score": "-8.184998"
            }
        ]
    },
    {
        "mention": "magee",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Samantha Magee",
                "confidence_score": "-8.47456"
            }
        ]
    },
    {
        "mention": "davis",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Josh Davis (swimmer)",
                "confidence_score": "-10.500339"
            }
        ]
    },
    {
        "mention": "scelta",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Liselotta Lennartsson",
                "confidence_score": "-11.690816"
            }
        ]
    },
    {
        "mention": "mickayla dexter",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Sally Dexter",
                "confidence_score": "-10.85406"
            }
        ]
    },
    {
        "mention": "alyssa martin",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Alyssa Miller",
                "confidence_score": "-11.371221"
            }
        ]
    },
    {
        "mention": "littell",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Franklin Littell",
                "confidence_score": "-5.129702"
            }
        ]
    },
    {
        "mention": "flannery",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Flannery",
                "confidence_score": "7.5777273"
            }
        ]
    },
    {
        "mention": "kingston 93",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Volleyball at the 2010 Central American and Caribbean Games",
                "confidence_score": "-12.00921"
            }
        ]
    },
    {
        "mention": "rhinebeck",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Rhinebeck (town), New York",
                "confidence_score": "-4.6632495"
            }
        ]
    },
    {
        "mention": "vitale",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Jonathan Vitale",
                "confidence_score": "-9.491218"
            }
        ]
    },
    {
        "mention": "sistare, behrens, dahl",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sandris B\u0113rzi\u0146\u0161",
                "confidence_score": "-11.992166"
            }
        ]
    },
    {
        "mention": "taylor robey",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Taylor Spivey",
                "confidence_score": "-11.465317"
            }
        ]
    },
    {
        "mention": "miya vitale",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ami Vitale",
                "confidence_score": "-10.703291"
            }
        ]
    },
    {
        "mention": "kelly manley",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jake Manley",
                "confidence_score": "-11.340362"
            }
        ]
    },
    {
        "mention": "brianna stinemire",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "St. Francis Xavier women's ice hockey",
                "confidence_score": "-11.303053"
            }
        ]
    },
    {
        "mention": "helen wunderlich",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Renner Wunderlich",
                "confidence_score": "-11.234546"
            }
        ]
    },
    {
        "mention": "erin leirey",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Aliz\u00e9e Brien",
                "confidence_score": "-11.943175"
            }
        ]
    },
    {
        "mention": "kingston",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Kingston, Jamaica",
                "confidence_score": "8.18341"
            }
        ]
    },
    {
        "mention": "leirey",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Water polo at the 2011 World Aquatics Championships \u2013 Men's team rosters",
                "confidence_score": "-11.918557"
            }
        ]
    },
    {
        "mention": "mitchell",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mitchell County, North Carolina",
                "confidence_score": "-7.5610466"
            }
        ]
    },
    {
        "mention": "kathryn sistare",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Kathryn Deane",
                "confidence_score": "-11.492135"
            }
        ]
    },
    {
        "mention": "rhinebeck 0-1",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Rhinebeck (town), New York",
                "confidence_score": "-8.878014"
            }
        ]
    },
    {
        "mention": "minisink valley 112",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Minisink",
                "confidence_score": "-5.292115"
            }
        ]
    },
    {
        "mention": "grace murray",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Grace Murray",
                "confidence_score": "-4.2593913"
            }
        ]
    },
    {
        "mention": "abigail gibbons",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Abigail Hopper Gibbons",
                "confidence_score": "-4.7268777"
            }
        ]
    },
    {
        "mention": "jennifer borek",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jennifer Ketcham",
                "confidence_score": "-11.305047"
            }
        ]
    },
    {
        "mention": "erin clifford",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Brendan Clifford",
                "confidence_score": "-11.475499"
            }
        ]
    },
    {
        "mention": "melina blezy",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Melina Almodovar",
                "confidence_score": "-11.212278"
            }
        ]
    },
    {
        "mention": "madeline callahan",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Madeline Weeks",
                "confidence_score": "-10.990237"
            }
        ]
    },
    {
        "mention": "scanlon",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Scanlon, Minnesota",
                "confidence_score": "-7.4168653"
            }
        ]
    },
    {
        "mention": "losauro",
        "type": "ORG",
        "predictions": [
            {
                "entity": "2017\u201318 Biathlon World Cup \u2013 Nation Men",
                "confidence_score": "-11.837165"
            }
        ]
    },
    {
        "mention": "middletown 1-2",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Middletown, Orange County, New York",
                "confidence_score": "-8.173306"
            }
        ]
    },
    {
        "mention": "newburgh free academy",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Newburgh Free Academy",
                "confidence_score": "10.947212"
            }
        ]
    },
    {
        "mention": "roosevelt",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Theodore Roosevelt",
                "confidence_score": "-0.7063169"
            }
        ]
    },
    {
        "mention": "emma reining",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Eloise Wellings",
                "confidence_score": "-11.348949"
            }
        ]
    },
    {
        "mention": "jamie forbes",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "DJ Forbes",
                "confidence_score": "-9.272473"
            }
        ]
    },
    {
        "mention": "nfa",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Namibia Football Association",
                "confidence_score": "-8.300901"
            }
        ]
    },
    {
        "mention": "ashley patierno",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ashley Tesoro",
                "confidence_score": "-11.493422"
            }
        ]
    },
    {
        "mention": "fdr",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Athletics at the 2016 Summer Olympics \u2013 Women's 200 metres",
                "confidence_score": "-11.99917"
            }
        ]
    },
    {
        "mention": "mercedes dewitt",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mercedes Gonzalez",
                "confidence_score": "-11.6119"
            }
        ]
    },
    {
        "mention": "kaylee debenedictus",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Lisa DeBenedictis",
                "confidence_score": "-11.376528"
            }
        ]
    },
    {
        "mention": "sydney reede",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Danielle Drady-Harte",
                "confidence_score": "-11.871711"
            }
        ]
    },
    {
        "mention": "amber diehl",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Edith Diehl",
                "confidence_score": "-10.898139"
            }
        ]
    },
    {
        "mention": "mahoney, giesler",
        "type": "ORG",
        "predictions": [
            {
                "entity": "James E. Mahoney",
                "confidence_score": "-10.010175"
            }
        ]
    },
    {
        "mention": "colaizzo,",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Anthony Colaizzo",
                "confidence_score": "-7.8121324"
            }
        ]
    },
    {
        "mention": "dosio",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Dosolo",
                "confidence_score": "-11.105489"
            }
        ]
    },
    {
        "mention": "valley central 101",
        "type": "LOC",
        "predictions": [
            {
                "entity": "List of New York area codes",
                "confidence_score": "-10.60297"
            }
        ]
    },
    {
        "mention": "marlboro",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Marlboro Township, New Jersey",
                "confidence_score": "-7.3392353"
            }
        ]
    },
    {
        "mention": "callo",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Callabo",
                "confidence_score": "-11.269776"
            }
        ]
    },
    {
        "mention": "flynn",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Flynn",
                "confidence_score": "-1.6803088"
            }
        ]
    },
    {
        "mention": "benkert",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Joseph Benkert",
                "confidence_score": "-6.5621977"
            }
        ]
    },
    {
        "mention": "demassio",
        "type": "ORG",
        "predictions": [
            {
                "entity": "DeMeco Ryans",
                "confidence_score": "-11.537514"
            }
        ]
    },
    {
        "mention": "daniela lusuriello",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Daniela Sruoga",
                "confidence_score": "-12.154161"
            }
        ]
    },
    {
        "mention": "vc",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Paragliding at the 2018 Asian Games \u2013 Women's individual accuracy",
                "confidence_score": "-11.93056"
            }
        ]
    },
    {
        "mention": "erin carmody",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Erin Carmody",
                "confidence_score": "4.873707"
            }
        ]
    },
    {
        "mention": "emily boldrin",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Emily Bold",
                "confidence_score": "-9.389087"
            }
        ]
    },
    {
        "mention": "antonia demassio",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Nancy Dell'Olio",
                "confidence_score": "-11.596481"
            }
        ]
    },
    {
        "mention": "garcia",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Garc\u00eda Municipality",
                "confidence_score": "-10.249774"
            }
        ]
    },
    {
        "mention": "carmody",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Anson Carmody",
                "confidence_score": "-6.5828695"
            }
        ]
    },
    {
        "mention": "allison armistead",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Allison Schmitt",
                "confidence_score": "-11.507933"
            }
        ]
    },
    {
        "mention": "camuso",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Caluso",
                "confidence_score": "-10.479674"
            }
        ]
    },
    {
        "mention": "jason day",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jason Day",
                "confidence_score": "11.941165"
            }
        ]
    },
    {
        "mention": "rory mcilroy",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Rory McIlroy",
                "confidence_score": "11.389794"
            }
        ]
    },
    {
        "mention": "the bmw championship",
        "type": "ORG",
        "predictions": [
            {
                "entity": "BMW Championship (PGA Tour)",
                "confidence_score": "0.024830054"
            }
        ]
    },
    {
        "mention": "east lake",
        "type": "LOC",
        "predictions": [
            {
                "entity": "East Lake Golf Club",
                "confidence_score": "-3.0913558"
            }
        ]
    },
    {
        "mention": "jordan spieth",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jordan Spieth",
                "confidence_score": "10.039963"
            }
        ]
    },
    {
        "mention": "rickie fowler",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Rickie Fowler",
                "confidence_score": "10.169225"
            }
        ]
    },
    {
        "mention": "henrik stenson",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Henrik Stenson",
                "confidence_score": "5.985124"
            }
        ]
    },
    {
        "mention": "bubba watson",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bubba Watson",
                "confidence_score": "10.985373"
            }
        ]
    },
    {
        "mention": "stenson",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Henrik Stenson",
                "confidence_score": "2.7928336"
            }
        ]
    },
    {
        "mention": "ryder cup",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Ryder Cup",
                "confidence_score": "10.158486"
            }
        ]
    },
    {
        "mention": "jb holmes",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "J. B. Holmes",
                "confidence_score": "-9.298498"
            }
        ]
    },
    {
        "mention": "spieth",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Jordan Spieth",
                "confidence_score": "-5.241715"
            }
        ]
    },
    {
        "mention": "mcilroy",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Greg McElroy",
                "confidence_score": "-8.939454"
            }
        ]
    },
    {
        "mention": "cv",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "National symbols of Croatia",
                "confidence_score": "-11.440146"
            }
        ]
    },
    {
        "mention": "billy horschel",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Billy Horschel",
                "confidence_score": "6.675257"
            }
        ]
    },
    {
        "mention": "cheder chabad",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Cheder",
                "confidence_score": "-0.84406376"
            }
        ]
    },
    {
        "mention": "monsey",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Monsey, New York",
                "confidence_score": "0.4643005"
            }
        ]
    },
    {
        "mention": "torah",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Torah",
                "confidence_score": "8.946289"
            }
        ]
    },
    {
        "mention": "abundance",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Flourishing",
                "confidence_score": "-8.196713"
            }
        ]
    },
    {
        "mention": "parsha perspective",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Paroksha",
                "confidence_score": "-8.532312"
            }
        ]
    },
    {
        "mention": "binyomin philipson",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Binyomin Adler",
                "confidence_score": "-9.807174"
            }
        ]
    },
    {
        "mention": "ellen",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Ellen DeGeneres",
                "confidence_score": "-5.1869545"
            }
        ]
    },
    {
        "mention": "judaism",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Judaism",
                "confidence_score": "6.8220835"
            }
        ]
    },
    {
        "mention": "the land of\u00a0israel",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Land of Israel",
                "confidence_score": "4.692332"
            }
        ]
    },
    {
        "mention": "vayelech",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Vayelech",
                "confidence_score": "11.061332"
            }
        ]
    },
    {
        "mention": "devorim",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Devarim (parsha)",
                "confidence_score": "-1.3112812"
            }
        ]
    },
    {
        "mention": "mitzvah",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Mitzvah",
                "confidence_score": "8.151362"
            }
        ]
    },
    {
        "mention": "moses",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Moses",
                "confidence_score": "9.2273"
            }
        ]
    },
    {
        "mention": "shemittah",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Shmita",
                "confidence_score": "-3.483737"
            }
        ]
    },
    {
        "mention": "temple",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Temple",
                "confidence_score": "6.2415586"
            }
        ]
    },
    {
        "mention": "sinai",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Sinai Peninsula",
                "confidence_score": "-1.5670881"
            }
        ]
    },
    {
        "mention": "moshe",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Moses",
                "confidence_score": "-3.526408"
            }
        ]
    },
    {
        "mention": "chabad.org",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Chabad.org",
                "confidence_score": "12.246767"
            }
        ]
    },
    {
        "mention": "feldman",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Yael Feldman",
                "confidence_score": "-6.658401"
            }
        ]
    },
    {
        "mention": "shabbos",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Shabbat",
                "confidence_score": "9.610666"
            }
        ]
    },
    {
        "mention": "the gold king mine",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Newmont Goldcorp",
                "confidence_score": "-8.816589"
            }
        ]
    },
    {
        "mention": "david weindorf",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "David Weinberger",
                "confidence_score": "-10.766471"
            }
        ]
    },
    {
        "mention": "texas tech university",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Texas Tech University",
                "confidence_score": "11.029931"
            }
        ]
    },
    {
        "mention": "kevin lombard",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "John Lombard",
                "confidence_score": "-11.706773"
            }
        ]
    },
    {
        "mention": "new mexico state university",
        "type": "ORG",
        "predictions": [
            {
                "entity": "New Mexico State University",
                "confidence_score": "9.9945755"
            }
        ]
    },
    {
        "mention": "the animas river",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Animas River (Mexico)",
                "confidence_score": "-1.7553355"
            }
        ]
    },
    {
        "mention": "weindorf",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Wei\u00dfendorf",
                "confidence_score": "-7.3497753"
            }
        ]
    },
    {
        "mention": "animas",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Animas River",
                "confidence_score": "-7.026639"
            }
        ]
    },
    {
        "mention": "the environmental protection agency",
        "type": "ORG",
        "predictions": [
            {
                "entity": "United States Environmental Protection Agency",
                "confidence_score": "7.655483"
            }
        ]
    },
    {
        "mention": "epa",
        "type": "ORG",
        "predictions": [
            {
                "entity": "United States Environmental Protection Agency",
                "confidence_score": "10.967144"
            }
        ]
    },
    {
        "mention": "the navajo nation",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Navajo Nation",
                "confidence_score": "7.488562"
            }
        ]
    },
    {
        "mention": "weindorf and lombard",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Delta and Dawn",
                "confidence_score": "-11.402548"
            }
        ]
    },
    {
        "mention": "laura allen",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Laura Allen",
                "confidence_score": "1.1854126"
            }
        ]
    },
    {
        "mention": "the natural resources conservation service",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Natural Resources Conservation Service",
                "confidence_score": "8.929482"
            }
        ]
    },
    {
        "mention": "the u.s. department of agriculture",
        "type": "ORG",
        "predictions": [
            {
                "entity": "United States Department of Agriculture",
                "confidence_score": "6.2610035"
            }
        ]
    },
    {
        "mention": "the nmsu agricultural science center",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Texas Tech University College of Agricultural Sciences & Natural Resources",
                "confidence_score": "-11.344987"
            }
        ]
    },
    {
        "mention": "san juan river",
        "type": "LOC",
        "predictions": [
            {
                "entity": "San Juan River (Colorado River tributary)",
                "confidence_score": "1.7076807"
            }
        ]
    },
    {
        "mention": "russell begaye",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Russell Begaye",
                "confidence_score": "12.496789"
            }
        ]
    },
    {
        "mention": "begaye",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Russell Begaye",
                "confidence_score": "4.304874"
            }
        ]
    },
    {
        "mention": "u.s. house",
        "type": "ORG",
        "predictions": [
            {
                "entity": "United States House of Representatives",
                "confidence_score": "0.6208067"
            }
        ]
    },
    {
        "mention": "mark salley",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jerry Salley",
                "confidence_score": "-11.239314"
            }
        ]
    },
    {
        "mention": "the new mexico environment department",
        "type": "ORG",
        "predictions": [
            {
                "entity": "New Mexico Environment Department",
                "confidence_score": "6.416036"
            }
        ]
    },
    {
        "mention": "weindorf and lombard's",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Gable and Lombard",
                "confidence_score": "-10.788187"
            }
        ]
    },
    {
        "mention": "allison majure",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Allison Strong",
                "confidence_score": "-11.114906"
            }
        ]
    },
    {
        "mention": "follow dan elliott",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "2014 Dubai 24 Hour",
                "confidence_score": "-11.71274"
            }
        ]
    },
    {
        "mention": "santander",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Banco Santander",
                "confidence_score": "3.9781563"
            }
        ]
    },
    {
        "mention": "tesco bank",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Tesco Bank",
                "confidence_score": "9.60124"
            }
        ]
    },
    {
        "mention": "tesco bank current account",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Current account",
                "confidence_score": "-3.4769037"
            }
        ]
    },
    {
        "mention": "clubcard",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Tesco Clubcard",
                "confidence_score": "9.392434"
            }
        ]
    },
    {
        "mention": "david mccreadle",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "David McCorkell",
                "confidence_score": "-10.86278"
            }
        ]
    },
    {
        "mention": "santander's",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Banco Santander",
                "confidence_score": "0.8478794"
            }
        ]
    },
    {
        "mention": "tesco",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Tesco",
                "confidence_score": "11.188067"
            }
        ]
    },
    {
        "mention": "nationwide flexdirect",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Flexible Support Fund",
                "confidence_score": "-11.492518"
            }
        ]
    },
    {
        "mention": "tsb",
        "type": "ORG",
        "predictions": [
            {
                "entity": "TSB Bank (United Kingdom)",
                "confidence_score": "-2.0135615"
            }
        ]
    },
    {
        "mention": "lloyds bank",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Lloyds Bank",
                "confidence_score": "9.460269"
            }
        ]
    },
    {
        "mention": "bank of scotland's",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Bank of Scotland",
                "confidence_score": "9.490842"
            }
        ]
    },
    {
        "mention": "vantage",
        "type": "ORG",
        "predictions": [
            {
                "entity": "VantageScore",
                "confidence_score": "-7.7273316"
            }
        ]
    },
    {
        "mention": "halifax",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Halifax (bank)",
                "confidence_score": "3.4195416"
            }
        ]
    },
    {
        "mention": "the world health organization",
        "type": "ORG",
        "predictions": [
            {
                "entity": "World Health Organization",
                "confidence_score": "5.650107"
            }
        ]
    },
    {
        "mention": "the state of the world's",
        "type": "ORG",
        "predictions": [
            {
                "entity": "State of the World (book series)",
                "confidence_score": "-8.3590765"
            }
        ]
    },
    {
        "mention": "antibiotics",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bacteria",
                "confidence_score": "-5.8078856"
            }
        ]
    },
    {
        "mention": "the sunday times",
        "type": "ORG",
        "predictions": [
            {
                "entity": "The Sunday Times",
                "confidence_score": "6.922102"
            }
        ]
    },
    {
        "mention": "victim support",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Victim Support",
                "confidence_score": "5.111811"
            }
        ]
    },
    {
        "mention": "lucy hastings",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Lucy Hastings",
                "confidence_score": "-0.8045176"
            }
        ]
    },
    {
        "mention": "bbc",
        "type": "ORG",
        "predictions": [
            {
                "entity": "BBC",
                "confidence_score": "10.874385"
            }
        ]
    },
    {
        "mention": "thornhill trojans",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Thornhill Trojans",
                "confidence_score": "8.277642"
            }
        ]
    },
    {
        "mention": "national conference league division",
        "type": "ORG",
        "predictions": [
            {
                "entity": "National Conference League",
                "confidence_score": "-3.074806"
            }
        ]
    },
    {
        "mention": "crosfields",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "John Crosland",
                "confidence_score": "-11.448372"
            }
        ]
    },
    {
        "mention": "mindaugas bendikas",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mindaugas Lukauskis",
                "confidence_score": "-11.73491"
            }
        ]
    },
    {
        "mention": "liam morley",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Chris Morley",
                "confidence_score": "-10.95963"
            }
        ]
    },
    {
        "mention": "thornhill",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Thornhill, Southampton",
                "confidence_score": "-4.7125278"
            }
        ]
    },
    {
        "mention": "danny ratcliffe",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Samuel D. Ratcliffe",
                "confidence_score": "-10.872925"
            }
        ]
    },
    {
        "mention": "joel gibson",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Joel Reynolds",
                "confidence_score": "-11.764317"
            }
        ]
    },
    {
        "mention": "crosfield",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "John Crosfield",
                "confidence_score": "-7.11296"
            }
        ]
    },
    {
        "mention": "nathan taylor",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Nathan Taylor (footballer)",
                "confidence_score": "-4.915008"
            }
        ]
    },
    {
        "mention": "taylor",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Taylor Caniff",
                "confidence_score": "-10.620693"
            }
        ]
    },
    {
        "mention": "jack reid",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jack Reid",
                "confidence_score": "6.0778146"
            }
        ]
    },
    {
        "mention": "ben kendall",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "FC Kendall",
                "confidence_score": "-11.33881"
            }
        ]
    },
    {
        "mention": "luke haigh",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Luke Haines",
                "confidence_score": "-10.785933"
            }
        ]
    },
    {
        "mention": "gibson",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "William Gibson",
                "confidence_score": "-5.961008"
            }
        ]
    },
    {
        "mention": "morley",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Morley, West Yorkshire",
                "confidence_score": "-1.9889169"
            }
        ]
    },
    {
        "mention": "ian bowden",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Andrew Bowden",
                "confidence_score": "-10.010528"
            }
        ]
    },
    {
        "mention": "joe moore",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Joe Moore (musician)",
                "confidence_score": "-6.87881"
            }
        ]
    },
    {
        "mention": "pleasanton - hall of fame",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Pleasanton, California",
                "confidence_score": "-2.8786216"
            }
        ]
    },
    {
        "mention": "john madden",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "John Madden",
                "confidence_score": "11.682927"
            }
        ]
    },
    {
        "mention": "rose hotel",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Rose Hotel",
                "confidence_score": "1.7670436"
            }
        ]
    },
    {
        "mention": "nfl",
        "type": "ORG",
        "predictions": [
            {
                "entity": "National Football League",
                "confidence_score": "8.221288"
            }
        ]
    },
    {
        "mention": "emmy award",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Emmy Award",
                "confidence_score": "10.869496"
            }
        ]
    },
    {
        "mention": "jim tomsula",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jim Tomsula",
                "confidence_score": "11.082054"
            }
        ]
    },
    {
        "mention": "jim harbaugh",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jim Harbaugh",
                "confidence_score": "10.54531"
            }
        ]
    },
    {
        "mention": "kaepernick",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Colin Kaepernick",
                "confidence_score": "5.9220734"
            }
        ]
    },
    {
        "mention": "the university of nevada",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Nevada Wolf Pack football",
                "confidence_score": "-2.3625658"
            }
        ]
    },
    {
        "mention": "patriots",
        "type": "ORG",
        "predictions": [
            {
                "entity": "New England Patriots",
                "confidence_score": "-2.5284922"
            }
        ]
    },
    {
        "mention": "navorro bowman",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "NaVorro Bowman",
                "confidence_score": "8.659656"
            }
        ]
    },
    {
        "mention": "bowman",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bowman Yachts",
                "confidence_score": "-8.199382"
            }
        ]
    },
    {
        "mention": "smiths",
        "type": "LOC",
        "predictions": [
            {
                "entity": "The Smiths",
                "confidence_score": "-3.8831403"
            }
        ]
    },
    {
        "mention": "aldon",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Aldon Inc.",
                "confidence_score": "0.4691499"
            }
        ]
    },
    {
        "mention": "justin",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Justin Smith (American actor)",
                "confidence_score": "-2.9675663"
            }
        ]
    },
    {
        "mention": "justin smith",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Justin Smith (American actor)",
                "confidence_score": "1.785391"
            }
        ]
    },
    {
        "mention": "patrick",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Patrick Kane",
                "confidence_score": "-9.736839"
            }
        ]
    },
    {
        "mention": "matt millen",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Matt Millen",
                "confidence_score": "12.192708"
            }
        ]
    },
    {
        "mention": "frank gore",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Frank Gore",
                "confidence_score": "9.271398"
            }
        ]
    },
    {
        "mention": "jarryd hayne",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jarryd Hayne",
                "confidence_score": "-0.44328946"
            }
        ]
    },
    {
        "mention": "reggie bush",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Reggie Bush",
                "confidence_score": "5.226647"
            }
        ]
    },
    {
        "mention": "bruce ellington",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bruce Ellington",
                "confidence_score": "4.821603"
            }
        ]
    },
    {
        "mention": "jimmie ward",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jimmie Ward",
                "confidence_score": "3.2252228"
            }
        ]
    },
    {
        "mention": "mike davis",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mike Davis (wrestler)",
                "confidence_score": "-3.1626894"
            }
        ]
    },
    {
        "mention": "reggie",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Reggie King",
                "confidence_score": "-8.994649"
            }
        ]
    },
    {
        "mention": "eddie debartolo",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Edward J. DeBartolo Jr.",
                "confidence_score": "3.492275"
            }
        ]
    },
    {
        "mention": "the pro football hall of fame",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Pro Football Hall of Fame",
                "confidence_score": "8.166419"
            }
        ]
    },
    {
        "mention": "eddie",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Eddie George",
                "confidence_score": "-7.871673"
            }
        ]
    },
    {
        "mention": "paul tagliabue",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Paul Tagliabue",
                "confidence_score": "11.653808"
            }
        ]
    },
    {
        "mention": "follow cam",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Cam (singer)",
                "confidence_score": "-9.775342"
            }
        ]
    },
    {
        "mention": "twitter.com/caminman",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Twitter",
                "confidence_score": "0.78926146"
            }
        ]
    },
    {
        "mention": "rob quinn",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Rob Quinn",
                "confidence_score": "-1.8450193"
            }
        ]
    },
    {
        "mention": "mcdonald",
        "type": "ORG",
        "predictions": [
            {
                "entity": "McDonald's",
                "confidence_score": "0.22948773"
            }
        ]
    },
    {
        "mention": "mcdonald's",
        "type": "ORG",
        "predictions": [
            {
                "entity": "McDonald's",
                "confidence_score": "8.3873825"
            }
        ]
    },
    {
        "mention": "bloomberg",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bloomberg Television",
                "confidence_score": "1.7210228"
            }
        ]
    },
    {
        "mention": "panda express'",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Panda Express",
                "confidence_score": "6.147899"
            }
        ]
    },
    {
        "mention": "wildcats",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Northwestern Wildcats football",
                "confidence_score": "-7.7032313"
            }
        ]
    },
    {
        "mention": "axtell",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Axtell, Kansas",
                "confidence_score": "-6.10211"
            }
        ]
    },
    {
        "mention": "marysville",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Marysville, Kansas",
                "confidence_score": "-3.769655"
            }
        ]
    },
    {
        "mention": "north central kansas league",
        "type": "ORG",
        "predictions": [
            {
                "entity": "North Central Kansas",
                "confidence_score": "-7.4253964"
            }
        ]
    },
    {
        "mention": "wamego",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Wamego, Kansas",
                "confidence_score": "-2.3050907"
            }
        ]
    },
    {
        "mention": "tvl",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Pacific View League",
                "confidence_score": "-11.492096"
            }
        ]
    },
    {
        "mention": "valley heights",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Valley Heights, New South Wales",
                "confidence_score": "-9.663199"
            }
        ]
    },
    {
        "mention": "onaga",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Onaga, Kansas",
                "confidence_score": "-8.279366"
            }
        ]
    },
    {
        "mention": "washington co.",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Washington County, Oregon",
                "confidence_score": "-5.774222"
            }
        ]
    },
    {
        "mention": "doniphan west 7",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Doniphan County, Kansas",
                "confidence_score": "-9.576579"
            }
        ]
    },
    {
        "mention": "tescott",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Tescott, Kansas",
                "confidence_score": "-3.2031944"
            }
        ]
    },
    {
        "mention": "blue valley randolph",
        "type": "LOC",
        "predictions": [
            {
                "entity": "Bluefield, Virginia",
                "confidence_score": "-11.162382"
            }
        ]
    },
    {
        "mention": "linn",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Linn County, Missouri",
                "confidence_score": "-4.0896344"
            }
        ]
    },
    {
        "mention": "hodgeman co.",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Hodgeman County, Kansas",
                "confidence_score": "-6.6432586"
            }
        ]
    },
    {
        "mention": "lincoln",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Lincoln, Illinois",
                "confidence_score": "-2.6626446"
            }
        ]
    },
    {
        "mention": "clifton-clyde",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Clifton, Michigan",
                "confidence_score": "-9.982541"
            }
        ]
    },
    {
        "mention": "beloit",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Beloit, Wisconsin",
                "confidence_score": "-0.0074373754"
            }
        ]
    },
    {
        "mention": "sacred heart",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Sacred Heart Church (Eau Claire, Wisconsin)",
                "confidence_score": "-10.85104"
            }
        ]
    },
    {
        "mention": "herington",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Herington, Kansas",
                "confidence_score": "0.05050886"
            }
        ]
    },
    {
        "mention": "nemaha central 72",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Nemaha County, Kansas",
                "confidence_score": "-9.036435"
            }
        ]
    },
    {
        "mention": "holton",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Holton, Kansas",
                "confidence_score": "-7.705096"
            }
        ]
    },
    {
        "mention": "lacrosse",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Lacrosse",
                "confidence_score": "6.935642"
            }
        ]
    },
    {
        "mention": "royal valley 40",
        "type": "LOC",
        "predictions": [
            {
                "entity": "U.S. Route 40 in Colorado",
                "confidence_score": "-9.92212"
            }
        ]
    },
    {
        "mention": "santa fe trail",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Santa Fe Trail",
                "confidence_score": "4.840313"
            }
        ]
    },
    {
        "mention": "solomon",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Solomon, Kansas",
                "confidence_score": "-8.1391115"
            }
        ]
    },
    {
        "mention": "rural vista",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Alta Vista Township, Lincoln County, Minnesota",
                "confidence_score": "-11.813685"
            }
        ]
    },
    {
        "mention": "conned",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Connexions Card",
                "confidence_score": "-10.615592"
            }
        ]
    },
    {
        "mention": "kidderminster",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Kidderminster",
                "confidence_score": "7.823537"
            }
        ]
    },
    {
        "mention": "moneygram",
        "type": "ORG",
        "predictions": [
            {
                "entity": "MoneyGram",
                "confidence_score": "4.0431647"
            }
        ]
    },
    {
        "mention": "west mercia",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Mercia",
                "confidence_score": "-3.9671574"
            }
        ]
    },
    {
        "mention": "pok\u00e9mon rumble u 3d publisher",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Pok\u00e9mon Rumble U",
                "confidence_score": "3.670584"
            }
        ]
    },
    {
        "mention": "arena battle",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Multiplayer online battle arena",
                "confidence_score": "-6.7238603"
            }
        ]
    },
    {
        "mention": "launch trailer",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Teaser campaign",
                "confidence_score": "-2.0563421"
            }
        ]
    },
    {
        "mention": "pok\u00e9mon rumble u",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Pok\u00e9mon Rumble U",
                "confidence_score": "9.702672"
            }
        ]
    },
    {
        "mention": "nfc",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Near-field communication",
                "confidence_score": "6.076406"
            }
        ]
    },
    {
        "mention": "toy pok\u00e9mon",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Toy",
                "confidence_score": "-6.1236506"
            }
        ]
    },
    {
        "mention": "defeated pok\u00e9mon",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "List of generation II Pok\u00e9mon",
                "confidence_score": "-10.143596"
            }
        ]
    },
    {
        "mention": "pok\u00e9mon",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Pok\u00e9mon",
                "confidence_score": "9.174415"
            }
        ]
    },
    {
        "mention": "swot",
        "type": "ORG",
        "predictions": [
            {
                "entity": "SWOT analysis",
                "confidence_score": "-0.3342776"
            }
        ]
    },
    {
        "mention": "seattle public schools",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Seattle Public Schools",
                "confidence_score": "11.890267"
            }
        ]
    },
    {
        "mention": "the teachers union",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Trade union",
                "confidence_score": "-5.1864924"
            }
        ]
    },
    {
        "mention": "phyllis campano",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Phyllis Young",
                "confidence_score": "-11.43614"
            }
        ]
    },
    {
        "mention": "bureau of labor statistics",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Bureau of Labor Statistics",
                "confidence_score": "11.860587"
            }
        ]
    },
    {
        "mention": "the national compensation survey",
        "type": "ORG",
        "predictions": [
            {
                "entity": "National Compensation Survey",
                "confidence_score": "11.324277"
            }
        ]
    },
    {
        "mention": "richard vedder",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Richard Vedder",
                "confidence_score": "10.057748"
            }
        ]
    },
    {
        "mention": "bls",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Bureau of Labor Statistics",
                "confidence_score": "7.153271"
            }
        ]
    },
    {
        "mention": "ncs",
        "type": "ORG",
        "predictions": [
            {
                "entity": "National Council for Science and the Environment",
                "confidence_score": "-10.53114"
            }
        ]
    },
    {
        "mention": "the manhattan institute\u2019s",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Manhattan Institute for Policy Research",
                "confidence_score": "-4.041488"
            }
        ]
    },
    {
        "mention": "jay greene",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jay Greene",
                "confidence_score": "7.676336"
            }
        ]
    },
    {
        "mention": "marcus winters",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Dean Winters",
                "confidence_score": "-10.798395"
            }
        ]
    },
    {
        "mention": "seattle teachers on strike",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Seattle General Strike",
                "confidence_score": "-8.813672"
            }
        ]
    },
    {
        "mention": "redstate",
        "type": "ORG",
        "predictions": [
            {
                "entity": "RedState",
                "confidence_score": "11.244413"
            }
        ]
    },
    {
        "mention": "bermuda",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Bermuda",
                "confidence_score": "8.81509"
            }
        ]
    },
    {
        "mention": "john fredriksen",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "John Fredriksen",
                "confidence_score": "10.60584"
            }
        ]
    },
    {
        "mention": "kate blankenship",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jane Blankenship",
                "confidence_score": "-10.726731"
            }
        ]
    },
    {
        "mention": "georgina e. sousa",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Georgina Mellor",
                "confidence_score": "-11.727206"
            }
        ]
    },
    {
        "mention": "jens martin jensen",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Jens Fink-Jensen",
                "confidence_score": "-11.143422"
            }
        ]
    },
    {
        "mention": "paul m. leand",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Sonja Ludvigsen",
                "confidence_score": "-11.659074"
            }
        ]
    },
    {
        "mention": "hans petter aas",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Hans Aas",
                "confidence_score": "-9.328599"
            }
        ]
    },
    {
        "mention": "the company's board of directors",
        "type": "ORG",
        "predictions": [
            {
                "entity": "Board of directors",
                "confidence_score": "0.49285454"
            }
        ]
    },
    {
        "mention": "hamilton",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Hamilton, Bermuda",
                "confidence_score": "3.0234861"
            }
        ]
    },
    {
        "mention": "frontline",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Frontline (American TV program)",
                "confidence_score": "-4.1603403"
            }
        ]
    },
    {
        "mention": "investors",
        "type": "PERSON",
        "predictions": [
            {
                "entity": "Investor",
                "confidence_score": "4.1453314"
            }
        ]
    },
    {
        "mention": "nasdaq omx corporate solutions",
        "type": "ORG",
        "predictions": [
            {
                "entity": "NASDAQ",
                "confidence_score": "-4.8444333"
            }
        ]
    }
]
]
# Initialize a dictionary to store cumulative overlap and counts by type
overlap_stats = {}

# Process each entity sample
for sample in entity_samples:
    if sample["prediction"]:
        first_prediction = sample["prediction"]["entity"]
        overlap_percentage = calculate_overlap(sample["mention"], first_prediction)
        print(f"Overlap between '{sample['mention']}' and '{first_prediction}': {overlap_percentage:.2f}%")

        # Aggregate data for average calculation
        if sample["type"] in overlap_stats:
            overlap_stats[sample["type"]]['total'] += overlap_percentage
            overlap_stats[sample["type"]]['count'] += 1
        else:
            overlap_stats[sample["type"]] = {'total': overlap_percentage, 'count': 1}

# Calculate and print the average ratio for each category
for category, stats in overlap_stats.items():
    average_overlap = stats['total'] / stats['count']
    print(f"Average overlap ratio for {category}: {average_overlap:.2f}%")
```

```python
from google.colab import drive
import json
import difflib

drive.mount('/content/drive', force_remount=True)

def calculate_overlap(mention, entity):
    mention_normalized = mention.lower()
    entity_normalized = entity.lower()
    s = difflib.SequenceMatcher(None, mention_normalized, entity_normalized)
    match = sum(n for i, j, n in s.get_matching_blocks())
    average_length = (len(mention_normalized) + len(entity_normalized)) / 2
    return (match / average_length) * 100

def process_entity_samples(entity_samples):
    print("Starting to process entity samples...")
    overlap_stats = {}
    for sample in entity_samples:
        print("Processing sample:", sample)  # Check each sample's data
        if 'predictions' in sample and sample['predictions']:  # Ensure predictions key and content
            # Assuming we still take the first prediction for processing
            first_prediction = sample['predictions'][0]['entity']
            overlap_percentage = calculate_overlap(sample['mention'], first_prediction)
            print(f"Overlap between '{sample['mention']}' and '{first_prediction}': {overlap_percentage:.2f}%")

            if sample['type'] in overlap_stats:
                overlap_stats[sample['type']]['total'] += overlap_percentage
                overlap_stats[sample['type']]['count'] += 1
            else:
                overlap_stats[sample['type']] = {'total': overlap_percentage, 'count': 1}
        else:
            print(f"No predictions available for sample: {sample}")

    # Print the average ratio for each category
    for category, stats in overlap_stats.items():
        average_overlap = stats['total'] / stats['count']
        print(f"Average overlap ratio for {category}: {average_overlap:.2f}%")


file_path = '/content/drive/My Drive/cmput_656/processeddata.json'

try:
    with open(file_path, 'r') as file:
        entity_samples = json.load(file)
        print("Loaded entity samples:", entity_samples)
    process_entity_samples(entity_samples)
except Exception as e:
    print(f"An error occurred: {e}")
```
